## 1-1. 들어가기

샘플?
- 갖고 있는 샘플 자체 -> 기술통계
- 샘플의 모집단 -> 추론통계

- 기술통계: 수치, 시각화로 샘플 요약
- 추론통계: 추정과 검정으로 모집단 추론


## 1-2. 자료의 척도

- 양적 자료(구간, 비율): 수치
- 질적 자료(명목, 서열): 연산 x

- 명목척도: 이름으로 요소 속성 식별
- 서열척도: 항목 간 순서 또는 순위
- 구간척도: 덧셈, 평균 가능. 
- 비율척도: 임의의 측정단위로 구성. 0이면 없음.

## 2-2. 3가지 주요 도구 집합

1. 중심화 경향
- 평균: 이상치에 민감
- 절사 평균: 양끝 일정 비율 삭제
- 중앙값: 이상치에 덜 민감
- 최빈값: 주로 이산 데이터. 가장 많이 나타나는 값.
2. 자료의 산포
- 범위: 최댓값과 최솟값의 차이.
- 분산: 평균에서 얼마나 퍼져있는지. 클수록 많이 퍼진 것
- 표준편차: 분산에 제곱근. 단위의 일치성 덕분에 좋음.
- 사분위수: 데이터 표본을 네 개의 동일한 크기로 나눈 것.
  - Q1: 하위 25%
  - Q2: 하위 50%, 중앙값
  - Q1: 하위 75%
  - IQR: 데이터의 중앙 50%. 이상치 탐지에 유용함.
- 변동계수: 상대적 변동성을 나타냄.
(참고) 첨도와 왜도
- 첨도: 꼬리가 두꺼운 정도
- 왜도: 비대칭 정도.
- 평균 == 중앙값이면 왜도는 0이다.
3. 자료 정리 및 시각화
- 히스토그램, 박스플롯, 파이차트

- 확률분포: 확률변수가 특정 값을 가질 확률
  - 이산과 연속으로 나뉨.


## 3-3. 대표적인 분포 모형
### 정규분포
  - 표본의 크기가 커질수록 표본평균의 분포는 모집단의 분포와 상관없이 정규분포에 가까워짐.
  - 30보다 크면 근사적으로 따름
- 대수의 법칙: 표본의 크기가 커질수록 표본 평균이 모평균에 근사 현상
- 중심극한 정리: n개의 평균의 분포는 n이 적당히 크다면 정규분포에 가까워진다.
### T분포
- 자유도가 커짐에 따라 N(0,1)에 수렴
- t-분포는 모집단 분산을 모를 때, 표본의 분산을 사용해 모분산을 추정하여 만들어짐. 
- 모집단 분산을 모를 때 표본 크기가 30보다 작으면 t, 30 이상이면 표준정규분포 씀.
### 카이제곱분포
- 오른쪽으로 긴 꼬리를 가진 비대칭 분포
### F분포
- 독립적인 두 카이 제곱 분포의 비율
- 두 집단 간의 분산 차이를 비교할 때

## 5-1. 가설 검정의 절차
- 가설 검정: 가설의 합당성 여부 판정
- 가설 설정 - 유의 수준 설정 - 검정통계량 산출 - 가설 기각/채택 판단
1. 가설설정: H0. H1 설정
- H0이 기각하고 싶은 거, H1은 주장하고 싶은 거
2. 유의 수준 설정
- 연구자가 결과를 판단할 때 허용할 수 있는 오류의 최대치
- 제1종 오류: 귀무가설이 참인데 기각할 확률
3. 검정통계량 산출
- Z score, T score, F score 사용
4. 가설 기각/채택 판단
- 유의 확률(p-value)

## 5-2. 데이터의 자료형에 따른 검정 방법
1. [수치형 - 수치형] (피어슨)상관분석
- 두 변수간 피어슨 상관계수가 얼마나 유의미한지
- 공분산: 두 변수가 함께 어떤 방향으로 변화하는지와 그 크기를 표현
- 상관계수: -1~1, 공분산을 각 변수의 표준편차로 나누어 -1과 1사이로 표준화한 것

2. [범주형 - 수치형] F-test(ANOVA), T-test
- 범주 2개 -> T test
  - t값은 두 평균 간 차이를 표준 오차로 나눈 값
  - 절대값 2를 기준으로 커질수록 차이 큼
- 범주 3개 이상 -> F test(ANOVA)
  - F-통계량: 분모 그룹과 분자 그룹의 분산비.
  - F값이 2~3이면 차이가 있다. 클수록 차이가 커짐.
  - 사후 분석을 통해 어떤 범주가 튀는지 확인함.

  3. [범주형 - 범주형] Chi2-test
  - 기대빈도와 관측빈도 차이가 크다면 독립적이지 않을 가능성이 큼.
  - 자유도는 n-1
  

## 6. 마치며

중심 극한의 정리가 중요하다고 생각한다. 중심 극한의 정리란 표본의 개수가 30개 이상으로 충분히 크면 표본 평균의 분포가 정규분포에 가까워진다는 정리이다. 모집단을 정규분포로 가정할 수 있어서, 모집단을 파악하기 어려운 경우에도 표본의 수만 확보하면 정규분포로 분석을 진행해 나갈 수 있다는 게 굉장히 큰 장점이라 생각한다. 
또, 정규분포는 z값이 다 알려져 있고 함수로 직접 계산할 수도 있으니 이 개념이 더욱 유용하다고 생각한다.
