# 1. 데이터 기반 의사결정
## 1.1. 무엇을 위한 데이터 분석인가

- 데이터 분석의 장점은 목적 달성 프로세스의 질이 높다는 것이다.
- 효율적이고 확실한 프로세스를 가동할 수 있다.
- 객관성이 높아진다.

### 액션이란 무엇인가?
- 액션은 결정, 계획, 행동으로 나뉜다.
- 결정: 뒷받침을 위한 데이터 분석 수행
- 계획: 계획 수립을 위한 데이터 분석 수행
- 일별 계획에 활용할 수 있는 분석

## 1.2. Data-Driven-Decision-Making (DDDM)
- 비지니스 결정을 위해 직관 대신 데이터 분석을 사용하는 접근 방식
- 고객 피드백 등의 소스를 활용하여 프로세스 안내하는 것이 포함됨.

## 1.3. 6 Steps of DDDM
DDDM = data driven decision making
1. 목표 정의
- 목표 명확히 하기
2. 데이터 식별, 준비, 수집
- 데이터 소스를 평가하고 준비하여 체계적으로 검증함.
3. 구성 및 탐색
- 새로운 패턴 인사이트를 발견할 수 있도록 구조화된다. 정확성과 신뢰성을 보호할 수 있다.
- 시각화를 하면 이상치를 식별할 수 있다.
4. 데이터 분석 수행
- 비지니스 전략에 영향을 주는 패턴, 상관관계, 추세를 파악한다.
5. 결론 도출
- 데이터 분석 결과 검도, 비지니스 상황에 알맞게 배치하여 권장 사항 구성
6. 구현 및 평가
- 핵심성과지표(KPI)를 기준으로 분석 결과를 측정하고 피드백을 수집한다.
- 지속적인 모니터링을 통해 개선한다.
- 경험적 증거와 피드백을 기반으로 계획을 조정할 수 있다.


## 1.4. 데이터 분석 유형
1. 설명분석: 과거 데이터를 설명 및 요약. (월별 판매 보고서 등)
2. 진단분석: 특정 사건이 발생한 이유를 파악하는 것을 목표로 함. 상관관계 식별이 포함되며 추세나 사고의 근본 원인을 파악함
3. 예측분석: 과거를 기반으로 미래를 예측함. 고객 행동을 효과적으로 예측할 수 있음.
4. 처방분석: 데이터 기반 조치를 권장함. 최상의 행동 방침을 제안함.
5. 탐색분석: 데이터에서 패턴, 이상치, 예상치 못한 상관관계를 발견할 수 있게 함.
6. 추론 분석: 샘플로 모집단을 추론함.
7. 정성적 분석: 비수치 데이터에 중점을 둠.
8. 정량적 분석: 수치 데이터를 분석하여 패턴을 발견함.
9. 실시간 분석: 즉각적인 통찰력을 제공한다.
- 넥슨의 data-driven이 데이터 분석을 잘 활용한 경우이다.


# 2. 목표 정의
## 2.1. 지표

측정하고 싶은 것을 숫자로 표현한 것이다.

- 효율성 = 산출 / 투입 (노동생산성 = 생산량 / 노동시간)
- 효과성 = 달성 결과 / 목표 (목표 달성률 = 실제 매출 / 목표 매출)
- 수익성 = 이익 / 자원 (ROE = 순이익 / 자기자본)
- 성장성 = 증가분 / 기존 규모 (매출 성장률 = (당기매출 - 전기매출) / 전기매출)
- 안정성 = 보유자산 / 부채 (부채비율 = 부채 / 자기자본)


예외
- 절대치 지표: 매출액, 직원 수, 생산량
- 지수형 지표: CSI, NPS
- 복합 지표

## 2.2. 지표를 정의하기
절대적인 수치를 구할 수 없다면 어떻게 해야 할까? 바로 Proxy Metric을 설정하면 된다.
- 추상적이어서 가정과 정의, 논의가 필요한 지표를 말한다.
- 직접 추천할 수 없으니 대체물(Proxy)을 정하여 측정한다!!
- 이를 잘 활용한 게 Stibee라는 이메일 발송 서비스이다.
  - Aha moment 후로 잔존 곡선이 달라진다!!

## 2.3 북극성 지표, L1/L2 Metrics
- 북극성 지표: 기업이 가장 중요하게 생각하는 지표(에어비엔비-예약일 수, 페이스북-DAU)
- L1: 고수준의 성과 지표, 전략 관점
- L2: L1의 세부 사항을 측정하는 전술 관점 지표

## 2.4. AARRR(해적지표)
1. Acquisition(유입): 대체로 조회수, 가입 사용자 수 등의 지표를 통해 측정한다.
- 앱 설치 수, DAU, 실행 수
2. Activation(활성화): 중요 기능을 직접 사용하는 것. 회사마다 분야마다 지표를 다양하게 설정할 수 있다.
- 상품 조회수, 장바구니 담기 수, 좋아요 수
3. Retention(재방문): 신규 사용자를 장기 사용자로 전환시키는 것.
- 전환/이탈 데이터
4. Revenue(수익): 제품/서비스를 유료로 구매하기로 결정한 단계. 소비를 유도하는 것을 중요하게 여겨야 한다. 고객의 평생 가치(LTV)를 높이고, 유치 비용은 줄이는 것이 중요하다.
- 전체 매출, LTV, ARPPU
5. Referral(추천): 어떻게 잠재 고객들에게 추천하도록 만들 것인가를 다룬다.
- 리뷰 작성 수, 공유 수, 초대 수
