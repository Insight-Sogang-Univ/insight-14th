### 판다스
- 데이터 분석에 특화된 파이썬 라이브러리로, 테이블 형태의 DataFrame을 기본 구조로 사용한다.
- NumPy의 확장판

참고) NumPy는 다차원 배열의 계산에 강하고, Pandas는 데이터 조작 및 분석에 강함

### 라이브러리
- 미리 만들어진 기능들이 있음.
- 필요에 따라 적당한 라이브러리를 불러오면 됨.
- 설치가 필요하면 !pip install 명령어 사용
- 다양한 확장자의 파일을 지원하여 활용성이 높음
- read_xxx(), to_xxx() 함수를 사용하여 불러오기 및 저장 실행

### 데이터프레임
- 판다스에서 사용하는 데이터 구조

1) Index와 Column
- index + column으로 이루어져 있음.
- 데이터프레임 내 위치를 나타낼 수 있다.
2) Series
- value + index, index가 있는 1차원 배열
- 데이터프레임에서 컬럼 하나를 꺼내면 시리즈 객체가 됨.
- 시리즈 내 요소는 모두 동일한 데이터 타입이어야 함.
- 섞여있다면 전체 데이터가 object(문자열)로 처리되며, 산술 연산이 불가능함. 자료형 변환 후 사용 가능
3) DataFrame
- 여러 개의 시리즈가 모인 2차원 배열 구조
- 컬럼 내 타입은 동일해야 함. 컬럼 관은 달라도 무관함.
- 인덱스와 칼럼명은 필요에 맞게 지정할 수 있음.

### 데이터 확인
- 분석을 위해서는 주어진 데이터에 대해 이해가 필요함.
- 대체로 데이터 크기가 방대하기 때문에 일부만 미리 확인할 수 있는 함수를 통해 데이터를 파악함.

1) head, tail
- 앞/뒤 일부 데이터를 미리 확인할 수 있음.
- df.head(n): 앞 n개 확인
- df.tail(n): 뒤 n개 확인

2) 데이터 정보 확인
- df.shape: 행, 열 개수
- df.info(): 각종 기본 정보를 반환함
- df.describe(): 수치형 열들의 기초 통계량 반환
- df.isnull(): NaN이 있는지 판단
- df.dtypes: 열들의 데이터 타입 반환

3) 고유값 및 개수 확인
- df.count(): 결측치를 제외한 데이터의 행 수 반환
- df[열].value_counts(): 특정 열에 포함된 값의 빈도수 반환
- df[열].unique(): 특정 열에 포함된 값을 종류별로 하나씩 반환

### 데이터 추출

- 데이터의 크기가 매우 크기 때문에 필요한 부분만 골라내 가공해야 함

1) Indexing & Slicing
- 인덱스와 컬럼을 사용해 데이터를 추출함.
- 인덱싱: 지목할 때 / 슬라이싱: 범위 지정

2) loc & iloc
- 행 데이터를 선택하기 위해 사용함!!
- loc: 인덱스의 이름 기준으로 선택할 때.
  - 단위 지정에 따라 행, 값, 열 모두 추출할 수 있음.
  - 인덱싱, 슬라이싱 모두 가능함.
  - 끝 인덱스 포함 반환: 이름 기준이라
- iloc: 숫자 기반 위치로 선택할 때
  - 데이터가 커지면 이게 속도 측면에서 더 유리함.
  - 끝 인덱스 포함 반환 X

3) 열단위 추출: 컬럼명
- 열단위로 데이터를 추출하고 싶으면 열의 이름으로 인덱싱하는 게 더 직관적임

4) 불리언 인덱싱
- 특정 조건에 맞는 걸 추출하고 싶을 때
- true, false의 배열로 데이터를 필터링함
- 조건 여러개를 and, or로 묶어 사용할 수도 있음.
- loc도 불리언 인덱싱이 가능함.
- iloc는 불리언 인덱싱 불가. 정수형 순서만 돼서

5) 인덱스 설정
- 인덱스를 원하는 대로 변경할 수 있음.
- df.set_index(열): 기존 열 중 하나를 인덱스로 설정
- df.reset_index(): 인덱스 리셋, 다시 열로 돌아감.
- df.reindex(인덱스 배열): 인덱스 재배열에 따라 기존 행도 인덱스를 따라 감.
- df.index: 인덱스 재배열, 기존 행은 그대로


### 데이터 조작
1) 데이터 추가, 삭제
- df['새로운 열'] = 리스트: 이 리스트를 Dataframe에 추가함.
- df.drop(labels, axis): 삭제

- column - column: 같은 인덱스끼리 연산된다. 산술 연산
- column - 숫자: 모든 값에 같은 수를 덧셈/
- NaN: NaN과의 산술연산 결과는 NaN. 사전에 처리하는 게 좋다.
- Dataframe - Dataframe: 같은 위치끼리 연산. 라벨이 불일치 하면 NaN이 생성됨.

2) 데이터 정렬
- df.sort_index(): 인덱스 기준 정렬
- df.sort_values(by='열'): 특정 열 기준 정렬

3) 데이터 변환
- .astype: 컬럼 내 자료형을 변환할 수 있음.
- .replace: 값 자체를 변환할 수 있음.
- .transpose/.T: 행과 열을 맞바꿀 수 있음.

4) 데이터프레임 병합
- concat: 단순히 이어 붙이는 방식. 동일한 구조의 df 합칠 때 사용
  - axis=0: 행 방향, axis=1: 열 방향
- merge: 공통항목 기준 병합! 형태가 달라도 겹치는 데이터 기준으로 합쳐짐.
  - on: 통해 기준 열을 정해줄 수 있음.
  - how: 결합 방식을 정해줄 수 있음.
    - how='inner': AND 연산. NaN이 생기지 않음.
    - how='outer': OR 연산. NaN이 생길 수 있음.

### 집계함수
- 여러 행으로부터 하나의 결괏값을 반환하는 함수.
- 대표값을 구할 때 사용한다: 평균, 표준편차, 합 등
- mean, median, std, sum, max, min
- 어느 방향으로 연산을 진행할지 axis=를 통해 지정해야 함.
