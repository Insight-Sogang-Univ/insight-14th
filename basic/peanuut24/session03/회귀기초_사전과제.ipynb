회귀 기초
===
*****

 머신러닝과 모델링
---
*****

### 1. 머신러닝

(1) 머신러닝: 모델을 학습하여 새로운 데이터를 예측, 결정할 수 있도록 하는 기슬.    
이를 통해 **데이터에서 패턴**을 찾아내고, **새로운 데이터에 대해 예측** 가능하게 한다.

(2) 머신러닝의 종류    
* 지도학습: 정답이 있는 데이터 사용, **예측값을 이미 만들어둔 정답과 같아지도록** 기계를 학습시키는 알고리즘.
  + labele data로 학습.
  + 회귀, 분류
  + labeled data: 독립변수(feature)와 종속변수(target/label)가 모두 존재하는 데이터.


> 지도학습의 단계
> > 1) 데이터 나누기: traning set(모델을 훈련 시킬 때 쓰는 데이터) / test set(학습된 모델 검증을 위해 쓰는 데이터)
> > 2) feature(독립변수)를 통해 추정한 예측값이 주어진 target/label(종속변수)과 같아지도록 모델을 훈련시킴.
> > 3) 기계의 예측이 우리가 의도한 정답이 되도록 지도함.
> 


* 비지도 학습: 정답이 없는 데이터 사용, 데이터 속 패턴과 유사도를 기계가 학습하도록 하는 알고리즘.
  + target, lable 없이 feature만 있는 데이터
  + 군집화(비슷한 데이터를 묶어 큰 단위로 만듦), 밀도 추정(데이터의 분포 예측), 차원 축소(데이터 차원 간추리기)

* 강화 학습: 보상을 통해 상은 최대화, 벌은 최소화 하는 방향으로 행위를 강화하는 학습.

(3) 회귀와 분류
1) 회귀(Regression): 데이터(X)를 기반으로 정답(y)을 잘 맞추게 하는 함수 찾기.    
   종속변수(예측하고자 하는 값)가 연속형일 시 사용.

2) 분류(Classification): 기존 데이터가 어떤 레이블에 속하는지 패턴을 알고리즘으로 예측, 새롭게 관측된 데이터에 대한 레이블을 판별.

*****

회귀와 모델링
---
*****

### 1. 회귀란?
* 데이터에서 패턴을 찾아내어 미래 값(연속 값)을 예측하는 것.

### 2. 회귀의 종류
* 선형 회귀, 비선형 회귀, 릿지 회귀, 라쏘 회귀, 다항 회귀 등 ...
-> 오늘 다룰 것은 선형회귀

### 3. 선형 회귀
* 선형 회귀: 데이터를 가장 잘 설명하는 회귀선(y = ax + b)을 찾는 과정.
    + 선형 회귀 기본가정: 선형성, 독립성, 등분산성, 정규성
* 종류: 단순 선형 회귀/ 다중 선형 회귀

#### (1) 단순선형회귀분석
* 하나의 독립변수만 존재하는 선형 회귀.
* 데이터를 가장 잘 설명하는 직선을 찾는 과정. (= 회귀계수와 절편을 찾는 과정)
* y=ax+b 에서 기울기 a는 회귀계수, b는 y절편이다.
* 잔차(residual, (실체값-예측값)간의 차이)를 줄여나가면 된다.


> 최소제곱법(Least Squares Method): 최적의 회귀선 찾는 방식.
> * 잔차 제곱의 합 (RSS, Sum of Squared Residuals)이 최소가 되는 지점으로 최적의회귀선을 구하는 방식.
> * 제곱하는 이유: 양수-음수 오차의 상쇄, 큰 오차에 큰 벌, 최적화 할 때 미분가능한 형태가 되도록 하기 위함이다.

##### 1. 경사하강법(Gradient Descent)
* 최소제곱법이 계산이 복잡하여 최소값 구할 때 사용. 
* 회귀계수와 y절편을 찾고, 잔차제곱을 줄이려는 목적은 같다.
* 경사 하강법 사용 경우
  1) 함수가 닫힌 상태일 때
  2) 함수가 너무 복잡해 미분계수를 구하기 어려울 때
  3) Gradient Descent 구현이 미분 계수 구하기보다 쉬울 때
  4) 데이터 양이 너무 많은 경우 효율적 계산 위해.

> 목적함수, 비용함수, 손실함수
> * 목적함수: 우리가 달성하고 싶은 목표. 최적화를 위해 오차를 최소화하는 것이 목적인 함수를 말한다. (최적화 하려는 대상을 수학적으로 표현한 함수)    
> 
> * 손실함수: 개별 데이터 샘플에 대한 오차를 측정하는 함수. 예측값과 실제값의 차이를 계산한다.    
> 
>     $ (MSE of one sample)={(y-y\hat)^2}$     
> 
> * 비용함수: 전체 데이터셋에서 평균적인 손실을 측정하는 함수. 손실함수를 모든 데이터 포인트에 대해 계산한 후 평균 of 합산. 머신러닝은 대부분 비용함수를 최적화하는 것이 목표이다. (대부분 목적함수 = 비용함수)     
> 
>     $ (MSE)=1/N \sum_{i=1}^{N}{(y_i-y_i\hat)^2}$, N은 전체 샘플 수.

(1) 손실함수
*****

> $t_i$: i번째 데이터 포인트 실제값    
> $y_i$: i번째 예측 값

> y식 대입하여 쓴 손실함수. 여기서 E를 최소화 하는 w와 b 찾는게 목표.

1. W로 x축을 구성한 경우

> 만약 꼭짓점보다 w값이 오른쪽에 있다면    
> -> 미분계수(기울기)는 양수, W는 감소해야함.    
> -> 따라서 $현재값-미분값(양수)$ 하면 왼쪽으로 점 이동.  
> 만약 꼭짓점보다 w값이 왼쪽에 있다면    
> -> 미분계수(기울기)는 음수, W는 증가해야함.    
> -> 따라서 $현재값-미분값(음수)$ 하면 오른쪽으로 점 이동.

> 이렇게 W값을 업데이트 해나가서 최솟값에 도달하면 미분계수가 0이 되므로 더이상 변화되지 않고 멈춤.

> 동일한 원리로 x축을 b로 두어 최적의 b 찾기 가능.

(2) 학습률(learning rate, $\alpha$)
*****
* 얼만큼 이동해야 하는지. (단순히 미분값만 빼면 오히려 최솟값에서 멀어지는 경우도 있음.)
* 학습률이 너무 작다면, 모델이 수렴하기까지 너무 시간 오래걸림.     
* 학습률이 너무 크다면, 모델이 수렴하지 못하고 발산(손실값이 계속 커짐)하게 됨. (값이 업데이트 될 때마다 parameter의 부호가 계속 바뀌니까)
* 학습률 설정은 개발자가 직접 설정하는 하이퍼파라미터. (경험적인 작은 값-> 실험으로 조정)

(3) Local Minima 문제
*****
* 작은 기울기에서는 조금씩, 큰 기울기에서는 많이 이동하여 발생하는 문제.
* 다음 볼록으로 가야하는데 한 곳에 갇혀 못 나옴

(4) 해결법: 모멘텀(momentum)
* 기존 경사하강법은 과거 기울기를 잊고, 현재 기울기만 기준으로 움직임    
* -> 이전의 기울기, 이동 방향을 고려하도록 관성을 부여하는 것.

#### (2) 다중선형회귀분석
*****
* 독립변수가 2개 이상인 경우의 선형 회귀.
* $y=W_1*x_1+W_2*x_2+...+W_n*x_n+b$

> * 독립변수가 2개인 경우 이렇게 3차원 공간에 표현됨. 회귀선은 평면이 된다.
> * 최소제곱법을 사용할 수 있다. 잔차는 관측치와 평면간의 거리가 됨.
> * 마찬가지로 파악이 힘들 땐 경사하강법 사용.
> * 독립변수가 3개 이상일 때에도 비슷한 원리로 확장됨.

(1) 다중공선성 (Multicollinearity)
* 독립변수끼리 강한 상관관계를 가질 때, 두 변수 간 다중공선성이 발생함.   
  
* 다중공선성이 높은 경우 어떤 독립변수가 종속변수에 얼마나 영향을 미치는 지 파악이 어려움 -> 회귀 분석의 정확도 감소    
  
* 확인 방법
  + 상관계수(r): -1과 1사이에 존재. 절댓값 1에 가까울수록 두 변수의 상관성 높음
  + VIF지수(분산 팽창 인수): 회귀모델의 결정계수 R square를 사용하여 계산된다. VIF가 높으면 다중공선성이 존재한다고 판단한다.    

* 대처 방법
  1) 변수 제거(변수 선택법): 독립변수로 사용할 변수 선택.
  2) 변수 변환: 변수들을 더하거나 빼서 새로운 변수 생성.
  3) 규제 선형 모델 활용: 릿지,라쏘,엘라스틱넷 등의 방법으로 모델의 복잡도 줄임
  4) PCA(주성분 분석): 데이터의 차원을 축소하는데 사용하는 기법. 정보는 보존하면서 데이터의 복잡성을 줄임. dataset에 별로 영향을 주지 않는 차원을 없앤다.

> * PCA 과정   
> 1) 데이터 전처리(정규화): 평균 0으로 조정, 표준화(분산 1로)
> 2) 주성분 찾기: 공분산 행렬 계산, 고유벡터와 고유값 추출 (고유벡터가 데이터가 가장 많이 퍼진 방향으로, 주성분 축임)
> 3) 차원 축소: 주성분 축에 데이터 투영하여 중요 정보만 유지
> 4) 복원 가능(선택): 원래 좌표계로 변환

### 4. 규제 선형 모델   
*****
* 모델이 학습 데이터에 과적합(overfitting) 되지 않도록 규제를 가하는 모델.    
  
* 선형 회귀 모델에서는 계수(기울기, $W_n$)의 크기를 조정하는 것. (n번째 독립변수에 $W_n$만큼의 가중치를 가진다는 뜻으로, 이 회귀계수가 클수록 해당 독립변수에 과하게 의존한다는 뜻이다.)     
  ->회귀계수에 패널티 부여

* 즉 비용함수는 최소제곱법과 회귀계수 조절이 서로 균형을 유지해야 함.    
  -> 오류는 최대한 줄이고, 특정 변수에 집착하지 않도록 균형 맞춰 과적합 방지.
        
* 방식에 따라 3가지 규제선형모델 존재 (릿지 회귀, 라쏘 회귀, 엘라스틱넷 회귀.)

#### (1) 릿지회귀 (L2 규제)
* $CostFunction=RSS+\alpha*\sum_{j=1}^{p}{{w_j}^2}$
* 회귀계수(&W_j$)의 제곱을 패널티로 주어, 큰 회귀계수에는 큰 패널티 작은 회귀계수에는 작은 패널티를 줌.
* 모든 계수의 크기는 작아지지만, 사라지지는 않음(회귀계수가 0되진 않음)
* 모든 변수 유지, 크기 적절히 줄임
*  **예측 변수 많고, 다중공선성 존재할 때**

#### (2) 라쏘회귀 (L2 규제)
* $CostFunction=RSS+\alpha*\sum_{j=1}^{p}{|w_j|}$
* 회귀계수의 절댓값을 패널티로 주어, 회귀계수의 크기와 상관없이 같은 패널티 줌.
* 작은 계수는 0이되어 일부 계수는 사라질 수 있음 -> 변수 선택 기능
* 비중이 낮은 변수(작은 계수)는 사라지고, 중요한 변수만 남음
* **일부 변수만 실제 중요하고, 모델의 해석을 간단하게 유지하고자 할 때**

#### (3) 엘라스틱넷 (L1 + L2 규제)
* $CostFunction=RSS+\alpha_1*\sum_{j=1}^{p}{{w_j}^2}+\alpha_2*\sum_{j=1}^{p}{|w_j|}$
* L1의 변수 선택 기능, L2의 가중치 규제 가능을 동시에 활용.
* 불필요한 변수는 제거하고, 안정적인 예측을 할 수 있다.
* **변수 간의 상관관계가 높을 때 릿지보다 낫다** 

### 5. 모델 평가 방법
*****
#### (1) 성능 평가 지표
*****
##### (1) 평균 제곱 오차 (Mean Square Error, MSE)
* 모델의 예측값과 실제 관측값 사이의 오차 제곱 평균.
* $(MSE)=1/n \sum_{i=1}^{n}{(y_i-y_i\hat)^2}$
* 이상치에 민감 (큰 오차에 민감하니까)

##### (2) 평균 절대 오차 (Mean Absolute Error, MAE)
* 모델의 예측값과 실제 관측값 사이의 절대값 오차 평균.
* $ (MAE)=1/n \sum_{i=1}^{n}{|y_i-y_i\hat|}$
* * 이상치가 존재해도 예측 성능 유지 (모든 오차 동일하니까)

#### (2) 변수 유의성 평가
*****

##### (1) t 검정
* 독립변수의 회귀계수가 유의미한지 검정시 사용    

* t값 계산:    
  $t=\frac{\beta\hat-0}{SE(\beta\hat)}$

    $\beta$: 추정 회귀계수,
    $SE(\beta\hat)$: 표준오차

* p-value 확인 및 판단:
  + p-value < 0.05: 유의수준 5%에서 H0 기각 -> 해당 변수 유의미함
  + p-value >= 0.05: H0 기각 불가 -> 해당 변수 유의미하지 않음.

* 해석: 
  + 회귀계수 0이면 해당 독립변수 의미X
  + p-value 작을수록 종속변수에 영향 줄 확률 상승
