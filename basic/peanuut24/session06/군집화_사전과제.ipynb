-----
### (1) 머신러닝: 비지도 학습

  -> 패턴이 아직 알려지지 않은 경우/ 패턴이 계속 변하는 경우/ 열린 문제를 해결하고 지식을 일반화해야 하는 경우에 적합!

### (2) 군집화
* 데이터를 비슷한 특성을 가진 그룹(군집)으로 나누는 비지도 학습 기법 (서로 다른 군집에 속한 개체들은 비슷하지 않아야한다.)
* 사전에 정해진 목표 없이 데이터셋의 고유한 패턴을 발견하고 추출하믄로 비지도학습이다.
* 목표
   1. 응집도(cohension) 최대화: 같은 군집에 속하는 데이터끼리는 최대한 비슷하도록 함
   2. 분리도(seperation) 최대화: 서로 다른 군집은 최대한 분리되도록 함.

### (3) 군집화 과정

2. 군집화 알고리즘 선책: 데이터의 특성, 목표에 따라 군집화 알고리즘 선택
3. 군집 유효성 검증: 얼마나 잘 되었는가
4. 결과 해석: 각 군집이 지닌 특징 분석 및 해석 
           

## 2. 군집화를 위한 데이터 준비
-----
### (1) 주요 고려 사항
1. 변수 유형 이해
   * 군집화에 사용될 피처의 종류와 특성(연속/명목, 변수 개수) 명확하게 이해. 이에따라 군집화 알고리즘을 달리 쓴다.
2. 거리/유사도 정의와 측정
   * 군집화는 데이터 간의 거리, 유사도를 기반으로 그룹을 형성함
   * 변수 유형에 따라 적합한 거리 측정 방식이 달라지므로, 그 정의가 중요함
3. 차원 축소
   * 모델의 성능 향상 위해 유사한 변수들 묶어 처리해야 함.

### (2) 변수 유형 이해: : 변수 유형별 군집화 고려사항
1. 연속형 변수
   * 거리 측정: 유클리디안 거리, 맨하탄 거리 등
   * 알고리즘: K-means, 계층적 군집화 등
   * 주의사항: 스케일링 필수
        
2. 명목형 변수
   * 거리 측정: 해밍 거리, 자카드 거리 등
   * 알고리즘: K-modes, 계층적 군집화(특별한 거리 함수 사용) 등
   * 처리 방법: 원-핫 인코딩, 더미 변수 변환
       
3. 혼합형 변수 (연속+명목)
   * 알고리즘: K-prototypes, Gower distance 기반 계층적 군집화
   * 처리 방법: 각 변수 유형에 맞는 거리 함수 조합



### (3) 거리/유사도 정의와 측정
* 데이터 간 거리 측정 방법: 가까울수록 유사/ 멀 수록 비유사


    1. 유클리디안 거리 (Euclidean Distance)
      * 가장 일반적인 거리 측정법.
      * 직선 거리를 의미, K-means에서 사용

    2. 맨하탄 거리 (Manhattan Distance)
      * 각 차원의 차이의 절댓값의 합
      * 이상치에 덜 민감함

    3. 코사인 유사도 (Cosine Similarity)
      * 벡터간 각도를 측정
      * 텍스트 데이터, 고차원 데이터에 효과적임

* 6가지 연결 방법: 계층적 군집화에서 두 군집 간의 거리 계산하는 다양한 방식들. 저마다 거리 측정 기준이 달라 다른 군집화 결과 만듦.

1. Single linkage (단일 연결법/최단 연결법)
   * 거리 정의: 두 군집 간 **가장 가까운 두 점 간 거리**
   * 고립된 군집을 찾는데 중점
   * 이상치에 취약 (이상치와 연결될 수 있어서 연쇄적으로 잘못된 클러스트링 일어남)

2. Complete linkage (완전 연결법/최장 연결법)
   * 거리 정의: 두 군집 간 **가장 먼 두점 간 거리**
   * 군집들의 내부 응집성에 중점
   * 이상치에 민감

3. Average linkage (평균 연결법)
   * 거리 정의: 두 군집 간 **모든 점 쌍 사이 거리의 평균**
   * 이상치에 덜 민감 but 계산량 증가

4. Centroid Method (중심 연결법)
   * 거리 정의: 두 군집의 **중심점(centroid) 간의 거리**
   * 두 군집이 결합될 때 새로운 군집의 평균은 가중평균으로 구함(가중치 줌)
   * 시간 소요 있음
   * 덴드로그램에서 거리 축이 위에서 아래로 줄어드는 inversion 발생 가능성 있음

5. Median (중앙 연결법)
   * 거리 정의: (두 군집을 합친 뒤 중심점) 과 (기존 군집의 중심점) 간 거리
   * 군집간의 거리를 두 군집 내 모든 샘플의 중앙값으로 정의함
   * 극단값에 덜 민감 but 기하학적 구조 파악 어려움

6. Ward's Procedure (Ward 연결법)
   * 거리 정의: 군집 병합 후 군집 내 SSE(오차 제곱합)의 증가분이 최소인 것 선택

### (4) 차원 축소
#### 1. 차원의 저주
* 군집화 알고리즘은 거리 기반이므로, 데이터의 차원(피처)이 높아질 수록 차원의 저주 발생함.

  - 선형 투영: 고차원 -> 저차원 공간으로 선형적으로 데이터 투영, ex) PCA(주성분 분석), 특잇값 분해, 랜덤 투영 등

## 3. 군집화 알고리즘
-----
### (1) 계층적 군집화
-----
#### 1. 계층적 군집화의 특징
* 계층적 군집화: 데이터 간의 유사성을 기반으로 트리 구조(dendrogram)를 형성해가며, 상향식 또는 하향식 방식으로 군집을 형성해 나가는 방법.
* Dendogram
  - 가까운 두 개체 또는 군집을 점차적으로 병합해가는 과정을 표현한 트리 구조
  - 수직축(높이) = 두 군집이 병합될 때의 군집 간 거리 (=유사성)

* 군집의 개수를 사전에 설정하지 않고, 클러스트링을 마친 후 선택 가능
* 트리구조가 상향식 -> 응집형 계층적 군집화 / 하향식 -> 분리형 계층적 군집화

#### 2. 응집형 / 분리형 계층적 군집화
* 응집형 클러스터링: 각 샘플이 독립적인 클러스터 -> 하나만 남을때까지 가까운 애들 병합
* 분할형 클러스터링: 전체 샘플을 포함하는 하나의 클러스터 -> 유사성이 낮은 데이터들을 더 작은 클러스터로 나눔
* 두 방법 모두 데이터를 proximity matrix(거리 행렬, 유사도 행렬)를 기반으로 함
* 결과는 이진트리나 dendrogram으로 표현

#### 3. 응집형 계층적 군집화; 자세히 알아보기
* 모든 데이터 포인트를 각각 하나의 클러스터로 간주하고 시작 -> 점차적으로 가까운 클러스트들을 병합하는 방식.
* '가장 가까운 클러스터'를 어떻게 정의하느냐에 따라 알고리즘이 나뉨
* 단일연결/완전연결이 대표적
* 완전 연결을 사용한 응집형 군집화 과정
  1. 거리 행렬 계산: 모든 데이터 포인트 간의 거리를 계산하여 행렬 만듦
  2. 단일 클러스터 시작: 각 데이터 포인트를 개별적인 클러스터로 초기화
  3. 가장 가까운 클러스터 병합: 거리 기준에 따라 가장 가까운 두 클러스터를 병합하여 새로운 클러스터 만듦
  4. 거리 행렬 업데이트: 새로 생성된 클러스터와 나머지 클러스터들 간의 거리를 다시 계산하여 거리 행렬 업데이트
  5. 반복: 모든 데이터 포인트가 하나의 클러스터에 속할 때 까지 3,4번 과정 반복
* 계층적 군집화 시각화: 덴드로그램으로 확인. 소규모 데이터의 구조 파악에 용이함

### (2) k-means
#### 1.k-means
* k-means: 데이터를 정해진 개수(k)의 그룹으로 나누면서, 그룹의 중심점(centroid)과의 거리가 가장 가까운 데이터끼리 묶는 군집화 알고리즘.
* 프로토타입 기반 군집화. (프로토타입 기반 군집화: 각 클러스터의 중심을 대표점으로 삼아 데이터를 묶는 방식. 연속데이터는 평균이 대표점, 범주데이터는 대표값이나 최빈값 사용)
* 유클리디안 거리지표 사용-> 실제 데이터 적용시 특성이 같은 스케일로 측정되었는지 확인해야 함.
* 장점: 직관적, 구현 쉬움, 대용량 데이터에도 적용 가능
* 단점: 초기 중심값에 민감함, 군집수(k) 결정이 어려움, 아웃라이어에 민감, 기하학적 모양의 군집은 파악이 어려움

* k-means의 주요 단계
1. 데이터 표본들 중 랜덤하게 k개의 중심점을 초기 클러스터 중심으로 선택
2. 각 표본들을 가장 가까운 중심점에 할당
3. 각 클러스터에 할당된 표본들의 데이터 평균을 계산하여 중심점을 이동
4. 클러스터 할당이 변하지 않거나 사용자가 지정한 허용 오차 또는 최대 반복 횟수에 도달할 때 까지 2,3 번을 반복

#### 2.k-means++ 알고리즘
* k-means++: k-means에서 초기 중심점들을 서로 멀리 떨어진 곳에 위치시켜 더 일관되고 좋은 결과를 도출하는 개선된 초기화 방법
* 즉, k-means와 다르게 초기 중심점들을 랜덤X 멀리 위치시킴

> k-means++ 알고리즘 초기화 과정
> 1. 선택한 k개의 중심점을 저장할 빈 집합 M 초기화
> 2. 입력 샘플에서 첫번째 중심점(u)을 랜덤 선택, M에 할당
> 3. M에 없는 각 샘플 x에 대해 M에 있는 중심점까지의 최소제곱 거리 d(x,M)^2을 찾음
> 4. 다음 식과 같은 가중치가 적용된 확률 분포를 사용해 다음 중심점(u^p)을 랜덤하게 선택함 $\frac{d(u^(p),M)^2}{\sum_{i}{d(x^(i),M)^2}}$
> 5. k개의 중심점을 선택할 때까지 3,4 번 반복
> 6. 기본 k-means 알고리즘 수행

#### 3.엘보우 방법(elbow method)
* 엘보우 방법: 클래스 내 SSE를 바탕으로 그래프를 활용해 최적 클러스터 개수 k를 추정하는 방법.
* 왜곡이 빠르게 감소하는 지점의 k값을 찾아 이를 최적 클러스터 개수로 추정한다.
* 장점: 직관적, 구현이 간단
* 단점: 주관적으로 판단하게 됨, 모호한 경우가 있음

> k=3 정도가 적절

### (3) DBSCAN (density-based spatial clustering of applications with noise)
* DBSCAN: 밀도가 높은 지역의 데이터를 하나의 군집으로 묶고, 밀도 기준을 만족하지 못하는 점은 군집에 포함시키지 않는 군집화 알고리즘 (밀도에 따른 군집화)
* 밀집도: 특정 거리 내에 존재해야 하는 데이터 포인터의 최소 개수
* 특징
  - 명시적으로 이상치를 지정하여 클러스트링에서 제외함
  - 이상치 영향 적음
  - 어떤 데이터가 여러 군집의 특정 거리 내에 있으면 가장 밀도가 높은 군집과 그룹화
* 주요 단계
  1. 샘플 분류

> * 핵심 샘플(Core point): 특정 반경 $\epsilon$안에 있는 이웃한 샘플이 지정된 개수(MinPts) 이상인 경우
> * 경계 샘플(Border Point): $\epsilon$ 이내에 MinPts보다 이웃이 적지만 다른 핵심 샘플 반경 $\epsilon$ 안에 있는 경우
> * 잡음 샘플(Noise Point): 위 두 조건에 해당하지 않는 모든 샘플

  2. 클러스터 생성: 개별 핵심 샘플 또는 $\epsilon$ 이내에 있는 핵심 샘플을 연결한 그룹을 클러스터로 생성함
  3. 경계 샘플 할당: 경계 샘플은 해당 핵심 샘플의 클러스터에 할당.

* 설정해야 하는 하이퍼 파라미터
  - eps: 두 포인트가 이웃이 되기 위한 포인트 사이의 최대 거리
  - min_samples: 한 포인트가 군집이 되기 위한 eps 거리 내 최소 포인트 개수
* 장점:
  - 기하학적 분포 갖는 데이터 처리 가능
  - 이상치에 덜 민감
  - 잡음 샘플을 구분해낼 수 있음
* 단점: 
  - eps와 min_samples 설정에 많은 영향을 받음
  - eps를 조절하기 쉽지 않음
  - 연산량 많아 속도 느림
  - 다른 밀도 분포 가진 데이터의 군집 분석은 잘 못함(noise point로 구분할 가능성 있음)
  - 차원의 저주 있음.              
                


* **HDBSCAN**: 계층적 DBSCAN으로, DBSCAN을 계층적 군집화로 변환한 버전.
  - 밀도 기반으로 1차 군집화 -> 거리 기준으로 밀도 기반 군집들을 반복적 연결
  - DBSCAN의 단일 밀도 기준 한계 보완이 가능하다

### (4) GMM (Gaussian Mixture Model)
* GMM: 데이터가 여러 다른 모양의 가우시안 분포로 구성되었다고 가정, 각 분포를 클러스터로 인식하는 군집화 방법
* 모델 기반 군집화방법(데이터를 생성하는 통계적 모델을 가정하여, 그 전제 하에 군집화 수행하는 것)이다.
* 즉 각 군집을 확률 분포로 간주하여, 전체 데이터 분포를 여러 확률 분포의 혼합으로 모델링한다.         
  -> 데이터 생성 메커니즘까지 모델링이 가능하다.

* GMM의 가정
  1. 관측된 데이터: 특정 가우시안 확률 분포에 의해 생성되었다.
  2. 전체 데이터셋: 여러 개의 다변량 가우시안 분포가 섞여있다.
  3. 개별 데이터: 우도에 따라 k개의 가우시안 분포 중 하나에 속한다.
              
* GMM 진행 과정

> 주어진 전체 데이터셋 분포 확인

> 전체 데이터셋은 서로 다른 정규 분포 형태의 확률 분포 곡선으로 구성되어있다고 가정

> 전체 데이터셋을 구성하는 여러 개의 정규분포 곡선 추출, 개별 데이터가 이중 어디 속하는지 결정. 이때 각각의 분포가 하나의 군집이 됨

> 3개의 군집이 형성된 것을 볼 수 있음

* 모델의 파라미터 추정: GMM은 기댓값-최대화(EM)알고리즘 사용하여 모델 파라미터 추정함. (군집의 모양, 크기, 방향 추정 가능.)
  - 파라미터 구성

  - EM 알고리즘 단계
        1. Initialization: 필요한 파라미터에 대해 초기값 설정
        2. E(Expectaion) step: 현재 $\theta$ 통해 x가 특정 군집에 속할 사후확률 계산
        3. M(Maximization) step: 계산된 사후 확률을 통해 파라미터 다시 추정
        4. 수렴 조건이 만족될 때까지 반복 E,M Step

* 장점
  - k-means 보다 유연하게 적용 가능
* 단점
  - 오래 걸림
  - 가정한 분포에 맞지 않은 데이터일 경우 성능 저하

## 4. 군집화 평가 방법
-----
### (1) 외부평가와 내부평가
: 정답 레이블의 유무에 따라 나뉨.

#### 1. 외부 평가
* 정답 레이블이 존재하는 경우 사용
* 군집화 결과와 실제 레이블(정답) 비교하여 얼마나 유사한지 평가
* ARI가 대표적.
* 분류 할 수 있는데 왜 군집화?: 데이터에 대한 새로운 인사이트 얻기, 이해 심화용

#### 2. 내부 평가
* 정답 레이블이 없는 경우 사용
* 군집 내부 응집도와 군집 간 분리도 기반으로 군집의 품질을 정량 측정
* 알고리즘 별 내부 평가 지표, 목표
  - k-means: SSE 최소화하기
  - GMM: 로그 우도 최대화 하기
* 대표적인 범용 지표: 실루엣 계수, 던 지수

### (2) 실루엣 계수
:클러스터 내 샘플들이 얼마나 조밀하게 모여있는지 측정하는 도구
* 실루엣 계수 계산

> * $s^(i)=\frac{b^(i)-a^(i)}{max(b^(i),a^(i))}$
> 1. 샘플 x^i와 동일한 클러스터 내 모든 다른 포인트 사이의 거리를 평균하여 응집력 a^i 계산
> 2. 샘플 x^i와 가장 가까운 클러스터의 모든 샘플 간 평균 거리로 최근접 클러스터의 클러스터 분리도 b^i 계산
> 3. 클러스터 응집력과 분리도 사이의 차이를 둘 중 큰 값으로 나누어 실루엣 s^i 계산

* 실루엣 계수 해석
  - (-1,1) 값 가짐
  - 분리도 > 응집도 인 경우 1에 가까워짐 -> 군집화 잘 됨
  - 분리도 = 응집도 인 경우 0에 가까워짐 -> 군집 간 거리 모호
  - 분리도 < 응집도 인 경우 -1에 가까워짐 -> 군집화 잘 안됨

* 사이킷런 패키지로 계산 가능

### (3) Dunn Index
: 클러스터 간 최소거리(분리도)와 클러스터 내 최대거리(응집도)의 비율을 계산해 클러스터링 품질 평가하는 지표.
* Dunn Index 계산

> * $d(C_i,C_j)$: 클러스터 $C_i$와 $C_j$ 사이의 거리 (분리도)
> * $\delta(C_k)$: 클러스터 $C_k$내의 최대 거리 (응집도)

* Dunn Index 해석
  - 값이 클수록 좋은 군집화 결과를 의미함
  - 군집 간 거리 크고 (분리도 높고), 군집 내부 거리 작을수록 (응집도 높을수록) 좋은 군집화임
  - $(0,\infty)$
* Dunn Index 한계
  - 군집 수 많아질수록 계산 비용 증가.
  - 극단적인 값에 큰 영향 받음
  - 이상치에 민감.
