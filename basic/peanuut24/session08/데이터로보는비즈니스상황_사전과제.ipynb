# 데이터로 보는 비즈니스 상황_사전과제
-----
## 1. 데이터 기반 의사결정
-----
### (1) 데이터 분석의 목적
: 목적 달성 과정의 질 향상. 데이터에 근거한 객관적인 미래 전망을 제시할 수 있다.

* 액션?
  - 결정 (ex.신제품 출시에 관한 의사결정)
  - 계획 (ex. 마케팅 전략 세우기)
  - 행동 (ex. 판매 예측 모델을 위한 시스템 개발)

### (2) DDDM (Data-Driven-Decision-Making)
: 비즈니스 의사 결정을 제공하기 위해 직관 대신 데이터와 분석 사용하는 것을 강조한 접근 방식.

### (3) 6steps of DDDM
1. 목표 정의: 조직의 목표 명확하게 표현. 집중적이고 목적이 명확한 노력을 기울일 수 있도록 한다.  
          
2. 데이터 식별, 준비, 수집: 데이터 요구 사항을 판별해 데이터 소스를 준비, 체계적으로 수집하고 검증한다.           

3. 구성 및 탐색: 데이터를 새로운 패턴, 추세 및 인사이트 발견할 수 있도록 구조화함. 데이터를 정리하여 정확성 보전, 시각화하여 이상치 및 추세 식별화     
       
4. 데이터 분석 수행: 비즈니스 전략에 영향을 미치는 패턴, 상관관계, 추세 파악.        

5. 결론 도출: 주요 데이터 분석 결과 검토, 올바른 비즈니스 상황에 배치하여 실행 가능한 인사이트와 권장 사항 구성함. 전략적 결정을 한다.        

6. 구현 및 평가: 인사이트를 검증, 결과 측정. 실행 계획 개발, 리소스 할당, 이니셔티브의 진행 상황 지속적으로 모니터링 함.
   -> KPI(핵심성과지표) 기준으로 결과를 측정, 분석, 피드백 수집해 지속적인 조정을 수행한다.


### (4) 데이터 분석 유형

1. 설명 분석: 데이터 취합 및 마이닝을 통해 과거 데이터 설명, 요약 -> 과거 데이터에 대한 인사이트를 제공한다. (월별 판매 보고서, 고객 만족도 설문 조사, 웹 사이트 트래픽 분석)        

2. 진단 분석: 특정 사건이 발생한 이유를 파악하는 분석. (판매 감소, 고객 불만 급증 등 사고의 근본 원인, 추세 파악)      

3. 예측 분석: 과거 데이터 기반으로 미래 추세, 결과 예측 (판매 예측, 위험 관리)       

4. 처방 분석: 더 나아가 데이터에 기반하여 조치를 권장. 예측 분석과 최적화 알고리즘을 결합. (공급망 최적화, 마케팅 캠페인 전략, 리소스 할당 결정 지원)

## 2. 목표 정의
-----
### (1) 지표
: 방향, 목적, 기준 등을 수치적으로 설명한 것. (지수, 척도 등) 대게 분수 형태로 표현됨.
|유형|공식 구조|예시 지표|
|------|------|------|
|효율성(Efficiency)|산출/투입|노동생산성= 생란량/노동시간|
|효과성(Effectiveness)|달성 결과/목표|목표 달성률=실제 매출/목표 매출|
|수익성(Profitability)|이익/자원|ROE=순이익/자기자본|
|성장성(Growth)|증가분/기존 규모|매출 성장률=(당기매출-전기매출)/전기매출|
|안정성(Stabiilty)|보유자산/부채|부채비율=부채/자기자본|

### (2) 지표 정의
: 분자, 분모에 들어갈 절대적인 수치를 구할 수 없는 경우 어떻게 해야할까?      
-> **Proxy Metric** 설정; 활성화와 인게이지 같이 추상적이어서 산물마다 나름의 가정과 정의, 논리가 필요한 지표를 말한다.

### (3) 북극성 지표 & L1/L2 Metrics
* 북극성 지표: 기업이 가장 중요하게 생각하는 지표. (ex. 에어비엔비: 예약된 박 수)
* L1,L2 지표
  - L1: 전사적, 고수준의 성과 지표 (전략 관점)
  - L2: L1 지표의 세부 사항을 측정하는 지표 (전술 관점)

### (4) AARRR (Acquisition, Activation, Retention, Revenue, Referral)
: 사용자 증가를 위한 5개의 로드맵으로 구성됨. 사용자 운영 과정에 활용하는 데이터 분석 모델이다.           

1. Acquisition(유입): 사용자가 여러 채널을 통해 기업의 앱 혹은 웹사이트에 방문하는 것.       
   - 중요 포인트: 신규 고객을 어떻게, 어디로부터, 어떤 채널 통해 획득할 지
   - 지표: 조회수, 가입 사용자 수 등
            
2. Activation(활성화): 사용자가 앱 또는 웹사이트에 첫 방문하여 주요 기능을 직접 사용하는 것.     
   - 중요 포인트: 사용자에게 제품의 가치를 알려 활성화 정도 높여야 함
   - 지표: 재품 핵심 기능의 활용도, 사용자 온라인 접속 시간 등 (ex. 페이스북: 회원가입 10일 내 최소 7명의 친구를 가진 신규 사용자를 지표로 삼음)
           
3. Retention(재방문): 신규 사용자를 장기 사용자로 전환 시키는 것.       
   - 중요 포인트: 장기 고객을 양성 방법, 고객의 지속적 구매 행위 증가 방법
   - 지표: 익일 유지율, 1주차 유지율, 월 유지율, EDM 메일 개설율 등
          
4. Revenue(수익): 사용자가 기업의 제품 혹은 서비스를 유료 구매하기로 결정한 단계.
   - 중요 포인트: 사용자의 소비를 유도해 매출 증가시키기, 최대한 사용자를 소비하게끔 만들기. 고객의 평생 가치를 지속적으로 높이고, 고객 유치 비용은 줄여야 한다.
   - 지표: ex. 한 게임이 유료화 상품 판매 중심의 수익구조 일 때, 사용자가 유료로 구매한 게임 아이템 등이 주요 측정 지표임
           
5. Referral(추천): 사용자가 타인에게 기업의 서비스, 상품을 추천하는 단계.
   - 중요 포인트: 어떻게 충성도 높은 사용자를 브랜드 홍보자로 만들 지, 어떻게 제품을 잠재 고객들에게 추천하도록 만들게 할 지.
   - 지표: 순 추천고객 지수 (NPS), 바이럴 지수 등.
