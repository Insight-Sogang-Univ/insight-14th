# 통계_사전과제

## 1. 기술통계

### 1) 중심화 경향
 
* 평균(mean): 데이터 값의 총합/데이터의 개수.
  이상치(outlier)에 민감할 수 있다 -> 절사 평균(데이터의 양 끝을 잘라낸 평균) 사용
    series.mean()

* 중앙값(median): 데이터를 크기 순으로 정렬했을 시 한 가운데에 위치하는 값. 이상치에 덜 민감.
    series.median()

* 최빈값(mode): 데이터에서 가장 자주 나타나는 값.
    series.mode()

### 2) 자료의 산포

* 범위(range): 최대값과 최소값의 차이.
    series.max() - series.min()

* 분산(variance): 데이터 값들이 평균으로부터 얼마나 멀리 퍼져있는지를 나타낸 지표. 분산 값이 클수록 데이터가 많이 퍼져있다.
    series.var()

* 표준편차: 분산의 제곱근
    series.std()

* 사분위수(quartile): 데이터 표본을 네 개의 동일한 크기로 나눈 값
  + Q1: 하위 사분위수 (하위 25%)
  + Q2: 중앙값 (하위 50%)
  + Q3: 상위 사분위수 (하위 75%, 상위 25%)
    series.quantile(0.25)  # 25th percentile
    series.quantile(0.5)   # 50th percentile (중앙값)
    series.quantile(0.75)  # 75th percentile

    numpy로 실행은
    np.percentile(series, 75)

* 사분위간 범위(IQR): Q3에서 Q1을 제외한 범위, 데이터의 중앙 50%가 퍼져 있는 범위. 이상치 탐지 시 유용.
  + 아래쪽 이상치: Q1 − 1.5×IQR 보다 작은 값
  + 위쪽 이상치: Q3 + 1.5×IQR 보다 큰 값
  
* 변동계수(coefficient of variation): 데이터의 상대적 변동성을 나타내는 통계적 지표.
    변동계수=(표준편차/평균)*100

* 첨도: 확률분포의 꼬리가 두꺼운 정도를 나타내는 척도
    series.kurt()
* 왜도: 비대칭도. 확률 변수의 비대칭성을 나타내는 지표
    series.skew()

## 2. 확률과 분포

### 1) 확률 분포

* 확률변수: 경우의 수를 특정 값으로 대응시키는 함수.

* 확률분포: 확률 변수가 특정 값을 가질 확률을 나타낸 함수
  + 이산확률분포: 확률 변수가 이산적임
  + 연속확률분포: 확률 변수가 연속적임

### 2) 분포 모형

* 정규분포
  - 평균을 기준으로 대칭이다
  - 많은 자연현상과 사회 현상이 정규분포를 따르는 것으로 알려져 있다.
  - 표준편차가 클수록 곡선이 평평해진다.
  - 중심극한정리: 표본의 크기가 커질수록 표본평균의 분포가 모집단의 분포와 상관없이 정규분포에 가까워진다는 정리.
  n>=30이면 표본평균의 분포는 근사적으로 정규분포를 따른다.

* T분포
  - 평균을 기준으로 대칭이다.
  - 자유도가 커질수록 표준정규분포 N(0,1)에 수렴한다.
  - 표본으로 모평균을 추정할 때 사용, 표본 크기가 작을 때 정규분포보다 정확한 결과를 제공한다.
  - 자유도(df): 독립적인 정보의 수. 자유도가 높을 수록 T분포가 정규분포에 가까워지고 분석의 정확도가 높아진다. 그러나 오버피팅(overfitting) 문제가 생길 수 있다.
  - 오버피팅(overfitting): 모델이 학습 데이터에 너무 맞추어져서 새로운 데이터에 대한 일반화 성능이 저하되는 현상.

* 카이제곱분포
  - 0에서 시작하여 오른쪽 긴 꼬리를 가진 비대칭 분포.
  - 독립적인 표준 정규 분포를 따르는 변수들의 제곱 합으로, 표본분산의 분포와 관련된 지표이다.

* F분포
  - 0에서 시작하여 오른쪽 긴 꼬리를 가진 비대칭 분포.
  - 두 독립적인 카이 제곱 분포의 비율로, 두 집단 간의 분산 차이를 비교할 때 쓰인다.

### 2) 구간추정

## 4. 추론통계 - 검정
: 표본통계랑으로 추정한 모집단의 모수에 대한 가설이 옳다고 판단할 수 있는지 평가하는 과정.

### 1) 가설 검정의 절차

#### 1) 가설 설정
: 귀무가설(기각하고자 하는 명제) H0, 대립가설(내가 주장하고자 하는 명제) H1을 설정.

#### 2) 유의 수준 설정
* 유의수준: 결과 판단 시 허용할 수 있는 오류의 최대치. 주로 0.05로 설정함. 어떤 결정이 얼마나 신뢰할 수 있는지 판단하는 기준이 된다.

- 1종 오류: 귀무가설이 실제로 참이었는데 기각할 확률 (=유의수준)
- 2종 오류: 귀무가설이 실제로 거짓인데 기각하지 못할 확률

#### 3) 검정통계량 산출
* 검정통계량: 가설 검정의 결과를 판단하는 데 사용되는 값. 데이터 분포나 특징에 따라 어떤 통계량을 사용할 지 정해야 함.
  - Z값: Z=(X−μ)/σ. 큰 표본 크기에 대한 가설 검정시 사용.
  - T값: T=(X−μ)/s/sqrt(n). 모집단의 평균에 대한 가설 검정, 두 표본 평균 간의 차이를 검정할 때 사용.
  - F값: F값=분산_1/분산_2. 주로 분산 분석(ANOVA)에서 두 개 이상 그룹 간 평균의 차이를 검정할 때 사용.

#### 4) 가설 기각/채택 판단
* 유의확률(p-value): 가설 검정에 쓰일 수 있는 또다른 척도. 유의 확률이 유의수준보다 작을 시 귀무가설을 기각할 수 있다.
 
  

### 2) 데이터 자료형에 따른 검정 방법

#### 1) [수치형-수치형] (피어슨)상관분석
* 피어슨 상관 분석: 두 변수간 피어슨 상관계수가 얼마나 유의미한지 검정하는 방법.
* 가설:
  - H0: 두 변수간 선형 상관관계가 존재하지 않는다.
  - Ha: 두 변수간 선형 상관관계가 존재한다.

* 공분산: 두 변수가 함께 어떤 방향으로 변화하는지와 그 크기를 표현하는데 사용.
  + Cov (X,Y) > 0이면 X와 Y는 양의 관계, X가 커지면 Y도 커짐
  + Cov (X,Y) < 0이면 X와 Y는 음의 관계, X가 커지면 Y는 작아짐
  + Cov (X,Y) = 0이면 X와 Y는 양도 음도 아닌 관계

    np.cov(df['col1'], df['col2'])

* 상관계수: 두 변수 간의 선형적인 관계를 나타내는 통계적 지표. -1과 1 사이 값.
  공분산을 각 변수의 표준편차로 나누어 표준화 한 것이다.
  + Corr (X,Y)이 양수면 양의 상관관계 (1이면 완전한 양의 상관관계)
  + Corr (X,Y)이 음수면 음의 상관관계 (-1이면 완전한 음의 상관관계)

    df[['col1', 'col2']].corr()

    코드
    	import scipy.stats as spst
	    spst.pearsonr(df['col1'], df['col2'])


#### 2) [범주형-수치형] F-test(ANOVA), T-test

* T-test: 범주가 2개일 때.
  + 가설
    - H0: 두 그룹의 평균에는 차이가 없다.
    - Ha: 두 그룹의 평균에는 차이가 있다.

```python
# 민초 섭취와 수명 평균에 영향이 있는지
mincho = df.loc[df['민초여부'] == 'Yes', '수명']
no_mincho = df.loc[df['민초여부'] == 'No', '수명' ]

spst.ttest_ind(mincho, no_mincho)
```

    - Ha: 각 그룹 별 평균은 차이가 있다.

```python
# mbti 별로 수명 평균에 차이가 있는지
ISTP = MBTI.loc[MBTI['mbti'] == 'ISTP', '수명']
ESFP = MBTI.loc[MBTI['mbti'] == 'ESFP', '수명']
INFP = MBTI.loc[MBTI['mbti'] == 'INFP', '수명']

spst.f_oneway(ISTP, ESFP, INFP)
```
