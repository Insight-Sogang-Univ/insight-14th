## <span style="background-color:#E0F7FA; color:#1565C0; padding:4px; border-radius:5px;"> 자료의 척도 </span>

레이블이나 이름으로 요소의 속성을 식별하려고 하는 척도   
ex) 펀드 유형, 성별(남:0, 여:1)

##### <span style="background-color:#DCDCDC; color:#000000; padding:4px; border-radius:5px;"> 서열 척도 </span>

명목형의 분류 속성+항목 간의 순서 or 순위 특성 o       
ex. 직급

##### <span style="background-color:#DCDCDC; color:#000000; padding:4px; border-radius:5px;"> 구간(등간) 척도 </span>

값들 사이에 일정한 간격의 순서 o, 연산 가능    
but 곱셉과 나눗셈은 의미 x     
ex) 온도, 시간

##### <span style="background-color:#DCDCDC; color:#000000; padding:4px; border-radius:5px;"> 비율 척도 </span>


절대영점과 임의의 측정 단위로 구성, 동간성 유지    
**값 0**: 아무것도 존재 x

vs 종류    
종류: 데이터의 본질적 성격

## <span style="background-color:#E0F7FA; color:#1565C0; padding:4px; border-radius:5px;"> 기술 통계 </span>

데이터를 있는 그대로 설명/이해 = 샘플 자체에 대한 관심     

ex) 회사의 올해 평균 매출은? 제품의 불량률은? 고객의 연령 분포는?

### <span style="background-color:#FFFFE0; color:#CD5C5C; padding:4px; border-radius:5px;"> 중심화 경향 </span>

##### <span style="background-color:#DCDCDC; color:#000000; padding:4px; border-radius:5px;"> 평균(mean) </span>
- (데이터 값 총합) / (데이터 개수)     
- 총합 사용 -> 이상치에 민감       
`series.mean()`

##### <span style="background-color:#DCDCDC; color:#000000; padding:4px; border-radius:5px;"> 중앙값(median) </span>
- 데이터 크기 순서대로 정렬 시 한가운데에 위치하는 값  
- 평균보다 이상치에 덜 민감       
`series.median()`

##### <span style="background-color:#DCDCDC; color:#000000; padding:4px; border-radius:5px;"> 최빈값(mode) </span>
- 데이터에서 가장 자주 나타나는 값     
- 이산형 데이터, 데이터 분포 이해       
`series.mode()`


### <span style="background-color:#FFFFE0; color:#CD5C5C; padding:4px; border-radius:5px;"> 자료의 산포 </span>

##### <span style="background-color:#DCDCDC; color:#000000; padding:4px; border-radius:5px;"> 범위(range) </span>
- 데이터의 최대값 - 최소값 : 데이터의 전체 변동 범위           
`series.max() - series.min()`

##### <span style="background-color:#DCDCDC; color:#000000; padding:4px; border-radius:5px;"> 분산(variance) </span>
- 데이터 값들이 평균으로부터 얼마나 멀리 퍼져있는지  
- 분산값이 크다=데이터가 많이 퍼져 있다       

$\frac{\sum_{i=1}^{N}(x_i - μ)^2}{N}$        

`series.var()`

##### <span style="background-color:#DCDCDC; color:#000000; padding:4px; border-radius:5px;"> 표준편차(standard deviation) </span>
- 분산에 제곱근을 취한 값     
- 단위 일치를 위함       


$\sqrt{\frac{\sum_{i=1}^{N}(x_i - μ)^2}{N}}$

`series.std()`

##### <span style="background-color:#DCDCDC; color:#000000; padding:4px; border-radius:5px;"> 사분위수(quartile) </span>
- 데이터의 표본을 네 개의 동일한 크기로 나눈 값    
    + Q1: 하위 25% **`하위 사분위수`**       
    + Q2: 하위 50% **`중앙값`**    
    + Q3: 하위 75%(=상위 25%) **`상위 사분위수`**

```python
series.quantile(0.25)
series.quantile(0.5)
series.quantile(0.75)

np.percentile(series, 75)
```

##### <span style="background-color:#DCDCDC; color:#000000; padding:4px; border-radius:5px;"> 사분위간 범위(IQR) </span>
- Q3에서 Q1 값 제외한 범위: 데이터의 **중앙 50%** 범위     
- 이상치 탐지에 유용    

 

 ##### <span style="background-color:#DCDCDC; color:#000000; padding:4px; border-radius:5px;"> 변동계수 (coefficient of variation) </span>
- 데이터의 상대적 변동성    

$변동계수=(표준편차/평균)*100$

## <span style="background-color:#E0F7FA; color:#1565C0; padding:4px; border-radius:5px;"> 확률과 분포 </span>

### <span style="background-color:#FFFFE0; color:#CD5C5C; padding:4px; border-radius:5px;"> 확률분포 </span>

##### <span style="background-color:#DCDCDC; color:#000000; padding:4px; border-radius:5px;"> 확률변수 </span>
random한 세계에서 어떤 특정 사건이 발생할 때 나타날 수 있는 모든 경우의 수를 특정 값으로 대응시키는 함수    

##### <span style="background-color:#DCDCDC; color:#000000; padding:4px; border-radius:5px;"> 확률분포 </span>
확률변수가 특정한 값을 가질 확률을 나타내는 함수

- 이산확률분포: 확률 변수가 이산값

- 연속확률분포: 확률 변수가 연속값     
연속적: 무한한 범위, 특정값 선택 불가능

##### <span style="background-color:#DCDCDC; color:#000000; padding:4px; border-radius:5px;"> 정규분포(Normal Distribution) </span>



|형태|매개변수|특징|
|------|---|---|
|- 종 모양 <BR>     - 평균 기준 대칭    <BR>   - 표준편차가 클수록 곡선이 평평해짐(데이터가 넓게 분산)|- 평균(μ)   <BR> - 표준편차(σ)|- 가장 잘 알려진 연속 확률분포      <BR> - 많은 자연/사회 현상이 정규분표를 따름|



### **표본분포**
정규분포를 따르는 모집단에서 추출한 **표본** 통계량들이 따르는 분포
통계적 추론을 위한 도구

`추출한 표본이 '정규분포를 따르는 모집단'으로부터 왔다는 가정`

##### <span style="background-color:#DCDCDC; color:#000000; padding:4px; border-radius:5px;"> T분포(T-Distribution) </span>



|형태|매개변수|특징|
|------|---|---|
|- 종 모양 (정규분포보다 꼬리 두꺼움) <BR>     - 평균 기준 대칭    <BR>   - 자유도가 클수록 표준정규분포 N(0,1)에 수렴|자유도(df)|- 표본: 모평균(혹은 모평균 간의 차이) 추정할 때 사용      <BR> - 표본 크기 작을 때 정규분포보다 더 정확|

**자유도?** `얼마나 많은 정보가 독립적으로 변할 수 있는가를 나타내는 숫자`

**표준정규분포?** `평균이 0이고 표준편차가 1인 정규분포`



**결정 요령**



##### <span style="background-color:#DCDCDC; color:#000000; padding:4px; border-radius:5px;"> 카이제곱분포(Chi-Square Distribution) </span>

##### <span style="background-color:#DCDCDC; color:#000000; padding:4px; border-radius:5px;"> F분포(F-Distribution) </span>



|형태|매개변수|특징|
|------|---|---|

## <span style="background-color:#E0F7FA; color:#1565C0; padding:4px; border-radius:5px;"> 추론통계: 검정 </span>

### <span style="background-color:#FFFFE0; color:#CD5C5C; padding:4px; border-radius:5px;"> 가설 검정의 절차 </span>

표본 정보 활용해서 가설의 합당성 여부 판정

%가설 설정 - 유의 수준 설정 - 검정통계량 산출 -가설 기각/채택 판단

##### <span style="background-color:#DCDCDC; color:#000000; padding:4px; border-radius:5px;"> 가설 설정 </span>
**귀무가설 H0 : 주로 기각하고자 하는 명제** & **대립가설 H1 : 내가 주장하고자 하는 명제** 설정
|귀무가설|대립가설|
|------|------|
|현재 통용되고 있는 가설|새롭게 주장하고픈 가설|
|ex)~는 차이가 없다|ex)~는 차이가 존재한다|
|지지될 때 위험성이 비교적 약함|지지될 때 위험성 비교적 강함|

##### <span style="background-color:#DCDCDC; color:#000000; padding:4px; border-radius:5px;"> 유의 수준 설정 </span>
**유의수준?** 연구자가 결과를 판단할 때 허용할 수 있는 오류의 최대치, 주로 0.05 <BR>
귀무가설이 실제로 참이지만 5%의 확률로 귀무가설을 잘못 기각할 수 있다

- **1종 오류(𝛼)** : 귀무가설이 실제로 참이었는데 기각할 확률
- **2종 오류(𝛽)** : 귀무가설이 실제로 거짓인데 기각하지 못할 확률

유의수준: 1종 오류(더 치명적)의 확률

##### <span style="background-color:#DCDCDC; color:#000000; padding:4px; border-radius:5px;"> 검정통계량 산출 </span>
가설 검정의 결과 판단에 사용되는 값

##### **Z-score**

1. 정의
특정 데이터가 평균으로부터 얼마나 떨어져 있는지 표준편차 단위로 나타냄

2. 계산:

$Z=(X−μ)/σ$
X: 관측값 μ: 평균 σ: 표준편차

3. 사용
큰 표본 크기(>=30)에 대한 가설 검정에 사용 <BR>
ex) 모집단의 평균에 대한 가설 검정

4. 수치 해석
- Z값의 절댓값이 크면? <BR>
해당 데이터가 평균으로부터 멀리 떨어져 있으며, 이상치일 수도 o

- 임계점 찾기 <BR>
양측검정: Z=1.96일 때 유의수준 0.05, Z=2.58일 때 유의수준 0,01
-> Z값이 임계점보다 크면 귀무가설 기각

##### **T-score**

1. 정의
- T분포로 얻은 값 <BR>
- 정규분포와 유사하지만 꼬리 부분 두꺼워 작은 표본에서 발생할 수 있는 불확실성 잘 반영

2. 계산:

$T = \frac{X - \mu}{s / \sqrt{n}}$
X: 표본 평균 μ: 연구자 설정값 s: 표본 표준편차, n: 표본 크기

3. 사용
작은 표본 크기에 대한 가설 검정(T검정)에 사용 <BR>
ex) 모집단의 평균에 대한 가설 검정, 두 표본 평균 간 차이 검정

4. 수치 해석
- T값의 절댓값이 크면? <BR>
해당 데이터가 모집단 평균으로부터 멀리 떨어져 있으며, 관측 표본 평균이 모집단 평균과 유의미한 차이 o 가능성 높음

- 임계점 찾기 <BR>
T분포표 사용: 같은 유의수준에 대해 t값은 z값보다 커지고 자유도가 높아질수록 z값에 수렴

##### **F-score**

1. 정의
- F분포로 얻은 값 <BR>
- 두 표본 분산 사이의 차이 비교 <BR>
- 두 개의 독립적인 카이제곱 분포된 변수들의 비율로 정의

2. 계산:

$F값= 분산_1/분산_2$

두 독립적인 표본의 분산

3. 사용

분산 분석에서 그룹 간 평균차 검정에 사용 <BR>

4. 수치 해석

F값: 두 표본의 분산의 비
- 이 값이 1에 가까우면 두 표본의 분산이 비슷 : 즉 두 그룹이 퍼진 정도가 비슷

##### <span style="background-color:#DCDCDC; color:#000000; padding:4px; border-radius:5px;"> 유의확률(p-value) </span>
가설 검정에 사용, 검정통계량에서 유도 <BR>

유의 확률(p) < 유의 수준(a) : 귀무가설 기각

### <span style="background-color:#FFFFE0; color:#CD5C5C; padding:4px; border-radius:5px;"> 데이터 자료형에 따른 검정 방법 </span>

##### <span style="background-color:#DCDCDC; color:#000000; padding:4px; border-radius:5px;"> 수치형-수치형: (피어슨)상관분석 </span>
두 변수간 피어슨 상관계수가 얼마나 유의미한지를 검정



이런 거 관계 판단에 씀

> H0: 두 변수간 선형 상관관계 존재 x <BR>
> H1: 두 변수간 선형 상관관계 존재

상관계수
- 두 변수 간의 선형적 관계 나타냄 (-1~1) <BR>
- 공분산을 각 변수의 표준편차로 나누어 표준화함

`df[['col1', 'col2']].corr()`

공분산? 두 변수가 함께 어떤 방향으로 변화하는지와 그 크기를 표현 <BR>

`np.cov(df['col1'], df['col2'])`


```python

import scipy.stats as spst
spst.pearsonr(df['col1'], df['col2'])

```

상관계수 r
> |r|< 0.3 : 상관관계 약함 <BR>
> |r| >0.5: 강함

##### <span style="background-color:#DCDCDC; color:#000000; padding:4px; border-radius:5px;"> 범주형-수치형: F-test(ANOVA), T-test </span>
두 변수간 피어슨 상관계수가 얼마나 유의미한지를 검정

T-test: 범주 2개일 때

T값: 두 평균 간 차/표준 오차

>H0: 두 그룹의 평균에는 차이가 없다. <BR>
>Ha: 두 그룹의 평균에는 차이가 있다.

```python
mincho = df.loc[df['민초여부'] == 'Yes', '수명']
no_mincho = df.loc[df['민초여부'] == 'No', '수명' ]

spst.ttest_ind(mincho, no_mincho)
```
:
TtestResult(statistic = t-값, pvalue = p값, df = 자유도)

T값은 절대값 2 기준, 크면 차이가 o

F-test(ANOVA): 범주 3개 이상

F값: 집단 간 분산/집단 내 분산 <BR>
= (전체 평균과 각 집단의 평균을 갖고 만든 분산) / (각 집단의 평균과 개별 값으로 만든 분산)

> H0: 각 MBTI 그룹 별 수명 평균은 차이가 없다. <BR>
> Ha: 각 MBTI 그룹 별 수명 평균은 차이가 있다.

```python
ISTP = MBTI.loc[MBTI['mbti'] == 'ISTP', '수명']
ESFP = MBTI.loc[MBTI['mbti'] == 'ESFP', '수명']
INFP = MBTI.loc[MBTI['mbti'] == 'INFP', '수명']

spst.f_oneway(ISTP, ESFP, INFP)
```

카이제곱 통계량


= (관측빈도-기대빈도)^2/기대빈도 를 쌍에 맞게 더해 주기

범주 개수 상관 x -> 카이제곱 통계량: 자유도의 2배 기준으로 잡음

범주형 변수의 자유도: 범주의 수-1
여기서의 자유도 : (x 변수 자유도)*(y 변수 자유도)
