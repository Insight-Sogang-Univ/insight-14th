### <span style="background-color:#FFFFE0; color:#CD5C5C; padding:4px; border-radius:5px;"> 비지도 학습 </span>
답을 가르쳐 주지 않고 학습시키는 방법

- 데이터 자체에 내재된 구조 파악해서 학습
- 정답 레이블이 없으므로 모델 성능 명확히 측정 불가능
- 표현 학습을 통해 고유 패턴 식별 가능

> 목표가 명확하지 않고 새로운 특성을 찾을 때!

##### <span style="background-color:#DCDCDC; color:#000000; padding:4px; border-radius:5px;"> 적합 케이스 </span>
1. 패턴이 알려지지 않은 경우
2. 패턴이 계속해서 변하는 경우
3. 열린 문제 해결 및 지식을 일반화해야 하는 경우

ex) 고객 세그먼트 분석, 이상 거래 탐지, 자연어 처리

> 목표
- 응집도 최대화: 같은 군집에 속하는 데이터끼리는 최대한 비슷하도록 함
- 분리도 최대화: 서로 다른 군집은 최대한 분리되도록 함

## <span style="background-color:#E0F7FA; color:#1565C0; padding:4px; border-radius:5px;"> 데이터 준비 </span>

### <span style="background-color:#FFFFE0; color:#CD5C5C; padding:4px; border-radius:5px;"> 군집화의 고려 사항 </span>
##### <span style="background-color:#DCDCDC; color:#000000; padding:4px; border-radius:5px;"> 변수 유형 이해 </span>
- 변수 유형에 따라 적절한 군집화 알고리즘 달라짐

1. 연속형 변수
- 거리 측정: 유클리디안 거리, 맨하탄 거리 등
- 적합 알고리즘: K-means, 계층적 군집화 등
- 주의사항: **스케일링** 필수

2. 명목형 변수
- 거리 측정: 해밍 거리, 자카드 거리 등
- 적합 알고리즘: K-modes, 계층적 군집화 등
- 주의사항: **원-핫 인코딩** 또는 더미 변수 변환

3. 혼합형 변수
- 적합한 알고리즘: K-prototypes, Gower distance 기반 계층적 군집화
- 처리 방법: 각 변수 유형에 맞는 거리 함수를 조합하여 사용

##### <span style="background-color:#DCDCDC; color:#000000; padding:4px; border-radius:5px;"> 거리 / 유사도 정의와 측정 </span>

1. 유클리디안 거리
$d(\mathbf{x}, \mathbf{y}) = \sqrt{\sum_{i=1}^n (x_i - y_i)^2}$

> - 직선 거리 의미
> - K-means에서 기본 사용

2. 맨하탄 거리

$d(\mathbf{x}, \mathbf{y}) = \sum_{i=1}^n |x_i - y_i|$

> - 각 차원의 차이의 절댓값 합
> - 이상치에 덜 민감

3. 코사인 유사도

$\text{similarity} = \frac{\mathbf{x} \cdot \mathbf{y}}{||\mathbf{x}|| \times ||\mathbf{y}||}$

> - 벡터 간의 각도 측정
> - 텍스트/고차원 데이터에 효과적

##### <span style="background-color:#DCDCDC; color:#000000; padding:4px; border-radius:5px;"> 6가지 연결 방법 </span>

1. Single linkage(단일 연결법/최단 연결법)



2. Complete linkage(완전 연결법/최장 연결법)



3. Average linkage(평균 연결법)



4. Centroid Method(중심 연결법)



5. Median(중앙 연결법)

6. Ward's Procedure(Ward 연결법)



##### <span style="background-color:#DCDCDC; color:#000000; padding:4px; border-radius:5px;"> 차원 축소 </span>

> 변수를 제거하여 차원 낮추고, 고차원 데이터를 저차원 공간에 투영하면서도 핵심 정보는 지켜내도록 한다

- 선형 투영: 고차원 -> 저차원 (선형적)

ex) 주성분 분석(PCA), 특이값 분해(SVD), 랜덤 투영...

- 매니폴드 학습(비선형 차원 축소): 유클리드 거리 말고 데이터 포인트 사이의 곡선 거리 고려하며 학습

ex) t-SNE, UMAP, Isomap...

##### <span style="background-color:#DCDCDC; color:#000000; padding:4px; border-radius:5px;"> PCA </span>

> 다차원 데이터에서 가장 정보량(분산)이 큰 방향을 찾고, 그 방향을 기준으로 데이터를 투영



> 대각선 오른쪽 윗방향으로...

##### **주의 사항**
1. 표준화 필수 <BR>
2. 결측값 처리 <BR>
3. 직관적 해석 어려움 <BR>
4. 정보 손실

## <span style="background-color:#E0F7FA; color:#1565C0; padding:4px; border-radius:5px;"> 군집화 알고리즘 </span>

### <span style="background-color:#FFFFE0; color:#CD5C5C; padding:4px; border-radius:5px;"> 계층적 군집화 </span>

> 데이터 간의 유사성을 기반으로 트리 구조(dendrogram)를 형성하며, 상향식 또는 하향식 방식으로 군집을 형성해 나가는 방법

- 클러스터링 종류 이후에 원하는 군집의 개수를 선택할 수 o

- 트리 구조
+ 상향식: 응집형 계측적 군집화 (병합 클러스터링)
+ 하향식: 분리형 계층적 군집화 (분할 클러스터링)



##### <span style="background-color:#DCDCDC; color:#000000; padding:4px; border-radius:5px;"> 응집형 계층적 군집화 </span>



> 모든 데이터 포인트를 하나의 클러스터로 간주 -> 점차적으로 가까운 클러스터들을 병합
- 가장 가까운 클러스터를 어떻게 정의하느냐에 따라 다양한 알고리즘으로 나뉨 <BR>
ex) 단일 연결, 완전 연결...

##### **과정**
1. 거리 행렬 계산
2. 단일 클러스터 시작
3. 가장 가까운 클러스터 병합
4. 거리 행렬 업데이트
5. 반복: 모든 데이터 포인트가 하나의 클러스터에 속할 때까지 3,4번 반복



### <span style="background-color:#FFFFE0; color:#CD5C5C; padding:4px; border-radius:5px;"> k-means </span>



> 데이터를 정해진 개수(K)의 그룹으로 나누되, 각 그룹의 중심점(centroid)과의 거리가 가장 가까운 데이터끼리 묶는 군집화 알고리즘

_왠지 knn이 생각난다_

##### <span style="background-color:#DCDCDC; color:#000000; padding:4px; border-radius:5px;"> 주요 단계 </span>



##### <span style="background-color:#DCDCDC; color:#000000; padding:4px; border-radius:5px;"> 장단점 </span>

##### **장점**
- 직관적, 구현 쉬움
- 대용량 데이터에도 적용 가능

##### **단점**
- 초기 centroid 값에 민감
- 군집 수 (k) 결정 어려움 <BR>
-> 엘보우 방법, 실루엣 계수 통해 탐색
- 이상치에 민감
- 기하학적 모양의 군집은 파악 어려움

##### <span style="background-color:#DCDCDC; color:#000000; padding:4px; border-radius:5px;"> k-means++ 알고리즘 </span>
> k-means에서 초기 중심점들을 서로 멀리 떨어진 곳에 위치시켜 더 일관되고 좋은 결과를 도출하는 개선된 **초기화** 방법

##### <span style="background-color:#DCDCDC; color:#000000; padding:4px; border-radius:5px;"> 엘보우 방법 </span>
> 클래스 내 SSE를 바탕으로 그래프를 활용해 최적 클러스터 개수 k를 추정하는 방법

### <span style="background-color:#FFFFE0; color:#CD5C5C; padding:4px; border-radius:5px;"> DBSCAN </span>
> 밀도가 높은 지역의 데이터를 하나의 군집으로 묶고, 밀도 기준을 만족하지 못하는 점은 군집에 포함시키지 않는 군집화 알고리즘

- 명시적으로 이상치 지정하여 클러스터링에서 제외 <BR>
-> 비교적 이상치에서 자유로움

##### <span style="background-color:#DCDCDC; color:#000000; padding:4px; border-radius:5px;"> 단계 </span>
1. 샘플 분류
- 핵심 샘플: 특정 반경 ε 안에 있는 이웃 샘플이 지정된 개수(MinPts) 이상인 경우
- 경계 샘플: ε 이내에 MinPts보다 이웃이 적지만 다른 핵심 샘플의 반경 ε 안에 있는 경우
- 잡음 샘플: 위 두 조건에 해당하지 않는 모든 샘플

2. 클러스터 생성

3. 경계 샘플 할당

##### <span style="background-color:#DCDCDC; color:#000000; padding:4px; border-radius:5px;"> 하이퍼파라미터 </span>
- `eps` : 두 포인트가 이웃이 되기 위한 포인트 사이의 최대 거리
- `min_samples` : 한 포인트가 군집이 되기 위한 eps 거리 내 최소 포인트 개수

##### <span style="background-color:#DCDCDC; color:#000000; padding:4px; border-radius:5px;"> 장단점 </span>

##### **장점**
- 기하학적 분포 처리 가능
- 이상치에 민감 x
- 잡음 샘플 구분해냄

##### **단점**
- 하이퍼파라미터 설정에 많은 영향받음
- 속도 느림
- 차원의 저주 영향
- 다른 밀도 분포 가진 데이터의 군집 분석 잘하지 못함

##### <span style="background-color:#DCDCDC; color:#000000; padding:4px; border-radius:5px;"> HDBSCAN </span>
> DBSCAN을 계층적 군집화처럼 변환한 버전

### <span style="background-color:#FFFFE0; color:#CD5C5C; padding:4px; border-radius:5px;"> GMM </span>
> 데이터가 여러 다른 모양의 가우시안 분포로 구성되었다고 가정하고, 각 분포를 클러스터로 인식하는 군집화 방법

##### <span style="background-color:#DCDCDC; color:#000000; padding:4px; border-radius:5px;"> GMM의 기본 가정 </span>

1. 관측된 데이터는 특정 가우시안 확률 분포에 의해 생성됨

2. 전체 데이터셋에는 여러 개의 다변량 가우시안 분포가 섞여있음

3. 개별 데이터는 우도에 따라 K개의 가우시안 분포 중 하나에 속함

##### <span style="background-color:#DCDCDC; color:#000000; padding:4px; border-radius:5px;"> 과정 </span>

1. 전체 데이터셋 분포 확인

2. 전체 데이터셋이 서로 다른 정규 분포 형태의 확률 분포 곡선으로 구성되어 있다고 가정 

3. 전체 데이터셋을 구성하는 여러 개의 정규분포 곡선 추출

4. 개별 데이터가 이 중 어떤 정규분포에 속하는지 결정 <BR>
-> 이때 각각의 분포가 하나의 군집

##### <span style="background-color:#DCDCDC; color:#000000; padding:4px; border-radius:5px;"> 모델의 파라미터 추정 </span>
> 기댓값-최대화(EM) 알고리즘을 사용

- 추정 가능 요소: 각 군집의 중심, 모양, 크기, 방향

**EM 알고리즘 단계**



##### <span style="background-color:#DCDCDC; color:#000000; padding:4px; border-radius:5px;"> 장단점 </span>

##### **장점**
- 다양한 데이터셋에 적용 가능

##### **단점**
- 시간 오래 걸림
- 가정한 분포에 맞지 않은 데이터일 경우 성능 저하

## <span style="background-color:#E0F7FA; color:#1565C0; padding:4px; border-radius:5px;"> 군집화 평가 방법 </span>

### <span style="background-color:#FFFFE0; color:#CD5C5C; padding:4px; border-radius:5px;"> 외부 평가와 내부 평가 </span>
##### <span style="background-color:#DCDCDC; color:#000000; padding:4px; border-radius:5px;"> 외부 평가 </span>

- 데이터에 정답 레이블이 존재하는 경우에 사용
- 군집화 결과 vs 실제 레이블 비교 -> 유사도 평가
ex) 지표: ARI (-1~1, 1에 가까울수록 일치)

##### <span style="background-color:#DCDCDC; color:#000000; padding:4px; border-radius:5px;"> 내부 평가 </span>

- 정답 레이블 없는 경우
- 군집 내부 응집도(cohesion)와 군집 간 분리도(separation)를 기반으로 군집의 품질 측정
- 각 알고리즘의 내부 평가 지표 및 목포
+ K-means 알고리즘 : SSE(Sum of Squared Errors)를 최소화
+ GMM : 로그 우도(모델이 주어진 데이터를 얼마나 잘 설명하는지 나타내는 지표)를 최대화

ex) 실루엣 계수, 던 지수

### <span style="background-color:#FFFFE0; color:#CD5C5C; padding:4px; border-radius:5px;"> 실루엣 계수 </span>



> 클러스터 내 샘플들이 얼마나 조밀하게 모여 있는지를 측정하는 도구

$
s^{(i)} = \frac{b^{(i)} - a^{(i)}}{\max \left\{ b^{(i)}, a^{(i)} \right\}}$

1. 클러스터 응집력 a 계산
2. 클러스터 분리도 b 계산
3. 차이를 둘 중 큰 값으로 나누어 실루엣 계산

|조건|실루엣 계수|의미|평가|
|------|----|--------|---|
|분리도 > 응집도|1에 가까움|샘플이 자신의 클러스터에 잘 속해 있고, 다른 클러스터와는 멀리 떨어져 있음|군집화가 잘 됨|
|분리도 = 응집도|0에 가까움|샘플이 클러스터의 경계에 위치하여 군집 간 거리가 모호|군집 간 분리가 명확하지 않음|
|분리도 < 응집도|-1에 가까움|샘플이 잘못된 클러스터에 할당되어 군집이 섞여 있는 상태|군집화가 잘못됨|

### <span style="background-color:#FFFFE0; color:#CD5C5C; padding:4px; border-radius:5px;"> Dunn Index </span>
> 클러스터 간 최소 거리(분리도)와 클러스터 내 최대 거리(응집도)의 비율을 계산해 클러스터링의 품질을 평가하는 지표

$\text{Dunn Index} = \frac{
\min\limits_{i \ne j} \; d(C_i, C_j)}{\max\limits_{1 \le k \le K} \; \delta(C_k)}$

d: 분리도, $\delta$: 응집도

- 클수록 good (0 이상에서 무한대)
: 군집 간 거리 크고 (분리도 high) 군집 내부 거리 작을수록(응집도 high)

##### **한계**
- 군집 수가 많아질수록 계산 비용 증가
- 극단적인 값에 큰 영향을 받음
- 이상치에 민감하게 반응
