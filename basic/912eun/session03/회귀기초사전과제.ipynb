## <span style="background-color:#E0F7FA; color:#1565C0; padding:4px; border-radius:5px;"> 머신러닝? </span>

모델을 학습하여 데이터에서 패턴 발견 -> 새로운 데이터 예측



### <span style="background-color:#FFFFE0; color:#CD5C5C; padding:4px; border-radius:5px;"> 지도 학습 </span>

입력/결과값 같이 주고 학습시키는 방법 <BR>
-> 예측값과 정답이 같아지도록 학습

##### <span style="background-color:#DCDCDC; color:#000000; padding:4px; border-radius:5px;"> 머신러닝 과정 </span>


1. **labeled data** 독립변수/종속변수가 모두 존재하는 데이터 <BR>
를 training set과 test set으로 나눔
2. training set을 통해 모델 학습: 독립변수로 추정한 예측값이 종속변수와 같아지도록 훈련

### <span style="background-color:#FFFFE0; color:#CD5C5C; padding:4px; border-radius:5px;"> 회귀와 분류 </span>
##### <span style="background-color:#DCDCDC; color:#000000; padding:4px; border-radius:5px;"> 회귀 </span>
- 주어진 데이터(X) 기반으로 정답(Y)를 잘 맞히는(Fit) 함수를 찾는 문제 <BR>
- **연속형 변수** 예측

##### <span style="background-color:#DCDCDC; color:#000000; padding:4px; border-radius:5px;"> 분류 </span>
- 기존 데이터가 어떤 레이블에 속하는지 패턴을 알고리즘으로 인지 -> 관측 데이터 레이블 판별
- **범주형 변수** 예측

## <span style="background-color:#E0F7FA; color:#1565C0; padding:4px; border-radius:5px;"> 선형회귀 </span>


- 데이터를 가장 잘 설명하는 회귀선을 찾는 과정
- 현재 데이터 요약 / 미래값 예측

### <span style="background-color:#FFFFE0; color:#CD5C5C; padding:4px; border-radius:5px;"> 단순 선형 회귀 </span>

- 하나의 독립변수만 존재하는 선형 회귀
- y=aX+b (a: 회귀계수, b: y절편) : a, b 찾는 과정

### <span style="background-color:#FFFFE0; color:#CD5C5C; padding:4px; border-radius:5px;"> 최소제곱법 </span>


- 잔차와 제곱의 합이 최소가 되는 지점: 최적의 회귀선 구하는 방식
- 잔차? 실제값-예측값

### <span style="background-color:#FFFFE0; color:#CD5C5C; padding:4px; border-radius:5px;"> 경사하강법 </span>
최적의 회귀선 찾는 방법: 계산 과정이 복잡할 때 사용

##### **목적함수**
최적화를 위해 오차를 최소화하는 것이 목적인 함수

##### **손실함수**
개별 데이터 샘플에 대한 오차를 측정하는 함수

##### **비용함수**
전체 데이터셋에서 평균적인 손실 측정하는 함수 <BR>
: 손실함수 평균 or 합산값 <BR>
-> 머신러닝에서 비용함수 최적화하는 것이 목표
: 비용함수=목적함수

##### <span style="background-color:#DCDCDC; color:#000000; padding:4px; border-radius:5px;"> 1) 손실함수 </span>

ti: i번째 실제값 <BR>
yi: i번째 예측값


> E를 최소화하는 W와 b값 찾기

- W로 x축을 구성하는 경우


E값이 그래프에서 최소 지점으로 이동: W값이 왼쪽으로

: 최솟값에 도달 시 미분값=0

##### <span style="background-color:#DCDCDC; color:#000000; padding:4px; border-radius:5px;"> 2) 학습률(Learning Rate) </span>
얼만큼 이동하냐? '적당한' 학습률이 가장 좋다



##### <span style="background-color:#DCDCDC; color:#000000; padding:4px; border-radius:5px;"> 3) Local Minima 문제 </span>


##### <span style="background-color:#DCDCDC; color:#000000; padding:4px; border-radius:5px;"> 4) 해결법: 모멘텀 </span>
기존 경사하강법: 이전 기울기 잊음 -> 현재 기울기 중심으로만 이동
> 이전 기울기, 이동 방향 고려하도록 **관성** 부여



> **장점**
> - 작은 기울기는 쉽게 넘어갈 수 o
> - 지역 최소값 탈출 가능 (비교적 빠른 시간으로)

### <span style="background-color:#FFFFE0; color:#CD5C5C; padding:4px; border-radius:5px;"> 다중 선형 회귀 </span>


- 독립 변수가 2개 이상인 경우의 선형회귀
- 회귀선: 평면
- 최소제곱법 사용 가능: 관측치와 평면간의 차이

$y = W_1 x_1 + W_2 x_2 + ... + W_n x_n + b$

##### <span style="background-color:#DCDCDC; color:#000000; padding:4px; border-radius:5px;"> 다중공선성 </span>
- 독립 변수 간의 상관관계가 큰 경우 발생
- 회귀분석 정확도 저하 (어떤 독립 변수가 종속 변수에 얼마나 영향 미치는지 정확히 구분 어려움)

> 다중공선성 확인법
1) 상관계수 확인
2) VIF 지수 (분산 팽창 인수)

> 다중공선성 대처법
1. 변수 제거(변수 선택법)
2. 변수 변환
3. 규제 선형 모델 활용
4. PCA(주성분분석)
데이터 차원 축소에 사용하는 통계 기법



## <span style="background-color:#E0F7FA; color:#1565C0; padding:4px; border-radius:5px;"> 규제선형모델 </span>

모델이 학습 데이터에 과적합되지 않도록 규제를 가하는 모델 <BR>

> - 선형회귀에서는? 계수(기울기) 크기 조정
> - 회귀 계수가 커지면 과적합 발생: 특정 독립 변수(특징)에 과하게 의존
> - 과적합이 왜 안 됨? 학습 데이터에는 맞지만 예측 성능 크게 떨어짐

##### 비용 함수는 최소제곱법과 회귀 계수 값 커지지 않도록 하는 방법 균형 유지 필요

### <span style="background-color:#FFFFE0; color:#CD5C5C; padding:4px; border-radius:5px;"> 릿지 회귀(L2 규제) </span>
W(회귀계수)의 제곱에 패널티 부여 <BR>
-> W 클수록 강한 패널티 받음, 변수 제거 X

##### <span style="background-color:#DCDCDC; color:#000000; padding:4px; border-radius:5px;"> 이용 </span>
(1) 예측 변수 많을 때 <BR>
: 고차원 데이터에서 모든 변수에 작은 가중치 부여
(2) 다중공선성 존재 <BR>
: 변수 선택 불필요


### <span style="background-color:#FFFFE0; color:#CD5C5C; padding:4px; border-radius:5px;"> 라쏘 회귀(L1 규제) </span>
W의 절댓값에 패널티 부여 <BR>
-> W 크기와 상관없이 같은 패널티, 변수 선택

##### <span style="background-color:#DCDCDC; color:#000000; padding:4px; border-radius:5px;"> 이용 </span>
(1) 예측 변수 많고, 그중 일부만 중요할 때 <BR>
(2) 해석을 간단히... <BR>
: 모델 단순화 <BR>
> 최적화 복잡할 수 있고, 수치 불안정할 수 o

### <span style="background-color:#FFFFE0; color:#CD5C5C; padding:4px; border-radius:5px;"> 엘라스틱넷(L1+ L2) </span>
불필요한 변수 제거, 남은 변수 적절한 크기로 유지

##### <span style="background-color:#DCDCDC; color:#000000; padding:4px; border-radius:5px;"> 이용 </span>
예측 변수 수가 많고, 중요 변수 선택하면서도 다중공선성 관리 <BR>
> 변수 간 상관관계 높을 때 성능 good

## <span style="background-color:#E0F7FA; color:#1565C0; padding:4px; border-radius:5px;"> 모델 평가 방법 </span>

### <span style="background-color:#FFFFE0; color:#CD5C5C; padding:4px; border-radius:5px;"> 성능 평가 지표 </span>

##### <span style="background-color:#DCDCDC; color:#000000; padding:4px; border-radius:5px;"> 평균 제곱 오차(MSE) </span>

$\text{MSE} = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2$

: 예측값과 관측값 사이의 오차 제곱 평균 -> 이상치에 민감

##### <span style="background-color:#DCDCDC; color:#000000; padding:4px; border-radius:5px;"> 평균 절대 오차(MAE) </span>

$\text{MAE} = \frac{1}{n} \sum_{i=1}^{n} |y_i - \hat{y}_i|$

: 예측값과 관측값 사이의 절대값 오차 평균 -> 이상치 존재 시에도 예측 성능 유지 가능
