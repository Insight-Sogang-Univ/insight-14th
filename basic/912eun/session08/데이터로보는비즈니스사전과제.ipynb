## <span style="background-color:#E0F7FA; color:#1565C0; padding:4px; border-radius:5px;"> Data-Driven-Decision_Making (DDDM) </span>
> 비즈니스 의사 결정을 제공하기 위해 직관 대신 **데이터와 분석을** 사용하는 것을 강조하는 접근 방식

### <span style="background-color:#FFFFE0; color:#CD5C5C; padding:4px; border-radius:5px;"> 6 Steps of DDDM </span>


### 목표 정의 -> 데이터 식별, 준비 및 수집 -> 구성 및 탐색 -> 데이터 분석 수행 -> 결론 도출 -> 구현 및 평가

### <span style="background-color:#FFFFE0; color:#CD5C5C; padding:4px; border-radius:5px;"> 데이터 분석 유형 </span>


##### <span style="background-color:#DCDCDC; color:#000000; padding:4px; border-radius:5px;"> 설명 분석 </span>
> 과거 성과에 대한 인사이트 <br>
ex) 월별 판매 보고서, 고객 만족도 설문 조사, 웹 사이트 트래픽 분석

##### <span style="background-color:#DCDCDC; color:#000000; padding:4px; border-radius:5px;"> 진단 분석 </span>
> - 특정 이벤트가 발생한 이유 파악
> - 사고의 근본 원인/추세 파악(상관관계 식별)

##### <span style="background-color:#DCDCDC; color:#000000; padding:4px; border-radius:5px;"> 예측 분석 </span>
> - 과거 데이터 기반으로 미래 추세나 결과 예측
> - 통계 모델, 머신 러닝 활용

##### <span style="background-color:#DCDCDC; color:#000000; padding:4px; border-radius:5px;"> 예측 분석 </span>
> - 데이터 기반 조치 권장
> - 예측 분석+최적화 알고리즘

## <span style="background-color:#E0F7FA; color:#1565C0; padding:4px; border-radius:5px;"> 목표 정의 </span>

### <span style="background-color:#FFFFE0; color:#CD5C5C; padding:4px; border-radius:5px;"> 지표 </span>
> 방향이나 목적, 기준 등을 나타내는 표지

$metrics=\frac {molecule}{denominator}$

> 보통 분모 대비 분자로 표현

|유형|공식 구조|예시 지표|
|-----|-------|----------|
|효율성|$\frac {산출}{투입}$|$노동생산성=\frac{생산량}{노동 시간}$|
|효과성|$\frac {달성 결과}{목표}$|$목표 달성률=\frac {실제 매출}{목표 매출}$|
|수익성|$\frac {이익}{자원}$|$ROE=\frac {순이익}{자기자본}$|
|성장성|$\frac {증가분}{기존 규모}$|$매출 성장률=\frac {(당기 매출- 전기 매출)}{전기매출}$|
|안정성|$\frac {보유자산}{부채}$|$부채비율=\frac {부채}{자기자본}$|

##### <span style="background-color:#DCDCDC; color:#000000; padding:4px; border-radius:5px;"> 경영 기능별 지표 정리 </span>
|기능|효율성 지표(Output÷input)|효과성 지표(Actual÷Target)|수익성/가치 지표(Profit÷Resource)|성장성/안정성 지표|
|--|------|------|-------|-----|
|재무|자산회전율 = 매출액÷총자산|예산 집행률 = 실제 지출 ÷ 계획 지출|ROE = 순이익 ÷ 자기자본, ROA = 순이익 ÷ 자산|매출 성장률, 부채비율|
|마케팅|CAC 효율 = 신규고객 수 ÷ 마케팅비|전환율 = 구매자 ÷ 방문자|마케팅 ROI = 캠페인 이익 ÷ 캠페인 비용|시장점유율, 성장률|
|HR|노동생산성 = 산출량 ÷ 인력투입|교육 효과 달성률 = 실제 역량향상 ÷ 목표 역량향상|HR ROI = 교육 후 성과 증가 ÷ 교육비|이직률, 직원 유지율|
|운영|설비가동률 = 실제 가동시간 ÷ 총 가능시간|품질목표 달성률 = 합격품 ÷ 목표 품질수준|단위당 원가 절감율 = 절감액 ÷ 총원가|재고회전율, 공급망 안정성|

### <span style="background-color:#FFFFE0; color:#CD5C5C; padding:4px; border-radius:5px;"> 지표 정의 </span>
> 절대적 수치 구할 수 없을 때: Proxy Metric  (대체 지표)

### <span style="background-color:#FFFFE0; color:#CD5C5C; padding:4px; border-radius:5px;"> 북극성, L1/L2 지표 </span>

##### <span style="background-color:#DCDCDC; color:#000000; padding:4px; border-radius:5px;"> 북극성 지표 </span>

> 기업이 가장 중요하게 생각하는 지표 <br>
ex) 에어비엔비: 예약된 박 수, 페이스북: DAU

##### <span style="background-color:#DCDCDC; color:#000000; padding:4px; border-radius:5px;"> L1, L2 </span>
L1: 고수준의 성과 지표 (전략 관점) <br>
L2: L1 지표의 세부 사항을 측정하는 지표 (전술 관점)

### <span style="background-color:#FFFFE0; color:#CD5C5C; padding:4px; border-radius:5px;"> AARRR </span>



##### <span style="background-color:#DCDCDC; color:#000000; padding:4px; border-radius:5px;"> Acquistion </span>
> 사용자 획득, 유입

**지표 예시**
- CAC: 유저 획득 비용
- CTR: 클릭 한 건당 얼마를 지출하느냐
- CPC: 인스톨당 과금되는 상품
- CPL: 유저 행동당 드는 비용
- ROAS: 광고로 인한 매출액/광고비

##### <span style="background-color:#DCDCDC; color:#000000; padding:4px; border-radius:5px;"> Activation </span>
> 사용자가 첫 방문 시 주요 기능을 직접 사용하는 것

**단계 구성**
1. Sign-up: 유저가 제품에 들어온 시점
2. Setup Moment: 유저가 제품의 핵심가치를 경험할 준비 완료
3. A-HA! Moment: 유저가 핵심가치를 처음 경험한 순간
4. Habit Moment: 유저가 제품 사용에 대한 습관 형성

##### <span style="background-color:#DCDCDC; color:#000000; padding:4px; border-radius:5px;"> Retention </span>
> 신규 사용자 -> 장기 사용자, 유저 유지

**지표 예시**
- DN Retention: N일 안으로 재접속하는 경우
- MAU/DAU/WAU: 일/주/월마다 방문하는 횟수
- Stickness(DAU, MAU): 해당 기간 동안 활성화 수

##### <span style="background-color:#DCDCDC; color:#000000; padding:4px; border-radius:5px;"> Revenue </span>
> 사용자가 제품을 유료 구매

**지표 예시**
- PU(Paying User): 전체 사용자 중에 결제를 한 사용자 비율
- Average Check: 구매 1회당 평균 금액 (Revenue/Transaction)
- ARPU(Average Revenue Per User): 인당 평균 매출
- ARPPU(Average Revenue Per Paying User): 결제자당 평균 매출
- LTV(Life Time Value): 유저가 진입부터 이탈까지 누적 수익

##### <span style="background-color:#DCDCDC; color:#000000; padding:4px; border-radius:5px;" > Referral </span>
> 사용자가 제품을 타인에게 추천

**지표 예시**
- NPS: 순 추천고객 지수
- 바이럴 계수: 기존 고객 한 명당 데리고 온 새로운 고객의 수 (>1이면 Acquisition 생략에도 서비스 성장)
- 바이럴 주기: 기존 고객 1명이 새로운 고객을 데려오는 데 걸리는 시간
