    - 협업 필터링

# 추천시스템 사전과제

## 추천 시스템 개념 <br>
- 추천 시스템 = 사용자 정보 데이터를 분석하여 개인의 취향에 맞는 아이템을 추천하는 알고리즘 => 고객 만족 증가 & 매출 증가 기대    

#### <span style='background-color: #fff5b1'>1. 사용되는 데이터</span>
**명시적 데이터**   
= 고객이 입력한 데이터 => 사용자가 직접 선호도 표현    
ex. 평점, 투표, 스케일 점수   
- 명백한 데이터 얻기 어려움    
- 모든 사용자 선호도 반영되지 않음   
- 과거의 선호도가 현재와 동일한지 알 수 없음    

**암시적 데이터**   
= 고객 행동을 기반으로 수집한 데이터 => 사용자 선호도가 간접적으로 표현    
ex. 조회수, 구매 기록, 페이지 체류 기간 등   
- 사용자가 서비스를 이용하면 로그데이터나 클릭 기록 등은 무조건 남지만 높은 수치가 높은 선호도를 의미하는 건 아님   

#### <span style='background-color: #fff5b1'><u>2. 추천시스템 분류</u></span>
**내용 기반 필터링**   
- 콘텐츠 유사도 기반으로 추천    
- 사용자가 이용 / 선택한 아이템과 유사한 아이템 추천    
=> 한 사용자의 행동 이력 기준으로 추천   

**협업 필터링**    
- 서비스 내 다수의 사용자에게서 얻은 선호도 정보를 이용해 추천    
(1) 사용자 기반: 유사한 취향의 사용자가 선호한 아이템 추천 = 비슷한 사용자가 좋아한 상품 추천     
(2) 아이템 기반: 사용자 선호 아이템과 유사한 아이템 추천 = 선호한 아이템과 유사한 다른 아이템 추천  
=> 여러 사용자의 선호도를 활용해 추천  

**하이브리드 필터링**   
- 2가지 이상의 추천시스템 알고리즘 조합    

## <u>내용 기반 필터링</u> <br>
- 사용자 프로필과 아이템 특징과 일치도 계산 => 사용자 선호 아이템 추천 => 콘텐츠 유사도 기반 추천   
    - 사용자 프로필 = 사용자가 이전에 좋아했던 항목의 특징을 모아 놓은 데이터  

#### <span style='background-color: #fff5b1'><u>1. 유사도</u></span>
- 유사도 = 추천 시스템 이해에 중요한 수학적 개념   
    - 아이템 정보 데이터 벡터화하여 유사도 측정한 후, 유사도를 바탕으로 고객이 선호하는 아이템과 비교해 유사도 높은 아이템 선별   
    (1) 인풋 데이터 생성: 아이템 정보 활용해 테이블 형태로 데이터 정리한 후, 원-핫 인코딩해서 숫자로 변환    
    (2) 유저 정보 활용: 유저의 취향 파악한 후, 유저와 아이템을 동일한 기준의 피처로 표현   
    (3) 벡터 내적과 유사성: 내적 값이 클수록 두 벡터의 유사성이 높다는 사실을 바탕으로 추천   

#### <span style='background-color: #fff5b1'><u>2. 사용자 프로필</u></span>
**직접 지정형 데이터**   
= 사용자가 선호하는 아이템의 특징을 직접 지정하도록 함 ex. 서비스 가입 시 제시하는 서비스에 대한 선호도 체크   

**간접 지정형**   
= 사용자의 과거 행동 이력에 기반해 사용자 프로필 작성함 ex. SF 영화를 즐겨본 이력이 있는 유저에게 인터스텔라 추천   

**사용자 프로필 획득 방식**    
(1) 특성 가중치: 사용자가 가지고 있는 아이템 특성 가중치의 평균 값 활용   
(2) 사용자 기호 파악 or 설문조사 등으로 프로필 획득    
    - 평점, 댓글, 검색 로그 등    
(3) 아이템에 대한 유저의 기존 평가를 바탕으로 분류 문제 해결하는 것과 같음    
    - 사용자 프로필을 바탕으로 새로운 아이템을 유저가 평가한 metric 중 하나 선택    

#### <span style='background-color: #fff5b1'>3. 아이템 특징</span>
- 아이템을 벡터 형태로 표현, 도메인에 따라 다른 방법 적용      
- 컨텐츠 분석하는 아이템 분석 알고리즘: Clustering, Machine Learning, TF-IDF    

**<u>TF-IDF</u>**<br>
- TF = 단어 빈도: 특정 문서에서 특정 단어가 등장하는 빈도    
- IDF = 역문서 빈도: 전체 문서에서 특정 단어가 등장하는 빈도    
=> TF와 IDF를 통해 다른 문서에는 등장하지 않지만 특정 문서에만 자주 등장하는 단어를 찾아 가중치 계산     


#### <span style='background-color: #fff5b1'>4. 장단점</span>
(+)     
- 아이템에 대한 콜드 스타트 문제 해결 가능: 추천 시스템이 새로운 유저들에 대한 정보를 충분히 수집하지 못해서, 적절한 제품을 추천하지 못하는 문제   
- 아이템의 피처로 추천을 하기 때문에, 추천 이유 설명 용이    

(-)    
- 아이템 피처를 추출하고 이를 통해 추천하기 때문에, 정확한 피처를 추출하지 못하면 정확도 하락    
- 기존 아이템과 유사한 아이템만 추천해 새로운 장르의 아이템 추천 어려움   

## <u>협업 필터링</u> <br>
= 기존 사용자들과 아이템의 상관 관계를 기반으로 새로운 사용자와 아이템의 상관 관계 예측   
- 가정: 과거에 비슷한 선호도를 가진 사용자는 앞으로도 비슷한 선호도를 가질 가능성이 높음   
=> 여러 사용자의 아이템에 대한 평점 기반의 추천     

#### <span style='background-color: #fff5b1'><u>1. 메모리 기반 협업 필터링</u></span>
= 이웃 기반 협업 필터링    
- 사용자 / 아이템 간 유사도 기반 => 비슷한 취향의 사용자가 선호하는 아이템 추천    

**사용자 기반 접근**   
= 사용자 유사도가 높을수록 가중치 부여 => 같은 그룹의 다른 유저가 선호하는 아이템 추천   
- 유사한 사용자를 TOP-N으로 선정해 해당 TOP-N 사용자가 선호하는 아이템 추천   

**아이템 기반 접근**   
- 추천하고자 하는 타겟 아이템과 유사한 아이템 평점 활용 => 아이템 간 속성X 사용자들의 아이템에 대한 평가 척도가 유사한 아이템 추천      

**사용자 기반 vs 아이템 기반 비교**    

비교 기준 | 사용자 기반 [UBCF] | 아이템 기반 [IBCF]    
---------- | ---------- | ----------     
추천 정확성 | 아이템 수 > 사용자 수 | 사용자 수 > 아이템 수    
시스템 공간 및 효율성 | 아이템 수 > 사용자 수 | 사용자 수 > 아이템 수    
안정성 | 아이템이 수시로 바뀌는 경우 | 아이템 리스트 변화가 적고 고정적인 경우    
다양성 | 특정 사용자가 독특한 아이템 평가한 경우 독창적 아이템 추천 | 유사한 아이템 위주로 추천되어 다양성, 참신함 부족   

<br>**메모리 기반 협업 필터링 한계**    
- [scalability] 아이템 / 사용자가 계속 늘어나면 확장성 ↓    
- [sparsity] 주어진 선호도 데이터가 적으면 성능 저하 가능    
- 사용자가 늘어나면 연산은 늘어나고 오히려 성능이 떨어지기도 해서 KNN CF 이용    


#### <span style='background-color: #fff5b1'><u>2. 모델 기반 협업 필터링</u></span>

## 추천시스템 평가 <br>
**오프라인 평가**   
- 사용자 행동 이력에서 얻은 과거 로그로 모델의 예측 정밀도 등 평가 = 과거 사용자 행동 학습 후 미래 행동 예측     
    (+) 평가 비용 ↓ , 데이터 양 풍부해 결과 분산 ↓     
            - 결과 분산이 작으면 평가의 신뢰성이 높음. 분산이 작다는 것은 같은 모델을 여러 번 / 다른 데이터셋으로 평가해도 비슷한 성능 수치가 나온다는 의미

    (-) 과거 로그 기반이므로 실제 비즈니스 목표 / 만족도에 기여하지 못할 가능성 있음   

**온라인 평가**    
- 추천 모델 / 새로운 UI 일부 사용자에게 실제로 표시하여 평가   
- A/B 테스트, 인터리빙     
(+) 비즈니스 목표 기여 여부를 즉각적으로 확인     
(-) 구현 비용 ↑ , 성능이 좋지 않으면 만족도 하락 가능성    

**사용자 스터디**    
- 사용자 인터뷰 or 설문    
(+) 서비스 로그만으로 알 수 없는 개선점 발견 가능    
(-) 주관에 기반하는 경우가 많아 결과의 분산 ↑ , 데이터 양이 충분하지 않아 재현성 떨어질 가능성  

## 최근 추천시스템 연구 동향 <br>
- 자연어 처리 기술 기반의 추천 시스템 연구 활발
