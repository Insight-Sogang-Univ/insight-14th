# 연관 분석 기초 사전 과제

## 연관 분석 <br>
#### <span style='background-color: #fff5b1'>1. 정의</span>
= 거래 / 사건 안에 존재하는 항목 간 일정한 연관 규칙을 발견하는 분석   
- OTT, 질병 진단, 금융 기간, 요식 업계 등 다양한 분야에 사용 가능   
- 비지도 학습   

#### <span style='background-color: #fff5b1'>2. 연관 규칙 분석</span>
**연관규칙 형태**   
- if-then 형식 & A+B로 표현 ex. 라면을 구매하는 사람들은 달걀도 함께 산다      
    = 조건절: '만일 -라면' ex. 라면 구매     
    - 결과절: 뒷부분 ex. 달걀 구매     
    - 아이템 집합: 조건절 or 결과절을 구성하는 아이템들의 집합 ex. 라면, 달걀      
        - 상호배반이나, 여러 아이템이 들어가도 됨     

**연관규칙 주의**    
= 인과관계는 알 수 없다 (오직 상관관계만 파악)    
- 실험X 관찰 데이터 
- 방향성은 수학적 비율일 뿐   
- 상관관계 != 인과관계       

**연관 분석 vs 협업 필터링**    

구분 | 상품기반 협업 필터링 | 연관 분석     
------- | ------- | -------    
연산 방식 | 상관관계 기반 | 교집합 기반    
알고리즘 구조 | 상품마다 타 상품과 1:1 계산 | 상품마다 주문내역 1:n 계산     
특징 | 사용자 선호도에 따른 선호 상품 추출 | 연관성이 높은 상품을 세트로 추출    
추천 방식 | 타 상품의 유사도 순위대로 노출 | 인관 지표가 높은 세트 상품 노출     

## 주요 개념 <br>
#### <span style='background-color: #fff5b1'>1. 연관 규칙 조건</span>
**<u>지지도 (Support)</u>**    
= 전체 거래 중 A와 B가 동시에 포함하는 거래의 비율 = 두 품목을 함께 구매한 경우의 수가 일정 수준 이상이어야 함    
- 조건절이 발생하는 확률   
- 지지도가 낮다면 우연일 수 있는 규칙이 포함되거나 실행 가능한 인사이트 도출 어려움    
=> 이 규칙이 현실에서 의미를 가지는지 판단하는 1차적 기준   

**<u>신뢰도 (Confidence)</u>**    
= A를 포함하는 거래 수 중 A와 B가 동시에 포함되는 거래의 비율 = A를 포함하는 거래 중 B를 구입하는 경우의 수가 일정 수준 이상이어야 함   
- 신뢰도가 낮다면 의미 없는 여관 규칙이거나 추천 실패로 이어질 수 있음   

#### <span style='background-color: #fff5b1'>2. 연관 규칙 평가 측도</span>
**<u>향상도 (Lift)</u>**    
= A가 주어지지 않았을 때 B으 확률 대비 A가 주어졌을 때 B의 확률의 증가 비율 = A와 B가 우연히 함께 등장할 확률과 실제 함께 등장할 확률 비교    
(1) 독립일 때 기댓값   
- 사건 A와 B가 독립이면 P(A∩B) = P(A)*P(B)    
=> 향상도 = 1    
- A와 B가 독립이 아니라면 P(A∩B) = P(A)*P(B|A) = P(B)*P(A|B)    
=> A를 산 확률이 B를 살 확률에 영향을 주거나, B 산 확률이 A 살 확률에 영향을 줌   

(2) 실제 동시 발생 확률   
- 실제로 관측한 P(A∩B) > P(A)P(B) => 양의 연관성    
- 실제로 관측한 P(A∩B) < P(A)P(B) => 음의 연관성   

(3) 비율로 비교    

비율 | 내용    
------- | -------   
향상도 > 1 | 양의 상관관계 = 우연적 기회보다 높은 확률   
향상도 = 1 | 독립    
향상도 < 1 | 음의 상관관계 = 우연적 기회보다 낮은 확률    

<br>

##### ✅연관 규칙의 효용성은 지지도, 신뢰도, 향상도를 모두 반영해 평가

## 활용 사례 <br>
#### <span style='background-color: #fff5b1'>넷플릭스 case</span>
**'하우스 오브 카드' 제작 투자**   
(1) 지지도    
- 영국판 '하우스 오브 카드'를 시청한 사람 중 상당수가 케빈 스테이시 출연작 시청함 = 영국판 ∩  케빈 스페이시 지지도가 높음    
=> 사용자 데이터에서 이 조합이 빈번한 것을 활용해 사용자 기반의 취향 파악     

(2) 신뢰도     
- P(시청 완료율 높음|핀처 감독 작품 시청) 이 높음 = 핀처 감독 콘텐츠를 본 사람들은 끝까지 시청하는 경향이 있음    
=> 핀처 감독 콘텐츠는 몰입도가 높아 오리지널 시리즈 제작 시 리스크가 낮음   

(3) 향상도    
- 영국판 시청과 케빈 스페이시 콘텐츠 시청의 향상도 > 1 = 독립이 아님   
=> 두 요소가 결합되면 사용자 반응이 좋을 것으로 보고 조합을 기반으로 콘텐츠 기획    

## 연관 분석 알고리즘 <br>
#### <span style='background-color: #fff5b1'><u>1. Apriori 알고리즘</u></span>
**배경**    
- 연관 규칙을 찾고자 모든 조합을 시도하는 것은 매우 비효율적 = 무차별 탐색의 비효율   
- 상품 개수가 많을수록 조합의 수가 기하급수적으로 증가    

**Apriori 알고리즘**    
= 상위 조합부터 차례로 스캔하며 특정 조합이 자주 발생하지 않는다면 이 결과물로 탄생한 후속 조합들도 모두 후보에서 배제하는 방식     
(+) 하나의 조합만 검사해도 파생 조합까지 배제할 수 있어 시간과 연산량 효과적으로 감소    

**프로세스**     
- 빈발 항목 집합[일정 최소 지지도를 넘는 항목의 조합]을 찾고 단계적으로 조합을 생성, 필터링    

(1) **단일 항목 집합 생성**: 모든 개별 상품 지지도 계산해서, <u>최소 지지도 넘는 항목만 다음 단계로</u>    
(2) **2개 항목 집합 생성**: (1)에서 선별한 상품들로 <u>2개 항목의 조합을 만든 후, 해당 조합의 지지도를 계산하여 최소 지지도를 넘지 않는 조합 제거</u>   
(3) **더 큰 항목 집합 생성**: (2)에서 남은 조합을 바탕으로 <u>3개 이상으로 구성된 조합 만든 후, 최소 지지도를 넘지 않는 조합 제거</u>. 더 이상 조합 생성할 수 없을 때까지 과정 반복.      
(4) **최종 빈발 항목 집합 선정**: 마지막으로 남은 빈발 항목 집합을 바탕으로 <u>연관 규칙 생성</u>하고 각 규칙에 대해 <u>지지도, 신뢰도, 향상도 계산해 규칙 선택</u>    

**장단점**    
(+)     
- 수많은 상품 연관 구매 패턴 발견   
- 다른 연구 가설 탐지 가능   
- 간단한 원리, 쉬운 이해 및 분석    

(-)    
- 비즈니스 측면 중요한 현실적 중요 연관 규칙 부족    
- 데이터가 클수록 속도 느림    

#### <span style='background-color: #fff5b1'><u>2. FP Growth 알고리즘</u></span>
- Apriori 연산 속도 문제를 해결할 수 있도록 자료구조를 이용한 알고리즘    

**배경**    
- Apriori 알고리즘은 각 단계에서 데이터를 다시 스캔하기 때문에 여전히 데이터가 클수록 속도가 느려짐     

**FP-Growth 알고리즘 소개**     
- 빈발 품목 집합 효율적으로 탐색    
- FP-Tree 구조로 Apriori 효과적으로 구현    
- Tree, Array, Linked_list 합친 구조    
- 모든 거래를 확인해 아이템별 지지도 계산하고, 최소 지지도 이상의 아이템만 선택    

**프로세스**     
(1) 빈발 항목 필터링: 모든 항목의 빈도 계산 후, 최소 지지도 이상인 항목만 남김   
(2) 데이터 재정렬: 남은 항목을 거래별 빈도순으로 재정렬    
(3) FP-트리 생성: 재정렬 데이터로 FP-트리 생성 [거래 = 노드, 새로운 항목 = 새로운 노드]    
(4) 빈발 항목 집합 추출: FP-트리를 바탕으로 특정 항목 기준의 부분 트리 생성해 빈발 항목 집합 추출    
(5) 연관 규칙 도출: (4) 바탕으로 연관 규칙 생성하고, 각 규칙에 대한 ㅈ지도, 신뢰도, 향상도 계산해 규칙 선택   

**장단점**      
(+)     
- 데이터를 두 번만 스캔하는 tree 구조라 빠름   
- 후보 아이템셋 생성 없이 tree만 구성하면 됨    

(-)    
- 대용량 데이터셋에서 메모리 효율적으로 사용하지 않음     
- 설계하기 어렵고, 지지도 계산은 무조건 FP-Tree가 만들어져야 가능함
