 Number_of_kids  None/Unknown(=4)        .

```python
# 1. Number_of_kids    (None/Unknown=4)
def determine_kids(row):
    if pd.isnull(row['Number_of_kids']) or row['Number_of_kids'] == 4:

        # Household_Composition == 1, 3, 4   ->    
        if row['Household_Composition'] in [1, 3, 4]:  # No Kids
            return 0
        
        # Household_Size      
        # Household_composition == 2,  2   
        if row['Household_Composition'] == 2:          # 2 Adults Kids
            # (  - 2)      max   
            return max(row['Household_Size'] - 2, 0)
        
        # Household_Composition == 5,  1   
        if row['Household_Composition'] == 5:          # 1 Adult Kids
            # (  - 1)      max   
            return max(row['Household_Size'] - 1, 0)

        # Household_Size == 1 1    0 
        if row['Household_Size'] == 1:
            return 0
        
    return row['Number_of_kids']

# axis ?
df['Number_of_kids'] = df.apply(determine_kids, axis=1)
```

 Household_Composition  Unknown(=6)        .

```python
# 2. Household_Composition  Unknown(=6)   
def determine_composition(row):
    if row['Household_Composition'] != 6:
        return row['Household_Composition']
    
    size = row['Household_Size']
    kids = row['Number_of_kids']

    # 1 
    if size == 1:
    # 50%          
       return np.random.choice([3, 4], p=[0.5, 0.5])  # Single Female / Single Male

    # 2 
    if size == 2:
        if kids == 0:
            return 1  # 2 Adults No Kids
        else:
            return 5  # 1 Adults Kids

    # 3 
    if size >= 3:
        if kids == 0:
            return 1  # 2 Adults No Kids
        elif size - kids >= 2:
            return 2  # 2 Adults Kids
        else:
            return 5  # 1 Adult Kids

# axis ?
df['Household_Composition'] = df.apply(determine_composition, axis=1)
```

### Z-score
 Z-score          .   3         Z-score  3   -3        .

 Z-score     .

```python
def find_outlier_z(data, threshold=3):
    mean = np.mean(data)
    std = np.std(data)
    
    z_score = [(y-mean)/std for y in data] #

    masks = np.where(np.abs(z_score)>threshold)
    
    return masks
```

### Tukey Fences
![alt text](image-1.png)

Tukey Fences( )      ,   (IQR, Interquartile Range)  . IQR   50%  ,    25%  25%    .

(Q1  1.5 * IQR)   (Q3 + 1.5 * IQR)     . 1.5   3               .

 Tukey Fances     .

```python
def find_outlier_by_Tukey(data):
    q1, q3 = np.percentile(data,[25,75])
    iqr = q3-q1
    
    lower_bound = q1 - (iqr * 1.5)
    upper_bound = q3 + (iqr * 1.5)
    
    print(f'lower_bound {lower_bound.round(3)}, upper_bound {upper_bound.round(3)} .\n')
    
    # np.where   
    mask = np.where((data>upper_bound) | (data<lower_bound)) 
    
    # np.where Dataframe.index    
    #mask = data[(data['feature'] > upper_bound) | (data['feature'] < lower_bound)].index
    return mask
```

```python
sns.boxplot(x=df["Customer_paid"])
plt.show()

print(f"find_outlier_by_Tukey()    : {find_outlier_by_Tukey(df['Customer_paid'])}")
```
