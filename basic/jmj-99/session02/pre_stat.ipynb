# 통계

### 1.  
- 샘플(표본집단)만 추출해 분석<BR> → **[기술통계]** 샘플 자체에 대한 관심 / **[추론통계]** 모집단에 대한 관심
<BR> → 수치와 시각화를 통해 샘플 요약 / 추정과 검정을 통해 샘플에서 모집단 추론  
<BR>
- 자료: 설명과 해석을 위해 수집, 분석, 요약되는 사실과 수치  
    1. 양적 자료  
        (1) 구간 척도: 값 사이에 순서가 있고, 간격이 일정해 연산 가능    
        (2) 비율 척도: 절대영점과 임의의 측정단위로 구성되며 동간성 유지   
    2. 질적 자료   
        (1) 명목 척도: 레이블 / 이름으로 속성 식별    
        (2) 서열 척도: 명목 척도 + 항목 간 순서 / 순위     


### **2. 기술통계**   
- 데이터를 있는 그래도 설명하고 이해하는 데 중점 <BR>
1. **중심화 경향** <BR>
    (1) 평균 = 데이터 값의 총합 / 데이터 개수 => series.mean()  
    - 이상치에 민감, 절사 평균 개념 등장<BR>  

    (2) 중앙값 = 크기 순서로 정렬했을 때 한 가운데 위치한 값 => series.median()   
    - 이상치에 평균에 비해 덜 민감    
    - 데이터 개수가 짝수면, 가운데 두 값의 평균이 중앙값<BR>

    (3) 최빈값 = 데이터에서 가장 자주 나타나는 값 => series.mode()<BR>   

2. **자료의 산포** = 데이터가 퍼져있는 정도 <BR>
    (1) 범위 = 데이터의 최대값과 최소값의 차이 [전체 변동 범위] => series.max() - series.min()<BR>    

    (2) 분산 = 값이 평균에서 얼마나 멀리 퍼져있는지 =>series.var()<BR>  

    (3) 표준편차 = 분산의 제곱근 => series.std()<BR>   

    (4) 사분위수 = 데이터 표본을 4개의 동일한 크기로 나눈 값 => series.quantile()<BR>  

    (5) 변동계수 = (표준편차/평균)*100<BR>  
3. **자료의 정리 및 시각화** <BR>

### 3. 확률과 분포   
- 확률변수: 랜덤한 세계에서 어떤 사건이 발생할 때 나타나는 모든 경우의 수를 대응시키는 함수  
- 확률분포: 확률변수가 어떤 값을 가질 확률을 나타내는 함수  
1. 이산확률분포: 확률 변수가 이산적잊 값  
2. 연속확률분포: 확률 변수가 연속적인 값  <BR>  

- **대표적 분포 모형**  
1. 정규분포  
2. T 분포  
3. 카이제곱분포  
4. F 분포  

### **4. 추론통계-추정**  
- 수집한 데이터를 바탕으로 미래 / 조사하지 못한 전체 정보를 추론하고 결론을 도출하는 방법을 다루는 분야<BR>
  
1. 추정<BR>
- 모집단을 대표하는 일부를 표본으로 추출해 분석한 통계량으로 모집단의 모수를 예측하는 과정  
    - 점추정: 추종하려는 모수를 하나의 수치로 추정 =   
    - 구간추정: 추종하려는 모수가 있을 것으로 예상되는 구간을 정해 추정<BR>

2. 검정<BR>
- 표본을 추출해 얻은 표본통계량으로 모집단 모수에 대한 새로운 가설이 옳다고 판단할 수 있는지에 대한 평가 과정

### **5. 추론통계-검정**  
**가설 검정 절차** <BR>
- 가설 검정 = 표본의 정보를 사용해 모집단의 실제 값과 관련한 가설의 합당성 여부를 판단하는 과정 <BR>
1. 가설 설정
    - 귀무가설[현재 통용, 주로 기각하고자 하는 명제]과 대립가설[주장하고자 하는 명제] 설정  
2. 유의 수준 설정
    - 결과를 판단할 때 허용하는 오류의 최대치  
    - 주로 5% (0.05)로 설정  
    (1) 1종 오류: 귀무가설이 실제로 참이었는데 기각할 확률  
    (2) 2종 오류: 귀무가설이 실제로 거짓이었는데 기각하지 못할 확률  
3. 검정통계량 산출
    - 가설 검정 결과를 판단하는 데 사용되는 값  
    (1) Z값: 표준정규분포를 사용해 얻은 값. 데이터가 평균에서 얼마나 떨어져 있는지 std 단위로 나타냄. => 표본 30 이상인 경우  
    (2) T값: T 분포를 사용해서 얻은 값. 정규분포보다 꼬리가 두꺼워 작은 표본의 불확실성 반영 => 크기가 작은 표본에 대한 가설을 검정하는 경우  
    (3) F값: F 분포를 사용해서 얻은 값. 두 독립적인 카이제곱 분포된 변수들의 비율. => 분산 분석에서 두 개 이상 그룹 간 평균의 차이를 검정하는 경우   
4. 가설 기각 / 채택 판단
    - 유의 확률 (p-value): (p < 유의 수준)인 경우 귀무가설 기각  

**자료형에 따른 검정 방법** <BR>
1. [수치형-수치형] (피어슨) 상관분석
    - 두 변수 간 피어슨 상관계수가 얼마나 유의미한지 검정  
2. [범주형-수치형] F-test (ANOVA), T-test  
    (1) T-test: 범주가 2개인 경우에 사용하며, t값은 평균 간 차이 / 표준 오차  
    (2) F-test (ANOVA): 범주가 3개 이상인 경우에 사용하며, f값은 집단 간 분산 / 집단 내 분산  
3. [범주형-범주형] Chi2-test
    - 범주 갯수에 제한 없으며, 범주가 많을수록 카이제곱 통계량이 커짐  
    - 카이제곰 통계량 = (관측빈도-기대빈도)^2/기대빈도  
    - 카이제곱 통계량이 자유도의 2배 이상이면 차이가 존재한다고 해석 (범주형 변수의 자유도 = 범주의 수 - 1)  
4. [수치형-범주형] 수치형을 구간별로 범주화하지 않으면 검정 방법을 찾기 어려움
