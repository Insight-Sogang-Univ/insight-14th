# 데이터로 보는 비즈니스 상황 사전과제
## 데이터 기반 의사결정<br>
#### <span style='background-color: #fff5b1'>1. 데이터 분석 목적</span>    
- 목적 달성 프로세스의 질 ↑   
- 데이터 분석과 직감 / 경험치를 연계해 객관성 ↑   

**액션**   
(1) 결정    
- 신제품 출시에 관한 의사결정   
    - 출시 시기, 예상 판매량 등을 뒷받침하기 위한 데이터 분석

(2) 계획   
- 마케팅 전략 수립 / 미디어 관고, 전단지 배포 등 프로모션 계획    
    - 매출-미디어 관계 분석, 매출 시뮬레이션 등 계획 수립을 위한 데이터 분석    
(3) 행동    
- 판매 예측 모델을 위한 시스템 개발    
    - 일별 계획 실행에 활용하는 데이터 분석 ex. 매출-전단지 크기 관계, 매출-전단지 배포 요일 관계 등   
    
=> 객관적으로 과거를 돌아보고, 미래를 전망함

#### <span style='background-color: #fff5b1'><u>2. DDDM (Data-Driven Decision Making)</u></span>
= 경영적 의사결정을 위해 데이터와 분석을 사용하는 것을 강조하는 접근 방식   
- 데이터 소스 활용해 의사결정 프로세스 안내: 고객 피드백, 시장 동향, 재무 데이터 등    
<br>
#### <span style='background-color: #fff5b1'><u>3. 6 steps of DDDM</u></span> 

**(1) 목표 정의**    
- 조직 목표 명확하게 표현 & 달성을 위한 노력   

**(2) 데이터 식별, 준비 및 수집**    
- 데이터 요구 사항을 판별하며 데이터 소스를 평가 및 준비하여 데이터 수집, 검증   

**(3) 구성 및 탐색**    
- 패턴, 추세, 인사이트 발견을 위해 데이터 구조화 / 시각화    

**(4) 데이터 분석 수행**    
- 다양한 기술과 방법론으로 원시 데이터를 실행 가능한 인사이트로 변환   
- 전략에 영향을 미치는 패턴, 상관관계, 추세 파악 후 전략 결정    

**(5) 결론 도출**    
- 데이터 분석 결과 검토 후 비즈니스 상황에 배치하여 인사이트와 권장 사항 구성  

**(6) 구현 및 평가**    
- 실행 계획 개발과 리소스 할당, 지속적 모니터링   
- KPI 기준 결과 측정 후 피드백 수집 및 반복적 개선   

#### <span style='background-color: #fff5b1'><u>4. 데이터 분석 유형</u></span>
**설명 분석**    
- 데이터 취합 + 마이닝으로 과거 데이터 설명 및 요약 → 과거 성과 인사이트 제공   
- 월별 판매 보고서, 고객 만족도 설문 조사, 웹사이트 트래픽 분석 등   

**진단 분석**    
- 데이터 디스커버리, 마이닝, 상관관계 식별 등으로 이벤트 발생 원인 파악    
- 판매 감소, 고객 불만 급증 등의 추세 / 사고의 근본 원인 파악   

**예측 분석**    
- 과거 데이터 기반의 미래 추세 / 결과 예측   
- 통계 모델, 머신러닝과 예측 기술로 고객 행동 효과적 예측, 위험 관리    

**처방 분석**   
- 데이터 기반 조치 권장   
- 예측 분석 + 최적화 알고리즘 => 공급망 최적화, 마케팅 캠페인 전략 및 리소스 할당 결정 지원   


+) 탐색 분석, 추론 분석, 정성적 분석, 정량적 분석, 실시간 분석 등  

## 목표 정의<br>
#### <span style='background-color: #fff5b1'><u>1. 지표</u></span>
= 방향, 목적, 기준을 나타내는 표지    
= 추상적 개념과 관련되는 지수, 척도 등으로 측정하고자 하는 것을 숫자로 표현한 것   
- 주로 '분자 / 분모' 형식으로 표현    
    ex. ROE = 순이익/자기자본, 노동생산성 = 생산량 / 노동시간 등    
    - 절대치 지표, 지수형 지표, 복합 지표 등 예외 구조도 O

#### <span style='background-color: #fff5b1'>2. 지표 정의하기</span>
**Proxy Metric**    
= 추상적이라 프로덕트마다 나름의 가정과 정의, 논의가 필요한 지표   
- 설정 자체에 많은 논의가 필요하며 정답은 X   


#### <span style='background-color: #fff5b1'>3. 북극성 지표, L1/L2 Metrices</span>
**북극성 지표**     
= 기업이 가장 중요하게 생각하는 지표    

**L1/L2 지표**    
- L1 지표: 전사적, 고수준의 성과 지표   
- L2 지표: L1 지표의 세부 사항 측정하는 지표   

- 지표를 위한 프레임워크는 다양하지만 결론은 하나로 모임

#### <span style='background-color: #fff5b1'>4. AARRR</span>
- 프로덕트 별 가치 제공 방식에 따라 측정 지표도 달라져야 함   
    - 넷플릭스: 컨텐츠 감상 시간   
    - 우버: 택시 요청 중 실제 매칭 여부    

**AARRR 모델 = 해적 지표**   
(1) Acquisition    
= 사용자 획득 = 앱 / 웹사이트에 사용자가 방문하는 것   
- 신규 고객 획득 방법, 획득 채널 등에 대해 고민   
- 프로덕트에 따라 고객 획득 측정 기준이 다르지만, 대체로 조회 수, 가입 사용자 수 등으로 측정   
ex. CAC, CTR, ROAS 등   

(2) Activation     
= 사용자 활성화 = 앱 / 웹사이트에 사용자가 방문해 주요 기능을 직접 사용하는 것   
- 사용자에게 제품의 가치를 알려 활성화 정도 높여나감   
- 제품 핵심 기능 활용도, 사용자 온라인 접속 시간 등   
ex. 페이스북 - 회원가입 10일 내 최소 7명의 친구를 가진 신규 사용자 / 이커머스 앱 - 회원가입 3일 내 최소 1건 주문한 신규 사용자    

(3) Retention    
= 유지   
- 신규 사용자 → 장기 사용자 전환    
    - 신규 사용자 획득 비용 >> 기존 사용자 유지 비용   
- 장기 고객 양성 방법, 지속적 구매 행위 증가 방안 고민   
- 익일 유지율, 1주차 유지율, EDM 메일 개설율 등    
ex. MAU/DAU/WAU , Stickness, DN Retention 등   

(4) Revenue   
= 수익   
- 사용자가 프로덕트를 유료 구매하기로 결정하는 단계   
- 사용자 소비를 유도해 매출 향상에 집중   
- 고객 평생 가치를 지속적으로 높이고, 고객 유치 비용은 줄여야 함   
- 프로덕트 성격에 따라 측정 지표 다름: 유료화 상품 판매 중심 수익 구조 - 게임 아이템, 스킨 소비액 / 광고 시청료 위주의 수익 구조 - 광고비   
ex. PU, ARPU, ARPPU, LTV 등    

(5) Referral   
= 프로덕트에 만족한 사용자가 타인에게 추천하는 단계    
- 충성도 높은 사용자를 보이지 않는 브랜드 홍보자로 만드는 방법, 잠재 고객에게 제품 추천하는 방법 등 고민    
- 측정 지표: 순 추천고객 지수, 바이럴 지수 등    
    - 바이럴 계수: 기존 고객 한 명당 데리고 온 새로운 고객 수    
    - 바이럴 주기: 기존 고객 한 명 당 새로운 고객을 데려오는 데 걸린 시간
