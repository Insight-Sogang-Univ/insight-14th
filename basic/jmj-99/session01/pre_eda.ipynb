## 1. EDA 개념
- Exploratory Data Analysis, 탐색적 데이터 분석 과정  
- 주어진 데이터에 대해 알아보는 것  
(1) 시각화 -> 패턴 발견  
(2) 데이터의 특이성 확인  
(3) 통계 / 그래프로 가설 검정  

=> 정확한 분석을 위해 데이터 해석 & 문제 정의가 중요  
  
  **□ EDA 대상**
  - 하나의 변수 or 여러 변수 간 관계  
  (1) 일변량 (Univariate): 파악하려는 변수가 한 개  
  => 데이터 설명 & 패턴 파악  
  (2) 다변량 (Multi-variate): 파악하려는 변수가 여러 개  
  => 변수들 간의 관계 파악  
  
  **□ EDA 종류**  
  - 시각적 or 수치적 파악  
  (1) 시각화: 차트, 그림 등으로 데이터 확인  
  => 한 눈에 대략적인 형태 파악  
  (2) 비시각화: 주로 Summary Statistics로 데이터 확인  
  => 정확한 값  
    
  **□ EDA 유형**  
  - EDA 대상과 종류에 따른 구분  
  (1) 일변량 비시각화: 주어진 데이터의 분포 확인 ex. 빈도표, 기술 통계량    
  (2) 일변량 시각화: 주어진 데이터를 전체적으로 탐색 ex. 파이차트, 막대그래프, 히스토그램, 박스플롯  
  (3) 다변량 비시각화: 둘 이상의 변수 간 관계 ex. 교차표, 상관계수  
  (4) 다변량 시각화: 둘 이상의 변수 간 관계를 전체적으로 탐색 ex. 모자이크플롯, 박스플롯, 평행좌표, 산점도  

## 2. 데이터 확인  
  **□ 데이터 읽기 = 불러오기**   
  (1) 절대경로: root 파일부터 목적지까지 전체적인 경로    
  (2) 상대경로: 현재 작업 중인 디렉터리를 기준으로 상하위 디렉터리와 같이 표현하는 경로  
      - 현재 디렉터리: .\  
      - 상위 디렉터리: ..\  
      - 최상위 디렉터리: \   
    
  **□ 데이터 입출력**  
  - 데이터 파일의 형식에 따른 함수    

  | 파일 형식 | Reader | Writer |  
  | -------- | ------- | ------- |  
  | CSV | read_csv | to_csv |  
  | JSON | read_json | to_json |  
  | HTML | read_html | to_html |  
  | MS EXCEL | read_excel | to_excel |  
  | SQL | read_sql | to_sql |  
    
=> CSV와 EXCEL 파일 모두 pandas를 통해 DataFrame 형식으로 읽어올 수 있음  
    - CSV: 데이터를 쉼표로 구분하는 텍스트 파일. 크기가 작고 압축이 용이.  
    - EXCEL: 행과 열이 데이터프레임의 행과 열로 일대일 대응됨. 여러 sheet는 list로 받음.

- 변수 유형 확인: 질적, 범주형 or 양적, 수치형

## 3. 전처리   
- 데이터를 분석에 적합한 형태로 가공  
- 이상값, 결측값, 노이즈 수정 등 분석 목적에 맞게 변형 => 데이터 품질 결정 => 데이터 분석 정확도 결정  

  **□ 결측값 처리**   
  - <u>결측값</U> = 데이터 수집 과정에서 측정되지 않거나 누락된 데이터  
    (1) 결측값 확인: NaN, ?, 0 등의 값으로 나타난 결측값 확인  
      - info(): 각 열에 속하는 유효한 값의 개수   
      - value_counts(dropna=False): 각 열의 결측값을 포함한 전체 데이터 확인  
      - isnull(): 누락 데이터면 True, 유효 데이터면 False 반환   
      - notnull(): 유효 데이터면 True, 누락 데이터면 False 반환  
      - replace(): 0이나 ?로 입력된 결측값을 NaN으로 변환  
      
    (2) 결측값 처리

    a. 삭제: 결측값이 무작위로 발생한 경우, 데이터가 있는 행 / 열 삭제 => dropna()  
    - 삭제 기준  
      - axis = 0: 결측값이 있는 row 삭제  
      - axis = 1: column 삭제  
    - 삭제 조건  
      - how = 'any': 하나라도 결측값 있으면 삭제  
      - how = 'all': 모든 값이 결측값일 때만 삭제  
    - 특정 column 기준 삭제: subset['column명']  
      - axis = 0 일 때만 적용 가능  
    - 원본 데이터 변경 여부  
      - inplace = False: 반영 내용을 새로운 DataFrame으로 반환  
      - inplace = True: 원본 DataFrame을 변경하여 반영     

    b. 대체: 결측값을 다른 값으로 대체 => fillna()   
    - 평균값, 최빈값 등 활용  
      - 일괄 대체: 모든 변수를 같은 값으로 일괄 대체  
      - 유사 유형 대체: 유사한 범주에 따라 다른 값으로 대체  
  
  **□ 이상치 처리**   
  - <u>이상치</U> = 관측된 데이터의 범위에서 많이 벗어난 값   
    
    (1) 이상치 확인

    a. 통계 => describe()  

    b. 시각화 => Boxplot    
    
    c. Z-score: 데이터를 평균(0)과 표준편차(1)로 정규화해, 평균에서 얼마나 떨어졌는지 나타냄  
    - 평균에 가까울수록 0에 가깝고, 멀수록 Z-score가 커짐  

    d. Tukey Fences: 사분위 범위(IQR) 활용해 아래 경우에 이상치라고 판단    
    - Q1-(1.5*IQR) 미만  
    - Q3+(1.5*IQR) 초과  

    (2) 이상치 제거 및 처리  

    a. 전체 삭제: 이상값이 human error에 의해 발생한 경우. ex. 단순 오타, 데이터 처리 과정에서의 오류 등  

    b. 다른 값으로 대체: 절대적인 관측치의 숫자가 작아, 단순 제거 시 관측치의 절대량이 작아 신뢰성 문제가 발생하는 경우 => 다른 변수들로 예측 모델 생성 & 이상값 예측 후 대체  

    c. 변수화: 이상값이 자연 발생한 경우.  

    d. 리샘플링: 자연발생한 이상값을 처리할 때, 해당 값을 분리해 모델을 만드는 방법

  **□ 피처 엔지니어링 (변수 가공)**   
  - <u>피처</U> = column = 입력 변수   
  - 문제를 컴퓨터가 잘 이해하도록 피처들의 형태를 변형 or 처리하는 과정  

    (1) 레이블인코딩, 원핫인코딩  
    - 문자열 값을 숫자형으로 변경하는 작업  

    a. Label Encoding: 범주형 변수를 0부터 n-1까지 숫자로 변환  
    
    b. One-hot Encoding: 범주형 변수를 0과 1 이진 벡터로 변환 (표현하고자 하는 단어는 1, 아니면 0)  

    (2) 구간화  
    - 연속 데이터를 일정한 구간으로 나눠서 분석  
    - pd.cut(df['컬럼명'], bins=[나누는 기준 리스트], labels=[지정할 label])  

    (3) 파생변수 생성  
    - 기존의 피처를 다른 피처로 변환해 변수 추가 => 기존 데이터를 다른 관점에서 분석  
    - 단위 변환, 표현 형식 변환, 요약 통계량 변환, 정보 추출, 변수 결합, 조건문 이용 등  

    (4) 스케일링  
    - 서로 다른 변수의 값 범위를 일정한 수준으로 맞추는 작업  
    - 숫자의 상대적 크기 차이로 인한 결과 왜곡 방지를 위해 정규화 작업 수행  
    a. 표준화: 각 피처를 평균 0, 분산 1로 변경 => 정규분포를 따른다고 가정할 때   
    b. MinMaxScaler: 모든 피처를 0과 1 사이에 위치하도록

    

## 4. 데이터 시각화   
- 데이터 이해를 돕기 위해 그래픽 요소로 데이터를 묘사하고 표현하는 과정  
- 일반적으로 Matplotlib과 Seaborn 사용  
(1) Matplotlib: NumPy를 활용한 플로팅 라이브러리  
(2) Seaborn: Matplotlib을 기반으로 통계 데이터 시각화에 최적화된 라이브러리  
  
  **□ 파라미터**   
  - <u>파라미터</u> = 매개변수 = 프로그래밍된 함수의 임력값  
  - 파라미터 = 인자 형식으로 전달  

  **□ 다양한 그래프**   
  (1) boxplot: 사분위수, 이상치  

  (2) countplot, histplot  

  a. countplot: 범주형 변수의 빈도수  

  b. histplot: 도수분포표를 그래프로 나타낸 것. 수치형 변수의 구간별 빈도수.   

  (3) displot, kdeplot   
  - 히스토그램을 연속적으로 곡선으로 연결한 그래프   

  (4) barplot, pointplot  
  - barplot은 범주형 데이터 값 x에 따른 수치형 데이터 값 y의 평균 값 제공. pointplot은 모양만 다르고, 동일한 정보 제공  

  (5) scatterplot, regplot  
  
  a. scatterplot: 두 연속형 변수 간 관계 시각화 => 변수 간 패턴, 상관관계, 분포 등 쉽게 파악  

  b. regplot: 두 연속 변수 사이의 산점도와 회귀선  

  (6) catplot   
  - 수치형 데이터와 범주형 데이터의 관계   

  (7) pieplot  
  - 데이터의 부분과 전체 간 비율 => 모든 데이터를 합치면 전체를 이루는 경우에 효과적  

  (8) heatmap  
  - 변수 간 상관계수를 직관적으로 => corr() 활용  
  - <u>상관계수</u>: 두 수치형 변수 사이의 상관 관계 정도를 수치적으로 나타낸 계수  
  
  (9) violinplot: (1)과 (3) 합친 그래프  

  (10) pairplot  
  - 여러 그래프 한 번에
