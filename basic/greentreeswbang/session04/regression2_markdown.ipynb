# 1. 선형회귀

## 다중선형회귀란?

- 종속 변수들이 독립변수들이랑 싹 다 선형적으로 연관되어 있다 보고 더하여 생각한 것. 각 독립변수들에 가중치를 다르게 해서 그 모델을 정립
- 즉, 종속변수에 대한 **가중평균** 임!


## 다중선형회귀의 기본가정

- 항상 선형회귀를 쓸 수 있는 것이 아닌, 일정 가정을 충족하고 만족해야함

1. 선형성
- 선형회귀니까 선형적인 데이터에 있어서 사용해야 할 것임

2. 독립성
- 변수들이 독립적이지 않으면, 하나의 요인이 필요 이상의 가중치를 가지게 될 수 있음

3. 오차항의 평균은 0이다
- 실제 값과 예측 값의 차이인 오차항들의 모든 합은 0. 
- 백색 잡음 : 어찌 보면 혼돈 속 비혼돈이라 할 수 있음. 0이 아니면 offset이 발생

4. 등분산성
- 오차항의 분산 또한 일정해야 한다.
- 왜? 각 지점별로 오차가 큼 작음에 따라 예측값과 실제값 차이 천차만별이고, 이에 따라 예측 성능도 널뛸수 있다.

5. 오차항은 자기상관되어있지 않다
* 오차항의 공분산 (두 변수가 상관된 정도) 은 0. 즉, 다른 오차 값이 또다른 오차 값에 영향을 아예 안 주어야 함
* 자기상관이 있으면, 오차 값이 또 오차를 만들고 이 오차가 또 다른 오차를 계속 만들게 됨

6. 정규성 (선택)
* 오차항이 정규분포를 따른다는 가정 
* 회귀를 검증하는 방식이 정규성을 따른다는 가정을 기반으로 진행하기 때문

분석 전에 검증 가능한 것 : 선형성, 독립성, 오차항 자기상관 X

나머지 3개는 진행을 해봐야 알 수 있음!

## 회귀분석 평가방법

### 시각화
* 직접 보면서 데이터와 회귀 선이 잘 맞는지 보기 가능
* 잘 진행되었는지는 확인하기 좋겠으나 그 정밀도는 구분하기 어려움, 따라서 통계적 지표 필

### 통계지표

* 이전에 사용했던 **T검정, F검정**!

 **F검정**
* 회귀 모델 자체가 유의미 한가? 즉 이 값들을 가지고 결과를 예측하는게 의미를 가지는가? 그렇다면 다시 한 번, 적어도 하나의 변수는 종속값에 영향을 주긴 하는가?
* 값이 높을 수록, 모델링이 잘 된 것! 즉, 독립변수 set을 잘 구성했다고 보면 될 듯!

**T검정**
* 독립 변수 하나가 종속 변수와 유의미한 영향을 미치는가 확인
* 단순하다! pvalue가 일정수치보다 낮으면 영향이 있는 것!

### 성능 평가 : 결정 계수 R^2

**R^2 (결정 계수)**
* 데이터의 변동 중 모델이 설명한 비율 (설명되는 편차 / 총 편차) -> 모델이 데이터를 얼마나 잘 설명하는가!
* 즉, 1이면 모든 편차가 모델이 예측한 범주 내에 있고, 0이면 아예 편차를 설명하지를 못하는 것

**조정된 결정 계수**
* R^2는 독립 변수의 갯수와 관련이 있음. 즉, 필요 없는 변수이더라도, 추가할수록 커진다. (독립변수가 데이터를 설명하면 설명하지, 방해하지는 않을 것이기 때문)
* 이를 보정해서 필요 없는 독립변수를 제거한 결정 계수

**AIC, BIC**
* 낮을수록 좋은 지표들! ,BIC가  더 엄격하다.

# 2. 비선형 회귀

* 데이터 분포가 항상 선형적인 것은 아님, 따라서 선형적이지 않은 식은 어떻게 되야 할까!

## 1. 다항식 회귀모델   
* 독립 변수들을 일차식으로만 두지 않고, 상황에 따라 2차, 3차 식으로 모델링한다.
* 선형 모델에 비해 회귀선을 유연하게 그을 수 있다
* 그러나 당장 하나의 독립변수에만도 파라미터가 매우 많아지고 이에 따라 독립변수가 많아지면 찾기 힘드며, 오버피팅의 가능성 또한 존재한다.

## 2. 지수 회귀모델

* expoenetial 함수 안에 x를 합성시키는 방법임. (지수 함수 형태의 커브가 나타날것. 지수 함수는 1:1 대응이라 독립 변수들이 파라미터를 많이 가지게 되진 않음)
* 자연로그를 취하면 변수를 바꿔서 보았을 때 선형 회귀 식과 비슷해짐
* 즉, 파라미터가 엄청 많아지지 않으면서 비선형 회귀가 가능

## 3. 로그 회귀모델

* 선형 회귀식의 x 부분에 logX를 놓고 연산을 취함.
* 식의 형태가 유사하므로, 선형 회귀와 동일하게 진행하면 된다.
* 지수 회귀모델과의 차이점? 지수 모델은 차원이 익스포넨셜이라 전체에 로그를 걸어주어서 y가 로그스케일이 되었고, 이를 돌리는 과정이 필요
    그러나 로그 모델은 y 스케일에는 변화가 없으므로, 그래도 받아들이면 된다.

## 4. 스플라인 회귀

* 구간별로 회귀 방법이 다르게 요구될 때 사용
* 반대로 생각하면, 하나의 모델로 표현하는 것이 옳지 않다고 판단 후 구간화 하는 것이 맞을듯
* 구간 경계에서 연속을 유지하는 것이 중요

# 전체 회귀 분석 프로세스

1. 사전 검증
    * 데이터 자체에 대한 검증 및 모델을 써도 될 지에 대한 검증
    * 즉 모델 적용 이전의 모든 사전 준비

2. 모델 생성 및 fit
    * 적합하다고 생각되는 모델 설정 후 training set으로 training 진행   
    * 이후 f검정, t검정을 통해 유의성 검정 진행 (회귀 계수가 의미를 가지는가?)

3. 모델 성능 평가
    * 얼마나 모델이 잘 예측하는지 검정, test 셋 이용
    * R^2 가 대표적
    * 검증 가능한 가설들을 검토

4. 모델 성능 개선
    * 변수 제거 및 모델 변경 (비선형 등)
