앞서 서술한 숫자로 표현된 사실을 **통계**라고 부릅니다. 이처럼 통계는 우리가 다양한 비즈니스와 경제 상황을 이해하는데 도움을 주는 숫자로 표현된 사실을 나타냅니다.

다만 **통계학**이라는 분야는 이러한 숫자로 표현된 사실보다 훨씬 더 많은 것을 내포합니다. 폭넓은 의미에서 통계학은 자료를 수집, 분석, 표현, 해석하는 기술이자 과학입니다.

비즈니스와 경제 분야에서 자료를 수집, 분석, 표현, 해석하는 과정에서 얻어진 정보는 경영자와 의사결정자들이 환경을 보다 잘 이해하고, 더 많은 정보를 바탕으로 **더 나은 의사결정**을 내리는 데 도움을 줄 것입니다.

## 1-1. 우리나라 흡연자는 몇 명?
현재 우리나라 흡연율은 **19.6%**로 알려져 있습니다. 그런데 이 숫자 어떻게 나온걸까요?

가장 확실하게 파악하는 방법은 역시 우리나라 전 인구(**모집단**)을 대상으로 흡연 여부를 물어보는 것입니다.

다만 시간적, 비용적인 측면에서 이는 다소 비현실적인 방법입니다.

그렇기에 일반적으로 모집단에 대해 일부 샘플**(표본집단)**만 추출하여 분석하게 됩니다.
우리가 분석하기 위해 보고 있는 데이터도 대부분 표본의 일종인 셈이죠!

~~~~샘플을 구했을 때 우리의 관심사는 크게 두 가지입니다.

첫 번째는 내가 갖고 있는 샘플 자체에 대한 관심

두 번째는 이 샘플의 모집단에 대한 관심.

통계학은 이 지점에서 두 갈래 길로 나뉩니다. 전자는 **기술통계**, 후자는 **추론통계**의 영역에 해당한다고 보면 됩니다.

기술통계는 수치와 시각화를 통해 샘플을 요약하여 표현하는 방법론입니다.
추론통계는 추정과 검정을 통해 샘플로부터 모집단을 추론하는 방법론입니다.

지난 시간에 공부했던 EDA는 기술통계의 범주에 해당하고, 앞으로 배우게 될 회귀, 분류 등의 머신러닝은 추론통계의 연장선이라고 볼 수 있겠습니다.

## 1-2. 자료의 척도
자료란 설명과 해석을 위하여 수집되고 분석되며, 요약되는 사실과 수치를 말합니다.

이러한 자료는 **수치로 표현되는 양적 자료**와, 수치로 표현되더라도 **연산의 의미를 가지지 못하는 질적 자료**가 존재합니다.

양적 자료는 다시 구간척도와 비율척도, 질적 자료는 명목척도와 서열척도로 구분됩니다.

- **명목척도**
    
    : 레이블이나 이름으로 요소의 속성을 식별하려고 하는 척도
    
    - ex. 펀드 유형(국내주식: 1, 국제주식: 2, 채권: 3), 성별(남: 0, 여: 1)

- **서열척도**
    
    : 명목형의 분류 속성과 더불어, 항목 간의 순서 또는 순위를 나타내는 특성을 지니는 척도
    
    - ex. 만족도(`좋음` `보통` `나쁨`), 직급 체계(사원 < 대리 < 과장 < 차장 < 부장)

- **구간(등간)척도**
    
    : 값들 사이에 순서가 있고, 그 간격이 일정하여 덧셈이나 평균과 같은 연산이 가능한 척도
    
    - **단, 곱셈과 나눗셈은 의미를 지니지 못한다.**
    - ex. 온도(섭씨, 화씨), IQ, 시간

- **비율척도**
    
    : 절대영점과 임의의 측정단위로 구성하고, 동간성이 유지되는 척도
    
    - 이 척도에서는 `0` 값이 그 변수에 대해 아무것도 존재하지 않는 상태를 나타냄. 가격이 0이면, 비용이 전혀 안 들거나 공짜라는 사실을 나타냄.
    - ex. 거리, 높이, 무게, 시간

이러한 측정척도는 자료 속에 포함된 정보의 양을 결정하며, 가장 적합한 자료요약 방법과 통계분석 방법을 결정합니다.

- 잠깐? 척도랑 종류랑 어떻게 다른거에요?
    
    자료의 종류와 척도는 서로 관련된 개념이지만, 정보를 바라보는 관점이 다릅니다. 간단히 말해, 자료의 종류는 데이터의 본질적 성격을 구분하고, 척도는 그 데이터를 측정하는 방식의 규칙을 구분합니다.
    
    예컨대 온도(섭씨, 화씨)는 무한히 세분할 수 있기에 연속형 자료로 구분합니다. 다만 0℃는 절대적인 ‘온도의 없음’을 뜻하지 않기에 비율을 계산할 수 없고, 그렇기에 구간척도로 분류됩니다.
    
    무게(kg)는 어떤가요? 무게 역시 무한히 세분할 수 있으니 연속형 자료입니다. 그러나 0kg은 절대적인 영점이기에 비율척도로 구분됩니다.
    
    이렇듯 같은 연속형 자료더라도 데이터의 척도는 충분히 달라질 수 있습니다.
    
    결론적으로, 자료의 종류는 데이터의 대략적인 성격을 파악하는 **큰 그림**이고, 척도는 그 데이터를 다루기 위한 **더욱 구체적인 규칙과 기준**이라고 할 수 있습니다.

## 2️⃣ 기술통계

## 2-1. 기술통계란?

데이터를 있는 그대로 설명하고 이해하는 데 중점을 둔 통계 방법

마치 사진작가가 아름다운 풍경을 카메라에 담아내듯이, 우리가 수집한 데이터의 특성과 패턴을 있는 그대로 보여주는 역할을 합니다.

**기술통계가 답해주는 질문들**

---

- 우리 회사의 올해 평균 매출은 얼마인가?
- 제품 A의 불량률은 어느 정도인가?
- 고객들의 연령 분포는 어떻게 되는가?

## 2-2. 3가지 주요한 도구 집합

1️⃣ 중심화 경향

- 평균, 중앙값, 최빈값…

2️⃣ 자료의 산포

- 분산, 표준편차, 범위…

3️⃣ 자료의 정리 및 시각화

- 히스토그램, 박스플롯, 파이차트…

**중심화 경향**

데이터의 중심적인 경향을 나타내는 기술 통계량

- **평균 (mean)**
    - **데이터 값의 총합을 데이터의 개수로 나눈 값**. 데이터의 중심 경향성을 나타내는 가장 대표적인 지표
    - 총 데이터가 `2, 2, 4, 5, 27`이라면, 데이터의 총합은 40, 데이터의 개수는 5개이므로 평균값은 40 / 5 = **8**이 된다.
    - 총합을 사용하기에, 이상치(outlier)에 민감할 수 있다.

- **중앙값 (median)**
    - 데이터를 크기 순서대로 일렬로 정렬했을 때, **한가운데에 위치하는 값** *(if 데이터가 짝수개라면 가운데 있는 두 값의 평균)*
    - 총 데이터가 `2, 2, 4, 5, 27`이라면, 가운데에 있는 숫자가 4이므로 중앙값은 4이 된다.
    - 평균에 비해 이상치에 덜 민감하며, 데이터의 중심 경향성을 나타내는 **대안적인 지표**이다.

- **최빈값 (mode)**
    - 데이터에서 **가장 자주 나타나는 값**
    - 총 데이터가 `2, 2, 4, 5, 27`이라면, 가장 자주 나타나는 값은 2번 등장한 2이기에 최빈값은 2이다.
    - 주로 **이산형 데이터**(Discrete Data), 데이터의 분포를 이해하는데 도움을 준다.

### 자료의 산포

---

데이터가 퍼져있는 정도를 설명하는 기술 통계량

- **범위 (range)**
    - 데이터의 최대값과 최소값의 차이를 나타내는 지표로, 데이터의 전체 변동 범위를 나타낸다.

- **분산 (variance)**
    - 데이터 값들이 평균으로부터 얼마나 멀리 퍼져있는지를 나타내는 중요한 통계적 지표
    - 분산값이 크다는 것은 데이터가 많이 퍼져있음을 의미한다.

- **표준편차 (standard deviation)**
    - 분산에 제곱근을 취한 값
    제곱근을 왜 취하나요?

 - 분산의 단위는 관측값 단위의 제곱이기에 대상 데이터와 일치하지 않습니다. 그렇기에 제곱근을 씌워 단위를 일치시켜주는 것입니다.
    - 예컨대 기존 데이터 단위가 cm라면 분산은 cm² 단위가 됩니다.
    - cm² 단위였던 분산에 제곱근을 취하면 다시 cm 단위가 됩니다. 이 단위 일치성 덕분에 표준편차는 데이터를 해석하고 비교하는 데 직관적입니다.

- **사분위수 (quartile)**
    - 데이터의 표본을 네 개의 동일한 크기로 나눈 값
        - Q1 = 하위 25퍼센트에 위치한 값, `하위 사분위수`
        - Q2 = 하위 50퍼센트에 위치한 값, `중앙값`
        - Q3 = 하위 75퍼센트(상위 25%)에 위치한 값, `상위 사분위수`
    
- **사분위간 범위 (IQR)**
    - Q3에서 Q1 값을 제외한 범위, 즉 데이터의 중앙 50%가 퍼져 있는 범위. 이상치를 탐지할 때 유용하다. (Tukey Fence)
        - **아래쪽 이상치:** Q1 − 1.5×IQR 보다 작은 값
        - **위쪽 이상치**: Q3 + 1.5×IQR 보다 큰 값
    
- **변동계수 (coefficient of variation)**
    - 데이터의 절대적인 변동성이 아닌 상대적인 변동성을 나타내는 통계적 지표입니다.
    - 예를 들어 서울 지역 아파트 전세가(평균 5억, 표준편차 1억)와 원룸 전세가(평균 1억, 표준편차 0.5억)를 비교해 봅시다.
    - 표준편차만 보면 아파트 전세가가 원룸보다 더 많이 변동하는 것처럼 보입니다.
    - 하지만 변동계수를 사용하면 아파트 전세가의 변동계수는 20%(1/5)이고, 원룸은 50%(0.5/1)입니다.
    - 즉, 평균 대비 변동성은 원룸이 훨씬 더 크다는 것을 알 수 있습니다.

### cf. 첨도와 왜도

---

- **첨도: 확률분포의 꼬리가 두꺼운 정도를 나타내는 척도**
- **왜도: *비대칭도,* 실수 **값 확률 변수의 확률 분포 비대칭성을 나타내는 지표**
    - 왜도의 값은 양수나 음수가 될 수 있으며 정의되지 않을 수도 있다.
    - 왜도가 음수일 경우에는 확률밀도함수의 왼쪽 부분에 긴 꼬리를 가지며 중앙값을 포함한 자료가 오른쪽에 더 많이 분포해 있습니다.
    - 반대로 왜도가 양수일 때는 확률밀도함수의 오른쪽 부분에 긴 꼬리를 가지며 자료가 왼쪽에 더 많이 분포해 있다는 것을 나타냅니다.
    - 평균과 중앙값이 같으면 왜도는 0이 된다.

## 3️⃣ 확률과 분포

**3-1. 들어가며**

이런 상황을 한번 생각해봅시다.
여러분이 감귤 농장을 운영한다고 가정해 보세요. 농장에서 수확한 감귤 전체의 평균 당도를 알고 싶어서, 감귤 한 박스를 골라 측정했더니 평균이 21 brix가 나왔습니다.

그렇다면 과연 이 한 번의 측정 결과를 근거로, **“우리 농장의 전체 평균 당도는 21 brix다!”**라고 단정해도 괜찮을까요?

조금만 생각해 보아도 문제가 있어 보입니다. 혹시 선택한 박스가 우연히 유난히 달콤한 박스일 수도 있지 않을까요? 또, 다른 박스를 한 번 더 측정했더니 역시 평균이 21이더라도, 나머지 감귤들의 당도가 실제로는 훨씬 낮을 수도 있습니다.

추론통계학의 논리는 사실 끝없는 의심을 견디며 세워진 체계라고 할 수 있습니다.

예를 들어 다음과 같은 질문들이 차례로 던져집니다.

- **표본의 불확실성**: 혹시 운 없이 이상한 표본만 뽑힌 건 아닐까?
→ **표본분포**(Sampling Distribution)
- **모수 추정**: 그 표본의 통계량이 정말로 모수를 잘 반영한다고 할 수 있나?
→ **점추정**(Point Estimation)
- **추정의 현실성**: 점추정 값이 모수와 정확히 일치하는 건 비현실적이지 않나?
→ **구간추정**(Interval Estimation)
- **구간추정의 오류 관리**: 이 구간이 틀릴(모수를 포함하지 못할) 가능성은 어떻게 관리하지?
→ **신뢰수준**(Confidence Level)
- **적극적 주장 검증**: 특정 주장이 맞는지 틀리는지 직접 판단을 내릴 수는 없나?
→ **가설검정**(Hypothesis Testing)
- **가설검정의 오류 관리**: '이 주장은 틀렸다'는 결론을 내렸는데, 사실은 내 결론이 틀렸을 가능성은 어떻게 관리하지?
→ **유의수준**(Significance Level, 제1종 오류)

이처럼 추론통계는 “*끊임없는 의심과 도전*”을 견뎌내며, 점차적으로 더 정교한 논리 체계로 발전해 온 것이라고 볼 수 있습니다..

## 3-2. 확률분포

- **확률변수**는 random한 세계에서 어떤 특정 사건이 발생할 때 나타날 수 있는 모든 경우의 수를 특정 값으로 대응시키는 함수입니다. 예를 들어, 동전을 던지는 실험에서 "앞면"과 "뒷면"이라는 결과를 각각 1과 0으로 대응시킬 수 있습니다.
- **확률분포**(probability distribution)는 확률변수가 특정한 값을 가질 확률을 나타내는 함수를 의미합니다.

<aside>
💡 확률분포에는 다양한 종류가 있으며, 확률 변수의 속성에 따라 이산확률분포와 연속확률분포로 범주를 나눌 수 있습니다.

**이산확률분포**는 확률 변수가 이산적인(discrete) 값을 가지는 경우에 해당합니다. 
이산적이라는 것은 유한하거나 셀 수 있다는 뜻이며, 특정 값을 선택할 수 있습니다. 

**연속확률분포**는 확률 변수가 연속적인(continuous) 값을 가지는 경우에 해당합니다. 
연속적인 값은 무한한 범위를 가지며, 특정 값을 선택하는 것이 불가능합니다.

</aside>

- **이산확률분포의 종류**
    
    베르누이분포 (Bernoulli Distribution)
    
    이항분포 (Binomial Distribution)
    
    기하분포 Geometric Distribution)
    
    음이항분포 (Negative Binomial Distribution)
    
    초기하분포 (Hypergeometric Distribution)
    
    포아송분포 (Poisson Distribution)

- **연속확률분포의 종류**
    
    정규분포 (Normal Distribution)
    
    T분포 (T-Distribution)
    
    F분포 (F-Distribution)
    
    감마분포 (Gamma Distribution)
    
    지수분포 (Exponential Distribution)
    
    카이제곱분포 (Chi-squared Distribution)
    
    균등분포 (Uniform Distribution)

## 3-3. 대표적인 분포 모형

### **정규분포 (Normal Distribution)**

| 형태 | - 종(bell) 모양
- 평균을 기준으로 대칭
- 표준편차가 클수록 곡선이 평평해짐 (데이터가 넓게 퍼짐) |
| --- | --- |
| 매개변수 | 평균(μ), 표준편차(σ) |
| 특징 | - 가장 잘 알려진 (연속)확률분포
- 많은 자연 현상과 사회 현상이 정규분포를 따르는 것으로 알려져 있음 |

- 원래는 모집단의 분포가 정규분포를 따를 때 표본평균의 분포도 정규분포를 따른다고 말할 수 있습니다.
- 그런데 중심극한정리에 따르면, 표본의 크기가 커질수록 표본평균의 분포는 **모집단의 분포와 상관없이** 정규분포에 가까워집니다.
    - 즉, **모집단의 분포를 몰라도 표본의 크기가 충분히 크다면(n>=30)** **포본평균의 분포는 근사적으로 정규분포를 따른다**고 가정할 수 있습니다.

- **대수의 법칙**
    
    표본의 크기가 커질수록 표본 평균이 모평균에 근사하는 현상을 말하며, 취합하는 표본의 수가 많을수록 통계적 정확도는 올라가게 됩니다. 따라서 데이터가 많을수록 좋습니다.   

- **중심극한정리**
    
    동일한 확률분포를 가진 독립 확률 변수*(independent and identically distributed)* 
    
    n개의 평균의 분포는 n이 적당히 크다면 정규분포에 가까워진다는 정리입니다.

    <aside>
💡 **정규분포**는 많은 데이터가 따르는 가장 대표적인 확률분포입니다.

하지만 지금부터 소개해드릴 **T분포, 카이제곱분포, F분포**는 약간 성격이 다릅니다. 이 분포들은 정규분포를 따르는 **모집단**에서 추출한 **표본(sample) 통계량**들이 따르는 분포, 즉 '표본분포'입니다.

즉, 이 분포들은 모집단 자체가 갖는 분포라기보다는, 정규분포의 통계적 성질을 이용해 가설 검정이나 구간 추정 같은 통계적 추론을 하기 위해 만들어진 도구라고 할 수 있습니다. 🧐

따라서 이 분포들을 이론적으로 올바르게 사용하기 위해서는 한 가지 중요한 전제 조건이 있습니다. 바로 우리가 추출한 표본이 **'정규분포를 따르는 모집단'**으로부터 왔다는 가정입니다.

이걸 확인하기 위해 **정규성 검정**이라는 것이 있다는 것 까지만 알아주세요. 자세한 내용은 다음 시간에 배우게 되실 겁니다!

</aside>

T분포 (T-Distribution)

| 형태 | - 종(bell) 모양, 정규분포보다 꼬리가 두꺼움
- 평균을 기준으로 대칭
- 자유도가 커짐에 따라 표준정규분포 N(0,1)에 수렴 |
| --- | --- |
| 매개변수 | 자유도(df) |
| 특징 | - 표본으로 모평균(혹은 모평균 간의 차이)을 추정할 때 사용되는 분포
- 표본 크기가 작을 때 정규분포보다 더 정확한 결과를 제공함 |

자유도는 통계에서 "얼마나 많은 정보가 독립적으로 변할 수 있는가"를 나타내는 숫자입니다. 특정 상황에서는 일부 정보가 이미 결정되어 있어서, 다른 정보들이 그만큼 덜 자유롭게 변할 수 있다는 의미입니다. 여기서 "독립적인 정보의 수"라는 것은 이렇게 생각해 볼 수 있습니다.

### ex) 간단한 평균 계산

- 예를 들어, 세 개의 숫자 4, 6, 8이 있다고 해봅시다. 이 세 숫자의 평균은 6입니다.
- 만약 이 중 두 개의 숫자만 알고, 평균이 6이라는 것을 안다면, 나머지 하나의 숫자는 자동으로 결정됩니다. 예를 들어, 4와 6을 알면, 나머지 숫자는 8이어야 평균이 6이 됩니다.

이 상황에서 첫 번째 숫자(4)와 두 번째 숫자(6)는 자유롭게 정할 수 있지만, 세 번째 숫자는 이미 정해진 다른 숫자들과 평균 조건 때문에 자유롭게 변할 수 없습니다. 그래서 여기서 **자유도**는 3개의 숫자에서 1을 뺀 값, 즉 **2**가 됩니다. 이는 2개의 숫자가 자유롭게 변할 수 있다는 의미입니다.

→ T분포의 자유도는 표본 크기에 비례합니다. 자유도가 높을수록 (즉, 더 많은 독립적인 정보를 사용할수록) T분포는 정규분포에 가까워 지고, 통계적 분석의 정확도가 높아질 가능성이 있습니다. 그러나 너무 많은 자유도를 가지면 **오버피팅(Overfitting)**과 같은 문제가 발생할 수 있으므로, 적절한 자유도를 선택하는 것이 중요합니다.

카이제곱분포 (Chi-Square Distribution)

- 형태
    - 0에서 시작하여 오른쪽으로 긴 꼬리를 가진 비대칭 분포 
    - 자유도가 증가함에 따라 중심이 오른쪽으로 이동하고 모양이 더 대칭적으로 변화하며, 정규분포에 가까워짐

- 매개변수 : 자유도

- 특징
    - 독립적인 표준 정규 분포를 따르는 변수들의 제곱 합
    - 즉, 표본분산의 분포와 관련된 지표
    - 주로 독립성 검정, 적합도 검정 등에 사용됨


F분포 (F-Distribution)

- 형태
    - 0에서 시작하여 오른쪽으로 긴 꼬리를 가진 비대칭 분포 
    - 자유도가 증가함에 따라 중심이 오른쪽으로 이동하고 모양이 더 대칭적으로 변화하며, 정규분포에 가까워짐

- 매개변수 : 자유도(df1, df2)

- 특징
    - 두 독립적인 카이 제곱 분포의 비율
    - 두 집단 간의 분산 차이를 비교하는 데 사용됨

## 4️⃣ 추론 통계 - 추정

## 4-1. 추론통계란?

수집한 데이터를 기반으로 미래의, 혹은 조사하지 못한 전체 정보를 추론하고 결론을 도출하는 방법을 다루는 분야로, **통계적 가설검정**과 **신뢰구간** 등이 여기에 속합니다.

- 추정
모집단을 대표할 수 있는 일부를 표본으로 추출하여 분석한 통계량을 이용해서 모집단의 **모수**를 예측하는 과정
    - 점추정 : 추정하고자 하는 모수를 하나의 수치로 추정
    - 구간추정 : 추정하고자 하는 모수가 존재하리라 예상하는 구간을 정하여 추정
    - 모수?
        
        *Parameter.* 모집단의 특성에 관계되는 평균, 비율 또는 표준편차 등 관심의 대상이 되는 미지의 참값
        
        - 통계량?
            
            *Statistc.* 모집단으로부터 추출된 표본에 기초하여 계산되어지는 수치함수. 예컨대 표본에서 계산되는 평균, 중앙값, 사분위수, 분산, 표준편차 등등…
            
- 검정
표본을 추출하여 얻은 표본통계량으로 모집단 모수에 대한 새로운 예상·가설 등이 옳다고 판단할 수 있는지를 평가하는 과정

추정과 관련해서 알아야 할 개념은 추정량과 추정치이다. 추정량이란 표본정보에 의존하는 확률변수로서 모수를 추전하는 데 사용되는 표본통계량을 말하고, 추정치란 표본이 추출된 후 표본자료를 이용하여 계산한 통계량의 특정한 값입니다.

표본평균은 확률변수로서 표본에 포함되는 요소에 의존하기 때문에 표본마다 그의 값이 다르게 되어 모평균과 같지 않을 가능성이 매우 높다. 이러한 변동은 표본오차 때문에 발생하는데 모수를 추정할 때는 이를 고려해야 한다. 즉 점추정의 경우 오차에 대한 정보가 전혀 고려되지 않습니다.

따라서 어느 정도 **오차를 포함하는 신뢰구간 추정 방법이 일반적으로 널리 사용**됩니다..

## 5️⃣ 추론 통계 - 검정

## 5-1. 가설 검정의 절차

가설 검정은 모집단의 실제 값이 얼마가 된다는 주장과 관련해, 표본의 정보를 사용해서 가설의 합당성 여부를 판정하는 과정을 의미합니다.

가설 설정 - 유의 수준 설정 - 검정통계량 산출 - 가설 기각/채택 판단

**1️⃣ 가설 설정**

귀무가설 H0와 대립가설 H1를 설정합니다.

귀무가설은 주로 **기각하고자 하는 명제**로 설정합니다. 보통 “~는 차이가 없다”라는 명제입니다.

같은 말로, 대립가설은 **내가 주장하고자 하는 명제** 로 설정합니다. 보통 “~는 차이가 존재한다”라는 명제입니다.

**2️⃣ 유의 수준 설정**

<aside>
💡 **유의수준**이란 연구자가 결과를 판단할 때 허용할 수 있는 오류의 최대치로, 주로 0.05, 즉 5%로 설정합니다. 유의 수준은 연구자가 어떤 결정이 얼마나 신뢰할 만한지를 판단하는 기준이 됩니다.

유의 수준을 0.05로 설정한다는 것은, "귀무가설이 실제로 참인데도 불구하고, 5%의 확률로 귀무가설을 잘못 기각할 수 있다"는 것을 의미합니다. 다시 말해, 연구자가 5%로 발생할 수 있는 오류까지는 감수하면서, 귀무가설을 기각할 준비가 되어 있다는 뜻입니다.

예를 들어, 어떤 약물의 효과를 검증할 때, 유의 수준을 0.05로 설정하면, 이 약물이 실제로는 효과가 없더라도 5%의 확률로 약물이 효과가 있다고 잘못 결론 내릴 가능성을 수용하는 것입니다.

</aside>

**1종, 2종 오류**

- **1종 오류(𝛼)** : 귀무가설이 실제로 참이었는데 기각할 확률
- **2종 오류(𝛽)** : 귀무가설이 실제로 거짓인데 기각하지 못할 확률

가설검정은 1종 오류를 더 치명적인 것으로 파악합니다.

따라서 유의수준은 1종 오류의 확률을 의미하게 됩니다.

**3️⃣ 검정통계량 산출**

검정통계량은 가설 검정의 결과를 판단하는 데 사용되는 값입니다. 

데이터의 분포나 특징에 따라 T값, F값 등 어떤 통계량을 사용할지 정해야합니다.

- **Z값 (Z-Score)**
    1. **정의**:
        - Z값 또는 Z-점수는 표준 정규 분포(평균이 0이고 표준편차가 1인 분포)를 사용하여 얻은 값입니다.
        - 특정 데이터가 평균으로부터 얼마나 떨어져 있는지를 표준편차 단위로 나타냅니다.
    2. **계산**:
        - Z값은 다음 공식으로 계산됩니다:
            
            **$Z=(X−μ)/σ$**
            
            *X*는 관측값, *μ*는 평균, *σ*는 표준편차
            
    3. **사용**:
        - 큰 표본 크기(일반적으로 30 이상)에 대한 가설 검정에 주로 사용됩니다.
        - 예를 들어, 모집단의 평균에 대한 가설을 검정할 때 사용됩니다.
    4. **수치 해석:**
        - Z값의 절댓값이 크면?
        → 해당 데이터가 **평균으로부터 멀리 떨어져있으며**, 때로는 비정상적이거나 이상치일 가능성이 있음을 뜻합니다.
        - 임계점 찾기
        일반적으로 시행하는 양측검정(양쪽 꼬리에서 임계점을 찾을 때)에서는
            
            Z = 1.96일 때 유의수준 0.05(=신뢰수준 95%)
            
            Z = 2.58일 때 유의수준 0.01(=신뢰수준 99%)
            
            을 기준으로 판단합니다.
            
            → Z값이 임계점보다 크다면 귀무가설 기각
            
            (신뢰구간= 1-유의 수준(α))

- **T값 (T-Score)**
    1. **정의**:
        - T값은 T분포를 사용하여 얻은 값으로, 작은 표본 크기에 대한 가설 검정(T검정)에 주로 사용됩니다.
        - t분포는 정규분포와 유사하지만, 꼬리 부분이 더 두꺼워 작은 표본에서 발생할 수 있는 불확실성을 보다 잘 반영합니다.
    2. **계산**:
        - T값은 다음 공식으로 계산됩니다:
        $T = \frac{X - \mu}{s / \sqrt{n}}$
            
            *X*는 표본 평균, *μ*는 연구자가 설정하는 값, *s*는 표본 표준편차, *n*은 표본 크기
            
    3. **사용**:
        - 주로 **모집단의 평균**에 대한 가설을 검정하거나, **두 표본 평균 간의 차이**를 검정할 때 사용됩니다.
    4. **수치 해석:**
        - T값의 절댓값이 크면?
        → 해당 데이터가 **모집단 평균으로부터 멀리 떨어져있으며**, 관측된 표본 평균이 모집단 평균과 유의미한 차이가 있을 가능성이 높다는 것을 나타냅니다.
        - 임계점 찾기
        T분포는 자유도(df)에 따라 모양이 달라지므로 자주 쓰는 자유도와 유의수준에 따른 [t분포표](https://blog.naver.com/unirone/221082561573)라는 것이 사용되고 있습니다.
            
            
            → 같은 유의수준에 대해서 t값은 z값보다 커지며 자유도가 높아질수록 z값에 수렴합니다

- **F값 (F-Score)**
    1. **정의**:
        - F값은 F분포를 사용하여 얻은 값으로, 주로 두 표본 분산 사이의 차이를 비교하는 데 사용됩니다.
        - F분포는 두 개의 독립적인 카이제곱 분포된 변수들의 비율로 정의됩니다.
    2. **계산**:
        - F값=분산_1/분산_2
        분산_1과 분산_2는 두 독립적인 표본의 분산
    3. **사용**:
        - 주로 분산 분석(ANOVA)에서 두 개 이상의 그룹 간 평균의 차이를 검정하는 데 사용됩니다.
    4. **수치 해석:**
        - F값은 두 표본의 분산의 비이므로, 이 값이 1에 가까우면 두 표본의 분산이 서로 비슷하다는 것을 의미합니다. 즉 두 그룹이 퍼진 정도가 비슷하다고 볼 수 있습니다.

**4️⃣ 가설 기각/채택 판단**

**유의확률(p-value)**

- 가설 검정에 쓰일 수 있는 또 다른 척도입니다. 유의확률은 검정통계량에서 유도할 수 있으며 **유의 확률(p)가 유의 수준(α)보다 작을 시에 귀무가설을 기각**, 클 시에는 귀무 가설을 기각하지 못합니다.

어떤 학교에서 새로운 수학 교육 프로그램이 기존 프로그램보다 학생들의 성적을 더 높일 수 있는지 알아보려고 합니다. 연구자는 새로운 프로그램을 사용한 학생 그룹과 기존 프로그램을 사용한 학생 그룹의 수학 시험 점수를 비교하려고 합니다.

### 단계 1: 가설 설정

먼저, 귀무가설과 대립가설을 설정합니다.

- **귀무가설 (H0)**: 새로운 프로그램의 평균 점수와 기존 프로그램의 평균 점수에는 차이가 없다.
- **대립가설 (H1)**: 새로운 프로그램의 평균 점수가 기존 프로그램의 평균 점수보다 높다.

### 단계 2: 유의 수준 설정

가설 검정을 하기 전에 유의 수준 α를 정해야 합니다. 일반적으로 α=0.05로 설정합니다. 이 유의 수준은 "5%까지의 오류 가능성을 감수하고 귀무가설을 기각할 준비가 되어 있다"는 의미입니다.

### 단계 3: 데이터 수집 및 검정 통계량 계산

두 그룹의 학생들에게 시험을 치르게 한 후, 그 점수를 수집합니다.

- **새로운 프로그램 그룹**: 평균 점수 Xnew=78, 표준편차 snew=10, 표본 크기 nnew=30
- **기존 프로그램 그룹**: 평균 점수 Xold=74, 표준편차 sold=8, 표본 크기 nold=30

이제 t-검정을 수행합니다. t-검정 통계량을 계산하기 위해 다음 공식을 사용합니다:
이 수식을 이용해 t 값을 계산합니다. 계산 결과 t 값이 2.05로 나왔다고 가정합니다.

### 단계 4: p-값 계산 및 해석

p-값은 귀무가설이 참일 때, 현재와 같은 데이터가 나올 확률입니다. T분포표나 통계 소프트웨어를 사용해 이 T값(2.05)에 해당하는 p-값을 찾습니다.

예를 들어, p-값이 0.02로 나왔다고 가정합시다.

### 단계 5: 결론 도출

- **p-값과 유의 수준 비교**: p-값이 0.02이고, 유의 수준이 0.05로 설정되어 있으므로, p-값이 유의 수준보다 작습니다.
- **결론**: p-값이 유의 수준보다 작으므로, 귀무가설을 기각합니다. 이는 새로운 교육 프로그램이 기존 프로그램보다 학생들의 수학 성적을 더 높일 가능성이 있다고 결론지을 수 있다는 의미입니다.

## 5-2. 데이터의 자료형에 따른 검정 방법

**피어슨 상관 분석**

두 변수간 피어슨 상관계수가 얼마나 **유의미한지**를 검정하는 방법입니다. 귀무, 대립 가설은 아래와 같이 세워집니다.

가설

> H0: 두 변수간 선형 상관관계가 존재하지 않는다.
> Ha: 두 변수간 선형 상관관계가 존재한다.

- **공분산 (covariance)**
    
    두 변수가 **함께 어떤 방향으로 변화하는지와 그 크기**를 표현하는데 사용됩니다.
    
    - Cov (X,Y) > 0이면 X와 Y는 양의 관계, X가 커지면 Y도 커짐
    - Cov (X,Y) < 0이면 X와 Y는 음의 관계, X가 커지면 Y는 작아짐
    - Cov (X,Y) = 0이면 X와 Y는 양도 음도 아닌 관계
        - 선형적인 관계는 없으나, 비선형적인 관계가 존재할 수 있음
        ex) X가 Y의 제곱에 비례하는 경우

- **상관계수 (correlation)**
    
    두 변수 간의 선형적인 관계를 나타내는 통계적 지표입니다.
    
    상관계수는 **-1과 1 사이의 값**을 가지며, 상관관계와 인과관계는 다르다는 점을 주의해야 합니다.

    공분산을 각 변수의 표준편차로 나누어 -1과 1사이로 표준화 한 것이므로 다음과 같이 해석합니다.

        - Corr (X,Y)이 양수면 양의 상관관계 (1이면 완전한 양의 상관관계)
        - Corr (X,Y)이 음수면 음의 상관관계 (-1이면 완전한 음의 상관관계)

## 2️⃣ [범주형 - 수치형] F-test(ANOVA), T-test

**T-test : 범주가 2개일 때**

범주가 2개 (남/여, 민초/반민초 등등) ⇒ T-test 사용!

T-test 수행 시 t값은 두 평균 간 차이를 표준 오차로 나눈 값입니다.

가설

> H0: 두 그룹의 평균에는 차이가 없다.

> Ha: 두 그룹의 평균에는 차이가 있다.

- 근데 전 두 그룹의 평균에 차이가 없음을 주장하고 싶은데요!
    
    주장하고 싶은 것을 대립가설에 넣으라고 했는데, 차이가 없다는 것을 주장하고 싶으면 그걸 대립가설에 넣으면 되는 걸까요?
    
    답은 No입니다. t-test 검정법 자체가 두 변수간 차이를 기준으로 ‘얼마나 멀리 떨어져 있는지’를 체크하기 때문에, ‘차이가 없음’이라는 가설을 주장할 수 없습니다.
    
    간접적으로, H0에 차이가 없음, Ha에 차이가 있음을 넣고 ‘귀무가설을 기각하지 못하였다’는 식으로 주장할 수 있습니다.

코드

```python
mincho = df.loc[df['민초여부'] == 'Yes', '수명']
no_mincho = df.loc[df['민초여부'] == 'No', '수명' ]

spst.ttest_ind(mincho, no_mincho)
```

결과는 TtestResult(statistic = t-값, pvalue = p값, df = 자유도)

의 형태로 나옵니다.

t값은 보통 절대값 2를 기준으로, 그보다 크면 차이가 있으며 그보다 클수록 차이가 크다고 생각합니다.

p값은 역시, 이 결과가 얼마나 믿음직한지를 나타내며 보통 0.05를 기준으로 합니다.

- ttest_ind? ind가 뭐죠?
    
    independent의 약자입니다. 두 범주가 서로 독립적인 경우를 나타냅니다.
    
    두 범주가 독립적이지 않을 경우엔 related를 의미하는 ttest_rel이 있습니다.
    
    ind는 남녀, 민초반민초 와 같이 두 그룹이 독립적인 집단일 때를 나타냅니다.
    
    rel은 약을 먹기 전과 후, 정책을 시행하기 전과 후 처럼 같은 그룹이 다른 상태로 변화했을 때를 의미합니다.
    
    둘이 분리되어 있는 경우는, 각 경우 t-값을 계산하는 방법이 다르기 때문입니다. 정확히는 표준 오차의 값이 달라집니다.

- 저는 단측 검정을 진행하고 싶어요!
    
    양측 검정은 두 그룹 간 “차이”가 존재하는가,
    
    단측 검정은 두 그룹 간 차이의 ”한 쪽“만 보고 싶을 때 사용하는 방법입니다
    
    > H0: 반민초단의 수명이 민초단의 수명보다 작거나 같을 것이다.
    Ha: 반민초단의 수명이 민초단의 수명보다 길 것이다.(← 내가 증명하고 싶은 것!)
    > 
    
    를 증명하고 싶을 때에 해당합니다.
    
    t분포는 좌우대칭형이므로, 코드를 똑같이 하되 결과로 나온 p값을 2로 나눈 것이 단측 검정의 결과라고 생각하시면 됩니다.

**F-test(ANOVA) : 범주가 3개 이상일 때**

안타깝게도 t검정은 범주가 2개일 때 사용할 수 있습니다.

범주가 3개라면 2개씩 묶어서 3번, 4개라면 6번 진행하면 알 수 있지만, 범주가 매우 많아지면 어떨까요?

<aside>
💡 범주가 MBTI라고 해봅시다. MBTI별 수명의 평균이 다르다는 것을 증명하고 싶을 때, 16개의 mbti를 2개씩 묶어서 120번 진행하면…

</aside>

그래서 존재하는 것이 분산분석, Analysis Of Variance(**ANOVA**)입니다. ANOVA는 F-test의 일종이며, 지금은 가장 기본적인 버전인 one way ANOVA에 대해서만 살펴보겠습니다.

F-통계량은, 위에서 언급된 것처럼 분모 그룹과 분자 그룹의 분산 비입니다.

ANOVA 검정을 할 때엔, F값은 **집단 간 분산 / 집단 내 분산**으로 계산합니다.
즉 (전체 평균과 각 집단의 평균을 가지고 만든 분산) / (각 집단의 평균과 개별 값으로 만든 분산)으로 계산합니다.

가설

> H0: 각 MBTI 그룹 별 수명 평균은 차이가 없다.
> Ha: 각 MBTI 그룹 별 수명 평균은 차이가 있다.

코드

```python
ISTP = MBTI.loc[MBTI['mbti'] == 'ISTP', '수명']
ESFP = MBTI.loc[MBTI['mbti'] == 'ESFP', '수명']
INFP = MBTI.loc[MBTI['mbti'] == 'INFP', '수명']

spst.f_oneway(ISTP, ESFP, INFP)
```

결과는 F_onewayResult(statistic = f값, p-value = p값)

f값은 보통 2~3 이상이면 차이가 있으며 클수록 차이가 크고,

p값은 역시 마찬가지로 f값 결과가 얼마나 유의미한지를 의미합니다.

주의

- anova는 전체 평균 대비 각 그룹간 차이가 있는지만 알 수 있습니다. 즉 모두가 서로 다르다가 아니라, 하나가 튀고 나머지가 비슷하더라도 귀무가설을 기각하는 결과가 나올 수 있습니다.
(ex ISTP의 평균 수명만 혼자 300살인 경우)

위의 보라색 값들을 **기대빈도** 라고 합니다.

그런데 실제로 관측된 값이 아래와 같았다고 합시다.



위의 빨간색 값들을 **관측빈도** 라고 합니다.

같은 위치에 있는 **기대빈도**와 **관측빈도**가 서로 차이가 크다면, 둘이 독립적이지 않을 가능성이 높겠죠?

이제 카이제곱 통계량을 구할 수 있습니다!

식은 간단합니다. (**관측빈도** - **기대빈도**)^2 / **기대빈도** 를 쌍만 맞게, 순서 상관 없이 쭉 더해주면 됩니다. 

**5, 15    15, 5    20, 10   60, 70**

→ (**5**-**15**)^2 / **5** + (**15**-**5**)^2 / **15** + (**20**-**10**)^2 / **20** + (**60**-**70**)^2 / **60** = 약 33.3

카이제곱 분석은 범주 갯수에 제한이 없습니다.

범주가 많아질수록 카이제곱 통계량이 커질 수밖에 없겠죠?

따라서 카이제곱 통계량 해석은 어떠한 특정 숫자가 아닌, **자유도의 약 2배**가량을 기준으로 잡습니다. 
(2배 이상이면 차이가 존재)

범주형 변수의 자유도는 범주의 수 - 1이 됩니다.

카이제곱 변수에서의 자유도는 (x변수의 자유도) * (y변수의 자유도)로 계산합니다.

방금의 경영학과와 갈색머리 예시에서는 (2-1) * (2-1) = 1 이 됩니다.

만약 학과 10개, 머리색 빨주노초파남보 예시였다면 (10-1) * (7-1) = 54 가 될 것입니다.

즉 카이제곱 통계량이 약 108을 넘어야 의미있는 차이가 있다고 판단할 수 있습니다.
