**2-1) Index와 Column**

- 함수의 (x,y) 좌표처럼, 행과 열 또한 **데이터프레임 내에서 데이터의 위치를 나타내는 역할**을 맡습니다.

**2-2) Series**

**4-1) 인덱싱(Indexing) & 슬라이싱(Slicing)**
**인덱스와 컬럼을 사용해 데이터를 추출**하는 것을 **‘인덱싱 & 슬라이싱**’이라 합니다.

- **인덱싱(Indexing)**
    
    : 단일 항목을 **지목**할 때 사용
    
    → 예: `df[1]`
    
- **슬라이싱(Slicing)**
    
    : **범위**를 지정해 여러 항목을 추출할 때 사용
    
    - `df[start:end]`
        
        ```python
        # 범위 슬라이싱
        data[:1] # 0번째 데이터 반환
        data[1:3] # 1~2번째 데이터 반환
        data[2:] # 2번째~끝까지 데이터 반환
        ```

- **인덱스 이름(index label)**을 기준으로 행을 선택할 때는 **`loc`**을 이용하고,
- **정수형 위치(integer position**)를 사용할 때는 **`iloc`**을 이용합니다

**5-4) 데이터프레임 병합**

**concat**

- **단순히 두 데이터 프레임을 이어붙이는** 방식입니다.
- 데이터프레임을 **행 혹은 열 방향**으로 이어붙일 수 있습니다. 보통 **동일한 구조의 데이터프레임**을 합칠 때 사용합니다.

**merge**

- **공통 항목을 기준으로 병합**하는 방식입니다.
- 데이터프레임 간 공통 항목(열이나 인덱스)이 있어야 합칠 수 있습니다.
- 따라서 형태가 다르더라도, 겹치는 데이터를 기준으로 데이터프레임을 합칠 수 있
- on 인자를 통해 기준이 될 열을 지정해줄 수 있습니다.
- 집합 연산과 유사하게 작동하며, how 인자에 따라 어떤 방식으로 결합할지 결정할 수 있습니다.
- `df = pd.merge(df1, df2, how = 'inner', on = '~~')` ⇒ 교집합(and) 연산
- `df = pd.merge(df1, df2, how = 'outer', on = '~~')` ⇒ 합집합(or) 연산습니다.

### 6-1) 집계함수란?

- **여러 행으로부터 하나의 결괏값을 반환**하는 함수입니다.
- 각 컬럼의 평균, 표준편차, 합계 등.. 연산을 통해 대표값을 구할 때 주로 사용합니다.
- 결측치가 있을 경우 결측치는 무시하고 계산합니다.

**대표적인 집계함수**

<aside>

- `df['컬럼 명'].mean()` : 평균
- `df['컬럼 명'].median()` : 중앙값
- `df['컬럼 명'].std()` : 표준편차
- `df['컬럼 명'].sum()` : 합계
- `df['컬럼 명'].max()` : 최대값
- `df['컬럼 명'].min()` : 최소값
</aside>

<aside>
💡 집계함수는 기본적으로 모든 행을 대상으로 연산을 진행합니다. 그러나 모든 행보다 **일부 행을 대상으로 집계**할 때 더 유용한 인사이트를 얻는 경우도 있습니다.

가령 전체 데이터의 노트북 무게 평균을 구하면 그 이상의 정보를 얻을 수 없지만, 노트북 브랜드별로 그룹을 묶어 평균을 구하면 어떤 브랜드가 가벼운 노트북을 추구하는지 등을 알 수 있습니다.

</aside>

- **groupby** 메소드는 특정 기준으로 그룹을 묶고, 연산 처리를 할 수 있게 해줍니다.
- groupby 자체만으로는 아무 일도 일어나지 않고, 보통 **집계함수와 함께 사용하게 됩니다.**
    - **`df.groupby('column명1')[’column명2'].집계함수`**
     column1로 그룹화를 하고, 그룹별로 column2에 대해 집계함수를 적용해 결과를 시리즈 형태로 반환
    - **`df.groupby('column명1')[[’column명2']].집계함수`**
     그룹별로 column2에 대해 집계함수를 적용해 결과를 데이터프레임 형태로 반환
