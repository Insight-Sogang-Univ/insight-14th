# 1. 데이터 기반 의사결정

## 1.1 무엇을 위한 데이터 분석인가?

* 근본적인 질문 : 데이터의 장점은 무엇일까? -> 목적 달성 프로세스의 질!
    * 데이터에 근거해서 과거 및 미래를 분석해서 '지금 해야 할 일에 집중'을 할 수 있다.
    * 즉, 프로세스를 더 확실하고 효율적으로 진행할 수 있다!
    * 현장의 직감과 경험치를, 더 객관적으로 나타내고, 이에 옳고그름 및 집중할 액션 선택 가능!


### 액션이란 무엇인가?

* 결정 : '의사결정'을 내리기 위해 결정을 뒷받침 할 수 있는 데이터 분석을 수행
* 계획 : '계획' 수립을 위해서는, 수치적 근거에 기반해서 계획 수립을 위한 데이터 분석 수행
* 행동 : 행동을 위해서는 계획 실행에 활용할 수 있는 데이터 분석 진행

* 공통점 : 객관적으로 과거를 돌아보고 미래를 전망한다. 이를 기반으로, 지금 해야 할 일에 집중!

## 1.2 Data-Driven-Decision-Making

* 데이터 소스를 활용해서 분석을 진행해 직관보다 더 목표 및 목적에 부합하는 의사 결정을 내릴 수 있도록 하는 접근 방식

## 1.3 6 steps of DDDM

1. 목표 정의 : 조직의 목표를 명확하게 표현
2. 데이터 식별, 준비 및 수집 : 목표 설정과 데이터 요구 사항을 판별해서 데이터 소스를 평가 및 준비 후 체계적으로 데이터 수집 및 검증
3. 구성 및 탐색 : 데이터를 정리해서 인사이트 발견을 위해 구조화
4. 데이터 분석 수행 : 데이터를 인사이트로 변환에 전략에 영향을 미치는 패턴 및 상관관계, 추세 파악
5. 결론 도출 : 데이터 분석 결과를 검토하고 올바르게 비지니스 상황에 배치
6. 구현 및 평가 : 인사이트를 검증하고 결과 측정하는 과정. 경험적 증거와 피드백을 기반으로 전략 및 계획을 조정하기 위한 기초를 제공

## 1.4 데이터 분석 유형

1. 설명분석 : 과거 데이터를 설명 및 요약해서 과거 성과에 대한 인사이트 제공
2. 진단분석 : 특정 사건이 발생한 이유에 대한 파악에 중점
3. 예측분석 : 과거 데이터 기반으로 미래에 대한 정보를 예측하는 분석
4. 처방분석 : 데이터를 기반으로 앞으로 어떻게 조치를 취해야 할지 처방까지 제공하는 분석

* 이외에도, 탐색 분석, 추론 분석, 정성적 분석, 정량적 분석, 실시간 분석이 존재

# 2 목표 정의

## 2.1 지표

* '목표'를 달성하기 위해서는, 이에 맞는 목표 설정이 필요하다
* 그렇지만 글로만 표현하게 되면 이는 명확한 설정이 불가
* 이러한 것을 수치적으로 설명하기 시작했고, 이를 지표라고 부른다
* 지표 : 방향이나 목적, 기준 따위를 나타내는 표지. 대개 분모 대비 분자로 표현된다.

## 2.2 지표를 정의하기

* 절대적인 수치를 분모 혹은 분자에 넣을 수 있다면 좋겠지만(명확하니까), 그렇지 못한다면...?

* Proxy Metric : 활성화와 인게이지와 같이 추상적이어서 직접 측정하기엔 어려우니, 대체물을 정하여 측정한다. 설정에도 많은 논의가 필요하고, 정답이 존재하지 않음

* 스티비 예시 : 회원가입한 사용자 중 이메일 편집기에 진입한 사용자의 비율을 이용해서 쉬운 이메일 편집이라는 핵심 가치에 도달한 인원을 확인!


## 2.3 북극성 지표 L1/L2 Metrics

* 북극성 지표 : 기업이 가장 중요하게 생각하는 지표

* L1 지표 : 전략 관점의 전사적으로 높은 수준의 성과 지표
* L2 지표 : L1 지표의 세부적인 방향을 측정하는 지표, 전술 중심

## 2.4 AARRR (해적지표)

* 기업 및 집단별로 가치를 제공하는 방식이 다르면, 측정하는 지표에도 차이가 존재
* AARRR : 사용자 운영과정에서 자주 활요하는 데이터 분석 모델, 5가지 로드맵으로 구성됨

1. Acqisition : 사용자가 여러 방법을 통해 앱 및 웹사이트에 방문하는 것. 중요 포인트는 신규 가입자를 어떻게, 어디로 , 어떤 채널로 유도할 지 고민하는 것!
2. Activation : 사용자가 앱 또는 웹사이트에서 주요 기능을 사용하는 것. 사용자로 하여금 기능을 어떻게 알리고 사용해야 할 지 고민하며 활성화 정도를 올려야 한다.
    * sign-up, setup, a-ha, habit 의 4단계로 구성
3. Retention : 신규 사용자를 장기 사용자로 전환시키는 것. 기존 사용자를 만들어 두면 신규 사용자 획득에 비해 적은 비용으로 유지가 가능. 지속적으로 사용하도록 유도해야 한다.
4. Revenue : 사용자가 돈을 내고 재화를 구매할 것임. 이를 위해 구매를 많이 하도록 유도해야 한다. 매출과 관련된 정보가 주요 지표가 된다.
5. Referral : 사용자가 지인들에게 재화를 추천하는 단계. 지속적 성장을 위해 중요한 단계로, 기존 고객이 스스로 신규 고객을 가져온다. 충성도 높은 고객, 그리고 잠재 고객들에게 용이한 추천 방식을 만드는 방법을 고안해야 함
