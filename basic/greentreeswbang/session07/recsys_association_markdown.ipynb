**추천 시스템 기초**

# 1. 추천 시스템이란?

* 추천 시스템 : 사용자의 정보 데이터를 분석해서 개개인의 취향에 맞는 아이템을 추천하는 알고리즘
    * 고객 : 원하는 것 쉽게 접근 가능
    * 기업 : 많이 팔 수 있음


## 1.1 사용되는 데이터 종류 특징

1. 명시적 데이터 
* 직접적으로 원하는 것이 보이는 데이터
* 확실하지만 모든 사람들의 정보를 알기는 어려움
* ex. 평점, 리뷰 등

2. 암시적 데이터
* 행동 기반으로 간접적으로 추측한 데이터
* 쉽게 접근이 가능하지만, 정확성 및 직접성이 부족함
* 조회수, 시청시간 등

## 1.2 추천시스템의 분류
1. 내용 기반 필터링
* 콘텐츠 유사도 기반 추천, 사용하던 제품과 유사한 아이템! 
* 한 명을 스코프
2. 협업 필터링
* 서비스 안의 다수 사용자로부터 얻은 정보를 기반으로 추천
    * 사용자 기반 : 비슷한 사용자를 중심으로 추천
    * 아이템 기반 : 유사한 아이템을 추천
3. 하이브리드 필터링
* 추천시스템의 알고리즘을 여러가지 사용한다
* 내용 및 협업 각각의 장단점 융함

# 2. 내용 기반 필터링
* 사용자 & 아이템의 일치도를 이용해서 사용자에게 선호할 만한 아이템을 추천해주는 알고리즘
* 유사도가 핵심!
* 사용자 프로필 : 한 사용자에 대한 정보를 모아놓은 것. 이전 행동을 기반으로 선호를 예측한다

## 2.1 유사도

* 수학적으로 추천 시스템을 이해하는 데에 필요
1.  다양한 형태로 존재하는 데이터들을 벡터화한 후 벡터 사이에 유사도를 측정하는 방법으로 진행
2. 유사도를 중심으로 고객이 선호하는 아이템과 비교해서 유사도가 더 높은 아이템을 선별한다.

how?
*  코사인 유사도 : 두 점 사이의 각도를 측정
*  유클리드 거리 : 두 점 사이의 물리적인 거리를 측정


## 2.2 사용자 프로필
* 추천 시스템에 사용되는 데이터
1. 직접 지정형 : 사용자에게 원하는 걸 직접 말하라고 하는 방법
2. 간접 지정형 : 사용자의 행적을 기반해서 프로필 생성

how?
1. 특성 가중치는 사용자가 가진 아이템 특성 가중치 평균 이용
2. 사용자의 직접 혹은 간접적인 행동으로 프로필 획득
3. 유저들의 기존 평가를 바탕으로 분류 문제 해결
-> 좋은 프로필을 구성해야 다양한 방법으로 추천을 진행하는 것이 가능하다!

## 2.3 아이템 특징
* 이번에는 아이템을 벡터로 사용하는데, 도메인에 따라 어떻게 적용시키는 지에 방법이 다른
* 알고리즘 3가지가 존재! : 클러스터링, 머신러닝, 그리고.,.

**TF-IDF**
* 특정 단어가 등장하는 빈도를 특정 문서 안에서와 전체 문제 안에서 확인해서 다른 문서에 비해 얼마나 특정 문서에서 자주 나타는 문자를 찾아 단어의 가중치 계산
-> 각 단어가 문서에서 얼마나 중요한지 알 수 있다!

## 2.4 장단점

* 장점 : 아이템에 대한 콜드 스타트 문제 해결 가능, 추천 이유 명백하게 나타냄
    * 정보가 충분하지 못해 적절한 제품을 추천해주지 못하는 문제를 해결할 수 있다는 것!
* 단점 : 특징이 정확하지 않으면 정확도 급락, 새로운 장르로의 개척이 어려움

# 3. 협업 필터링

* 기존에의 사용자들을 기반으로 새로운 사용자와 아이템의 관계 예측하는 것
* 여러 사용자의 평점 기반으로 이루어짐(아이템에 대한)

## 3.1 메모리 기반 협업 필터링
* 사용자 혹은 아이템 간 유사도를 기반
* 비슷한 취향을 가진 사용자가 선호하는 아이템 추천
* 두가지 접근법!
1. 사용자 기반 : 너와 비슷한 취향을 가진 사람들은 요런 것을 샀어.
    * 아이템 기반에 비해 추천 범위 다양, 뜻밖 아이템 추천 가능 
2. 아이템 기반 : 이 물품을 산 사람들은 주로 다른 이런 것들도 샀어.
    * 빠른 계산으로 많은 내용을 처리 가능!

* 한계 
    1. 수가 많아질수록 확장성 떨어짐
    2. 데이터가 적으면 성능이 떨어짐
    3. 유저가 많아지면 계산이 많아짐 -> KNNCF!



## 3.2 모델 기반 협업 필터링

* 유저 - 아이템 평점 매트릭스 내부의 잠재적 요인을 통해 추천 예측을 하는 방법!

motivation
* 메모리 기반은 유사도에만 의존하기 때문에, 그 특성 및 이유를 찾는 데에는 취약하다
* 메모리 기반의 정보들만으로는 진짜 취향을 찾는 데에는 어려움이 많다
* 따라서, 잠재 요인을 수치적으로 학습해보고 예측을 진행해서 추천하자!

**잠재 요인**
* 실제 데이터는 아니지만, 관측 데이터 바탕으로 수치적으로 추정이 가능한 값
* latent factor model : 숨겨진 특성을 찾기 위해 행렬을 사용자 - 잠재요인, 아이템 - 잠재요인으로 분해해서 학습하는 모델!
    * 일반 특징 모델은 축을 알 수 있고 이를 기반으로 유사도를 파악할 수 잇음
    * 그러나 잠재 요인 모델은 축의 내용 및 차원의 개수를 알 수 없지만, 사이 각도 및 거리를 파악해 잠재적으로 어떤 점이 있는지 없는지 볼 수 있다.

**행렬 분해**
* 유저 - 아이템 행력을 분해하여 잠재 요인들을 설명할 수 있다면, 평점 예측이 가능하다
* 아이템에 대한 선호도는 유저 매트릭스와 아이템 매트릭스의 내적으로 분해가 가능하다
* SVD, SGD, NMF, PMF 등으로 가능하다

**연관분석**

# 1. 연관 분석이란?

* 구매 과정에서 항목 간의 일정한 연관 규칙을 발견하는 분석
* 따라서 상품 거래에만 사용되는 것이 아닌, 타 분야에도 폭넓게 사용이 가능하다!
* 인사이트 도출에 주로 쓰이고, 비지도 학습!

## 1.2 연관규칙 분석

* A를 샀다면 B 또한 구매할 것이다 형태로 표시
* 조건절 : 만일 ~라면
* 결과절 : ~ 일 것이다!
* 아이템 집합 : 조건 및 결과절을 구성하는 아이템들의 집합
* 아이템 집합은 구성은 상관없으나 상호 배반적이여야 함 (관련없)

* 주의 : 상관관계를 보는 것일 뿐, 이를 인과관계로 생각하면 안된다!
* 그냥 함꼐 행동하는 비율이 많았다는 것이지, 어떤 한 행동에 의해 나머지 행동을 한 것은 아님!

* 연관 분석 Vs 협업 필터링
    * 협업 필터링은 상품마다 다른 상품과 1:1로 계산하고 점수를 이용해서 구체적인 수치를 계산하고, 연관분석은 다양한 세트상품을 뽑아낼 수 있다
    - 이에 따라 협업 필터링은 개인화 상품 추천, 연관 분석은 세트 구성 및 노출에 자주 사용

# 2. 주요 개념

## 2.1 연관규칙의 조건

1. 지지도 : 함꼐 구매한 경우가 일정 수준 이상, 전체 거래 중 동시에 포함하는 거래의 비율!
    * 현실에 의미가 있는 규칙인가? 판단
2. 신뢰도 : 하나의 품목에 대한 거래에서 다른 품목의 거래가 수반되는 비율
    * 낮으면 의미 없는 연관 규칙

## 2.2 연관 분석의 평가 측도

* 향상도 : 우연히 등장할 확률과 실제 확률을 비교하자
* 독립일 때는 교집합이 두 확률의 곱이므로, 1의 값이 나온다, 반대로 1이 아니면 종속적 관계가 있다는 것!
* 그래서 실제 동시 발생 확률이 독립일 때에 비해...
    * 크면? 향상도 1보다 크고, 우연보다 둘을 같이 살 확률 큼
    * 작으면? 향상도 1보다 작고, 우연보다 둘을 같이 살 확률 적음

* 지지도, 신뢰도, 향상도가 모두 커야 효과적 규칙이라고 결론 내리기 가능

# 3. 활용 사례

* 넷플릭스의 하우스 오브 카드는 어떻게 잘 되었나?
1. 지지도 : 영국판 하오카를 본 사람 대부분이 케빈 스페이시가 출연한 콘텐츠 시청
2. 신뢰도 : 핀터의 콘텐츠를 본 사람들은 끝까지 시청하는 경향이 높음
3. 향상도 : 영국판 하오카를 본 집단과 케빈의 작품을 본 집단의 향상도가 높으므로, 둘 사이에는 우연보다 높은 관계가 있다

-> 이를 기반으로 오리지널 하오카를 제작!

# 4. 연관 분석 알고리즘

## 4.1 Aprori 알고리즘
* 선험적 경험 이용
* motivation : 가능한 모든 조합을 진행하면 너무 경우가 많아지므로, 이를 해결하고 싶음
* 개념 : 스캔을 진행하다가 특정 조합이 잘 발생하지 않으면 이후 이를 통한 조합들을 싹 다 제외하는 방법
* 장점 : 하나의 조합만 검사해도 이후 조합들을 굳이 해봐야할지 안해도될지 판단가능, 이에 따른 연산량 감소

* how? : 최소 지지도 넘는 항목인 빈발 항목 집합 찾고, 단계적으로 조합 생성 및 필터링
1. 단일 항목 집합을 생성 : 모든 개별 상품에 대해 지지도 계산 후 임계 넘는 애들만 선정
2. 2개 항목 집합 생성 : 1단계 가지고 조합 진행하고, 임계 안 넘는 애들 제고
3. 더 큰 항목 집합 생성 : 3개 넘는 항목 구성을 바탕으로 반복
4. 최종 빈발 항목 집합 선정 : 남은 항목 기반 연관 규칙 생성 후 지지도, 신뢰도, 향상도 계산

* 장점 : 연관 구매 패턴을 많이 발견할 수 있으며, 원리가 쉽고 이해분석 용이, 다른 가설까지 탐지 가능
* 단점 : 현실적 중요 연관 규칙 부족 및 여전히 느림

## 4.2 FP-Growth 알고리즘
* 여전이 Aprori 방법이 느려서 , 더 빠르게!
* FP-Growth 알고리즘은 효율적으로 빈발 품목 집합을 FP-Tree 구조 이용해서 Aprori를 구형하고 찾음
* 모든 거래를 확인 후 아이템별 지지도 계싼하고, 최소 지지도 이상의 템만 선택한다.

* how?
1. 최소 지지도 이상 항복만 남기는 빈발 항목 필터링
2. 빈도순으로 데이터 재정렬
3. FP - TRee 생성
4. 이를 기반으로 특정 항목 기준 부분 트리를 생성에 빈발 항목 집합 추출
5. 연관 규칙 도출

* 장점 : 데이터를 두번만 슥 보니 매우 빠르고, 후보 아이템 말고 Tree만 구성하면 됨
* 단점 : 대용량에서는 효율적인 메모리 사용을 못하고, 설계의 어려움 및 FP-Tree가 만들어져야 지지도 계산이 가능함
