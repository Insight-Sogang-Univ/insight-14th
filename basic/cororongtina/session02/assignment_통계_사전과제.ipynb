# 통계

## 1. 자료의 척도
1) 양적 자료
- 구간 척도: 순서 존재, 연산 가능
- 비율 척도: 임의의 측정 단위, 동간성 유지
2) 질적 자료
- 명목 척도: 레이블, 이름
- 서열 척도: 순서, 순위

## 2. 기술 통계
1) 중심화 경향
- 평균: 이상치에 민감
- 중앙값: 이상치에 덜 민감
- 최빈값: 이산형 데이터

2) 자료의 산포
- 분산
- 표준편차
- 범위
- 사분위수: 25%, 50%, 75% 단위로 나눈 값
- 변동계수: 데이터의 상대적 변동성

3) 자료의 정리 및 시각화
- 히스토그램
- 박스플롯
- 파이차트

## 3. 확률분포
- 확률변수: 특정 사건 발생 시 가능한 모든 경우의 수 -> 특정 값으로 대응
- 확률 분포: 확률변수가 특정값 가질 확률 나타내는 함수

1) 이산확률분포: 확률 변수 = 이산적
2) 연속확률분포: 확률 변수 = 연속적

### 대표적인 분포
1) 정규분포
- 평균 기준 대칭
- 중심극한정리: 표본평균 분포 -> 정규분포에 근사

2) T분포
- 표본으로 모평균 추정 시 사용
- 평균 기준 대칭
- 자유도 증가 -> 표준정규분포에 수렴

3) 카이제곱분포
- 독립적 표준정규분포의 제곱합
- 자유도 증가 -> 정규분포에 근사

4) F분포
- 두개의 독립적 카이제곱분포의 비율
- 두 집단 간 분산 차이 비교

## 4. 추론통계
- 수집 데이터 기반 전체 정보 추론
- 통계적 가설검정, 신뢰구간

### 추정
- 표본 추출 -> 모수 예측

1) 점추정: 하나의 값으로 추정
2) 구간추정: 일정한 범위로 추정

### 검정
- 표본 추출 -> 모수에 대한 새로운 가설/예상 평가

#### 가설 설정
- 귀무가설 H0: 기각하고자 하는 명제
- 대립가설 H1: 주장하고자 하는 명제

#### 오류
- 1종 오류: 귀무가설 = 참 but 기각
- 2종 오류: 귀무가설 = 거짓 but 기각 x
=> 검정통계량: Z값, T값, F값

#### 유의수준, 유의확률
- 유의수준: 허용 가능한 오류 최대치
- 유의확률: 유의확률 < 유의수준 -> 귀무가설 기각

#### 검정 방법
1) 수치형-수치형: 피어슨 상관 분석 -> 상관관계 파악
2) 범주형-수치형: F-test, T-test
3) 범주형-범주형: Chi2-test
4) 수치형-범주형: 유의성 검정 -> p값으로 판단
