# 회귀 기초

## 1. 머신러닝

> ## 개념
> - 인공지능의 한 분야
> - 모델 학습 -> 새 데이터 예측/결정
> - 데이터에서 패턴 찾음 -> 새로운 데이터 예측

> ## 종류
> * 지도 학습
>   * 정답 O
>   * 정답과 같아지도록 기계 학습
>   * ex) 회귀, 분류
> * 비지도 학습
>   * 정답 X
>   * 데이터 속 패턴 찾아 유사도 학습
>   * ex) 군집화, 밀도 추정, 차원 축소
>   > 군집화란?
>   > - 데이터의 비슷한 패턴 찾음 -> 그룹으로 나눔
>   > - 클러스터 수 지정 가능
>   > 
> * 강화 학습
>   * 시행 착오 -> 정답 찾음
>   * ex) 알파고

# 2. 지도 학습

> ## 개념
> - 입력값 & 결과값(정답) -> 학습
> - 에측값 <-> 정답 같아지도록 학습

> ## 모델링 과정
> #### 1. 전체 데이터(labled data) 분리
>   - 훈련 데이터(training set): 모델 학습에 사용
>   - 테스트 데이터(test set): 모델 평가에 사용
>   
> #### 2. 모델 학습
>   - 독립변수 통해 추정한 예측값 -> 종속변수에 가까워지도록
>   - 경사하강법 이용
> #### 3. 모델 평가
>   - 테스트 데이터로 반복 평가
> #### 4. 최종 모델 생성
> 

> ## 회귀, 분류
> #### 회귀 (Regression)
> - 연속형 변수 예측
> - 주어진 데이터(X) -> 정답(Y) 맞히는 함수 찾음
> - 종류: 선형 회귀, 비선형 회귀, 릿지회귀, 라쏘회귀, 다항회귀 등
> 
>
>  #### 분류 (Classification)
> - 범주형 변수 에측
> - 데이터를 주어진 항목들로 나눔
> - 기존 데이터가 어떤 레이블에 속하는지 패턴 학습 -> 새 데이터 레이블 판별


# 3. 선형 회귀
> ## 개념
> - 데이터를 가장 잘 설명하는 회귀선 찾음
> - 회귀선 = 현재 데이터 요약 & 미래 값 예측
> - 변수
>   - 독립 변수(x): 예측 위해 사용
>   - 종속 변수(y): 예측하고자 하는 것
> - 사용 조건: 선형성, 독립성, 등분산성, 정규성

> ## 종류
> ### 1. 단순 선형 회귀 분석
>   > #### 개념
>    > - 독립변수 1개 -> 직선으로 표현
>    > - y = ax + b에서 a, b 찾음
>    >   - a: 회귀계수
>    >   - b: y절편
>    >
>
>   > #### 최소제곱법
>   > - 최적 회귀선 찾는 방식
>   > - 잔차(실제값 - 예측값 차이)의 제곱의 합 최소값 지점
>   > 
>
>   > ### 경사하강법
>   > #### 개념
>   > - 함수가 닫힌 상태, 미분 계수 구하기 어려움, Gradient Descent 구현이 더 쉬움, 데이터 양 너무 많음 -> 최소제곱법 대신 사용
>   > - 잔차 제곱 최소값 지점
>   > #### 원리
>   > - -(미분값) * 학습률을 현재 값에 더함
>   >   - 현재 값 < 최솟값: 기울기 음수 -> 현재 값에서 오른쪽으로 이동
>   >   - 현재 값 > 최솟값: 기울기 양수 -> 현재 값에서 왼쪽으로 이동
>   >   - 현재 값 = 최솟값: 기울기 0 -> 변화 멈춤
>   > 
>   >   > #### 학습률이란?
>   >   > - '얼만큼' 이동하는지 결정 = 보폭
>   >   > - 작은 값부터 시도하며 적당한 값 찾음
>   >   > 
>   > #### 문제점
>   > - 작은 기울기: 조금씩, 큰 기울기: 많이 이동 -> local minima에서 탈출 어려움 (why? 현재 기울기의 절댓값에 따라 움직이기 때문)
>   > - 이전 기울기의 정보가 없음
>   > 
>   > #### 해결법 - 모멘텀
>   > - 이전 기울기, 이동해오던 방향 고려 -> 관성 부여
>   > - local minima 빠르게 탈출     
>   >      
>   > 
>
> ### 2. 다중 선형 회귀 분석
>   > #### 개념
>   > - 독립 변수 2개 이상 -> 평면 회귀선 찾기
>   > - 최소제곱법, 경사하강법 사용
>   > 
>   > #### 다중공선성
>   > - 독립변수 간 상관관계 큰 경우 발생
>   > - 변수끼리 겹침 -> 독립변수 - 종속변수 영향 구분 어려움
>   > - 확인 방법
>   >   - 상관계수: 수치로 나타낸 상관관계 정도, -1~1 사이    
>   >   
>   >   - VIF지수 (분산 팽창 인수): R square 사용, 높을수록 다중공선성 존재
>   > - 대처 방법
>   >   - 변수 제거: 독립변수로 사용할 변수 선택
>   >   - 변수 변환: 변수 더하거나 뺌 -> 새 변수 생성
>   >   - 규제 선형 모델 활용: 릿지, 라쏘, 엘라스틱넷 등 사용
>   >   - PCA(주성분분석): 상관 있는 변수들 묶어서 대표 축으로 바꿈 -> 축 수 줄어듦     
>   >   

> ## 규제 선형 모델
>   > ### 개념
>   > - 과적합 방지 규제 = 기울기 크기 조정
>   > - 잔차 제곱합 최소화 -> 회귀 계수 쉽게 커짐 (특정 변수에 과하게 의존) -> 과적합 발생
>   > ### 종류
>   > - 릿지 회귀
>   >   - L2규제
>   >   - 회귀계수 제곱에 페널티    
>   >   - 계수 크기 조절 기능     
>   >   
>   > - 라쏘 회귀
>   >   - L1규제
>   >   - 회귀계수 절댓값에 페널티
>   >   - 변수 선택 기능      
>   >   
>   > - 엘라스틱넷 회귀: L2 규제 + L1 규제
>   >   - L2 규제 + L1 규제
>   >   - 불필요한 변수 제거, 남은 변수 적절 크기로 유지     
>   >    

> ## 모델 평가 방법
>   > ### 성능 평가 지표
>   > - 평균 제곱 오차(MSE)
>   >   - 예측값 - 실측값 오차 제곱의 평균     
>   >   
>   > - 평균 절대 오차(MAE)
>   >   - 예측값 - 실측값 절댓값 오차 평균     
>   >   
>   > ### 변수 유의성 평가
>   > - t검정
>   >   - 독립 변수의 회귀계수가 유의미한지 검정
>   >   - 가설
>   >       - 귀무가설: 회귀계수 = 0
>   >       - 대립가설: 회귀계수 =/= 0
>   >   - 검정    
>   >   
>   >   - 판단: p-value가 -.-5 이하면 기각
