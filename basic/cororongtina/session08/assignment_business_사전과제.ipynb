# **데이터로 보는 비즈니스 상황**

## 1. 데이터 기반 의사결정 
<span style="color:black; background-color:#FFF099; padding:2px 4px; border-radius:4px">
<strong> 1-1. 무엇을 위한 데이터 분석인가 <strong></span>


- 과거 경험으로부터 얻은 지식 + 객관적 미래 전망 => 지금 해야 할 액션에 집중
- 목적 달성 프로세스의 질을 높임

### 액션이란?

- 결정: 의사결정을 내리기 위해 결정을 뒷받침할 데이터 분석 수행
- 계획: 프로모션 계획 등을 위해 수치적 근거에 기반한 데이터 분석 수행
- 행동: 전단지 배포 등을 위해 일별 계획 실행에 활용할 수 있는 데이터 분석 수행     
=> 객관적인 과거 성찰 + 미래 전망

<span style="color:black; background-color:#FFF099; padding:2px 4px; border-radius:4px">
<strong> 1-2. Data-Driven-Decision-Making (DDDM) <strong>
</span>


- 비즈니스 의사 결정 시 직관보다 데이터 & 분석 등 객관적 지표의 사용을 강조
- 고객 피드백, 재무 데이터 등 활용      
=> 더 나은 의사결정 가능!

<span style="color:black; background-color:#FFF099; padding:2px 4px; border-radius:4px">
<strong> 1-3. DDDM의 6단계 <strong>
</span>


1. 목표 정의
- 조직의 목표를 명확하게 표현
- 목표 달성 위해 목적이 명확한 노력 기울임

2. 데이터 식별, 준비 및 수집
- 데이터 요구 사항 판별 -> 데이터 소스 평가
- 체계적으로 데이터 수집 & 검증

3. 구성 및 탐색
- 새 패턴, 인사이트 발견할 수 있도록 구조화
- 정확성, 신뢰성 보호
- 시각화 -> 명확하지 않은 패턴, 이상치, 추세 식별

4. 데이터 분석 수행
- 원시 데이터 -> 실행 가능한 인사이트 변환
- 패턴, 상관관계, 추세 파악
- 전략적 결정 내림 & 성과 개선

5. 결론 도출
- 데이터 분석 결과 검토 -> 올바른 비즈니스 상황에 배치
- 실용적 결론 -> 성공적인 비즈니스!

6. 구현 및 평가
- 인사이트 검증 & 결과 측정
- 실행 계획 개발 & 리소스 할당 -> 진행 상황 지속적 모니터링
- 피드백 수집 & 반복적 개선 주기 -> 전략 계획 조정

<span style="color:black; background-color:#FFF099; padding:2px 4px; border-radius:4px">
<strong> 1-4. 데이터 분석 유형 <strong>
</span>



1. 설명 분석
- 데이터 취합 & 마이닝 -> 과거 데이터 바탕 인사이트 제공
- ex) 월별 판매 보고서, 고객 만족도 설문 조사 등

2. 진단 분석
- 특정 이벤트 발생 이유 파악 (사고의 근본 원인 파악)
- ex) 데이터 디스커버리, 마이닝, 상관관계 식별

3. 예측 분석
- 과거 데이터 기반 -> 미래 추세, 결과 예측
- 통계 모델, 머신 러닝 등 사용

4. 처방 분석
- 데이터 기반 조치 권장
- 예측 분석 + 최적화 알고리즘 -> 최상의 행동 방침 제안

etc. 탐색 분석, 추론 분석, 정성적 분석, 정량적 분석, 실시간 분석

## 2. 목표 정의

<span style="color:black; background-color:#FFF099; padding:2px 4px; border-radius:4px">
<strong> 2-1. 지표 <strong>
</span>

- 수치적으로 정보 설명 -> 방향, 목적, 기준 나타내는 표지
- 측정하고 싶은 것을 숫자로 표현
     
     
=> 분모 대비 분자로 표현!     
      
=> 경영학 기능별로 정리

<span style="color:black; background-color:#FFF099; padding:2px 4px; border-radius:4px">
<strong> 2-2. 지표 정의 <strong>
</span>

- 분모, 분자에 절대적인 수치를 넣을 수 없을 때
- Proxy Metric: 추상적 -> 프로덕트마다 가정, 정의, 논의 필요
- 추상적이므로 정답 별도로 존재하지 X
- ex) '스티비' 프로덕트: 사용자들이 핵심 가치를 느끼는지를 측정하기 위해 이메일 편집기에 진입한 사용자의 비율을 계산 -> 지표 개선 위해 노력    
<br>
** 핵심 가치란?
    사용자가 가치라는 것을 체감한 이후 행동이 눈에 띄게 좋아지는 결정적 순간
    => 재방문, 유료전환, 추천 등 긍정적 영향 미침

<span style="color:black; background-color:#FFF099; padding:2px 4px; border-radius:4px">
<strong> 2-3. 북극성 지표 & L1/L2 Metrics <strong>
</span>


- 북극성 지표: 기업이 가장 중요시하는 지표
    ex) 에어비엔비: 예약 박 수
<br><br>
- L1 지표: 전사적, 고수준 성과 지표 (전략 관점)     
    => 회사가 전략적으로 성공하고 있는지 확인하는 '큰 수치'     
    ex) 매출, 신규 가입자 수, 활성 사용자 수
<br><br>
- L2 지표: L1 지표의 세부 사항 측정 지표 (전술 관점)     
    => 큰 그림(L1)을 이루는 작은 목표들 (= 세부 점검표)     
    ex) 클릭률, 하루 접속 횟수 등

<span style="color:black; background-color:#FFF099; padding:2px 4px; border-radius:4px">
<strong> 2-4. AARRR <strong>
</span>

- 사용자 행동 측정 프레임워크
- 사용자가 제품 접하고, 계속 사용하고, 가치 만드는 과정 -> 5단계로 나눔!
- 프로덕트가 가치 제공하는 방식 다름 -> 측정하는 지표도 다름

### 과정
      
1. Acquisition
- 사용자가 채널 통해 서비스에 방문
- 유입 경로: 광고, 검색엔진 최적화, 오프라인 행사 등
- '어떻게 신규 고객을 어디로부터 획득할지' 가 중요!
- CAC, CTR, CPC, CPL, ROAS 등 지표 사용

2. Activation
      
- 사용자가 서비스에 방문에 주요 기능을 직접 사용
- '사용자에게 가치를 알려 활성화 정도 높이기'가 중요!
- sign-up(유저가 제품에 들어온 시점) -> Setup Moment(유저가 핵심가치 경험 준비 완료) -> A-HA! Momemt(유저가 핵심가치 처음 경험) -> Habit Moment(유저가 제품 사용 습관 성)

3. Retention
- 신규 사용자 -> 장기 사용자 전환
- '어떻게 장기 고객 양성하고 지속적 구매행위 증가할지'가 중요!
- DN Retention, MAU/DAU/WAU, Stickness 등 지표 사용 

4. Revenue
- 사용자가 서비스 유료 구매 결정
- '어떻게 사용자 소비 유도해서 매출 높일지'가 중요!
- 고객의 평생 가치 up, 고객 유치 비용 down
- PU, Average Check, jARPU, ARPPU, LTV 등 지표 사용

5. Referral
- 만족한 사용자가 타인에게 서비스 추천
- '어떻게 충성도 높은 사용자를 홍보자로 만들지'가 중요!
- 바이럴 계수(1 초과 시 Acquisition 단계 안 거치고 서비스 성장), 바이럴 주기 등 지표 사용
