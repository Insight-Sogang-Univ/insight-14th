### 1-2. 중심화 경향
- **평균(Mean)**: 데이터 값의 총합을 데이터의 개수로 나눈 값. 가장 대표적!!!, 총합을 사용하기에 <U>이상치(outlier)</U>에 민감.          
`series.mean()`
- **중앙값(Median)**: 데이터를 크기 순서대로 일렬로 정렬했을 때, 한가운데에 위치하는 값           
`series.median()`
- **최빈값(Mode)**: 데이터에서 가장 자주 나가는 값     
`seried.mode()`

### 1-3. 자료의 산포
ㄴ 데이터가 퍼져있는 정도를 설명하는 기술 통계량
- **범위(Range)**: 데이터의 최대값과 최소값의 차이, 데이터의 전체 변동 범위      
`series.max() - series.min()`
- **분산(variance)**: 데이터 값들이 평균으로부터 얼마나 멀리 퍼져있는지      
분산값이 크다 = 데이터가 많이 퍼져있다!     
`series.var()`
- **표준편차(Standard Deviation)**: 분산에 제곱을 취한 값     
`series.std()`
- **사분위수(Quartile)**: 데이터의 표본을 네 개의 동일한 크기로 나눈 값    
`series.quantile(0.25)  # 25th percentile
series.quantile(0.5)   # 50th percentile (중앙값)
series.quantile(0.75)  # 75th percentile`
- **사분위간 범위 (IQR)**: Q3에서 Q1 값을 제외한 범위
- **변동계수(Coefficient of variation)**: 데이터의 절대적인 변동성이 아닌 상대적인 변동성을 나타내는 통계적 지표

## 2. 확률분포

### 2-1. 확률변수
ㄴ random한 세계에서 어떤 특정 사건이 발생할 때 나타날 수 있는 모든 경우의 수를 특정 값으로 대응시키는 함수

### 2-2. 확률분포
ㄴ 확률변수가 특정한 값을 가질 확률을 나타내는 함수
1. 이산확률분포: 확률 변수가 이산적인(discrete) 값을 가지는 경우
2. 연속확률분포: 확률 변수가 연속적인(continuous) 값을 가지는 경우

### 2-3. 대표적인 분포 모형
1. 정규분포: 가장 잘 알려진 (연속)확률분포
- 종(bell) 모양
- 평균을 기준으로 대칭
- 표준편차가 클수록 곡선이 평평해짐 (데이터가 넓게 퍼짐)
       
2. T분포: 표본으로 모평균(혹은 모평균 간의 차이)을 추정할 때 사용되는 분포    
- 종(bell) 모양, 정규분포보다 꼬리가 두꺼움
- 평균을 기준으로 대칭
- 자유도가 커짐에 따라 표준정규분포 N(0,1)에 수렴
     
3. 카이제곱분포: 독립적인 표준 정규 분포를 따르는 변수들의 제곱 합, 표본분산의 분포와 관련된 지표
- 0에서 시작하여 오른쪽으로 긴 꼬리를 가진 비대칭 분포 
- 자유도가 증가함에 따라 중심이 오른쪽으로 이동하고 모양이 더 대칭적으로 변화하며, 정규분포에 가까워짐
     
4. F분포: 두 독립적인 카이 제곱 분포의 비율, 두 집단 간의 분산 차이를 비교하는 데 사용됨
- 0에서 시작하여 오른쪽으로 긴 꼬리를 가진 비대칭 분포 
- 자유도가 증가함에 따라 중심이 오른쪽으로 이동하고 모양이 더 대칭적으로 변화하며, 정규분포에 가까워짐

## 3. 가설검정 과정

### 3-1. 가설 검정의 절차
1. 가설 설정: 귀무가설 H0와 대립가설 H1를 설정
* 귀무가설: 주로 기각하고자 하는 명제 ex) “~는 차이가 없다”
* 대립가설: 내가 주장하고자 하는 명제 ex) “~는 차이가 존재한다”
2. 유의 수준 설정: 연구자가 결과를 판단할 때 허용할 수 있는 오류의 최대치로, 주로 0.05, 즉 5%로 설정
- **1종 오류(𝛼)** : 귀무가설이 실제로 참이었는데 기각할 확률
- **2종 오류(𝛽)** : 귀무가설이 실제로 거짓인데 기각하지 못할 확률

3. 검정통계량 산출: 검정통계량은 가설 검정의 결과를 판단하는 데 사용되는 값
- 데이터의 분포나 특징에 따라 T값, F값 등 어떤 통계량을 사용할지 정해야 함.

4. 가설 기각/채택 판단
- 유의확률(p-value): 가설 검정에 쓰일 수 있는 또 다른 척도     

유의 확률(p) < 유의 수준(α) => 귀무가설을 기각     

유의 확률(p) > 유의 수준(α) => 귀무 가설을 기각하지 못함

### 1. 수치형 - 수치형 **(피어슨)상관분석**: 두 변수간 피어슨 상관계수가 얼마나 유의미한지를 검정하는 방법         
⬇️ 시각화       

          
음..?
> 가설         

   H0: 두 변수간 선형 상관관계가 존재하지 않는다.     
   Ha: 두 변수간 선형 상관관계가 존재한다.         

> 공분산: 두 변수가 함께 어떤 방향으로 변화하는지와 그 크기를 표현하는데 사용     
- Cov (X,Y) > 0이면 X와 Y는 양의 관계, X가 커지면 Y도 커짐
- Cov (X,Y) < 0이면 X와 Y는 음의 관계, X가 커지면 Y는 작아짐
- Cov (X,Y) = 0이면 X와 Y는 양도 음도 아닌 관계        
 
> 상관계수: 두 변수 간의 선형적인 관계를 나타내는 통계적 지표
- Corr (X,Y)이 양수면 양의 상관관계 (1이면 완전한 양의 상관관계)
- Corr (X,Y)이 음수면 음의 상관관계 (-1이면 완전한 음의 상관관계)
