# **데이터로 보는 비즈니스 상황**

## 1. 데이터 기반 의사결정

### 1-1. 무엇을 위한 데이터 분석인가

데이터 분석이 지닌 가장 큰 특징 : 목적 달성 프로세스의 질    
액션 (결정, 계획, 행동)을 위한 데이터 분석 : 객관적으로 과거를 되돌아보고 미래를 전망하는 것 - 지금 해야 할 일에 집중!

### 1-2. Data-Driven-Decision-Making (DDDM)

DDDM : 비즈니스 의사 결정을 제공하기 위해 직관 대신 데이터와 분석을 사용하는 것을 강조하는 접근 방식

### 1-3. 6 Steps of DDDM

1. 목표정의 

2. 데이터 식별, 준비 및 수집
- 데이터 소스를 평가 및 준비, 데이터를 수집 및 검증

3. 구성 및 탐색
- 데이터의 새로운 패턴, 추세 및 인사이트를 발견할 수 있도록 구조화 및 시각화 (이상치 식별)

4. 데이터 분석 수행
- 다양한 기술과 방법론을 사용해 원시 데이터를 실행 가능한 인사이트로 변환
- 비즈니스 전략에 영향을 미치는 패턴, 상관관계 및 추세 파악

5. 결론 도출
- 주요 데이터 분석 결과 검토
- 올바른 비즈니스 상황에 배치

6. 구현 및 평가
- 데이터 기반 권장 사항을 기반으로 실행 계획 개발, 리소스 할당
- 핵심성과지표 (KPI)를 기준으로 결과 측정 및 분석, 피드백 진행

### 1-4. 데이터 분석 유형

1. 설명 분석
- 데이터 취합 및 마이닝을 통해 과거 데이터를 설명하고 요약, 과거 성과에 대한 인사이트 제공
2. 진단 분석
- 특정 이벤트가 발생한 이유를 파악하는데 중점
3. 예측 분석
- 과거 데이터를 기반으로 미래의 추세나 결과를 예측 (통계 모델, 머신 러닝)
4. 처방 분석
- 예측 분석과 최적화 알고리즘을 결합하여 최상의 행동 방침 제안

## 2. 목표 정의

### 2-1. 지표

지표 : 방향이나 목적, 기준 따위를 나타내는 표지 (추상적인 개념과 관련되는 지수, 척도 등을 말함)     
대부분 분모 대비 분자로 표현

$$

\text{metrics} = \frac{\text{molecule}}{\text{denominator}}

$$

- 효율성 : 산출/투입
- 효과성 : 달성 결과/목표
- 수익성 : 이익/자원
- 성장성 : 증가분/기존규모
- 안정성 : 보유자산/부채

### 2-2. 지표를 정의하기

분모, 분자에 절대적인 수치가 들어가지 않는 경우 Proxy Metric을 사용   
Proxy Metric : 활성화와 인게이지와 같이 추상적이여서 프로덕트마다 나름의 가정과 정의, 논의가 필요한 지표

### 2-3. 븍극성 지표 & L1/L2 Metrics

복극성 지표 : 기업이 가장 중요하게 생각하는 지표    
L1 : 전사적인, 고수준의 성과 지표 (전략 관점)    
L2 : L1 자표의 세부 사항을 측정하는 지표 (전술 관점)

### 2-4. AARRR

프로덕트가 가치를 제공하는 방식이 다르면, 측정하는 지표도 달라져야 함    
AARRR (해적지표) : 사용자 운영 과정에서 활용하는 데이터 분석 모델, 사용자 증가를 위한 5가지 로드맵으로 구성

1. Acquisition
- 사용자가 여러 채널을 통해 앱 또는 웹사이트에서 방문하는 것을 의미
- 고민해야 되는 포인트 : 신규 고객을 어떻게 획득할지, 어디로부터 획득할지, 어떤 채널을 통해 획득할지
- CAC (Customer Acquisition Cost): 유저 획득 비용
- CTR (Cost Per Click): 클릭 한 건당 얼마 지출
- CPC (Cost Per Install): 인스톨 당 과금되는 상품
- CPL (Cost Per Lead): 유저 행동 당 드는 비용
- ROAS (Return On Ads Spending): 광고로 인한 매출액/광고비

2. Activation (사용자 활성화)

- 사용자가 앱 또는 웹사이트에 첫 방문하여 주요 기능을 직접 사용하는 것
- 제품의 성격에 따라 다양한 기준으로 측정
- Sign-up: 유저가 제품에 들어온 시점
- Setup Moment: 유저가 제품의 핵심가치를 경험할 준비 완료
- A-HA! Moment: 유저가 핵심가치를 처음 경험한 순간
- Habit Moment: 유저가 제품 사용에 대한 습관 형성

3. Retention (유지)

- 신규 사용자를 장기 사용자로 전환시키는 것에 초점
- 어떻게 장기 고객을 양성할 것인지, 어떻게 고객의 지속적인 구매 행위를 증가시킬지
- 측정 지표 : 익일 유지율, 1주차 유지율, 월 유지율, EDM 메일 개설율
- DN Retention: N일 안으로 재접속하는 경우
- MAU/DAU/WAU: 일/주/월마다 방문하는 횟수
- Stickness(DAU, MAU): 해당기간 동안 활성화 수


4. Revenue (수익)

- 사용자가 제품 또는 서비스를 구매하기로 결정한 단계
- 소비를 우도해 매출을 높이는데 집중
- 고객의 평생 가치 (LTV)를 지속적으로 높여 나가고, 고객 유치 비용은 줄여나가야 함
- PU(Paying User): 전체 사용자 중에 결제를 한 사용자 비율
- Average Check: 구매 1회당 평균 금액 (Revenue/Transaction)
- ARPU(Average Revenue Per User): 인당 평균 매출
- ARPPU(Average Revenue Per Paying User): 결제자당 평균 매출
- LTV(Life Time Value): 유저가 진입부터 이탈까지 누적 수익

5. Referral (추천)

- 충성도 높은 사용자를 브랜드 홍보자로 만들 방법, 어떻게 제품을 잠재 고객에게 추천하도록 만들 것인지
- 순 추천고객 지수 (NPS), 바이럴 지수 (Viral Coefficient)
- 바이럴 계수: 기존 고객 한명당 데리고 온 새로운 고객의 수 (1이 넘으면 Acquisition 단계를 하지 않아도 서비스가 성장)
- 바이럴 주기: 기존 고객 1명이 새로운 고객을 데려오는데 걸리는 시간
