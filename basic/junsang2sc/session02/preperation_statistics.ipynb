# **통계 사전 학습**

자로의 척도
- 명목척도 : 레이블이나 이름으로 요소의 속성을 식별하려고 하는 척도
- 서열척도 : 명목척도 + 항목 간 순서나 순위 존재
- 구간(등간)척도 : 값들 사이에 순서가 있고, 그 간격이 일정하여 덧셈이나 평균과 같은 연산이 가능한 척도 (곱하기나 나누기 X)
- 비율척도 : 절대영점과 임의의 측정단위로 구성하고, 동간성이 유지되는 척도

## 1. 기술통계

### 1-1. 기술통계 

데이터를 있는 그대로 설명하고 이해하는 데 중점을 둔 통계 방법

### 1-2. 3가지 주요한 도구 집합

1. 중심화 경향
- 평균: 데이터 값의 총합을 데이터의 개수로 나눈 값, 이상치에 민감
- 중앙값: 데이터를 크기 순서대로 일렬로 정렬했을 때, 한가운데에 위치하는 값, 이상치에 평균보다 덜 민감
- 최빈값: 데이터에서 가장 자주 나타나는 값, 이산형 데이터의 분포 이해에 도움을 줌

2. 자료의 산포
- 범위: 데이터의 최대값 - 최소값
- 분산: 데이터 값들이 평균으로부터 얼마나 멀리 퍼져있는지
- 표준편차: 분산에 제곱근을 취한 값 (단위를 일치시키기 위해)
- 사분위수: 데이터의 표본을 네 개의 동일한 크기로 나눈 값
- 변동계수: 데이터의 상대적인 변동성, 변동계수 = (표준편차/평균)*100

3. 첨도와 왜도
- 첨도: 확률분포의 꼬리가 두꺼운 정도를 나타내는 척도
- 왜도: 비대칭도, 실수 값 확률 변수의 확률 분포 비대칭성을 나타내는 지표 (왜도가 양수일때 오른쪽으로 긴 꼬리를 가짐)

## 2. 확률과 분포

### 2-1. 확률분포

- 확률분포 : 확률변수가 특정한 값을 가질 확률을 나타내는 함수
1. 이산확률분포 : 확률변수가 유한하거나 셀 수 있음
2. 연속확률분포 : 확률변수가 무한하고 특정 값 선택이 불가능

### 2-2. 대표적인 분포 모형

1. 정규분포
- 종 모양, 평균을 기준으로 대칭, 표준편차가 클수록 곡선이 평평해짐
- 중심극한정리 : 모집단의 분포를 몰라도 표본의 크기가 충분히 크다면(n>=30) 포본평균의 분포는 근사적으로 정규분포를 따른다

2. T 분포
- 종 모양, 정규분포보다 꼬리가 두꺼움, 자유도가 커짐에 따라 표준정규분포 N(0,1)에 수렴
- 표본으로 모평균(혹은 모평균 간의 차이)을 추정할 때 사용되는 분포
- 표본 크기가 작을 때 정규분포보다 정확한 결과를 제공함

3. 카이제곱 분포
- 0에서 시작하여 오른쪽으로 긴 꼬리를 가진 비대칭 분포 
- 자유도가 증가함에 따라 중심이 오른쪽으로 이동하고 모양이 더 대칭적으로 변화하며, 정규분포에 가까워짐
- 표본분산의 분포와 관련된 지표
- 독립성 검정, 적합도 검정에 사용



4. F 분포
- 0에서 시작하여 오른쪽으로 긴 꼬리를 가진 비대칭 분포 
- 자유도가 증가함에 따라 중심이 오른쪽으로 이동하고 모양이 더 대칭적으로 변화하며, 정규분포에 가까워짐
- 두 독립적인 카이 제곱 분포의 비율
- 두 집단 간의 분산 차이를 비교하는 데 사용됨

## 3. 추론 통계 - 추정

### 3-1. 추론통계란

- 추론통계: 수집한 데이터를 기반으로 미래의, 혹은 조사하지 못한 전체 정보를 추론하고 결론을 도출하는 방법을 다루는 분야
- 추정: 모집단을 대표할 수 있는 일부를 표본으로 추출하여 분석한 통계량을 이용해서 모집단의 모수를 예측하는 과정
- 검정: 표본을 추출하여 얻은 표본통계량으로 모집단 모수에 대한 새로운 예상·가설 등이 옳다고 판단할 수 있는지를 평가하는 과정

### 3-2. 점추정

- 어떤 모수를 표본에 기초하여 계산된 하나의 값으로 추정하는 것

### 3-3. 구간추정

- 모수의 참 값이 포함되리라고 기대하는 추정치를 일정한 범위로 신뢰도를 가지고 나타내는 방법

## 4. 추론 통계 - 검정

### 4-1. 가설 검정의 절차

1. 가설 설정
- 귀무가설 H0 = 기각하고자 하는 명제 (~차이가 없다)
- 대립가설 H1 = 새롭게 주장하고 싶은 가설 (~차이가 존재한다)
2. 유의 수준 설정
- 유의수준 : 결과를 판단할 때 허용할 수 있는 오류의 최대치 -> 귀무가설이 참인데도 불구, (유의수준)의 확률로 1종 오류를 범할 수 있다
3. 검정통계량 산출
- 가설 검정의 결과를 판단하는데 사용되는 값
- Z값 : 표준정규분포를 사용하여 얻은 값, 특정 데이터가 평균으로부터 얼마나 떨어져 있는지
-       Z값의 절댓값이 클수록 평균으로부터 멀리 떨어져있음, Z값이 임계값보다 크면 귀무가설 기각
- T값 : T분포를 사용하여 얻은 값, 작은 표본 크기에 대한 가설 검정 때 사용
-       주로 모집단의 평균에 대한 가설 검정이나, 두 표본 평균 간의 차이를 검정할 때 
- F값 : F분포를 사용하여 얻은 값, 주로 두 표본 분산 사이의 차이를 비교하는 데 사용 (분산분석)
        F값 = 분산1/분산2, 따라서 1에 가까우면 두 표본의 분산이 비슷하다고 할 수 있음
