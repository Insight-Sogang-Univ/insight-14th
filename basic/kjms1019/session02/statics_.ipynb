#통계
 1.자료의 척도
    질적 자료
        -명목척도
        -서열척도
    양적 자료
        -구간척도
        -비율척도
2.기술통계 : 데이터를 있는 그대로 설명하는 방식
    -중심화 경향
        -평균, 중앙값, 최빈값
    -자료의 산포
        -범위, 분산, 표준편차, 사분위, 사분위간 범위, 
        변동계수(표준편차/평균*100)
3.확률과 분포
    -확률변수 
    -확률분포 : 확률변수가 특정한값을 가질 확률
        -이산확률분포
        -연속확률분포
            -정규분포
            -T분포 : 표본크기가 작을때 유용
            -카이제곱분포: 표본 분산의 분포와 관련
            -F분포: 두 집단간 분산차이를 비교
4.추론통계
    -정의 : 통계적 가설검정과 신뢰구간 추정!
    -점추정 
    -구간추정
    -검정
        -가설검정 : 귀무가설 설정, 유의 수준 설정, 검정통계랑 산출, 가설 기각 및 채택
    -데이터 자료형에 따른 검정방법
        -피어슨 상관분석
        -F-test, T-test
        -chi2-test
