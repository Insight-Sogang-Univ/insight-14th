## 다중선형회귀의 기본가정

# 다중선형회귀 
    1. 단순선형회귀와 근본적으로 같음. 변수만 더 많은 것 뿐이다. 
    2. 회귀분석 프로세스 또한 매한가지. (사전검증 -> 모델 생성 및 모델 fit -> 모델평가 -> 모델성능 개선)
    3. 사전검증 단계에서 회귀 분석의 기본가정 6가지 검토가 매우 중요!!
        (1) 선형성
            * 종속변수와 설명변수 간의 관계가 선형적이어야 한다
            * scattor plot을 찍어보면된다!
        (2) 독립성
            * 각각의 설명변수가 서로 선형독립적이어야 한다. 
            * 여기서 잠깐 선형독립적이란? 선형대수에서 말하는 선형독립이 아니라, 관측치들 사이에 상관관계가 없어야 함을 의미하는 것이다. 
            * 선형독립적이지 않다는 것은 즉 다중공선성이 있다는 의미. VIF로 쉽게 다중공선성을 체크할 수 있다.
            * 이후 변수제거 규제 선형모델 PCA의 방법등으로 대처 가능!
        (3) 오차항의 평균은 0이다.
            * 오차항이란 실제값과 예측값 사이의 차이!
            * 오차항이 0이라는 의미는 회귀모델이 특정 패턴이 없이 정규화되어있다는 가정이다!
            * 만약 오차항이 존재한다면 모델이 항상 실제값보다 오차항만큼 잘못 예측하는 문제가 발생한다!
            * np.mean(residuals)로 확인해 볼 수 있다.
        (4) 등분산성
            * 오차항의 분산이 일정해야한다
            * 만약 일정하지 않다면, 모델이 구간마다 정확도가 달라지는 문제가 발생한다!
            * 오차항 분산이 적은 구간에선 좋은 모델이었다가 갑자기 분산이 큰 구간에선 안좋은 모델이 될 수 있다. 
            * 등분산성을 만족하는지 알기위해선 잔차의 도표화, BP검정의 방법이 있다. 즉 표준화잔차 대 표준화 예측값의 산점도를 이용한다는 의미이다.
        (5) 오차항은 자기상관되어있지 않다
            * 오차항의 공분산은 항상 0이어야 한다
            * 만약 0이 아니라면 자기상관되어있다고 보며, 현재값이 과거 값과 상관관계가 있다는 의미이다.
            * 이런 조건을 필요로하는 이유는 모델이 예측할 때 이전 데이터의 패턴을 학습해서 반복적으로 오류를 만들어낼 수 있기 때문이다.
            * 자기상관을 알아보는 방법으로는 더빗-왓슨 검정, ACF PACF 함수 찍어보기가 있다.
        (6) 정규성
            * 오차항이 정규분포를 따른다는 가정이다. 필수는 아니지만 되면 좋다 
            * 왜냐하면 p값이 유의미하려면 오차항이 정규분포를 따라야하기 때문이다!
            * 정규성을 알아보는 방법으로 샤피로-월크 검정, 자퀴-베라 검정, Q-Qplot이 있다.
        (7) 회귀분석 전 검증가능한 가정
            * 선형성, 독립성, 오차항의 자기상관성
            * 회귀분석 후에는 모든 가정이 검증 가능!

    
## 회귀 분석의 평가방법
   
    1. 시각화를 먼저 해봄으로써 대략적인 유효성을 파악할 수 있다. 하지만 정밀한 분석은 통계지표를 활용해야 한다!
    2. 통계지표
        (1) R-square(결정계수)
            * 대표적인 지표이며 모델의 성능을 평가하는 지표이다!
            * 결정계수만큼 전체모델을 설명할 수 있다는 의미로 해석하고 0와 1사이 값으로 나타낸다.
            * 조정된 결정계수도 있는데, 기존 결정계수는 독립변수의 개수가 많을수록 자연스럽게 증가하기 때문에 이를 해결하기 위해 만들었다.
            * 기존 식에 변수의 개수를 분모로 곱하는 식으로 변수개수의 뻥튀기를 조정했다.
        (2) F-statistic 
            * 회귀모델이 유의미한지 검정하는 지표이다
            * 값이 클수록 모델이 통계적으로 유의미하다는걸 의미한다
        (3) Prob(F-statistic)
            * F-statistic의 p값으로 0.05보다 작아야 귀무가설을 기각할 수 있다!
        (4) T-statistic
            * 각 독립변수가 종속변수에 유의미한 영향을 미치는지 판단하는 지표!
            * p값을 기준으로 가설검정을 수행한다.
        (5) F랑 T 무슨차이냐?
            * F는 전체 회귀모델이 통계적으로 유의한지 검정하는 것이고, T는 개별 독립변수들이 종속변수에 얼마나 유의한지 검정하는 것이다!

## 비선형 회귀모델
    
    1. 다항식 회귀모델
        (1) 독립변수에 거듭제곱 항을 추가하여 데이터의 곡선적 특성을 모델링한다!
        (2) 다만 너무 많은 feature를 이용하면 오버피팅될 수 있다는 문제가 존재한다. 라쏘, 릿지를 활용해 문제를 극복하자!
        (3) 성장률, 감소율 등 시계열데이터에 적합할 수 있다.
    2. 지수 회귀 모델
        (1) 기존 지수 함수에 양변 로그를 취해 선형모형으로 만든 후 회귀를 적용한다
        (2) Y값 예측시엔 다시 지수함수를 적용시켜서 로그변환 전의 Y를 구한다!
    3. 로그 회귀모델
        (1) Y=a+b*log(X) 의 형태를 띈다. 따라서 x에 로그연산을 취하고 이후 선형회귀를 진행한다!
        (2) Y는 로그가 아니므로 따로 처리를 하지 않아도 된다!
        (3) 인구성장, 감염병 확산과 같이 초기 빠른 증가후 점진적 감소를 보이는 데이터에 적합
    4. 스플라인 회귀
        (1) 데이터 범위를 구간별로 나누고 각 구간에 대해 별도의 회귀 모델을 적용하는 방식
        (2) 데이터 패턴이 여러 구간에서 다르게 나타날 때 유용하다! 계절이나 특정 기간이 반영된 데이터를 예시로 들 수 있다!.
## 회귀 모델링 프로세스
   
    1. 사전검증 : 이 단계에선 결측치, 이상치 처리 및 기본가정 6가지를 검토해야 한다. 필요하다면 PCA 및 EDA를 진행한다!
    2. 모델생성 및 fit : model.fit , model.predict()를 이용한다!
    3. 모델 성능 평가 : .summary()를 이용해 F검정,T검정,상관계수,회귀계수,결정계수 등 분석을 진행하고 가설검정을 실행한다. 이 단계에서 다른 PCA나 다른 이상치 데이터들의 결과들을 서로 비교분석할 수 있다.
    4. 모델 성능 개선 : 최적의 데이터 구성을 택하고 필요없는 변수는 제거하거나, 비선형 모델을 활용할 수 있다.
