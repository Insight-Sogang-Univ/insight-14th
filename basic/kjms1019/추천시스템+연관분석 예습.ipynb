# 추천시스템 기초
## 1. 추천시스템이란
### (1) 데이터 종류

        a. 명시적 데이터 

                * 고객들이 입력한 데이터. ex) 평점, 좋아요, 투표 등

                * 모든 고객이 입력하진 않는다. 그리고 과거의 선호도와 지금의 선호도가 다를 수 있다.

        b. 암시적 데이터

                * 행동을 기반으로 간접적인 선호도를 알 수 있는 데이터 ex) 조회수, 영상시청기간,
                구매기록 등

                * 높은수치가 높은 선호도를 의미하는건 아니다.

                * 데이터를 얻긴 쉽지만 직접적인 선호도라고 단정짓긴 어렵다.

## 2. 내용기반 필터링

### 특징

        a. 사용자의 프로필과 아이템 특징의 유사도를 기반으로 추천하는 방식

        b. 사용자의 행동 이력을 기준으로 비슷비슷한 아이템을 계속 추천해주는 메커니즘이다.


        
### (1) 유사도

        a. 유사도

        
                ㄱ. 영상,이미지,텍스트 데이터를 벡터화한 이후 아이템 벡타간의 유사도를 측정한다. 이후 사용자가 좋아하는 아이템들의 벡터값의 유사도를 측정해 얼마나 비슷한지를 알아낸다.
        
                ㄴ. 유사도 방식의 종류 
        
                        * 자카드 유사도 : 아이템 집합 간의 교집합 크기를 이용
        
                        * 코사인 유사도 : 아이템 벡터 사이의 각도를 이용
        
                        * 유클리디안 거리 : 아이템 벡터 사이의 거리를 이용
        
                        * 피어슨 상관계수 : 아이템 벡터간 선형 상관관계를 이용
### (2) 사용자 프로필
        
        a. 직접 지정형
        
                ㄱ. 사용자가 선호하는 아이템 특징을 명시적으로 스스로 지정하도록 하는 방법 ex) 가입시 선호도 체크
        
        b. 간접 지정형
        
                ㄱ. 사용자의 과거 행동이력을 기반해 프로필을 작성 ex) sf를 즐겨보면 인터스텔라 추천
        
        c. 프로필 획득 방식
        
                ㄱ. 특정 가중치는 사용자가 가진 아이템 특성 가중치의 평균 값을 활용
        
                ㄴ. 사용자의 기호를 따로 파악하거나, 설문조사 혹은 자발적 키워드 입력 등으로 프로필 획득
        
                ㄷ. 아이템에 대한 유저의 기존 평가를 바탕으로 분류문제를 해결하는 과정과 비슷
### (3) 아이템 특징
        
        a. 아이템을 벡터 형태로 표현한다. 도메인에 따라 다른 방법이 적용된다
        
        b. 아이템 분석 알고리즘
        
                ㄱ. TF-IDF
        
                        * 특정 문서내에 특정 단어의 빈도(TF)와 전체 문서에서의 빈도인 역문서빈도(IDF)를 통해 다른문서에는 없고 특정문서에만 자주 등장하는 단어를 찾는다
        
                        * 각 단어가 문서에서 얼마나 중요한지를 나타내는 값으로 중요도를 가늠한다.
        
                        * 장점 : 아이템에 대한 콜드 스타트 문제 해결 가능, 피쳐를 통해 추천하기 때문에 추천 이유를 설명하기 쉽다
        
                        * 단점 : 피쳐를 추출하는 과정이 매우 중요해서 추출하지 못할 시 정확도가 떨어진다. 기존 item과 유사한 item위주로만 추천하기에 새로운 장르의 아이템을 추천하기 어렵다.

## 3. 협업 필터링
### 서비스내 다수의 사용자로부터 얻은 선호도를 기반으로 추천하는 방식. 비슷한 선호도를 가진 사용자들은 앞으로도 비슷한 선호도를 가질 가능성이 높다고 가정. 
### Input은 사용들이 매긴 등급이 되고 Output은 사용자와 비슷한 다른 사용자들을 판별, 그들의 아이템 등급이 됨.
### (1) 메모리기반
        
        a. 사용자 또는 아이템간의 유사도를 기반으로 한다. 사용자 a와 비슷한 취향을 가진 사용자가 선호하는 아이템을 추천해주는 방식이다.
        
        b. 사용자 기반 
        
                ㄱ. 특정 사용자와 유사한 다른 사용자를 TOP-N으로 선정해 이 TOP-N이 좋아하는 아이템을 추천해준다
        
                ㄴ. 사용자간 유사도를 측장하고 가장 유사도가 높은 top-n을 골라 아이템을 추천해준다.
        
                ㄷ. 유사도는 아이템에 준 평가 점수로 측정한다.
        
                ㄹ. 아이템 수가 사용자수보다 많을 때 추천정확성과 효율성이 올라간다.
        
                ㅁ. 아이템이 수시로 바뀔때 안정성이 올라가고, 다양성도 좋다.
        
        c. 아이템 기반
        
                ㄱ. 사용자들이 그 아이템을 좋아하는지/ 싫어하는지의 평가척도가 유사한 아이템을 추천하는 누기준이 되는 알고리즘 
        
                ㄴ. 사용자 수가 아이템보다 적을 때 추천정확성과 효율성이 올라간다.
        
                ㄷ. 아이템 수가 적고 변동성이 적을수록 안정성이 좋다. 참신성은 부족할 수 있다. 
        
        d. 한계
        
                ㄱ. 아이템/유저가 늘어날수록 확장성이 떨어진다.
        
                ㄴ. 주어진 평점/선호도 데이터가 적을 경우, 성능이 저하됨
                        * 왜냐하면 희소행렬이 99.5%보다 넘어버리면 성능이 현저히 낮게 나오기 떄문
        
                ㄷ. 아이템에 대한 평점을 계산하기 위해 모든 유저와의 유사도를 구하려면 연산이 너무 늘어남. >> KNN CF이용
### (2) 모델기반
        
        a. 사용자-아이템 평점 매트릭스 속의 잠재 요인을 추출해 추천 예측을 해준다. 
        
        b. 잠재요인
        
                ㄱ. 잠재요인이란 사용자와 아이템 사이의 숨겨진 특성이다. 숨겨져 있는데 어떻게 포착할까??
        
                ㄴ. 모델이 관측 데이터를 바탕으로 추정한다. 
        
                ㄷ. Latent Factor Model은 아이템, 유저를 모두 정해진 차원으로 하나의 벡터로 표현한다. 그리고 각각의 차원에 특성이 존재하는 방식이다.
        
                ㄹ. 실제 모델은 차원의 수가 정해지지 않아 벡터별로 명확한 특징을 파악하긴 어렵고, 아이템을 표현한 factor는 사람이 명시적으로 이해할 수 없다.
        
                ㅁ. 사용자와 아이템을 잠재축으로 사용해 나타냄으로써 메모리 기반의 한계를 극복할 수 있던 것이었다. 
        
        c. 행렬분해
        
                ㄱ. 유저-아이템 행렬 에는 유저취향과 아이템의 특성이 모두 녹아있으므로 이를 분해해 유저따로 아이템따로 두 행렬의 내적 형태로 만든다면 각각의 잠재요인을 알아 낼 수 있다.

# 연관분석 기초
## 1. 연관분석이란
### (1) 일련의 거래나 사건안에 존재하는 항목 간 연관규칙을 발견하는 분석! 장바구니 분석이라고도 한다.
### (2) 특징
        
        a. 정답값이 따로 없어 비지도학습에 해당한다.
        
        b. 주요 알고리즘으로 Apriori, FP-Growth가 있다.

        c. if-then의 형식을 띄고, A>>B로 표현된다. 예를들어 '라면을 구매하는 사람들은 달걀도 함께산다'는 문장에서 라면구매의 조건절과 달걀구매의 결과절이 있다.

        d. 연관규칙으로 상관관계는 알 수 있지만 인과관계는 알 수 없다. 실험이 아닌 관찰 데이터기 때문이다. 방향성을 인과관계로 오해하면 안된다

        e. 협업 필터링도 상관관계를 사용한다는 점에서 비슷하지만, 연관분석은 복잡한 수치는 계산하기 어려운 대신 다양한 세트상품을 뽑아내긴 쉽다.

## 2. 주요개념
### (1) 연관규칙 조건
        
        a. 일정 이상의 지지도 : 두 품목이 함께 구매한 경우의 수가 일정 수준은 이상이어야 한다
        
        b. 일정 이상의 신뢰도 : 품목 A를 포함하는 거래 중 품목B를 구입하는 경우의 수가 일정수준 이상이어야 한다. 
        
        c. 지지도란 전체 거래항목중 A와B의 교집합인 거래의 비율이다. 즉 조건이 발생할 확률이다.
        
        d. 신뢰도란 A를 포함하는 거래 수 중 A와B가 동시에 포함하는 거래의 비율이다. 즉 조건부 확률이라고 생각하자. 신뢰도가 낮다면 추천,진열에 의미가 없어진다. 
### (2) 평가 측도 (지지도, 신뢰도, 향상도 모두 고려)

        a. 향상도 

            ㄱ. 품목 A가 주어지지 않을때의 B의 확률 대비, A가 주어졌을떄의 B의 증가 비율. 즉 A와B가 우연히 함께 등장할 확률과 실제 함께 등장할 확률을 비교해보자는 아이디어.

            ㄴ. A와B가 독립일 땐 향상도는 1이며, 보통 향상도가 1보다 높다면 양의 상관관계, 1보다 낮으면 음의 상관관계라고 본다. 

## 3. 연관분석 알고리즘
### (1) Apriori 알고리즘
        
        a. 배경
        
            ㄱ. 연관 규칙을 찾자고 가능한 모든 조합의 무차별 탐색은 비효율적이다.
        
            ㄴ. 상품 개수가 많아질수록 조합은 더 많아지고 비용또한 늘어날 것.
        
        b. 개념
        
            ㄱ. 상위 조합에서부터 차례로 스캔하며 특정 조합이 자주 발생하지 않는다면 이의 결과물로 탄생한 후속 조합들 까지 모두 후보에서 배제
        
            ㄴ. 하나의 조합만 검사하고도 파생된 다른 조합들을 배제할 수 있으므로 효율적이다.
        
        c. 프로세스
        
            ㄱ. 빈발 항목 집합(최소 지지도를 넘는 조합)을 찾고 단계적 조합을 생성 및 필터링
        
            ㄴ. 단일 항목 집합 생성
        
                * 모든 개별 상품의 지지도 계산 후 최소 지지도를 넘는 애들만 뽑는다
        
            ㄷ. 2개 항목집합 생성
        
                * 1단계 상품들로 2개 항목 조합을 만들고, 최소 지지도를 넘지 않는 조합을 제거
        
            ㄹ. 더 큰 항목 집합 생성
        
                * 이전 단계에서 남은 조합들을 바탕을 3개이상의 항목으로 구성된 조합을 만듦
        
                * 지지도를 계산하고, 최소 지지도를 안넘는 조합 제거
        
                * 더이상 조합을 생성할 수 없을때까지 반복
        
            ㅁ. 최종 빈발 항목집합 선정
        
                * 마지막으로 남은 빈발 항목 집합들로 연관규칙을 생성
        
                * 각 규칙에 대해 지지도, 신뢰도, 향상도 계산해 가장 유용한 규칙 선택
        
        d. 장단점
        
            ㄱ. 원리가 간단하고 상품 수가 많아도 패턴발견이 쉽다.
        
            ㄴ. 비지니스 측면의 중요연관규칙은 부족할 수 있고, 데이터가 많아지면 속도가 느려진다.
### (2) FP Growth 알고리즘
        
        a. 배경
        
            ㄱ. Apriori는 여전히 데이터가 커질수록 속도가 느려진다는 단점 존재.
        
        b. 개념
        
            ㄱ. 알고리즘 빈발 품목 집합을 더 효율적으로 찾기 위해 FP-TREE 구조를 추가했다
        
            ㄴ. FP-tree란 tree, array, linked-list를 합쳐놓은 구조로 모든 거래를 확인해 아아템마다 
            지지도를 계산하고, 최소 지지도 이상의 아이템만 선택한다.
            
            ㄷ. 모든 거래에서 빈도를 계산해 높은순으로 정렬 후, 부모 노드를 중심으로 자식노드를 추가하며 tree를 생성한다
            
            ㄹ. 새로운 아이템이 생기면 부모노드부터 재시작하고 아니면 기존노드에서 확장한다
            
            ㅁ. 위 과정을 반복해 최소 지지도 이상 패턴만 추출한다. 
        
        c. 장단점
        
            ㄱ. 데이터를 두번만 스캔하는 구조기 때문에 훨씬 빠르고, 후보셋을 생성할 필요가 없어졌다.
        
            ㄴ. 대용량 데이터셋에서 메모리를 효율적으로 사용하지 않는다. 설계가 어렵고 지지도 계산은 FP-tree가 만들어진 후여야만 가능하다는 단점이 있다.
