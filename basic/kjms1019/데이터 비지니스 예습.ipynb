# 데이터로 보는 비지니스 상황
## 1. 데이터 기반 의사결정
### 데이터 분석은 왜할까? 궁극적인 목적이 무엇이길래?
- 경험과 감으로만 의존하던 의사결정에 데이터가 추가되면 생각지도 못했던 것들과 잘못된 편견들이 걷히면서 올바른 인사이트를 얻을 수 있다. 
- 인사이트는 액션에 필수적으로 이용된다. 액션에는 어떤것들이 있을까?
    1. 결정 : 출시에대한 의사결정, 프로젝트 진행 여부 결정 등
    2. 계획 : 마케팅 전략, 프로모션 계획, 매출 시뮬레이션 등
    3. 행동 : 판매예측 모델을 위한 시스템 개발, 전단지 배포 등
- DDDM : 데이터와 분석을 적극 이용한 비지니스 의사결정. 6 단계로 이루어진다
    1. 목표정의 : 조직의 목표를 명확히 표현.
    2. 데이터 식별, 준비, 수집 : 목표에 필요한 데이터를 평가 및 준비
    3. 구성 및 탐색 : 데이터를 뜯어보며 인사이트를 발견할 수 있도록 EDA 및 시각
    4. 데이터 분석 수행 : 패턴 및 상관관계 등 통계적 분석 실시
    5. 결론 도출 : 분석결과를 검토하고 인사이트를 뽑아낸다.
    6. 구현 및 평가 : 인사이트를 검증하고 결과를 측정한다. 결과를 실제로 실행하며 진행상황을 모니터링하고 피드백을 수집해 개선사항을 제공한다.
- 데이터 분석 유형
    1. 설명분석 : 과거 데이터를 설명하고 요약한다
    2. 진단분석 : 현재 이벤트가 발생한 이유를 파악한다. 사건의 근본 원인에 집중한다.
    3. 예측분석 : 미래의 추세나 결과를 예측한다. 과거 데이터를 기반으로 여러 모델을 사용한다.
    4. 처방분석 : 예측분석에 더해 최적화 알고리즘을 결합해 아직 일어나지 않은 상황에서도 최적화 및 자동화를 도모한다.
## 2. 목표정의
### 지표란 무엇일까?
- 기업의 의사결정에 구체적인 수치를 부과함으로써 목적과 기준을 세워준다. 추상적인 목표들이 눈앞의 목적으로 다가올 수 있게 해준다. 
- 지표의 종류
    1. 재무 : 효율성 지표(자산회전율 = 매출액/총자산), 효과성지표(예산 집행률=실제 지출/계획 지출), 수익성 가치 지표(ROE = 순이익/자기자본, ROA = 순이익/자산), 성장성안정성 지표(매출 성장률,부채비율) 
    2. 마케팅 : 효율성지표(CAC 효율 = 신규고객 수/마케팅비), 효과성지표(전환율 = 구매자/방문자), 수익성/가치지표(마케팅 ROI = 캠페인 이익/캠페인 비용), 성장성/안정성 지표(시장점유율 성장률)
    3. HR : 효율성지표(노동생산성 = 산출량/인력투입), 효과성지표(교육 효과 달성률 = 실제 역량향상/목표 역량향상), 수익성/가치지표(HR ROI = 교육 후 성과 증가/교육비), 성장성/안정성 지표(이직률, 직원 유지율)
    4. 운영 :  효율성지표(설비가동률 = 실제 가동시간/총 가능시간), 효과성지표(품질목표 달성률 = 합격품 /목표 품질수준), 수익성/가치지표(단위당 원가 절감율 = 절감액/총원가), 성장성/안정성 지표(재고회전율, 공급망 안정성)
- 절대적으로 수치내기 어려운 지표는 어떡할까?
    1. Proxy Metric을 이용한다.  대체물(Proxy)을 정하여 측정한다는 의미로, Proxy 지표들은 설정 자체에 많은 논의가 필요하고 정답이 별도로 존재하지 않다.
- 북극성 지표 L1/L2 Metrics
    1. 북극성지표: 기업이 가장 중요히 생각하는 지표 ex) 에어비엔비- 예약된 박 수, 페이스북 -DAU
    2. L1/L2 지표
        (1) L1 : 전사적인, 고수준의 성과지표 (전략 관점)
        (2) L2 : L1 지표의 세부 사항을 측정하는 지표 (전술 관점)
- AARRR 해적지표 
    1. Acquisition 
        - 사용자의 유입을 의미한다. 신규고객을 어떻게 획득하고 어디서, 어떤 채널로 획득할 지 고민해야한다. 대체로 조회수, 가입사용자 수 등 지표를 측정한다.
    2. Activation
        - 사용자가 첫 방문해 주욕능을 직접 사용하는 것. 사용자 활성화정도, ㅈ품 사용감, 서비스 체험등을 어떻게 향상시킬 수 있을 지 고민해야 한다. 어떤 사용자를 측정 지표로 사용할지는 제품의 성격에 따라 달라진다. 
    3. Retention
        - 신규 사용자를 장기 사용자로 전환시킨다. 장기고객을 양성하고, 지속적 구매를 유도하는 방법을 고민한다. 측정 지표로 익일유지율, 주차별 유지율, EDM 메일 개설율 등이 있다.
    4. Revenue
        - 사용자가 제품, 서비스를 유료 구매하기로 결정하는 단계. 사용자의 소비를 유도해 매출을 극대화 시키는것이 목표. LTV를 지속적으로 높여 나가고, 고객 유치 비용은 지속적으로 줄여나간다.
    5. Referral
        - 사용자가 타인에게 추천하는 단계. 충성도 높은 사용자를 브랜드 홍보자로 만들고, 잠재고객들에게 추천할 수 있을지 고민해야 한다. 측정 지표로 순추천고객지수, 바이럴지수 등이 있다.
