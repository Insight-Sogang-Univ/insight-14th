# 회귀 기초
 
 ## 머신러닝과 모델링
 # 1. 머신러닝이란 - 인공지능의 한 분야로 모델이라는 소프트웨어를 학습하는 과정이다. 그리고 그 목적은 새로운 데이터로 예측을 하기 위함이다!
 (1) 머신러닝의 종류는 3가지로 나뉜다
    1. 지도학습 : 데이터셋에 이미 정답이 있어 모델의 예측값과 정답이 같아지도록 학습시키는 방식이다.
    2. 비지도학습 : 정답을 쥐어주지 않고 데이터의 패턴이나 유사도를 기계가 학습하도록 한다.
    3. 강화학습 : 시행착오를 반복하며 정답을 찾는 과정이다.
    *지도학습에 대해 더 자세히 설명해보자면, 
        * 지도학습에는 regression과 classfication 두가지가 있는데 각자는 데이터셋의 형태에 따라 다르게 활용된다. 전자의 경우 연속적인 숫자값을 예측할 때 사용되고, 후자의 경우 범주형의 데이터셋이 결과값일 때 사용한다. 

 ## 회귀와 모델링
 # 1. 회귀란
    (1) 회귀는 데이터에서 패턴을 찾아내어 미래값을 예측하는 것이다. 집값, 주식가격, 피해금액등이 그 예시가 될 수 있다.

 # 2. 회귀의 종류
    (1) 선형회귀, 비선형회귀, 릿지회귀, 라쏘회귀, 다항회귀등이 있으며 선형회귀, 릿지, 라쏘, 다항회귀는 이번 세션에서 다루나 다항회귀를 제외한 비선형회귀는 이번세션에서 다루지 않는다. 따라서 추가 정보를 정리하자면 비선형 회귀에는 로그회귀, 로지스틱 회귀, local회귀등이 있다. (2)(2) 로그회귀의 경우 종속변수나 독립변수에 로그변환을 적용하여 관계를 분석하는 모형이다.
    (3) 로지스틱회귀분석은 종속변수가 범주형일 경우에만 사용되고, 사건발생가능성을 예측해서 회귀결과를 0과 1 사이의 확률값으로 변환한다.

 # 3. 선형 회귀
    (1) 선형회귀란 데이터를 가장 잘 설명하는 1차함수 회귀선을 찾는 과정이다. 당연히 현재 데이터를 기준으로 그린 직선이므로 현 데이터의 요약본이자 미래데이터의 예측도 가능하다.
    (2) y값인 결과가 종속변수, x값인 원인이 독립변수라고 한다.
    (3) 회귀선을 그리는 핵심적인 방식은 바로 잔차의 크기를 줄이는것이다!
        1. 최소제곱법
            * 잔차들은 음수가 나올수도 있으므로 이를 제곱으로 바꿔 양수화한다. 1변수 2차함수일 경우 최소값을 구하기 쉽다.
        2. 경사하강법
            * 세상에는 선형회귀만 있는 것이 아니다. 자연에는 오히려 비선형 울퉁불퉁 회귀선들이 더 많다. 당연히 잔차 최소화도 어렵게된다. 이럴 때 필요한것이 경사하강법이다!
            * 목적함수 : 최적화를 위해 오차를 최소화하는것이 목적인 함수
            * 손실함수 : 개별데이터 샘플에대한 오차를 측정하는 함수(예측값과 실제값의 차이)
            * 비용함수 : 전체 데이터셋에서 평균적인 손실을 측정하는 함수. 머신러닝과정에서는 비용함수는 목적함수로 봐도 무방하다.
            * 비용함수를 미분하면 미분계수가 나올 것이다. 이때의 미분계수의 음수 양수 판정으로 최소값로의 방향을 정할 수 있다. 이렇게 값을 업데이트하는 과정이 경사를 따라 내려가는것 같다고 하며 최소값에 도달하면 미분계수가 0이 될 것이다. 
            * 업데이트한다고 했는데 얼마나 자주, 크게 업데이트할지를 정하는것이 '학습률'이다.
            * 또한 만약 다변수함수나 비선형함수일 경우 봉우리가 1개가 아닐수도 있다! 이럴때 빠지는 문제가 local minima이다. 이를 golbal minima로 옮기려면 모멘텀이 필요하다!
            * 모멘텀은 학습에 해당 포인트의 미분계수와 모멘텀을 추가해 총 학습량을 정하도록 한다. 모멘텀은 관성이 붙은것처럼 작용해서 작은 언덕즘은 통과할 수 있는 장점이 있다. 
    (4) 다중선형회귀분석
        1. 독립변수가 2개이상이면 다중선형회귀분석이라고 한다.
        2. 다중공선성은 다중회귀에서 중요하게 체크해야한다
            * 아무래도 독립변수가 다양할때, 서로 비슷하게 움직이며 영향력이 큰 변수들도 있을 것이다. 예를 들어 집의 크기와 방의 수는 매우 비슷하게 움직일 것이다.
            * 체크방법으로는 기본적인 상관계수, 히트맵 등이 있고, VIF지수가 있다. 
            * VIF지수는 회귀모델 결정계수를 사용해 계산된다.
            * 체크를 했다면 대응해야하는데 대응방법으로는 변수제거, 변수변환, 규제선형모델사용, PCA 등이 있다. 규제선형모델은 아래서 다루며 pca에 대해 더 알아보자
            * pca는 데이터 차원을 축소하는데 사용되는 통계기법이며 중요정보는 보존하면서 복잡성을 줄이고자 한다. 차원이 높으면 계산이 어렵기 때문이다. 예를들어 5변수라 5차원이었다면, 이를 상관있는것들끼리 몇 개씩 묶어 2차원으로 바꿀 수 있다. 


 # 4. 규제선형모델
    (1) 규제선형모델은 다중선형회귀에서 다중공선성을 줄이기 위한 방법으로 등장했다.
    (2) 기존 선형모델에서는 제곱을 사용하다보니 회귀계수가 쉽게 커졌고 과적합이 발생했다. 이는 특정 독립변수에 과하게 의존한다는 뜻이다.
    (3) 릿지회귀
        1. L2규제가 들어간 회귀분석이고 W의 제곱에 패널티를 부여하는 방식이다. 
        2. 알파값으로 규제를 적용하며 제곱에 알파가 붙어서 반영되므로 제곱값이 클수록 크게 뚜드려맞는 방식이다.
    (4) 라쏘 회귀
        1. L1 규제가 들어간 회귀분석이고 W의 절댓값에 패널티를 부여하는 방식이다
        2. 잔차에 절대값을 씌워서 알파를 곱하기 때문에 알파가 크다고해서 더 크게 뚜드려맞는 릿지와는 다르다. 모두가 똑같은 크기의 알파만큼 뚜드려맞는다. 
        3. 작은기둥들은 아예 없어지기도 해서 변수선택가능 기능이 생긴다.
    (5) 엘라스틱넷 회귀
        1. L2,L1 규제가 결합되어 있다. 
        2. 불필요한변수는 제거하고, 남은 변수들을 적절한 크기로 조절해줄 수 있다. 

 # 5. 모델평가방법
    (1) 성능평가지표
        1. 평균제곱오차(mse)
            * mse는 모델예측값과 실제관측값 사이의 오차제곱의 평균을 의미한다.
        2. 평균절대오차(mae)
            * mae는 모델예측값과 실제관측값 사이를 절대값으로 평균한다.
        3. 이둘의 차이로 mse는 제곱값이므로 이상치에 영향을 크게 받는다. 이상치의 영향력을 분석할때는 mse를 쓰는것이 좋고, 혹시모를 이상치조차 영향받기 싫을때 mae를 쓰면 좋을 것 같다. 
    (2) 변수 유의성 평가
        1. t 검정
            * 이전 통계세션에서 배웠으나 t값계산은 안배웠다 하지만 이번에 등장한다
            * 베타햇은 추정된 회귀계수이고, SE(베타헷)은 표준오차일때, t값은 (추정회귀계수-0)/표준오차 로 나타낼 수 있다.
