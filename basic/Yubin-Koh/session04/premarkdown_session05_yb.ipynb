### Review: 회귀
* 회귀란? 
0. 정의: 연속적인 데이터에서 패턴(함수관계)을 찾아내는 통계적 방법 
1. 종류: 선형회귀, 비선형 화귀, 다향회귀 등
2. 활용: 데이터 예측, 요약, 시계열 등 변수 간 인과관계 발견 등
3. 평가지표: MSE, MAE, R-square, Adjusted R-square, AIC, BC(SC)

* 종류
0. 단순선형회귀분석: 하나의 독립변수만 존재
1. 다중선형회귀분석: 독립변수가 2개 이상 존재

* **최소제곱법 = 최소자승법 (OLS)**
0. 예측값 - 관측된 데이터값 = 잔차인데, 이 잔차 제곱의 합을 최소화하는 회귀선을 찾는 방법

### 다중선형회귀의 수식적 이해
* **y(종속변수)=b0(설명 변수가 모두 0일때 값, y절편)+ b1x1(회귀계수*설명변수)+b2x2 +...+error**
1. 선형회귀식은 결국 종속 변수 y에 대한 설명변수의 가중평균이라고 할 수 있음
2. 설명변수에 가중치를 곱해 더함으로써 y를 얼마나 잘 설명하는지를 나타냄
3. **그러나 모든 상황에 선형회귀를 쓸 수 있는 것은 아니다.**

### 다중선형회귀의 기본가정
1. 선형성
* 종속변수와 설명변수간의 관계가 선형적이어야 함
2. 독립성
* 설명변수끼리 서로 선형 독립적이어야 함 
* **그렇지 않으면 다중공선성의 문제가 발생** -> 상관계수나 VIF 구해서 변수 제거 등을 함!
3. 오차항의 평균은 0이다.
* **오차향=실제 값과 예측 값 사이의 차이**
* why? :  random 해야 함. 즉, 평균적으로 어느 방향에 치우치지 않아야 함. 오차항의 평균이 0이 아니라면 모델이 데이터의 패턴을 올바르게 반영하고 있다고 보기 어려움.
4. 등분산성
* 오차항의 분산이 일정해야 함
* why?: 모델 예측의 정확도가 일정하지 못하고, 특정 부분에서는 매우 정확하고, 특정 부분에서는 매우 정확하지 않은 상태가 됨. 모델 예측이 전체 데이터에서 일정한 정확도를 보여야 함!!
* how?: 잔차의 도표화, 검정
5. *오차항은 자기상관되어 있지 않다* (잘 이해 안됐으니 세션 열심히 듣기!!!)
* 오차항의 공분산은 항상 0이어야 한다. **공분산=두변수가 양/음/상관x 중 어떻게 변하는지 나타냄**
* why? : 0이 아니면 **자기상관(=한 변수의 현재값이 과거 값과 상관관계를 가지는 현상** 이 있음
* how?: to be continue... 더빈-왓슨검정 이나 ACF,PACF 함수 찍어보기 등이 있다네요!
6. 정규성
* 오차항이 정규분포를 따른다는 가정! 그러나, 이는 위배되어도 다중선형회귀분석의 결과에 큰 영향을 주지 않음.
* 그러나 p-value의 기본 가정이라... p-value가 잘못 계산될 수도 있다!! ㅠㅠ
* how?: 샤피로-윌크 검정, 자퀴-베라 검정, Q-Q plot 찍어보기

*회귀분석 후에는 모든 가정을 검증 가능하지만, 1,2,5만 회귀분석 전에 검증이 가능하다. 따라서!! 회귀 분석 전에 모든 가정을 검증해야 하는 것은 아니다!!*

### 회귀분석 평가방법
* **회귀선을 데이터와 함께 시각화하거나 통계지표를 이용하면 된다!**
1. 회귀선 : 기울기 차이가 크지 않으면, 객관적으로 비교하기 어렵다
2. 통계지표
* **모델의 유의성 검정**
1. F-검정 : 전체 모델 검정|회귀 모델이 유의미한지 검정| P<0.05이면 유의미
* HO: 모든 회귀계수가 0이다. = 독립 변수가 종속변수들과 관계가 없다.
* F값이 클수록, 통계적으로 유의미하다는 것을 의미
2. T-검정 : 개별 변수 검정|특정 독립 변수가 유의미한지 검정| P<0.05이면 유의미
* HO: 해당 변수의 회귀계수가 0이다. = 이 변수가 종속변수와 관계가 없다.
* **모델의 성능 평가**
1. 결정계수 = R-squared = 데이터의 평균을 예측값으로 사용할 때보다, 모델을 사용할 때 얼마나 더 정확한 예측인지 나타냄
* 1에 가까울 수록 모델이 데이터를 잘 설명한다고 할 수 있음
2. Adjusted R-squared
* R-squared는 독립 변수의 개수가 증가하면 자연스럽게 증가함!
* 따라서 변수 개수에 덜 민감하도록 하는 지표
3. AIC, BIC : 정보 기준이라 불리며, 값이 낮을수록 좋다고 평가함

## 비선형 회귀
1. 다항식 회귀모델 
* 독립변수에 거듭제곱 항을 추가하여 데이터의 비선형적인 특성을 모델링함.
* 장점: 선형 모델보다 회귀선을 잘 fit 하도록 그릴 수 있음
* 단점: 너므 많은 독립 변수를 이용하면, 결과가 overfitting ! -> 이를 해소하기 위해 릿지, 라쏘 규제 이용해도 됨!
* **적합한 상황: 관계가 곡선을 그릴 때 | 성장률, 감소율 등 시간에 따라 변화하는 속도가 다른 경우**

2. 지수 회귀모델
* 종속 변수가 지수적으로 변화하는 관계를 모델링할 때 사용
* 양변에 로그를 취한 뒤 선형 회귀를 적용하면 된다. 다만, lnY가 되므로, 원래 데이터 형태로 복구하려면, 예측 값에 지수함수를 적용시킴

3. 로그 회귀모델
* 종속변수와 독립변수의 관계가 로그함수로 더 잘 표현될 때 사용한다.
* x에 로그연산을 취한 뒤, 나머지는 선형 회귀와 동일하게 사용
* **지수회귀모델과 달리 예측 값인 y에 변함이 없으므로, 그대로 사용한다!**

4. 스플라인 회귀
* 데이터를 구간별로 나누고 각 구간에서 서로 다른 선형 또는 비선형 함수를 적용하여 예측하는 방법
* *구간경계에서의 연속성을 유지하는 것이 중요하다*
