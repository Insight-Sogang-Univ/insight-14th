# 데이터로 보는 비즈니스 상황

## 데이터 기반 의사결정

### 무엇을 위한 데이터 분석인가?
: 데이터 분석이 지닌 가장 큰 특징은 **목적 달성 프로세스의 질**이다. 데이터에 근거한 객관적인 지식을 활용해 '지금 해야 할 액션에 집중'함으로써 개별 프로세스의 질을 높인다. <br>
직감과 경험치를 항상 옳다고 하기 어렵기 때문에, '데이터 분석'으로 객관성을 확보하고 집중하게 해준다.

#### 액션이란 무엇인가?
1. 결정: 언제, 어떤, 얼마나 같은 결정을 뒷받침하기 위한 데이터 분석 수행
2. 계획: '프로모션을 계획한다'는 액션을 위해서는 수치적 근거에 기반한 계획 수립을 위한 데이터 분석 수행
3. 행동: 판매 예측 모델을 위한 시스템을 개발하고, 전단지 배포등 일별 계획 실행에 활용할 수 있는 데이터 분석 실행

* 세 가지의 공통점은 **데이터 분석을 통해 *객관적으로 과거를 되돌아보고 미래를 전망하는 것*** <br>
과거를 제대로 직시하지 못하면 현재 위치가 불분명해지고, 미래를 적절히 전망하지 않으면 잘못된 방향으로 나아가게 됨.

### Data-Driven-Decision-Making (DDDM)
: **비즈니스 의사 결정을 제공하기 위해 직관 대신 데이터와 분석을 사용하는 것을 강조하는 접근 방식**
* 고객 피드백, 시장 동향 및 재무 데이터와 같은 데이터 소스를 활용하여 의사 결정 프로세스를 안내하는 것 포함

### 6 Steps of DDDM
1. 목표 정의
* 목표가 정확히 정의되어야 집중적이고 목적이 명확한 노력 가능
2. 데이터 식별, 준비 및 수집
* 명확한 목표를 설정하고 데이터 요구 사항을 판별하며 데이터 소스를 평가 및 준비 후 데이터 수집 및 검증
3. 구성 및 탐색
* 데이터는 새로운 패턴, 추세 등 가치 있는 인사이트를 발견하도록 구조화
* 데이터를 정리하면 정확성과 신뢰성이 보호됨
* 데이터를 시각화하면 원시데이터에서 볼 수 없는 것들을 식별 가능
4. 데이터 분석 수행
* 다양한 기술과 방법론을 사용해 원시 데이터를 실행 가능한 인사이트로 변환해 비즈니스 전략에 영향을 미치는 패턴, 상관관계 및 추세를 파악
* 데이터 분석을 수행함으로써 전략적 결정을 내리고 전반적인 성과를 개선
5. 결론 도출
* 주요 데이터 분석 결과 검토 ㅣㅊ 올바른 비즈니스 상황에 배치해 실행 가능한 인사이트와 권장 사항을 구성
* 실용적인 결론 -> 비즈니스 성공을 이끄는 의미있는 조치
6. 구현 및 평가
* 인사이트를 검증라고 결과를 측정하는 데에 중요함
* 데이터 기반 권장 사항을 기반으로 실행 계획 개발, 리소스 할당, 이니셔티브의 진행상황을 지속적으로 모니터링
* 핵심성과지표를 기준으로 결과 측정, 분석, 피드백 수집 등 반복적인 개선 주기를 통해 조정 수행

### 데이터 분석 유형
1. 설명 분석 : 데이터 취합 및 마이닝을 통해 과거 데이터를 설명 및 요약해 과거 성과에 대한 인사이트 제공
2. 진단 분석 : 특정 이벤트가 발생한 이유 파악에 중점을 둠. 데이터 디스커버리, 마이닝 및 상관관계 식별이 포함되며 추세 또는 사고의 근본 원인을 파악
3. 예측 분석 : 과거 데이터를 기반으로 미래의 추세나 결과를 예측 + 위험 관리
4. 처방 분석 : 하나 더 나아가 데이터에 기반하여 조치를 권장함. 최상의 행동방침을 제안 <br>
<br>
추가로, 탐색 분석, 추론 분석, 정성적 분석, 정량적 분석, 실시간 분석 등이 있다.

## 목표 정의

### 지표
: 방향이나 목적, 기준 따위를 나타내는 표지. 추상적인 개념과 관련되는 지수, 척도 등을 말하며 측정하거 싶은 것을 숫자로 표현한 것.
<br>
대개 분모 대비 분자로 표현
1. 효율성 : 산출/투입
2. 효과성 : 달성 결과 / 목표
3. 수익성 : 이익 / 자원
4. 성장성 : 증가분 / 기존 규모
5. 안정성 : 보유자산 / 부채
<br>
물론 예외적인 구조도 있다...!

| 기능      | 효율성 지표 (Output ÷ Input)                | 효과성 지표 (Actual ÷ Target)                  | 수익성/가치 지표 (Profit ÷ Resource)                             | 성장성/안정성 지표         |
|-----------|---------------------------------------------|------------------------------------------------|------------------------------------------------------------------|-----------------------------|
| **재무**   | 자산회전율 = 매출액 ÷ 총자산                 | 예산 집행률 = 실제 지출 ÷ 계획 지출              | ROE = 순이익 ÷ 자기자본<br>ROA = 순이익 ÷ 자산                    | 매출 성장률, 부채비율       |

### 북극성 지표 & L1, L2 지표
1. 북극성 지표: 기업이 가장 중요하게 생각하는 지표
2. L1 / L2 지표:
    - L1: 전사적인, 고수준의 성과 지표(전략 관점)
    - L2: L1 지표의 세부 사항을 측정하는 지표(전술 관점)

### AARRR
1.  넷플릭스, 유튜브, 스포티파이
    - 컨텐츠 감상 시간이 중요한 지표
    - 오래 감상할수록 좋은 영향이니까!!
2. 우버
    - 공급자와 수요자를 매칭하는 서비스
    - 실제로 어느정도 매칭되는지가 중요함
    
**프로덕트가 가치를 제공하는 방식이 다르면, 측정하는 지표도 달라져야 함**

**AARRR 모델 또는 해적지표는 사용자 증가를 위한 5가지 로드맵을 구성**
: 사용자 운영 과정에서 자주 활용하는 데이터 분석 모델로
1. Acquisition (획득)
* 사용자가 여러 채널을 통해 여러분의 앱 혹은 웹사이트에 방문하는 것을 의미
* 신규 고객을 어떻게 획득하지? 어디로부터 획득하지? 어떤 채널을 통해 획득하지? 가 주고민
* 보통은 조회수, 가입 사용자 수 등 지표를 통해 측정
2. Activation (활성화)
* 앱 또는 웹사이트에 처음 방문하여 주요 기능을 직접 사용하는 것을 의미
* 어떻게 측정할까? 제품 성격에 따라 다양한 기준으로 측정
* 주로 4단계로 구성
    - Sign-up: 유저가 제품에 들어온 시점
    - Setup Moment: 유저가 제품의 핵심가치를 경험할 준비 완료
    - A-HA! Moment: 유저가 핵심가치를 처음 경험한 순간
    - Habit Moment: 유저가 제품 사용에 대한 습관 형성
3.  Retention
* 신규 사용자를 장기 사용자를 어떻게 전환시킬까?가 질문
* 신규 사용자 획득 비용 > 기존 사용자 유지 비용
* 따라서 어떻게 하면 고객의 지속적인 구매 행위를 증가시킬 수 있는지가 주요 고민
4. Revenue (수익)
* 사용자가 유료 구매를 결정한 단계로, 소비를 유도해 매출을 높이는데 총력을 기울여야 함
* 이 지표도 2번과 같이 제품 성격에 따라 측정 지표가 달라질 수 있음
* 중요: **고객의 평생 가치(LTV)를 지속적으로 높여 나가고, 고객 유치 비용은 계속해서 줄여나가야 한다는 것**
* 지표 예시
    - PU(Paying User): 전체 사용자 중에 결제를 한 사용자 비율
    - Average Check: 구매 1회당 평균 금액 (Revenue/Transaction)
    - ARPU(Average Revenue Per User): 인당 평균 매출
    - ARPPU(Average Revenue Per Paying User): 결제자당 평균 매출
    - LTV(Life Time Value): 유저가 진입부터 이탈까지 누적 수익
5. Referral
* 제품이 마음에 들었다면 타인에게 추천할 것임! 이처럼 **충성도 높은 사용자를 브랜드 홍보자로 만드는 것**이 매우 중요함! 
* 측정지표: ‘순 추천고객 지수(NPS)’, ‘바이럴 지수(Viral Coefficient)' = 브랜드에 대한 고객의 만족도와 사용자가 평균적으로 추천하는 고객의 수
* 지표 예시
    - 바이럴 계수: 기존 고객 한명당 데리고 온 새로운 고객의 수
        - 바이럴 계수 > 1,  Acquisition 단계 없이도 서비스 성장
    - 바이럴 주기: 기존 고객 1명이 새로운 고객을 데려오는데 걸리는 시간
