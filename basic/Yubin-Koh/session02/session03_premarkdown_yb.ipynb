# 통계

### 기술통계
1. 기술통계란?: 수치와 시각화를 통해 샘플을 요약하여 표현하는 방법론
2. 주요 도구: 중심화경향, 자료의 산포, 자료의 정리 및 시각화
3. 중심화 경향: 평균, 중앙값, 최빈값 등 자료의 중심적인 경향을 나타냄
4. 자료의 산포: 범위, 분산, 표준편차, 사분위수, 변동계수를 이용해서 데이터가 퍼져있는 정도를 설명함
* 변동계수: 데이터의 절대적인 변동성이 아닌 상대적인 변동성을 나타내는 통계적인 지표

### 추론통계
1. 추론통계란?: 추정과 검정을 통해 샘플로부터 모집단을 추론하는 방법론

### 자료의 척도
1. 명목척도: label으로 요소의 속성을 식별하려고 하는 척도
2. 서열척도: 명목형에 더해 순서, 순위를 나타내는 특성을 지니는 척도
3. 구간척도: 값 사이 순서가 있고, 그 간격이 일정한 척도
4. 비율척도: 절대영점과 임의의 측정단위로 구성하고, 동간성이 유지되는 척도

### 확률과 분포
1. 표본의 불확실성, 모수추정, 추정의 현실성, 구간추정의 오류관리, 적극적 주장 검증, 가설검정의 오류 관리 등 끝없는 의심과 도전을 하며 추론 통계를 해나가야 함
2. 이산확률분포: 확률변수가 이산적인 값을 가지는 경우
3. 연속확률분포: 확률변수가 연속적인 값을 가지는 경우

### 정규분포
1. 가장 잘 알려진 연속확률분포로, 평균을 기준으로 대칭하는 그래프, 표준편차가 클수록 곡선이 팽팽해짐
2. 표본 크기가 커질 수록 모평균에 근사해짐. 즉 통계적 정확도가 올라감
3. 중심극한정리 = 동일한 확률분포를 가진 독립확률 변수로, n이 적당히 크면 정규분포에 가까워진다는 뜻

### T분포 
1. df(자유도)가 커짐에 따라 표준정규분포에 수렴한다. 표본으로 모평균(혹은 모평균 간의 차이)을 추정할 때 사용되는 분포이다.
2. t분포는 모집단의 분산을 모를 때 불확실성이 커자고, 따라서 표준 정규분포보다 중심으로부터 더 퍼져있을 가능성이 크다.
3. 쉽게 말해 n이 30 미만이면 t분포, 30이상이면 표준정규분포를 이용한다.

### 카이제곱 분포
1. 독립적인 표준 정규 분포를 따르는 변수들의 제곱 합
즉, 표본분산의 분포와 관련된 지표이다.
이것도 t분포처럼 df(자유도)를 이용한다는 것을 알 수 있다.

### F분포
1. 두 독립적인 카이제곱분포 비율을 보여주며, 주로 두 집단 간의 분산 차이를 비교하는 데에 사용된다.
2. 오른쪽으로 긴 꼬리를 가진 비대칭 분포를 하며, df가 커질 수록 중심이 오른쪽으로 이동하며 정규분포에 가까워진다.

### 추론 통계란?
1. 추정: 모집단을 대표할 수 있는 표본을 추출하여 분석한 것으로 모수를 예측하는 과정
* 모수란? 모집단의 특성에 관계되는 평균, 비율, 표준편차
+ 오차가 있을 수 밖에 없기 때문에 어느정도 오차를 포함하는 신뢰구간 추정 방법이 일반적으로 널리 사용된다.
2. 점추정: 어떤 모수를 표본에 기초하여 계산된 하나의 값으로 추정하는 것
3. 구간추정: 모수의 참 값이 포함되리라고 기대하는 추정치를 일정한 범위로 신뢰도를 가지고 나타내는 방법

### 추론통계 - 검정 (가설검정과정)
1. 가설 검정을 하기 위해선 귀무가설과 대립가설을 설정
* 귀무가설이 주로 기각하고자 하는 명제이고, 대립가설이 내가 주장하고자 하는 명제
2. 유의수준 설정 (보통 0.05 혹은 0.01 중에 선택)
* 1종 오류 : 귀무가설이 실제로 참이었는데 기각할 확률 !!!! <유의수준의 확률>
* 2종 오류 : 귀무가설이 실제로 거짓인데 기각하지 못할 확률
3. 검정통계량 산출
z값, t값, f값이 있음
4. 가설 기각/ 체택 판단 -> 유의확률이 유의수준 보다 작으면 귀무가설을 기각하면서 대립가설을 체택한다.

### 데이터 자료형에 따른 검정 방법
1. 수치형-수치형 : 피어슨 상관분석
* 공분산, 상관계수 사용. 다만 상관계수는 선형관계만 파악할 수 있음
2. 범주형 - 수치형 : F-test, T-test
* 범주가 2개이면 T-test, 3개 이상이면 F-test를 사용한다.
3. 범주형- 범주형 : 카이제곱테스트
* 카이제곱 통계량으로 함. 자유도의 약 2배가량을 기준으로 잡고 계산
