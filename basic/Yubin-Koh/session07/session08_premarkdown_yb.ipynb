# 추천 시스템 기초

## 추천 시스템이란?
: 사용자의 정보 데이터를 분석하여, 개인의 취향에 맞는 아이템을 추천하는 알고리즘 <br>
고객 입장에서는 원하는 콘텐츠를 얻을 수 있어 **만족도 증가**를, 기업은 **매출 증가**를 기대할 수 있다

### 사용되는 데이터의 종류와 특징
1. **명시적 데이터** : 고객들에 의해 입력된 데이터. 사용도를 직접적으로 표현한 데이터. <br>
사용자의 (한 때) 취향을 정확하게 알 수 있지만, 데이터 자체를 얻기 어려움
2. **암시적 데이터** : 고객들의 행동을 기반으로 수집된 데이터. 사용자의 선호도가 간접적으로 표현된 데이터. <br>
데이터 얻기 굉장히 쉽지만, 사용자의 선호를 직접적으로 알 수 없다.

### 추천시스템의 분류
1. 내용 기반 필터링 <br>
* 콘텐츠의 **유사도**를 기반으로 사용자에게 추천하는 방법
* 사용자가 선택한 아이템과 유사한 아이템을 추천하는 방식
* **행동 이력**을 기준으로 **유사도 계산** 해서 추천
2. 협업 필터링 <br>
* 서비스 내에 있는 **다수의 사용자로부터 얻은 선호도 정보**를 기반으로 사용자에게 추천하는 방법 
* 두 명의 사용자가 비슷한 관심사를 가지고 있다면, 한 사용자의 데이터를 바탕으로 다른 사용자에게 추천해주는 방식 
* **여러 사용자의 선호도** 활용해서 추천
    1. 사용자 기반: 유사한 취향의 사용자가 선호한 아이템 추천
    2. 아이템 기반: 사용자가 선호한 아이템과 유사한 아이템을 추천
3. 하이브리드 필터링
* 2가지 이상의 다양한 종류의 **추천 알고리즘을 조합**하여 만들어진 알고리즘

## 내용 기반 필터링
: 사용자 프로필과 아이템 특징과의 일치도를 계산함으로써, 사용자가 선호할 만한 아이템을 추천해주는 알고리즘 
* 콘텐츠의 **유사도**를 기반으로 사용자에게 추천해주는 방식
* cf. 사용자 프로필 = 사용자가 이전에 좋아했던 항목들의 특징을 모아 놓은 데이터

### 유사도
1. 영상, 이미지, 텍스트 등의 형태로 존재하는 아이템 정보 데이터를 벡터화한 이후 **아이템 벡터간의 유사도를 측정**
2. 측정된 유사도를 바탕으로 **고객이 선호하는 아이템과 비교하여, 유사도가 높은 아이템을 선별**
<br>
유사도 측정 방식: 거리 기반 추천 < 각도 기반 추천
    1. 코사인 유사도: 두 점 사이의 각도 측정, 값이 작으면 두 데이터가 가까이 = 유사하다!
    2. 유클리드 거리: 두 점 사이의 거리를 측정한 지표로, 값이 작으면 두 데이터가 가까이 = 유사하다!

### 사용자 프로필
1. 직접 지정형: 사용자에게 자신이 선호하는 아이템의 특징을 명시적으로 지정하도록 하는 방식
2. 간접 지정형: 사용자의 과거 행동 이력에 기반하여 사용자의 프로필을 작성하는 방법
<br>
* 사용자의 프로필을 획득하는 방식
    1. 특성 가중치= 사용자의 아이템 트성 중 아이템 특성 가중치의 평균 값
    2. 사용자의 기호 따로 파악하거나, 설문조사 등으로 프로필 획득
    3. 아이템에 대한 유저의 기존 평가를 바탕으로 분류 문제를 해결 <br>
    => 좋은 사용자 프로필을 구성할 수록 정확도 업! 만족도 업! 
    <br> **명시적+암묵적을 혼합하여 사용**하는 것이 일반적

### 아이템 특정
* 아이템을 벡터 형태로 표현, **도메인에 따라 다른 방법 적용**
* 콘텐츠의 내용을 분석하는 아이템 분석 알고리즘
    1. 군집화
    2. 머신러닝
    3. TF-IDF : **단어빈도**=특정 문서 내에 특정 단어가 얼마나 자주 등장하는가? + **역문서 빈도** = 전체 문서에서 특정 단어가 얼마나 자주 등장하는가? 를 통해서 **다른 문서에서는 등장하지 않지만 특정 문서에서만 자주 등장하는 단어**를 찾아서 **문서 내 단어의 가중치**를 계산하는 방법 <br>
    => 각 단어가 그 문서에서 얼마나 중요한 지를 나타내는 값, 값이 높을 수록 특정 문서에서 더 중요한 역할을 함!

### 장단점
* 장점
1. 아이템에 대한 콜드 스타트(새로운 유저들에게 적절한 정보 제공이 어려움) 문제 해결 가능
2. 아이템에 피터를 통해서 추천하기에 추천이 잘된 이유 설명 용이함
* 단점
1. 아이템 피처 추출 해야하고, 정확하게 추출 못하면 정확도가 떨어짐
2. 기존 아이템과 유사한 것으로만 추천하기에 새로운 장르의 아이템 추천이 어려움

## 협업 필터링 
* 가정: 과거에 비슷한 선호도를 가진 사용자들은 앞으로도 비슷한 선호도를 가질 가능성이 높다.
* 기존 사용자들과 아이템의 상관 관계를 기반으로 새로운 사용자와 아이템의 상관 관계를 예측
* 아이템에 대한 **여러 사용자의 평점**을 기반으로 추천이 이루어짐

### 메모리 기반 협업 필터링 = 이웃 기반 협업 필터링
* 사용자 또는 아이템 간 유사도를 기반으로 함
* **비슷한 취향을 가진 사용자가 선호하는 아이템을 추천**
* 접근법이 두 가지임!  
    1. 사용자 기반: 특정 사용자와 타 사용자 간의 유사도를 측정한 뒤 가장 유사도가 높은 tOP-1 사용자를 추출해 그들이 선호하는 아이템을 추천
    * 유저 간의 유사도가 높을 수록 가중치를 부여 | **사용자의 평점**을 이용 <br>
    다양성이 있음 | 사용자 유입이 적을 때 유리
    2. 아이템 기반: 이 상품을 선택한 다른 고객들은 다음 상품도 구매했습니다~!
    * 타겟 아이템과 유사한 아이템의 평점을 이용 | **아이템의 평점**을 이용 <br>
    계산이 빠름 | 사용자 유입이 많은 환경에서 유리
* 메모리 기반 협업 필터링의 한계
    * 확장성이 떨어짐
    * 데이터가 적으면 성능이 저하됨
    * 유저가 많아지면 knn cf 활용!!

### 모델 기반 협업 필터링 = 잠재 요인 협업 필터링
: **사용자-아이템 평점 매트릭스** 속에 숨어있는 잠재요인을 추출해 **추천 예측** 하는 기법! <br>
메모리 기반 협업 필터링은 숨겨진 선호 요인이나 아이템의 특성을 반영하지 못하고, **유사도에만 의존**하기 때문에, 결측이 많거나 복잡한 환경에서 정확도가 떨어진다. <br>
<br>
=> 모델 기반 협업 필터링은 기계학습모델을 통해 **관측 데이터로부터 숨겨진 신호 요인 = 잠재요인을 수치적으로 학습 -> 백터 계산으로 평점 예측**  <br>
<br>
    1. 잠재요인 : 사용자와 아이템 관계를 설명해주는 비가시적 요인 = 관측된 데이터를 바탕으로 수치적으로 추정하는 값<br>
        차원의 수가 정해지지 않은 모르는 차원에 표현될 수 있고, 차원은 여러 개 일 수 있다! **잠재적 특성을 파악** 하는 것이므로, 아이템을 표현한 facter는 사람이 명시적으로 이해할 수 없어서 맵핑 등을 한다!<br>
    2. 행렬 분해 : 유저 취향 + 아이템의 특성이 녹아있지만, 그 결과값만 볼 수 있다.<br>

## 추천 시스템 평가
1. 오프라인 평가 <br>
사용자의 행동 이력에서 찾은 과거의 로그를 사용하여 모델의 예측 정밀도를 평가
2. 온라인 평가 <br>
추천 모델이나 새로운 사용자 인터페이스를 일부 사용자에게 표시하면서 평가를 수행
3. 사용자 스터디 <br>
사용자에게 직접 **인터뷰나 설문**으로 데이터를 얻는 것

### 최근 추천 시스템 연구 동향

# 연관 분석 기초

## 연관분석이란?

### 연관분석의 정의
: 상품이나 서비스를 구매하는 등 일련의 사건 안에 있는 **항목 간의 일정한 *연관규칙*을 발견하는 분석** 
* 굉장히 다양한 분야에 사용 가능함
* 비지도 학습에 해당함

### 연관규칙 분석
1. 연관 규칙의 형태
A을 구매하였을 때, B 또한 구매할 것이다 같이 if -then 형식을 띄고, A -> B 으로 표현!!
    1. 아이템 집합 : if 절
    2. 결과절: then 부분
    3. 아이템 절: 1,2 제외 연결 부분
    * 아이템 집합엔 여러 개가 들어가야 하지만, 상호 배반이어야 한다
2. 연관 규칙 주의: 인과관계는 알 수 없다 !!!! **JUST 상관관계 파악**
* 실험이 아닌 관찰 데이터임
* 방향성은 그냥 수학적 비율이지, 인과를 알 수는 없음
3. 연관분석 VS 협업 필터링 (관점에 따라 추천시스템의 하위 개념으로 분류하기도 함)<br>
    = 협업 필터링이 복잡한 수치를 계산할 수 있는 반면, 연관분석은 상품이 포함된 비율만 계산할 수 있지만, **다양한 세트상품을 뽑아낼 수 있음**

## 주요개념

### 적절한 연관규칙의 조건
1. 일정 이상의 **지지도**를 가져야 함!
* 지지도 : 전체 거래 항목 중 품목 A와 B를 동시에 포함하는 거래의 비율 (교집합) 
* 지지도가 낮으면, 현실에서 의미없는 우연일 가능성이 높음
2. 일정 이상의 **신뢰도**를 가져야 함!
* 신뢰도 : 품목 A를 포함하는 거래 수 중 품목 A와 품목 B가 동시에 포함되는 거래의 비율
* 신뢰도가 낮으면 의미없는 연관 규칙이거나, 추천 실패로 이어짐

### 연관 분석의 평가 측도
1. **향상도** : A와 B가 우연히 함께 등장할 확률과 실제 함께 등장할 확률
* 향상도 > 1 : 두 품목이 서로 양의 상관관계
* 향상도 = 1 : 독립
* 향상도 <> 1 : 두 품목이 서로 음의 상관관계

#### 위의 세 지표 모두 클 때에만 효과적이라는 결론을 내릴 수 있음!!

## 연관 분석 알고리즘

### **Apriori 알고리즘**
: **선험적 경험**을 이용한다.
1. 배경: 무차별 탐색은 매우 비효율적 + 상품 개수가 많아질 수록 조합의 수가 너무 크게 증가
2. Apriori 알고리즘이란? 상위 조합에서부터 스캔하면서, 특정 조합이 자주 안생기면, 후속 조합까지 모든 후보에서 배제하는 방식!
3. 장점: 하나의 조합만 검사하고도 파생된 조합까지 다 제거 가능! -> 연산량 매우 낮아짐

### Apriori 알고리즘 프로세스
: 빈발 항목 집합(일정 지지도를 넘는 항목의 조합)을 찾고 -> 단계적으로 조합 생성하고 -> 필터링!
1. 단일 항목 집합 생성<br>
    1. 모든 개별 상품에 대해 지지도 계산  <br>
    2. 최소 지지도 넘는 항목만 넘긴다<br>
2. 2개 항목 집합 생성<br>
    1. 1단계에서 선별된 상품으로 2개 항목 조합 만들기<br>
    2. 이 조합에서 지지도 계산하고, 또 최소 지지도 안넘는 것 배제<br>
3. 더 큰 항목 집합 생성<br>
    1. 3개 이상의 항목으로 구성된 집합 생성<br>
    2. 또 최소 지지도 안넘는 조합 제거<br>
    3. 더이상 조합을 생성할 수 없을 때까지 반복<br>
4. 최종 빈발 항목 집합 선정<br>
* 마지막까지 남은 빈발 항목 집합 바탕으로 **연관 규칙** 생성
* 각 규칙에 대해 지지도, 신뢰도, 향상도를 계산해서 가장 유용한 규칙 선택

### Apriori 알고리즘 장단점
1. 장점: 수많은 상품 연관 구매 패턴 발견 | 원리 간단
2. 단점 : 현실적 중요 연관 규칙 부족 | 데이터가 커질 수록 여전히 느려짐

### FP GROWTH 알고리즘
: Apriori의 연산 속도 문제를 해결할 수 있게 **자료 구조**를 잘 이용한 알고리즘 <br>
FP-growth 알고리즘이란? : 빈발 품목 집합을 효율적으로 찾아냄! by **FP-Tree** <br>
* 모든 거래를 확인해 각 아이템마다 지지도를 계산하고 , 최소 지지도 이상의 아이템만 선택

#### FP-Growth 알고리즘 프로세스
1. 빈발 항목 필터링 : 최소 지지도 이상인 항목만 남김
2. 데이터 재정렬 : 빈도순으로 정리
3. FP-tree 생성: 재정렬된 데이터를 바탕으로 FP-tree를 생성, 각 거래를 노드를 표현하고 동일한 항목을 노드 연결해 빈도 늘리고, 새로운 항목은 새로운 노드로 추가!
4. 빈발 항목 집합 추출 : Fp-트리를 바탕으로 **특정 항목을 기준으로 부분 트리를 만들어 빈발 항목 집합을 추출**함. 
5. 연관 규칙 도출 : 4에서 도출된 것을 바탕으로 연관 규칙을 만듦. 그 후, 지지도 + 신회도 + 향상도를 계산해 가장 유용한 규칙을 선택!

#### FP-Growth 알고리즘의 장단점
1. 장점 : 데이터를 두 번만 스캔해서 Apriori보다 빠름 + 후보 아이템 계산없이 트리만 생성하면 끝!
2. 단점 : 대용량 데이터셋에서 메모리 비효율적 사용 + 설계가 어렵고, 지지도 계산이 오직 FP-tree가 만들어져야만 가능!!
