# **1)데이터 기반 의사결정**  
- 데이터 분석 -> 목적 달성 프로세스의 질 향상  
- 액션: 결정 -> 계획 -> 행동  
  각 단계에 맞는 데이터 분석 수행 
  

### Data Driven Decision Making (DDDM)  
- 비즈니스 의사결정을 제공하기 위해 직관 대신 데이터와 분석을 사용하는 것을 강조하는 접근 방식  
- **DDDM의 단계**  
  
1. 목표 정의:   
   조직 목표 명확하게 표현  
2. 데이터 식별, 준비 및 수립:  
   데이터 요구 사항 판별, 데이터 평가 및 준비 -> 수집 및 검증  
3. 구성 및 탐색:  
   데이터 구조화, 정리 -> 정확성, 신뢰성 보호  시각화를 통해 직관적 해석 가능   
4. 데이터 분석 수행:  
   상관관계 및 추세 파악 -> 전략적 결정  
5. 결론 도출:  
   실행 가능한 인사이트 구성  
6. 구현 및 평가:  
   인사이트 검증 및 결과 측정   

### 데이터 분석 유형  
1. **설명 분석**   
- 데이터 취합 및 마이닝 -> 과거 데이터 설명   ex) 판매 보고서, 만족도 설문 조사 등  
2. **진단 분석**  
- 특정 이벤트가 발생한 이유(원인) 파악  
3. **예측 분석**  
- 과거 데이터 -> 미래 추세 결과 예측    ex) 통계 모델, 머신 러닝 및 예측 기술  
4. **처방 분석**  
- 데이터 기반 조치 (공급망 최적화, 마케팅 캠페인 전략 등)  
5. 탐색 분석  
- 가설 X 패턴, 관계 및 이상 현상 발견  
6. 추론 분석  
- 모집단에 대한 추론   ex) 회귀분석, 가설 검증 및 신뢰구간 -> 시장 조사 연구, 소비자 선호도 조사  
7. 정성적 분석  
- 비수치 데이터 활용   ex) 고객 피드백, 소셜 미디어 정서 분석  
8. 정량적 분석   
- 수치 데이터 -> 변수 정량화    ex) 재무 모델링, 운명 메트릭 분석 및 성과 측정  
9. 실시간 분석   
- 사기 탐지, 재고 관리 -> 실시간 고객 지원 모니터링  


# **2)목표 정의**  

### 지표  
- **지표**: 방향이나 목적, 기준 따위를 나타내는 표지 / 추상적인 개념과 관련되는 지수, 척도 등을 말하며 측정하고 싶은 것을 숫자로 표현한 것   
 $\text{metrics} = \frac{\text{molecule}}{\text{denominator}}$  
 주로 분모 대비 분자로 표현  

|유형|공식 구조|예시 지표|
|---|---|---|
|효율성|산출/투입|노동생산성=생산량/노동시간|
|효과성|달성결과/목표|목표 달성률=실제 매출/목표 매출|
|수익성|이익/자원|ROE=순이익/자기자본|
|성장성|증가분/기존 규모|매출 성장률=(당기매출-전기매출)/전기매출|
|안정성|보유자산/부채|부채비율=부채/자기자본|

*예외)  
- 절대치 지표: 매출액, 직원 수, 생산량  
- 지수형 지표: 고객만족지수(CSI), 순추천지수(NPS)  
- 복합 지표: 균형성과표(BSC)처럼 여러 지표를 묶어서 본 것  

| 기능 | 효율성 지표 (Output ÷ Input) | 효과성 지표 (Actual ÷ Target) | 수익성/가치 지표 (Profit ÷ Resource) | 성장성/안정성 지표 |
|---|---|---|---|---|
| **재무** | 자산회전율 = 매출액 ÷ 총자산 | 예산 집행률 = 실제 지출 ÷ 계획 지출 | ROE = 순이익 ÷ 자기자본<br>ROA = 순이익 ÷ 총자산 | 매출 성장률<br>부채비율 |
| **마케팅** | CAC 효율 = 신규고객 수 ÷ 마케팅비 | 전환율 = 구매자 ÷ 방문자 | 마케팅 ROI = 캠페인 이익 ÷ 캠페인 비용 | 시장점유율 성장률 |
| **HR (인사)** | 노동생산성 = 산출량 ÷ 인력투입 | 교육 효과 달성률 = 실제 역량향상 ÷ 목표 역량향상 | HR ROI = 교육 후 성과 증가 ÷ 교육비 | 이직률<br>직원 유지율 |
| **운영** | 설비가동률 = 실제 가동시간 ÷ 총 가동시간 | 품질목표 달성률 = 합격품 ÷ 목표 품질수준 | 단위당 원가 절감율 = 절감액 ÷ 총원가 | 재고회전율<br>공급망 안정성 |

### 지표 정의   
- Proxy Metric:   
활성화와 인게이지와 같이 추상적이여서 프로덕트마다 나름의 가정과 정의, 논의가 필요한 지표  
절대적인 수치를 알 수 없을 때 대체물(Proxy)을 정하여 측정 -> 정답 존재 X  

### 북극성 지표 L1/L2 Metrics  
- 북극성 지표: 기업이 가장 중요하게 생각하는 지표  
- L1 지표: 전사적인, 고수준의 성과 지표(전략 관점)  
- L2 지표: L1 지표의 세부 사항을 측정하는 지표(전술 관점)   
   

### **AARRR**  
  
- 사용자 운영 과정에서 자주 활용하는 데이터 분석 모델, 사용자 증가를 위한 5가지 로드맵으로 구성  
1. Acquisition  
- 사용자가 여러 채널을 통해 앱 혹은 웹사이트에 방문하는 것  
- 지표 예시  
  - CAC (Customer Acquisition Cost): 유저 획득 비용  
  - CTR (Cost Per Click): 클릭 한 건당 얼마 지출  
  - CPC (Cost Per Install): 인스톨 당 과금되는 상품  
  - CPL (Cost Per Lead): 유저 행동 당 드는 비용  
  - ROAS (Return On Ads Spending): 광고로 인한 매출액/광고비  
2. Activation  
- 사용자가 앱 또는 웹사이트에 첫 방문하여 주요 기능을 직접 사용하는 것  
- 구성  
  - Sign-up: 유저가 제품에 들어온 시점  
  - Setup Moment: 유저가 제품의 핵심가치를 경험할 준비 완료  
  - A-HA! Moment: 유저가 핵심가치를 처음 경험한 순간  
  - Habit Moment: 유저가 제품 사용에 대한 습관 형성  
3. Retention  
- 신규 사용자를 장기 사용자로 전환시키는 것  
- 지표 예시  
  - DN Retention: N일 안으로 재접속하는 경우  
  - MAU/DAU/WAU: 일/주/월마다 방문하는 횟수  
  - Stickness(DAU, MAU): 해당기간 동안 활성화 수 (*점성)  
4. Revenue 
- 사용자가 제품 혹은 서비스를 유료 구매하기로 결정한 단계  
- 지표 예시   
  - PU(Paying User): 전체 사용자 중에 결제를 한 사용자 비율  
  - Average Check: 구매 1회당 평균 금액 (Revenue/Transaction)  
  - ARPU(Average Revenue Per User): 인당 평균 매출  
  - ARPPU(Average Revenue Per Paying User): 결제자당 평균 매출  
  - LTV(Life Time Value): 유저가 진입부터 이탈까지 누적 수익  
5. Referral  
- 사용자가 상품 혹은 서비스 만족 후 타인에게 추천  
- 지표 예시  
  - 바이럴 계수: 기존 고객 한명당 데리고 온 새로운 고객의 수  
  - 바이럴 주기: 기존 고객 1명이 새로운 고객을 데려오는데 걸리는 시간
