
# **3) 데이터 확인**
## 3-1) head(), tail()
``` python
df.head(n): 앞 n개의 행 반환  
df.tail(n): 뒤 n개의 행 반환
```
## 3-2) 데이터 정보 확인
``` python
df.shape: (행 개수,열 개수) 반환
df.info(): 여러 정보들을 종합적으로 반환
df.describe(): 수치형 열들의 기본적인 통계값 반환
df.isnull(): 데이터에 결측치가 있는지 여부 반환
df.dtypes: 열들의 데이터 타입 반환
```
## 3-3) 고유값 및 개수 확인
``` python
df.count(): 결측치의 개수 제외 데이터 행 수 반환
df.[열 이름].value_counts(): 특정 열에 포함된 값들의 빈도수 반환
df.[열 이름].unique(): 특정 열에 포함된 값들을 종류별로 1개씩 반환
```


# **4) 데이터 추출**
## 4-1) 인덱싱&슬라이싱
-인덱싱: 
단일 항목 지목  
-슬라이싱:
범위 지정 여러 항목 추출
## 4-2) 행 단위 추출
### 4-2-1) loc
-**인덱스 이름**을 기준으로 행 선택
-인덱싱&슬라이싱 모두 가능
-**끝 인덱스 포함** 반환
``` python
df.loc[행 이름]: 행 단위 추출
df.loc[행 이름, 열 이름]: 값 단위 추출
df.loc[:, 열 이름]: 열 담위 추출
```
### 4-2-2) iloc
-**정수형 인덱스 번호**
-속도 측면에서 더 유리
-**끝 인덱스 포함 안함**
``` python
df.iloc[행 순서, 열 순서]
```
## 4-3) 열 단위 추출
-주로 **열 이름으로 인덱싱**
``` python
df.[열 이름]: 시리즈 반환
df.[[열 이름]]: 데이터 프레임 유지
df.[[열 이름1, 열 이름2...]]: 부분적인 데이터프레임 반환
df.[열 이름][행 이름]
```
## 4-4) 불리언 인덱싱
-특정 조건에 만족하는 데이터 추출  
-True/False 배열 통해 데이터 필터링
``` python
df[조건문]
df[(조건문) 논리연산자 (조건문2)]
df.loc[조건문, 열 이름]
```
*iloc은 정수형 순서 인덱스만 받을 수 있으므로 불가능*
## 4-5) 인덱스 설정
``` python
df.set_index(열 이름): 기존 열 중 하나를 인덱스로 설정
df.reset_index(): 인덱스를 리셋하고 기존 인덱스를 열에 추가
df.reindex(인덱스 배열): 인덱스 재배열, 기존 행데이터도 인덱스를 따라감
```


# **5) 데이터 조작**
## 5-1) 데이터 추가/삭제
``` python
df['새로운 열 이름']= 리스트 : 특정 리스트 데이터 프레임에 추가
df.drop(labels, axis): 열 삭제
```
      -axis=1로 지정하면 해당 열 삭제
## 5-2) 데이터 정렬
``` python
df.sort_index(): 인덱스 기준 정렬
df.sort_values(by="열 이름"): 특정 열의 값 기준 정렬
```
## 5-3) 데이터 변환
``` python
-df.dtypes 로 열들의 데이터 타입 확인
-df[열 이름]=df[열 이름].astype(바꿀 자료형)
 df.replace(바꾸고 싶은 값, 바꿀 값)
 df.transpose(): 행, 열 교환
 df.T: 행, 열 교환
 ```
## 5-4) 데이터프레임의 병합
### 5-4-1) concat: 동일한 구조의 데이터프레임 합칠 때 
``` python
df=pd.concat([df1,df2], axis=0): 행 방향 결합
df=pd.concat([df1,df2], axis=1): 열 방향 결합
```
### 5-4-2) merge: 공통 항목 기준으로 병합
``` python
df=pd.merge(df1, df2, how= 'inner', on= '열 지정'): 교집합
df=pd.merge(df1, df2, how= 'outer', on= '열 지정'): 합집합
```

# **6) 집계함수**
## 6-1) 집계함수:
        여러행으로부터 하나의 결과값 반환
``` python
df['칼럼 명'].mean(): 평균
df['칼럼 명'].median(): 중앙값
df['칼럼 명'].std(): 표준편차
df['칼럼 명'].sum(): 합계
df['칼럼 명'].max(): 최댓값
df['칼럼 명'].min(): 최솟값
```
참고) axis=0 (행방향, 열에 대한 값), axis=1 (열방향, 행에 대한 값)
## 6-2) 그룹화
-groupby: 특정 기준으로 그룹을 묶음
``` python
df.groupby('column명1')['column명2'].집계함수 : 시리즈 형태
df.groupby('column명1')[['column명2']].집계함수 : 데이터프레임 형태
```
