# **1) 통계 개요**  
### 1-1) 통계  
-기술통계: 샘플 자체에 대한 관심  
-추론통계: 샘플의 모집단에 대한 관심
### 1-2) 자료의 척도   
**1-2-1) 질적자료**    
-명목척도: 레이블이나 이름   
-서열척도: 명목형,항목 간의 순서/순위  

**1-2-2) 양적자료**   
-구간(등간)척도: 순서, 간격이 일정, 연산 가능  
-비율척도: 절대영점, 동간성 유지  

# **2) 기술통계**   
### 2-1) 기술통계란:  
수집한 데이터의 특성을 있는 그대로 설명/이해  

### 2-2) 3가지 도구  
1. 중심화 경향  
-평균.mean(): (데이터의 총합/데이터의 개수), 이상치에 민감  
-중앙값.median(): 한가운데 위치, 이상치에 덜 민감  
-최빈값.mode(): 이산형 데이터 분포 이해  

2. 자료의 산포  
-범위.max()-.min(): 최대값-최솟값  
-분산.var(): 데이터 값들이 평균으로부터 얼마나 퍼져있는지  
-표준편차.std(): 분산에 제곱근 취한 값  
-사분위수.quantile(): 데이터의 표본을 4개의 동일한 크기로 나눈 값  
                     사분위간 범위 (IQR)  
-변동계수: 데이터의 상대적인 변동성, (표준편차/평균)*100  

3. 자료의 정리 및 시각화  
-히스토그램, 박스플롯, 파이차트  

# **3) 확률과 분포**  
### 3-1) 고려할 점    
* 표본의 불확실성-표본분포   
* 모수 추정-점추정  
* 추정의 현실성-구간추정  
* 구간추정의 오류 관리-신뢰수준  
* 적극적 주장 검증-가설검정  
* 가설검정의 오류 관리-유의수준  

### 3-2) 확률분포  
*확률분포: 확률변수가 특정한 값을 가질 확률을 나타내는 함수   

**-이산확률분포:** 확률 변수가 이산적인 값을 가지는 경우  
베르누이분포 (Bernoulli Distribution)  
이항분포 (Binomial Distribution)  
기하분포 (Geometric Distribution)  
음이항분포 (Negative Binomial Distribution)  
초기하분포 (Hypergeometric Distribution)  
포아송분포 (Poisson Distribution)    

**-연속확률분포:** 확률 변수가 연속적인 값을 가지는 경우  
정규분포 (Normal Distribution)  
T분포 (T-Distribution)  
F분포 (F-Distribution)  
감마분포 (Gamma Distribution)  
지수분포 (Exponential Distribution)  
카이제곱분포 (Chi-squared Distribution)  
균등분포 (Uniform Distribution)  

### 3-3) 대표적인 분포  
1. 정규분포
- 종 모양, 평균 기준 대칭
- 중심극한정리->모집단의 분포와 관계없이 표본의 크기가 충분히 크다면 표본평균의 분포는 근사적으로 정규분포 따름  
- 매개변수: 평균, 표준편차  

2. T 분포  
- 종 모양, 평균 기준 대칭, 자유도가 커짐에 따라 표준정규분포 N(0,1)에 수렴  
- 표본으로 모평균 추정할 때 사용
- 매개변수: 자유도
          자유도란? 얼마나 많은 정보가 독립적으로 변할 수 있는가   
- 모집단의 분산을 모를 때, 표본의 분산을 사용해 모분산을 추정하여 만들어짐  

3. 카이제곱분포  
- 0에서 시작, 오른쪽으로 긴 꼬리, 비대칭 분포  
- 자유도가 증가함에 따라 중심이 오른쪽으로 이동, 정규분포에 가까워짐  
- 매개변수: 자유도  
- 독립적인 표준 정규 분포를 따르는 변수들의 제곱 합 (표본분산의 분포와 관련)  

4. F 분포  
- 0에서 시작, 오른쪽으로 긴 꼬리, 비대칭 분포  
- 자유도가 증가함에 따라 중심이 오른쪽으로 이동, 정규분포에 가까워짐  
- 매개변수: 자유도  
- 두 독립적인 카이 제곱 분포의 비율
- 두 집단 간의 분산 차이를 비교하는 데 사용됨  

# **4) 추론 통계-추정**  
### 4-1) 추론 통계란:  
수집한 데이터를 기반으로 전체 정보 추론, 결론을 도출하는 방법을 다루는 분야  
- 추정: 모집단을 대표할 수 있는 일부를 표본으로 추출하여 분석한 통계량을 이용해서 모집단의 모수를 예측하는 과정  
- 검정: 표본을 추출하여 얻은 표본통계량으로 모집단 모수에 대한 새로운 예상·가설 등이 옳다고 판단할 수 있는지를 평가하는 과정  

### 4-2) 점추정  
어떤 모수를 표본에 기초하여 계산된 하나의 값으로 추정하는 것  

### 4-3) 구간추정  
모수의 참 값이 포함되리라고 기대하는 추정치를 일정한 범위로 신뢰도를 가지고 나타내는 방법  

# **5) 추론 통계-검정**  
### 5-1) 검정 과정
1. 가설 설정  
-귀무가설, 대립가설 설정  

2. 유의 수준 설정  
-주로 5%, 1종 오류를 범할 확률  

3. 검정통계량 산출  
  1) Z값 
  - 표준정규분포 사용, 평균으로부터 얼마나 떨어져 있는지  
  - Z=(관측값-평균)/표준편차  
  - Z = 1.96일 때 유의수준 0.05(=신뢰수준 95%)  
    Z = 2.58일 때 유의수준 0.01(=신뢰수준 99%)  
    Z값이 임계점보다 크다면 귀무가설 기각
  2) T값  
  - T분포를 사용하여 얻은 값, 작은 표본 크기에 대한 가설 검정(T검정)에 사용  
  - T=(표본평균-설정값)/(표본 표준편차/sqrt(표본크기))  
  - 자유도에 따라 모양 달라짐, 자유도가 높아질수록 Z값에 수렴  
  3) F값  
  - F값=분산_1/분산_2  
  - 분산 분석(ANOVA)에서 두 개 이상의 그룹 간 평균의 차이를 검정하는 데 사용  
  - 1에 가까우면 두 표본의 분산이 서로 비슷

4. 가설 기각/채택 판단  
-유의 확률(p)가 유의 수준(α)보다 작을 시에 귀무가설을 기각  

### 5-2) 검정 방법  
- [수치형-수치형]  
  -피어슨 상관 분석  
  ```python 
  	import scipy.stats as spst
	spst.pearsonr(df['col1'], df['col2'])
    ```  

- [범주형-수치형]  
  -T-test: 범주가 2개일 때 사용, 보통 절대값 2 기준  
  -F-test(ANOVA): 범주가 3개 이상일 때, 집단 간 분산 / 집단 내 분산  

- [범주형-수치형]  
  -Chi2-test  
  -(관측빈도 - 기대빈도)^2 / 기대빈도
