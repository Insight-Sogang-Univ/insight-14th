```
머신러닝이란? : 쉽게 말해서, 데이터를 학습해서 미래를 예측하는 기술임.
```
ㄴ 모델에서 패턴을 찾아내고, 이를 통해 예측 및 결정을 가능하게 하는 거임.

#📍머신러닝의 종류


```python
- 지도 학습 : 문제 + 정답을 주고 학습시킴 > 예측 및 분류로 사용 가능함

ex) 스팸 메일 분류 : 메일 내용과 스팸/정상 라벨을 주고 학습 > 새로운 메일이 스팸 메일인지 판단
    이미지 인식 : 강아지 사진에 강아지 라벨을 주고 학습 > 새로운 사진이 강아지인지 판단


- 비지도 학습 : 문제만 주고 학습시킴 > 연관 규칙, 군집화로 사용 가능함

ex) 고객 세분화: 고객 데이터(구매 기록, 방문 기록 등)를 주고 비슷한 고객끼리 군집
    추천 시스템: 유저의 기록을 분석하여 비슷한 취향의 사용자들에게 콘텐츠 추천


- 강화 학습 : 아무것도 안 주고 학습 시킴 > 보상과 벌점을 통해 최적의 행동을 학습하는 방법

ex) 게임 AI: 게임 환경에서 점수를 높이는 행동을 하도록 학습 > 게임을 잘하는 AI가 탄생
    자율주행: 안전하게 운전하는 행동을 하도록 학습 > 사고 없이 운전하는 자율주행차

## <span style="background-color:#E0F7FA; color:#1565C0; padding:4px; border-radius:5px;"> 2. 선형회귀 (단순, 다중선형회귀) </span>

```
선형회귀란? : x가 y에 미치는 영향을 직선으로 표현하는 방법!
```
```python
ex) - 공부 시간이 점수에 미치는 영향
    - 광고비가 판매량에 미치는 영향
    - 집 크기가 집값에 미치는 영향

```
단순선형회귀 vs 다중선형회귀

단순선형회귀
- 독립 변수가 1개만 있는 경우
- 식: 예측하려는 값 = x가 0일 때 y의 예상값 + 회귀계수 + 오차항
- ex) 공부 시간 > 시험 점수의 관계

다중선형회귀
- 독립변수가 2개 이상인 경우
- 식: 예측하려는 값 = x가 0일 때 y의 예상값 + 회귀계수*독립변수 + ... + 오차항
- ex) 공부시간 + 수면시간 + 출석률 > 시험 점수의 관계

```
최소제곱법

최소제곱법이란? 선형회귀에서 최적의 회귀계수를 찾는 방법!

최소제곱법의 원리
- 실제값과 예측값 사이의 차이를 '잔차'라고 하는데, 이 잔차들을 제곱해서 모두 더한 값(RSS)을 구한다.
- RSS가 최소가 되는 회귀계수를 찾는 것!

```
선형회귀의 기본 조건

1. 선형성 : 독립변수와 종속변수 사이에 선형관계가 있어야 함
2. 독립성 : 관측값들이 서로 독립적이어야 함
3. 등분산성 : 오차의 분산이 일정해야 함
4. 정규성 : 오차가 정규분포를 따라야 함

## <span style="background-color:#E0F7FA; color:#1565C0; padding:4px; border-radius:5px;"> 3. 경사하강법 </span>

```
경사하강법이란? : 함수의 최솟값을 찾기 위한 최적화 알고리즘 (like 안개 낀 산에서 내려갈 때, 현재 위치에서 가장 경사가 급한 방향으로 한 걸음씩 내려가는 것)

경사하강법의 작동 원리
1. 현재 위치에서의 기울기(경사)를 구함
2. 기울기의 반대방향으로 이동 (내리막 방향)
3. 정해진 거리(학습률)만큼 이동
    * 학습률이란? 경사하강법의 하이퍼파라미터임.
      * 학습률이 너무 작으면 수렴 속도가 느리고, 자원이 낭비됨 but, 안전하게 최솟값 도달 가능!
      * 학습률이 너무 크면 overshoot되고, 최솟값을 찾을 수 없게 됨.
        * 적절한 학습률을 찾으려면?
      * 
4. 최솟값에 도달할 때까지 반복

경사하강법을 사용하는 이유
- 실제 문제에서는 함수가 매우 복잡하고 비선형적이기 때문에 효율적으로 구하고자!


그러나, 문제점도 존재한다.
- Local Minimum 문제 : 전역 최솟값이 아닌 지역 최솟값에 빠질 수도 있다.
- 실제로 우리는 전역 최솟값을 찾아야 하는데, 지역 최솟값에 빠지고 탈출하지 못하면 해당 위치로 수렴해 버린다. 이러면 Local Minimum이 Global Minimum으로 인식되어 버림!


  또한, Learning Rate의 문제도 존재.
  - 학습 시간 혹은 검증 세트의 손실 값에 따라 학습률을 조절하는 Scheduling 기법이 필요하다.


## <span style="background-color:#E0F7FA; color:#1565C0; padding:4px; border-radius:5px;"> 4. 다중공선성 </span>

```
다중공선성이란? 독립변수들 간에 높은 상관관계가 존재하는 것! > 하나의 독립변수가 다른 독립변수들의 선형결합으로 완벽하게 설명될 수 있는 것이다.

ex)
- 키(cm)와 키(인치)를 동시에 독립변수로 사용
- 총소득과 세후소득을 함께 사용
- 온도 섭씨, 화씨를 동시에 사용


다중공선성의 문제점
1. 회귀계수가 불안정해짐
2. 통계적 유의성 문제 (개별 변수의 유의성 검정이 부정확해지고, 실제로는 중요한 변수가 유의하게 나타나지 않을 수 있음 ㅜㅜ)
3. 모델의 일반화 성능이 떨어지고, 과적합 가능성이 증가함

다중공선성 진단방법
```
1. 상관행렬 분석: 독립변수들 간의 상관계수 확인
    일반적으로 |r| > 0.8 이면 다중공선성 의심

2. 분산팽창계수 (VIF: Variance Inflation Factor):
    VIF = 1/(1-R²ᵢ)
    VIF > 10: 심각한 다중공선성
    VIF > 5: 주의가 필요한 수준
    VIF의 의미:
        VIF = 1: 다른 독립변수와 전혀 상관없음
        VIF = 5: 해당 변수 분산의 80%를 다른 변수들로 설명 가능
        VIF = 10: 해당 변수 분산의 90%를 다른 변수들로 설명 가능

다중공선성 해결방법
```
1. 변수 제거:
상관관계가 높은 변수 중 하나 제거
이론적 중요성을 고려하여 선택

2. 변수 결합:
상관관계가 높은 변수들을 하나로 결합
예: 키와 몸무게 → BMI

3. 주성분분석 (PCA):
원래 변수들을 선형결합하여 새로운 독립변수 생성
서로 독립적인 주성분들로 변환

4. 규제선형모델 사용:
릿지회귀, 라쏘회귀 등 활용
자동으로 다중공선성 문제 완화

## <span style="background-color:#E0F7FA; color:#1565C0; padding:4px; border-radius:5px;"> 5. 규제선형모델 </span>

기존 선형회귀는
- RSS만 최소화하는 데 집중했고,
- 학습데이터에 과적합될 위험성이 높고,
- 회귀계수가 너무 커질 수 있고,
- 새로운 데이터에 대한 예측 성능이 저하될 수 있었다.

그래서 규제선형모델을 통해 <U>모델의 복잡도를 제한하여 과적합을 방지하고 일반화 성능을 향상시키고자 함</U><BR>

규제를 이해하기 위해서는 편향 > 분산 트레이드오프를 알아야 함

> 편향: 잘못된 가정으로 인한 오차, 모델이 너무 단순할 때 발생 (편향이 높다 = 모델이 데이터의 패턴을 제대로 학습하지 못한다)

> 분산: 훈련데이터의 작은 변동에 민감하게 반응하고, 모델이 너무 복잡할 때 발생 (분산이 높다 = 모델이 훈련데이터의 노이즈까지 학습해 버린다)

> 트레이드오프: 편향을 줄이면 분산이 증가, 분산을 줄이면 편향이 증가 --> 최적점을 편향과 분산의 균형이 맞는 지점이다

규제선형모델의 종류
- L2규제: W의 제곱에 대해 페널티를 부여하는 방식 > 릿지회귀
- L1규제: W의 절댓값에 대해 페널티를 부여하는 방식 > 라쏘 회귀
- LS규제 + L1규제 결합: 엘라스틱넷 회귀

## <span style="background-color:#E0F7FA; color:#1565C0; padding:4px; border-radius:5px;"> 6. 모델평가방법 </span>

- 평균 절대 오차: 회귀 분석에서 모델의 예측값과 실제 관측값 사이의 절대값 오차의 평균을 의미한다

    - 둘 다 모델의 예측 성능을 평가하는 지표 중 하나인데, <U> 전자는 큰 오차에 민감하게 반응하여 이상치에 민감하고 + 후자는 이상치가 존재해도 전반적인 예측 성능을 유지할 수 있음. </U>
