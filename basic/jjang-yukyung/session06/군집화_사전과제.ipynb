## 군집화

### <span style="background-color:#FFFFE0; color:#CD5C5C; padding:4px; border-radius:5px;"> 1. 군집화에 대하여</span>

군집화란?
- 데이터 간 유사성을 측정
- 비슷한 데이터들이 한 군집에 속하게 하는 것
- 서로 다른 군집 간 차이가 최대화되도록 분할하는 것

군집화의 목표
- 응집도(cohesion) ↑ → 같은 군집끼리 최대한 유사하게
- 분리도(separation) ↑ → 군집 간에는 최대한 다르게

```
즉, 비슷한 특성을 가진 데이터들을 하나의 cluster로 묶는 것

### <span style="background-color:#FFFFE0; color:#CD5C5C; padding:4px; border-radius:5px;"> 2. 군집화를 위한 준비사항</span>

```
데이터 수집 및 정제 > 특징 선택 및 변환 > 유사도(거리) 측정 기준 결정 > 군집화 알고리즘 선택 > 군집 수(k) 결정 방법 설정 > 평가 및 해석 기준 마련

1) 데이터 수집 및 정제
   - 군집화를 위한 데이터를 확보한다 (정형/비정형 모두 가능)
   - 결측치 처리, 이상치 제거, 노이즈 제거 등 우리가 흔히 아는 전처리를 수행한다




2) 특징 선택 및 변환
   - 군집화에 사용할 유의미한 변수를 선정


        - 1) 연속형 변수
            - ㄴ 수치로 측정 가능하며, k-means가 적합. 스케일링이 필수적임!
        - 2) 명목형 변수
        - 3) 혼합형 변수
  
   - 필요시 PCA로 차원 축소
     - 고차원 데이터의 시각화 및 탐색 용이
     - 불필요하거나 노이즈가 많은 차원을 제거하여 분석 성능 향상(예: 군집화, 분류 전처리)
     - 데이터의 '차원의 저주' 문제를 완화해 계산 효율을 높임
       - 차원의 저주: 학습 데이터 수 > 차원의 수 --> 성능이 저하

3) 유사도(거리) 측정 기준 결정
   - 데이터 간 유사성 판단 기준 정하기
   - 대표적 거리: 유클리드 거리, 맨해튼 거리, 코사인 유사도 etc


>>>>> 본 세션에서는 유클리드 거리(유클리디안 거리)를 사용

4) 군집화 알고리즘 선택
   - 데이터 특성에 맞는 알고리즘을 선택 (3번에서 더 자세히)

5) 군집 수(k) 결정 방법 설정
   - 최적 군집 수 결정에 엘보우 기법, 실루엣 분석 등을 활용
     - 엘보우 기법: SSE의 감소율이 급격히 줄어드는 지점을 찾는 것
     - 실루엣 분석: 군집이 효율적으로 잘 분리됐는지 확인

6) 평가 및 해석 기준 마련
   - 군집화 결과를 평가할 내부, 외부 지표를 설정

```
즉, 피처 선택/추출 > 군집화 알고리즘 선택 > 군집 유효성 평가 > 결과 해석

### <span style="background-color:#FFFFE0; color:#CD5C5C; padding:4px; border-radius:5px;"> 3. 군집화 알고리즘</span>

- 계층적 군집화
- K-평균 군집화
- 가우시안 혼합 모델

```
계층적 군집화

- 데이터 포인트들을 시작부터 군집 하나씩 떼어내거나 합치면서 트리 형태로 군집을 만들어가는 방법
- 병합적(agglomerative)과 분할적(divisive) 방법으로 구분
- 덴드로그램이라는 트리 구조로 시각화 가능 (사진이 덴드로그램)
- 장점: 군집 수(K-mean과 달리)를 미리 정하지 않아도 됨, 데이터의 계층적 관계 파악 가능
- 단점: 데이터 크기가 크면 계산량 증가

거리가 가까운 관측치들끼리 차례대로 군집으로 묶어 봄. 거리가 가장 짧은 것이 2이고 이에 해당하는 개체는 A와 D이므로 먼저 A와 D를 하나의 군집으로 엮으면 됨.

여기서 A와 D가 한 군집이니 개체-개체 거리를 군집-개체 거리로 계산해야 한다. 즉, AD / B or AD / C 이런 식으로!

위의 4가지 방법 중 하나를 택하여 거리행렬을 업데이트하면 됨.

거리 행렬을 업데이트해서 군집-개체 간 가까운 것들을 묶고 덴드로그램을 계속 엮어나가다 보면 분석 대상 관측치가 하나도 없어지는 순간이 옴. 그때 학습이 종료되는 것!

```
K-평균 군집화

- 데이터 집합의 k개의 군집으로 나누는 알고리즘
- 임의로 k개의 중심점을 정하고 각 데이터가 가장 가까운 중심점에 할당
- 각 군집 내 데이터들의 평균으로 중심점을 다시 계산하고 이를 반복해 중심점이 수렴할 때까지 수행
- 장점: 계산이 빠르고 직관적, 대규모 데이터에 적합
- 단점: k값(군집 개수)을 미리 정해야 하며, 초기 중심점 설정에 따라 결과가 달라질 수 있음 > 이는 현재 k-means++같은 알고리즘을 통해 보완하고 있다고 함!

```
가우시안 혼합 모델

- 각 군집을 여러 개의 가우시안(정규분포)의 혼합으로 모델링하는 군집화 방법
- 데이터가 각 군집에 속할 '확률'로 표현되어 부드러운 군집 경계가 형성됨
- 장점: 군집의 모양이 원형이 아닌 경우에도 잘 적합, 소프트 클러스터링 가능
- 단점: 계산량이 많고, 모델 학습이 복잡함

알고리즘별 강점

- k-means  → 대규모, 원형 데이터

- 계층적 군집화 → 소규모, 구조 이해

- DBSCAN → 이상치 포함, 임의 형태 군집

- GMM → 확률 기반, 타원형 데이터 처리

### <span style="background-color:#FFFFE0; color:#CD5C5C; padding:4px; border-radius:5px;"> 4. 군집화 평가 방법</span>

- 외부평가
- 내부평가
- 실루엣 계수
- 던 인 지수

```
외부평가

- 이미 알고 있는 정답과 군집화 결과를 비교해서 평가
- 정확도, 혼동 행렬, 정밀도, 재현율 등 사용 가능
- 비지도 학습의 군집화에서는 정답이 없을 때는 사용이 어려움


- ARI : 실제 레이블과의 일치도 측정

```
내부평가

- 데이터 자체의 특성만으로 군집화 품질을 평가
- 가장 널리 사용되는 지표는 실루엣 지수임
- 군집 간 거리와 군집 내 거리의 차이를 계산해 군집이 잘 분리되었는지 평가
- 실루엣 계수, Dunn Index

```
실루엣 계수

###### 헐... 이사진 인사이트 oopy에서 받았어요.....

- 각 데이터 포인트에 대해 군집 내 응집력과 이웃 군집과의 분리도를 평가
- 값은 -1~1 사이로, 높을수록 좋은 군집화

상기 이미지처럼 10개의 데이터 포인트들이 3개로 나누어졌다고 설정해보자.

이처럼 모든 데이터 포인트에 대한 실루엣 계수를 계산하고, 평균값을 계산한다.

장점
- 클러스터링이 수행된 후 실제 구분된 클러스터에 따라 실루엣 계수를 구하기 때문에, 클러스터링 알고리즘에 영향을 받지 않는다. 
- 적절한 클러스터 개수를 정하거나 더 나은 클러스터링 기법을 선택하는 기준으로 삼을 수 있다. 
- 클러스터링 결과 값 시각화 가능하다.

단점
- 데이터 양이 많아질수록 수행 시간이 오래 걸린다.
*만일 100개의 데이터 포인트가 있다면 99개의 다른 데이터 포인트들과 거리를 구해 실루엣 계수를 구해야 하고, 이런 동일한 연산을 100번 수행해야 한다.
- 전체 데이터 포인트의 실루엣 계수 평균값만으로 클러스터링 결과를 판단할 수 없으며 개별 클러스터의 평균값도 함께 고려해야 한다.

```
던 인 지수(Dunn Index)

- 군집 간 거리의 최솟값과 군집 내 거리의 최댓값 비율로 군집 품질 평가
- 증가할수록 군집 간 간격이 크고, 군집 내 응집력이 좋음

````python
import numpy as np
import random
import matplotlib.pyplot as plt
 
from itertools import combinations
 
np.random.seed(100)
num_data = 20
 
x1 = np.linspace(0.3,0.7,num_data)
error = np.random.normal(1,0.5,num_data)
x2 = 1.5*x1+2+error
 
X = np.stack([x1, x2], axis=1)
````

````python
fig = plt.figure(figsize=(7,7))
fig.set_facecolor('white')
plt.scatter(x1, x2, color='k')
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()
````

````python
def complete_diameter_distance(X):
    res = []
    for i, j in combinations(range(X.shape[0]),2):
        a_i = X[i, :]
        a_j = X[j, :]
        res.append(np.linalg.norm(a_i-a_j))
 
    return np.max(res)
````

````python
def average_diameter_distance(X):
    res = []
    for i, j in combinations(range(X.shape[0]),2):
        a_i = X[i, :]
        a_j = X[j, :]
        res.append(np.linalg.norm(a_i-a_j))
 
    return np.mean(res)
````

````python
def centroid_diameter_distance(X):
    center = np.mean(X, axis=0)
    res = 2*np.mean([np.linalg.norm(x-center) for x in X])
 
    return res
````
