### ğŸ”¹ Python ì‹¤ìŠµ
```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats

# ìƒ˜í”Œ ë°ì´í„° ìƒì„±
data = np.random.normal(loc=50, scale=10, size=100)  # í‰ê·  50, í‘œì¤€í¸ì°¨ 10, í‘œë³¸ í¬ê¸° 100
df = pd.DataFrame(data, columns=["value"])

# ê¸°ìˆ  í†µê³„
print("í‰ê· :", np.mean(data))
print("ì¤‘ì•™ê°’:", np.median(data))
print("ìµœë¹ˆê°’:", stats.mode(data, keepdims=True).mode[0])
print("ë¶„ì‚°:", np.var(data))
print("í‘œì¤€í¸ì°¨:", np.std(data))
print("ìµœì†Ÿê°’:", np.min(data), "ìµœëŒ“ê°’:", np.max(data))

# ì‹œê°í™”
sns.histplot(data, kde=True)
plt.title("Histogram + KDE")
plt.show()

sns.boxplot(x=df["value"])
plt.title("Boxplot")
plt.show()

## 2. í™•ë¥ ê³¼ ë¶„í¬
- í™•ë¥  ê°œë…
  - ì¡°ê±´ë¶€ í™•ë¥ 
  - ë…ë¦½ ì‚¬ê±´
  - ë² ì´ì¦ˆ ì •ë¦¬
- í™•ë¥  ë¶„í¬
  - ì´ì‚°í˜• ë¶„í¬: ë² ë¥´ëˆ„ì´, ì´í•­, í¬ì•„ì†¡
  - ì—°ì†í˜• ë¶„í¬: ì •ê·œë¶„í¬, t-ë¶„í¬, F-ë¶„í¬, ì¹´ì´ì œê³±ë¶„í¬

### ğŸ”¹ Python ì‹¤ìŠµ ì˜ˆì œ
```python
# ì´í•­ë¶„í¬ (n=10, p=0.5)
binom_rv = stats.binom.rvs(n=10, p=0.5, size=1000)
sns.histplot(binom_rv, kde=False)
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.show()

# ì •ê·œë¶„í¬
x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x, loc=0, scale=1)
plt.plot(x, y)
plt.title("Standard Normal Distribution")
plt.show()

# í¬ì•„ì†¡ ë¶„í¬ (Î»=3)
poisson_rv = stats.poisson.rvs(mu=3, size=1000)
sns.histplot(poisson_rv, kde=False)
plt.title("Poisson Distribution (Î»=3)")
plt.show()


## 3. ì¶”ë¡  í†µê³„

3-1. ì ì¶”ì •
- í‘œë³¸ í‰ê· , í‘œë³¸ ë¹„ìœ¨ ë“±ìœ¼ë¡œ ëª¨ì§‘ë‹¨ ì¶”ì •

```python
sample_mean = np.mean(data)
print("í‘œë³¸ í‰ê·  (ëª¨ì§‘ë‹¨ í‰ê·  ì¶”ì •):", sample_mean)


3-2. êµ¬ê°„ ì¶”ì •
- ì‹ ë¢° êµ¬ê°„ ê³„ì‚°
    ex) ëª¨ì§‘ë‹¨ í‰ê· ì— ëŒ€í•´95% ì‹ ë¢° êµ¬ê°„

```python
confidence_interval = stats.t.interval(alpha=0.95, 
                                       df=len(data)-1,
                                       loc=np.mean(data),
                                       scale=stats.sem(data))
print("95% ì‹ ë¢°êµ¬ê°„:", confidence_interval)


3-3. ê°€ì„¤ ê²€ì •
- ê·€ë¬´ê°€ì„¤ : ì´ê²Œ ë§ë‹¤
- ëŒ€ë¦½ ê°€ì„¤ : ì´ê²Œ ì•„ë‹ˆë‹¤

```python
t_stat, p_val = stats.ttest_1samp(data, popmean=50)
print("t-í†µê³„ëŸ‰:", t_stat, "p-ê°’:", p_val)

if p_val < 0.05:
    print("ê·€ë¬´ê°€ì„¤ ê¸°ê°: í‰ê· ì´ 50ê³¼ ë‹¤ë¥´ë‹¤.")
else:
    print("ê·€ë¬´ê°€ì„¤ ì±„íƒ: í‰ê· ì´ 50ê³¼ ìœ ì˜í•˜ê²Œ ë‹¤ë¥´ì§€ ì•Šë‹¤.")
