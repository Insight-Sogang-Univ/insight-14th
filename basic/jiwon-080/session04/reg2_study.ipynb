### 1. 다중선형회귀의 기본가정

1. **선형성**  
- 종속변수와 설명변수 간의 관계가 선형적(선형식으로 설명 가능해야)

2. **독립성**  
- 설명변수들 간 선형독립적이어야 함 (=다중공선성이 없어야 함)  
- 다중공선성 판단: 상관계수, pairplot, VIF  
- 다중공선성 제거: 변수 제거, 규제 선형 모델(라쏘, 릿지, 결합), PCA(주성분분석)

3. **오차항의 평균은 0과 같거나 0에 가까워야 함**  
- 오차항=실제값과 예측값의 차이=백색잡음

4. **등분산성**  
- 오차항의 분산이 일정해야 함  
- 일정하지 못한 경우 이분산성이 있다고 함

오차항의 평균 0(백색잡음을 따르는지), 등분산성, 정규성은 회귀분석 후에 검증 가능

### 2. 회귀분석의 평가 방법

- 시각화로 평가  
회귀선과 데이터 포인트를 동시에 그려서 확인

- 통계지표로 평가  
    ```Python
    import statsmodels.api as sm  
    result=sm.OLS(Y,X).fit() #fit 
    result.summary() #에서 확인
    ```
    - 모델의 유의성 검정: Prob(F값), T값  
    F값은 모델 전체의 유의성을 검정: p-value가 0.05보다 작으면 유의미  
    H0: 모든 회귀 계수가 0(모델 자체의 유의성이 없다)  

        T값은 한 독립변수의 모델 안에서의 유의성을 검정: p-value가 0.05보다 작으면 그 독립변수는 유의미  
    H0: 이 변수의 회귀 계수가 0(이 변수의 유의성이 없다)

    - 결정계수(R^2 score)  
    (데이터의 평균으로 예측할 때보다) 모델을 통해 예측할 때 더 얼마나 정확한 예측인지  
    $R^2=SSR/SST=1-SSE/SST$  
    SST: 평균을 통해 나타난 편차(총 편차)  
    SSR: 회귀선을 통해 나타난 편차  
    SSE: 회귀선을 통해 설명되지 않는 편차  
    R^2가 1에 가까워질수록 설명력이 높다고 한다.  

    - Adjusted R^2  
    R2 score의 문제점: 독립변수 수가 증가하면 같이 증가한다.(유의성 적은 변수를 추가해도 오히려 높아질 수 있다.)  
    -> 변수의 개수 증가에 민감하지 않도록 조정  
    $Adjusted R^2=(SSR/SST)(n-1/n-k)$  
    $=1-(SSE/(n-k))/(SST/(n-1))$  
    n:샘플 수(데이터 수)    k:독립변수의 개수

    - AIC, BIC(정보기준)  
    값이 낮을수록 설명력이 좋다.


### 3. 비선형 회귀  


- 회귀선이 비선형적인 모양일 수 있다.  
$Y=a+bX+cX^2+dX^3+\cdot\cdot\cdot$  
의 거듭제곱을 추가한 형태의 비선형적 회귀선  
기존의 선형 모델을 다항식으로 만드는 방식  
```python
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree=2, include_bias=False)  
X_poly = poly.fit_transform(X)
```  
후에 X_poly를 회귀모델 학습의 X변수에 이용하는 방식으로 하면 된다  
- 비선형 회귀의 장점: 더 핏한(설명을 잘하는) 회귀선을 그릴 수 있다.  
비선형 회귀의 단점: 변수가 많으면 과적합이 될 수 있다.(규제 선형 모델을 고려해야 할수도)  

- 지수 회귀모델  
종속변수 Y에 로그를 취해, 회귀식은 지수함수 형태를 따르게 하는 것  
![지수회귀](https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcRkLsch-v5K8xC7LxHmfdMuVhdYbpdANY29Tw&s)  
예: 수확체증이 나타나는 데이터
```python
log_Y=np.log(Y)
```
후에 log_Y를 회귀모델 학습의 Y변수에 이용하는 방식으로 하면 된다  
그 후에 기존 데이터 형태로 복구 - **예측값에 지수함수를 적용해야 한다.**  

- 로그 회귀모델  
종속변수 X에 로그함수를 적용해, 회귀식은 로그함수 형태를 따르게 하는 것  
![로그회귀](https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcS7L2dPEOYyT_svgqtzqoJkvVpJNelhemsa_A&s)  
예: 한계체감이 나타나는 데이터
```python
log_X = np.log(X)
```
후에 log_X를 회귀모델 학습의 X변수에 이용하는 방식으로 하면 된다  

- 스플라인 회귀  
데이터를 구간별로 나누고 각각 구간들에 대해 다른 회귀함수(선형 or 비선형)를 사용하는 것  
**구간 경계에서 연속성을 유지하는 것이 중요**하다
반복문을 통해 나눈 구간들에 모델을 적용하는 방식으로 사용  
예: 구간에서 패턴이 달라지는 데이터, 계절성이 강한 데이터

### 4. 회귀 모형 모델링 과정 정리  
- 간단화  
사전 검증 -> 모델 생성, 모델 핏 -> 모델 평가 -> 모델을 개선해 나가기  

1. 사전 검증  
- EDA  
- 결측치와 이상치 처리  
- 회귀 분석의 기본 가정에 부합하는지 검토 (백색잡음 부합, 다중공선성 여부, 자기상관 여부)  
- 피쳐 수정, 결합, 삭제, PCA 등 피쳐에 대한 전처리

2. 모델 생성과 모델 핏  
- 모델 생성, 설정(파라미터 튜닝), 모델 학습  
- 모델의 유의성 검정(F검정으로 모델 전체의 유의성, T검정으로 특정 변수의 유의성 검정)  

3. 모델 성능 평가  
- R^2, Adjusted R^2, AIC, BIC로 평가  
- 회귀 분석을 통해, 이전에 세웠던 가설을 검토한다.

4. 모델 개선  
- 피처 병합 or 제거  
- 다시 파라미터 튜닝  
- 비선형 모델을 사용  
- 규제 선형 모델을 사용
