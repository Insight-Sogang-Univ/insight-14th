> ### 1. 데이터 기반 의사결정

데이터 분석의 장점: **목적 달성 프로세스의 질**  

데이터를 통한 과거 경험+미래 전망 -> **액션**에 집중  

- 액션이란?  
크게 결정/계획/행동으로 나뉨  
    - 결정: 프로젝트에 대한 의사결정을 내리는 일  
    <- 결정을 뒷받침하기 위한 데이터 분석  

    - 계획: 프로젝트의 과정 단위를 계획하는 일  
    <- 과정의 방법론과 수치 근거를 통해 계획 수립  

    - 행동: 프로젝트의 과정 안에서 특정한 '~~한다' 라는 행동  
    <- 일별 실행 계획에 사용할 수 있는 현상 분석으로 접근

> DDDM  
비즈니스의 의사 결정에서 직관 대신 데이터, 분석을 사용하는 접근 방식  

   - DDDM의 6가지 과정  
        1. 목표 정의: 조직의 명확한 목적을 정의  
        2. 데이터 식별, 준비 및 수집:
        목표에 맞는 데이터 소스를 평가하면서 데이터를 수집  
        3. 구성 및 탐색(구조화):  
        정리-> 정확성과 신뢰성 확보  
        시각화-> 어느정도의 패턴, 이상치, 추세 식별  
        4. 데이터 분석:  
        데이터를 '실행 가능한' 인사이트로 변환하는 과정  
        패턴, 상관관계, 추세를 파악-> 전략적 결정에 사용, 성과를 개선  
        5. 결론 도출: 분석 결과를 검토, 올바른 비즈니스 상황에 배치  
        6. 구현 및 평가:  
        인사이트 검증, 결과 측정  
        데이터를 통한 실행 계획을 개발, 자원을 할당, 이니셔티브 차원에서 모니터링  
        핵심성과지표(KPI)를 통해 결과를 측정, 분석  
        피드백 수집, 개선 주기를 통해 조정

> 데이터 분석 유형

(1->4로 갈수록 가치와 복잡도가 올라간다)  
1. 설명 분석(What Happened?)  
과거 데이터를 설명, 요약 -> 과거 성과에 대한 인사이트 제공  
예) 월별 판매 보고서, 고객 설문조사, 웹사이트 트래픽 분석  
2. 진단 분석(Why Did It Happen?)  
데이터 마이닝, 상관관계 식별 등을 통해 특정한 '이벤트'가 발생한 '이유'를 판단  
이벤트: 판매 감소, 고객 불만, 추세와 같은 현상  
3. 예측 분석(What Is Likely To Happen?)
데이터를 기반으로 미래의 추세나 결과를 예측  
통계 모델, 러닝 기법, 예측 기술 등을 통해 판매와 고객 행동을 예측  
4. 진단 분석(How Will It Happen?)  
데이터에 기반한, 다음 단계의 '조치'  
최적화, 홍보 캠페인, 자원 배치 등을 결정

> ### 2. 목표 정의

> 목표를 나타내는 지표  
방향, 목적, 기준을 나타내는 것  

- 효율성: 산출/투입  
예) 노동 생산성=생산량/노동시간  
- 효과성: 달성결과/목표  
예) 달성률=실제/목표  
- 수익성: 이익/자원  
예) ROE=순이익/자기자본  
- 성장성: 증가분/기존 규모  
예) 성장률=$(X_당-X_전/X_전)$  
- 안정성: 보유자산/부채  
예) 부채비율=부채/자산  

- 절대치 지표: 매출액, 직원 수, 생산량, 판매량, 재고 등  
- 지수형 지표: 고객만족지수, 순추천지수 등  
- 복합 지표: 여러 지표를 묶은 것  

> 지표를 정의하기

위의 지표들처럼 정확한 수치를 정의/측정할 수 없을 때?
-> **Proxy Metric**  
직접 측정할 수 없는 것의 대체물(프록시)을 정하여 측정한다는 의미  
예) 넷플릭스 초기의 프록시 메트릭  
넷플릭스 초기(DVD 대여)의 리텐션을 올리기 위한 측정 지표 
- 1개월 동안 최소 15분의 비디오를 스트림하는 멤버 비율  
- 한달 내, 대기 목록에 최소 6개의 DVD를 추가하는 멤버 비율  
- 첫 6주 동안 최소 50개의 영화를 평가한 신규 멤버 비율  

프록시 메트릭을 평가할 때, 프록시에 대해서는  
- 측정 가능해야 한다.  
- 변화할 수 있다.  
- 평균이 아니다.  
- 신규 고객 대 기존 고객을 명시한다.  
- 조작할 수 없다.  
- 성공적인 프록시 메트릭은 리텐션과 양의 상관관계가 있다.

> 북극성 지표(핵심 지표)

![북극성지표](https://static.waveon.io/img/blog/_posting/north.png)
기업이 가장 중요하게 생각하는 지표  
가장 높은 우선순위를 가지되, 핵심 지표의 성장이 다른 지표에 부정적인 영향을 주면 안 된다.

> L1/L2 지표

Level 1, Level 2는 **핵심 지표(북극성 지표)의 성장을 끌어내는 지표**  

- **L1**: 북극성 지표를 **보완**하는 지표  
핵심 지표에 직접적으로 기여 or 제품의 건전성 확보  
    1. Reach: 특정 기간 유입된 사용자 수  
    2. Activation: 신규 유저가 활성 사용자로 얼마나 전환되었는지  
    3. Engagement: 제품에 얼마나 몰입했는지(예: 활성 사용자가 어떤 핵심 액션을 얼마나 수행)  
    4. Retention: 사용자들이 제품으로 얼마나 다시 돌아오는지/유지하는지  
    5. Business-specific: 특정 비즈니스에 특화된 지표  

- **L2**: L1 **그리고** 북극성 지표의 성장에 기여하는 구체적인 지표  
(참조: https://brunch.co.kr/@mobiinside/3148)

> AARRR  
제품/서비스가 가치를 제공하는 방식이 다르면, 측정하는 지표 또한 달라져야 한다!  

1. Acquisition: 사용자 획득  
사용자가 여러 가지 채널(광고, 검색엔진, 지인 추천, 행사)을 통해 방문하는 것  
<- 어떻게/어디서/주로 신규 고객을 획득할 지를 고민

    - 지표 예시:  
    CAC(유저 획득 비용), CTR(클릭 1건당 지출), CPC(인스톨 당 지출), CPL(유저 행동 당 지출), ROAS(광고로 인한 매출액/광고비용)  

2. Activation: 사용자 활성화  
사용자가 처음 주요 기능을 사용하는 것  
-> 제품 성격, 도메인에 따라 다양한 기준으로 측정  
예) 첫 상품 구입, 페이스북 친구 추가, 접속 시간 등  

    - 구성:  
    Sign-up(유저가 제품에 들어온 시점), Setup moment(유저가 핵심가치를 경험할 준비가 완료), A-ha moment(유저가 처음 핵심가치를 경험), Habit moment(유저가 제품 사용에 대한 습관을 형성)  

3. Retention: 유지  
신규 사용자->장기 사용자로 전환하는 것을 고민  

    - 측정 지표 예시: n일/n주차 유지율, EDM 메일 개설율 등  
    DN Retention(N일 안으로 재접속), MAU/DAU/WAU(일/주/월마다 방문하는 횟수), Stickness(해당 기간 동안 활성화 수)  

4. Revenue: 수익  
사용자가 제품/서비스를 유료 구매  
소비를 유도, 매출을 높이도록 유도해야 함  
고객의 평생 가치(LTV)를 높이고, 고객 유치 비용은 줄여나가야 함  

    - 지표 예시:  
    PU(전체 사용자 중 결제한 사용자 비율), Average Check(구매 1회당 평균 금액), ARPU(인당 평균 매출), ARPPU(결제자 당 평균 매출), LTV(유저의 진입~이탈 까지의 누적 수익)  

5. Referral: 추천, 홍보  
사용자가 제품/서비스를 타인에게 추천  
충성 고객을 어떻게 홍보자로 만드는지를 고민해야 함  

    - 지표 예시: NPS(순 고객추천 지수), 바이럴 지수  
    바이럴 계수(기존 고객 한명 당 데리고 온 새로운 고객의 수/1이 넘으면 1. Acquisition 단계를 하지 않아도 서비스가 성장한다고 봄), 바이럴 주기(기존 고객 1명이 새 고객을 데려오는 데 걸리는 시간)
