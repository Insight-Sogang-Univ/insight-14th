## 1. 추론통계와 기술통계  
기술통계: 수치와 시각화 등으로 샘플을 요약  
추론통계: 샘플로부터 모집단을 추론(가설검정, 회귀 등0)

- 자료의 척도  
명목척도: 순서의 구분이 없는 명목형  
서열척도: 순서/순위의 구분이 있는 명목형  
구간(등간)척도: 값들 사이에 순서가 있으나 연산이 가능(온도, dt 등)  
비율척도: 측정이 가능하지만 0=아무것도 존재하지 않는 상태의 척도(거리, 높이, 무게 등)

## 2. 기술통계

- 중심화 경향의 기술통계량  
평균: 직관적이나 이상치에 민감하다  
중앙값: 이상치에 민감하지 않다, W자형, U자형일때는 유의
최빈값: 이산형 데이터를 볼때 유의하다  

- 산포의 기술통계량
범위: 최대값과 최소값의 차이  
분산: 데이터가 평균으로부터 퍼져있는 정도  
표준편차: 분산에 제곱근을 취한 값(분산은 계산할때 제곱을 합한 값이기 때문에 값이 비직관적)  
사분위수: Q1(하위 25퍼센트) Q2(중앙값) Q3(상위 25퍼센트)  
각각 .quantile(0.25) .quantile(0.5) .quantile(0.75)로 구함  
IQR: Q1~Q3의 범위, tukey fence로 이상치 판단할 때 활용  
변동계수: 데이터의 상대적인 변동성을 나타냄  
$(표준편차/평균)*100$

- 첨도(kurt)와 왜도(skew)  
첨도: 데이터가 몰려 있는 정도(중첨-정규 분포, 급첨-정규분포보다 뾰족, 완첨-정규분포보다 완만)  
왜도: 정규분포에 비해 왼쪽 혹은 오른쪽으로 데이터가 몰려 있는 정도

## 3. 확률과 분포  

- 확률분포  
확률변수: 특정 사건이 발생할 때(happen) 나타날 수 있는 모든 경우의 수  
확률분포: 확률변수가 특정한 값을 가질 확률을 나타내는 함수  
    - 이산확률분포  
확률변수가 이산적인 값을 가짐(예:성공/실패)  
종류: 베르누이, 이항, 포아송 ...  
    - 연속확률분포  
확률변수가 연속적인 값을 가짐(예:점수)  
종류: 정규, T, F, 카이제곱 ...

- 정규분포  
매개변수: 평균(μ), 표준편차(σ)  
종 모양을 따르는 (가장 보편적인) 연속확률분포  
    - 대수의 법칙  
    표본의 크기가 커질수록, 표본평균이 모평균과 비슷해짐(주사위 굴려서 평균을 구할때 굴리는 수가 많을수록 3.5와 가까워짐)  
    - 중심극한정리  
    동일한 확률분포를 가지고 독립적인 확률분포 n개의 평균의 분포는 n이 크다면 정규분포에 가까워짐

- T분포  
매개변수: 자유도(df)  
종 모양이지만 정규분포보다 완첨, 자유도가 커질수록 표준정규분포 N(0,1)에 수렴  
표본으로부터 모평균(모평균 간 차이)를 추정할 때 사용  
표본 크기가 작을 때 정규분포보다 유용 (보통 30을 기준으로)

- 카이제곱분포  
매개변수: 자유도(df)  
0에서 시작, 오른쪽으로 긴 꼬리, 자유도가 커질수록 모양이 정규분포에 가까워짐  
표본분산의 분포와 관련  
->독립성 검정, 적합도 검정에 사용

- F분포  
매개변수: 자유도(df1, df2)  
겉보기 특징은 카이제곱과 비슷하나  
두 독립적인 카이제곱 분포의 비율  
->두 집단 간의 분산 차이를 비교하는 데 사용

## 4. 추론통계  

### 추정  
표본의 통계량을 이용하여 모집단의 모수를 예측하는 과정
- 점추정: 하나의 '모수'를 추정  
- 구간추정: 그 모수가 존재할 수 있는 '구간'을 추정  
    - 신뢰수준이란?  
    예: (95% 확률로 신뢰가능)  
    =동일한 조사를 100번 했을 때, 그 중 95개의 신뢰구간이 실제 모수를 포함하는 구간이다

### 검정

- 가설 검정의 과정  
    1. 가설 설정  
    귀무가설 H0와(H0와 반대인)대립가설 H1를 설정  
    H0(보통 주장하려는 것과 반대)  
    -> H0를 기각하면 H1을 수용/H1을 기각하면 H0을 수용

    2. 유의수준 설정  
    유의수준: 허용할 수 있는 (1종)오류의 최대치. 보통 5%(0.05)로 사용  
    =1종 오류가 5% 확률로 발생할 수 있음  

    - 1종 오류(알파), 2종 오류(베타)  
    1종 오류: 실제 참을 기각할 확률, False Negative (질병이 있었는데 없다고 판단해서 병이 커짐)  
    2종 오류: 실제 거짓을 수용할 확률, False Positive (일반 메일을 광고 메일로 분류해버림)  
    예시에서도 나타나듯이 보통 1종 오류의 여파가 더 안좋다

    3. 검정통계량 산출  
    검정통계량: 가설 검정의 결과를 판단하는 데 사용되는 값  
        - Z score: 표준 정규분포 기반  
        $Z=(X−μ)/σ$  
        주로 큰 표본에 사용  
        - T score: T분포 기반  
        $T=(X-μ)/(s/\sqrt{n})$  
        더 작은 표본에서 사용, 두 표본 간 평균 차이 검정에 사용  
        - F score: F분포 기반  
        $F=분산1/분산2$  
        분산분석(ANOVA)에서 두 개 이상 그룹 간 평균의 차이 검정에 사용  

    4. 가설 기각 or 수용 ->  결론 도출

### 가장 중요한 통계 개념

1. 확률분포의 모양들(T, F, 정규)을 대략적으로 파악하는 것이 중요하다  
그 이유는 이후 가설검정에서 검정통계량 판단에 이용되고, 표본을 추출할 때 표본의 타당성을 대략 확인할 수 있다.
2. 가설의 유의수준과 1종, 2종 오류  
유의수준은 간단하게 1종 오류가 발생할 수 있는, 최대 허용 확률이다.  
1종 오류란 실제 참인 것을 거짓이라고 판단하는 오류이다.
