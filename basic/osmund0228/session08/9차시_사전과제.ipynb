# **데이터로 보는 비즈니스 상황**

## 1. **데이터 기반 의사결정**

### 1.1 무엇을 위한 데이터 분석인가

- 데이터 분석을 하면 객관적으로 과거를 돌아보고 미래의 전략을 수립하여 앞으로 해야 할 일에 자신있게 집중할 수 있는 근거가 마련됨.
---
### 1.2 Data-Driven-Decision-Making (DDDM)
- 비즈니스 의사 결정을 제공하기 위해 직관 대신 데이터와 분석을 사용하는 것을 강조하는 접근 방식
- 고객 피드백, 시장 동향, 재무 데이터와 같은 데이터를 수집, 분석, 해석하여 더 나은 의사결정을 내릴 수 있음!
---
### 1.3 6 Steps of DDDM

#### 1️⃣ 목표 정의
- 조직의 목표를 명확히 정하고 표현
#### 2️⃣ 데이터 식별, 준비 및 수집
- 데이터 요구사항 판별, 데이터 소스 평가, 체계적으로 데이터를 수집 및 검증
#### 3️⃣ 구성 및 탐색
- 데이터에서 새로운 패턴, 추세, 가치 있는 인사이트를 발견할 수 있도록 구조화 및 정리, 그리고 시각화
#### 4️⃣ 데이터 분석 수행
- 원시 데이터를 분석하여 비즈니스 전략에 영향을 미치는 패턴, 상관관계, 추세 파악, 그리고 전략적 결정을 내리고 전반적인 성과 개선
#### 5️⃣ 결론 도출
- 데이터 분석으로 얻은 인사이트를 비즈니스에 적용할 수 있는 권장사항으로 구성
#### 6️⃣ 구현 및 평가
- 데이터 기반 권장 사항을 기반으로 실행 계획을 개발하고 실행하며 상황을 모니터링
- 핵심성과지표(KPI)를 기준으로 결과를 측정하고 분석하여 피드백 및 개선, 조정 수행
---
### 1.4. 데이터 분석 유형

#### 1️⃣ 설명 분석(Descriptive Analysis)
- 데이터 취합 및 마이닝을 통해 과거 데이터를 설명, 요약하여 과거 성과에 대한 인사이트 제공
- 월별 판매보고서, 고객 만족도 설문 조사, 웹 사이트 트래픽 분석 등에 사용됨
#### 2️⃣ 진단 분석(Diagnostic Analysis)
- 특정 이벤트가 발생한 이유를 파악하는데 중점
- 데이터 디스커버리, 마이닝, 상관관계 식별 포함
- 판매 감소 or 고객 불만 급증과 같은 추세 또는 사고의 근본 원인 파악
#### 3️⃣ 예측 분석(Predictive Analysis)
- 과거 데이터를 기반으로 미래 추세나 결과를 예측
- 통계 모델, 머신 러닝 및 예측 기술을 사용하여 판매 및 고객 행동을 더욱 효과적으로 예측하고 위험을 관리할 수 있음
#### 4️⃣ 처방 분석(Prescriptive Analysis)
- 데이터에 기반하여 조치를 권장
- 예측 분석과 최적화 알고리즘을 결합해 공급망 최적화, 마케팅 캠페인 전략 및 리소스 할당 결정을 지원하는 최상의 행동 방침을 제안함
#### 부가적인 분석들
- 탐색 분석
- 추론 분석
- 정성적 분석(비수치 데이터)
- 정량적 분석(수치 데이터)
- 실시간 분석(사기 탐지, 재고 관리 등)
#### ex) 넥슨 케이스


- 기업 회계, 재무제표에 적용한다면 -> 오류 없는 재무제표 작성을 위한 내부회계관리제도의 구축, 오류 조기 탐지 시스템 구축의 중요성(애초에 재무제표가 잘 만들어질 수 있게끔 하는 것!)

## 2. **목표 정의**

### 2-1. 지표
- 지표: 방향, 목적, 기준 따위를 나타내는 표지. 측정하고 싶은 것을 숫자로 표현한 것. 보통 분모 대비 분자로 표현됨.

| 유형 | 공식 구조 | 예시 지표 |
| --- | --- | --- |
| 효율성(Efficiency) | 산출 / 투입 | 노동생산성 = 생산량 / 노동시간 |
| 효과성(Effectiveness) | 달성 결과 / 목표 | 목표 달성률 = 실제 매출 / 목표 매출 |
| 수익성(Profitability) | 이익 / 자원 | ROE = 순이익 / 자기자본 |
| 성장성(Growth) | 증가분 / 기존 규모 | 매출 성장률 = (당기매출-전기매출) / 전기매출 |
| 안정성(Stability) | 보유자산 / 부채 | 부채비율 = 부채 / 자기자본 |

- 경영학에서 분류하는 기능별 주요 경영 비율

| 기능 | 효율성 지표 (Output ÷ Input) | 효과성 지표 (Actual ÷ Target) | 수익성/가치 지표 (Profit ÷ Resource) | 성장성/안정성 지표 |
| --- | --- | --- | --- | --- |
| 재무 | 자산회전율 = 매출액 ÷ 총자산 | 예산 집행률 = 실제 지출 ÷ 계획 지출 | ROE = 순이익 ÷ 자기자본ROA = 순이익 ÷ 자산 | 매출 성장률, 부채비율 |
| 마케팅 | CAC 효율 = 신규고객 수 ÷ 마케팅비 | 전환율 = 구매자 ÷ 방문자 | 마케팅 ROI = 캠페인 이익 ÷ 캠페인 비용 | 시장점유율 성장률 |
| HR (인사) | 노동생산성 = 산출량 ÷ 인력투입 | 교육 효과 달성률 = 실제 역량향상 ÷ 목표 역량향상 | HR ROI = 교육 후 성과 증가 ÷ 교육비 | 이직률, 직원 유지율 |
| 운영 | 설비가동률 = 실제 가동시간 ÷ 총 가능시간 | 품질목표 달성률 = 합격품 ÷ 목표 품질수준 | 단위당 원가 절감율 = 절감액 ÷ 총원가 | 재고회전율, 공급망 안정성 |
---
### 2-2. 지표를 정의하기
- 분모, 분자에 들어갈 절대적인 수치를 구할 수 없을 때 어떻게 해야할까?
- Proxy Metric을 설정한다!
- Proxy Metric : '활성화', '인게이지'와 같이 추상적이어서 나름의 가정, 정의, 논의가 필요한 지표
- Proxy(대체물)을 측정한다는 의미로, 정답이 별도로 존재하지 않음.

#### ex) 스티비
- 회원 가입한 유저들 중 몇 퍼센트가 우리 프로덕트의 핵심가치(Aha Moment: 사용자가 가치를 체감하고 재방문,유료전환,추천이 눈에 띄게 좋아지는 결정적 순간)를 느끼고 있는지 알고 싶음
- 회원가입한 사용자 중 이메일 편집기에 진입한 사용자의 비율을 가지고 측정함!
---
### 2-3. 북극성 지표 L1/L2 Metrics
- 북극성 지표: 기업이 가장 중요하게 생각하는 지표
- L1/L2 지표
    - L1: 전사적인, 고수준의 성과 지표(전략 관점)
    - L2: L1 지표의 세부 사항을 측정하는 지표(전술 관점)
- 예시(Uber eats)  


### 2-4. AARRR
- 넷플릭스, 유튜브, 스포티파이
    - 컨텐츠 감상 시간이 중요한 지표
    - 오래 감상할수록 광고 수익, 유저 리텐션에 좋은 영향을 끼침
- 우버
    - 공급자와 수요자를 매칭하는 서비스
    - 택시를 부르는 사람들의 요청 중 몇 퍼센트가 실제로 매칭되는지가 중요
- -> **프로덕트가 가치를 제공하는 방식이 다르면, 측정하는 지표도 달라져야 함!**

- **AARRR의 로드맵**
  

1️⃣ Acquisition
- 사용자 획득, 즉 사용자가 여러 채널을 통해 앱 또는 웹사이트에 방문하는 것
- 채널은 광고, 검색엔진 최적화, 지인 추천, 오프라인 행사 등 다양함
- 신규고객 어떻게 확보? 어디로부터 획득? 어떤 채널을 통해 획득? 이 포인트
- 지표: 조회수, 가입 사용자 수, 유저 획득 비용, 클린 한 건당 지출비용, 광고로 인한 매출액/광고비 등

2️⃣ Activation
-
