# 1. EDA
- 데이터 분석(모델링)을 위해 데이터를 여러 가지 방식으로 파악하는 모든 과정
- 시각화 같은 도구를 통해 패턴 발견 / 데이터의 특이성 확인 / 통계와 그래프로 가설을 검증하는 과정 등
- 주어진 데이터를 분석하여 '탐색과 이해'과정을 거치는 것!

## (1) EDA 대상 (일변량/다변량)
 - 일변량 : EDA를 통해 한 번에 파악하려는 변수가 '한 개' / 데이터 설명, 패턴 발견 목적
 - 다변량 : EDA를 통해 한 번에 파악하려는 변수가 '여러 개' / 여러 변수들 간의 관계 파악 목적

 ## (2) EDA 종류 (시각화/비시각화)
 - 시각화 : 차트 혹은 그림 등을 이용하여 데이터 확인 / 대략적 파악
 - 비시각화 : 주로 SUmmary Statistics를 통해 데이터 확인 / 정확한 값 파악

 ## (3) EDA 유형
 - 일변량 비시각화 : 주어진 데이터의 Distribution 확인
 - 일변량 시각화 : 주어진 데이터를 전체적으로 살펴보는 것
 - 다변량 비시각화 : 주어진 둘 이상의 변수 간 관계를 확인
 - 다변량 시각화 : 주어진 둘 이상의 변수 간 관계를 전체적으로 살펴보는 것

# 3. 전처리
- 데이터를 분석에 적합한 형태로 만드는 과정
## <결측값 처리>
- 결측값 = 데이터 수집 과정에서 측정되지 않거나 누락된 데이터
### (1) 결측값 확인
 - 데이터프레임 상에서 `NaN`, `?`, `0` 등의 값으로 나타나 있는 결측값을 확인
 - 결측값 관련 다양한 pandas function
    - `info()` : 데이터 프레임의 요약 정보를 출력 -> 각 열에 속하는 유효한 값 개수를 보여줌
    - `value_counts(dropna=False)` : 각 열의 결측값을 포함한 전체 데이터 확인 가능
    - `isnull()` : 누락 데이터면 True, 유효한 데이터면 False 반환, `df.isnull().sum()` 형식으로 자주 사용!
    - `notnull()` : 유효 데이터면 True, 누락 데이터면 False 반환
    - `replace()` : 결측값이 NaN이 아니라 '0'이나 '?'등으로 입력되기도 하는데, 이때 replace를 활용하여 NaN으로 변환 가능
### (2) 결측값 처리
 **a. 삭제**
 :데이터가 있는 행 또는 열을 삭제
- `dropna()`: DataFrame 내의 결측값이 포함된 레이블을 제거하는 매서드
 ```python
DataFrame.dropna(axis=0/1, how='any'/'all', subset=[col1, col2, ...], inplace=True/False)
```
- <삭제 기준> `axis = 0` & `axis = 1`  
결측값이 있는 행 삭제 / 열 삭제
- <삭제 조건> `how = 'any'` & `how = 'all'`  
하나라도 결측값이 있으면 삭제(디폴트 값) / 모든 값이 결측값일때만 삭제
- <특정 column(열) 기준으로 삭제> `subset = ['column명']`  
`axis=0`(행 삭제)일 떄만 적용가능
- <원본 데이터 변경 여부> `inplace = False` & `inplace = True`  
반영한 내용을 새로운 DataFrame으로 반환(디폴트 값) / 원본 DataFrame 자체를 변경하여 반영  
 **b. 대체**
 : 결측값을 다른 값으로 대체
 - 일괄 대체 : 모든 변수들 일괄적으로 같은 값으로 대체
 - 유사 유형 대체 : 범주형 변수들을 활용해 유사한 범주에 따라 다른 값으로 대체
 * `fillan()` : 결측값 대체
    - 일괄 대체

## 이상치 처리
- 이상치 : 관측된 데이터의 범위에서 많이 벗어난 값
### (1) 이상치 확인
- a. 통계를 통해 확인: `describe()`
- b. 시각화를 통해 확인: BoxPlot
- c. Z-score을 통해 확인
    - 데이터를 평균(0)과 표준편차(1)로 정규화하여, 평균으로부터 얼마나 떨어져있는지 나타냄
    - 평균에 가까울수록 0에 가깝고 멀어질수록 Z-score가 커짐
    - Z-score가 특정 기준값을 넘어가는 데이터를 이상치로 간주
- d. Tukey Fences을 통해 확인: 사분위 범위를 기반으로, 두 가지 경우에 이상치라고 판단
    - Q1 - (1.5*IQR) 미만
    - Q2 + (1.5*IQR) 초과
### (2) 이상치 제거 및 처리
- a. 전체 삭제
- b. 다른 값으로 대체
- c. 변수화 : 이상값을 변수화하여 삭제하지 않고 모델에 포함
- d. 리샘플링 : 해당 이상값 분리해서 모델 만들기

## 피처 엔지니어링 (변수 가공)
: 해결하고자 하는 문제를 컴퓨터가 잘 이해할 수 있도록 피처(변수)들의 형태를 변형하거나 적절하게 처리하는 과정
* feature = 입력 변수 = column
- 기존의 데이터에 정보를 추가하는 일련의 과정
- 피처 엔지니어링 방식
    1. 레이블인코딩, 원핫인코딩: 문자열 값들을 숫자형으로 변경하는 작업
    2. 구간화(Binning): 연속적인 값을 일정한 구간(Bin)으로 구분
    3. 파생변수 생성: 기존의 피처를 변환해 새로운 피처 생성
    4. 스케일링: 서로 다른 변수의 값 범위를 일정한 수준으로 맞추는 작업

### 1) 레이블인코딩(Label Encoding) vs 원핫인코딩(One-hot Encoding)
- 텍스트를 숫자로 바꾸는 작업
- Label Encoding
    - 범주형 변수를 0부터 N-1까지의 숫자로 변환
- One-Hot Encoding
    - 범주형 변수를 이진 벡터(0과1)로 변환
    - `pd.get_dummies(데이터프레임, columns=[컬럼 리스트])` : 명목변수를 0과 1로 원-핫 인코딩

### 2) 구간화(Binning)
- 연속형 데이터를 일정한 구간으로 나눠서 분석
- ex)신용카드 플래티넘, 다이아몬드, 골드, 에메랄드 등급 구분
- `pd.cut(데이터프레임['컬럼명'], bins=[나누는 기준 리스트], labels=[지정할 label])`

### 3) 파생변수 생성
- 기존의 피처를 다른 피처로 변환(재정의)하여 변수를 추가함

### 4) 스케일링
- 서로 다른 변수의 값 범위를 일정한 수준으로 맞추는 작업
- 정규화 작업 수행
- a. StandardScaler(표준화)
    - 각 feature의 평균을 0, 분산을 1로 변경
- b. MinMaxScaler
    - 모든 feature가 0과 1사이에 위치하게 만듦
-`StandardScaler()` `MinMaxScaler()`
    - `sklearn` 라이브러리에서 원하는 스케일러 `import`  
원하는 스케일러를 `scaler` 변수에 저장  
해당 `scaler` 변수로 기존 컬럼을 `fit_transform` 해주면 됨 !



# 4. 데이터 시각화
- 데이터 시각화는 데이터에 대한 이해를 돕기 위해 그림, 도형 등 그래픽 요소들을 이용해 데이터를 묘사하고 표현하는 과정
- 숫자와 텍스트로 표현되는 정보를 그래프, 차트, 그림, 도표 등의 시각적 요소로 변환하고, 이를 통해 데이터 패턴, 관계, 추세 등을 쉽게 파악 가능
- How?: Matplotlib과 Seaborn을 사용함
    - Matplotlib: Python 프로그래밍 언어 및 수학적 확장 Numpy 라이브러리를 활용한 플로팅(그래프를 그려내기 위한) 라이브러리
    - Seaborn: matplotlib을 기반으로 만들어져 통계 데이터 시각화에 최적화된 인기 라이브러리
- 시각화를 하기 전 데이터가 범주형인지 수치형인지 파악하고, 결측값 및 이상치가 있는지 확인해야 함. 데이트 특성 파악 후 어떤 그래프를 그려 데이터를 바라볼지 고민해야 함.

## <파라미터>
- 컴퓨터 프로그래밍에서 매개변수(파라미터)란? -> 프로그래밍된 함수의 입력값
- 예시
    - `plt.figure(figsize=(5,4))`  
    `sns.countplot(data=data, x='Survived')`
    - 위 코드의 `data=data, x='Survived'` 에서,
        - **data**와 x는 각각 countplot의 매개변수(parameter)
        - data와 'Survived'는 실제 전달되는 인자
        - 즉, 그래프를 그릴 때 **data**로서 data를 활용하고 x축에 표현할 칼럼을 'Survived'로 지정한다는 의미
        - 이처럼 함수를 내가 원하는 조건에 맞춰 이용하기 위해 `파라미터=인자`형식으로 전달함
- `hue` 파라미터에는 성별(Sex)과 같은 범주형 변수를 넣어, 내가 원하는 변수를 기준으로 데이터를 구분하여 그래프에 표시해줄 수 있음
    - ex) `data=data, x='Survived', hue='Sex')`

## <다양한 그래프 톱아보기>

### (1) boxplot(상자 수염 그림)

boxplot은 아래와 같이 사분위수와 이상치를 보여주는 그래프



### (2) countplot, histplot

countplot은 범주형 변수의 빈도수를 확인하는 그래프



histplot은 도수분포표를 그래프로 나타낸 것으로, 수치형 변수의 구간별 빈도수를 보여줌



### (3) displot kdeplot (커널밀도추정 그래프)

displot과 kdeplot은 히스토그램을 연속적으로 곡선으로 연결한 그래프



### (4) barplot, pointplot

barplot은 범주형 데이터 값 x에 따른 수치형 데이터 값 y의 평균 값을 제공  
pointplot은 막대 그래프와 모양만 다르고 동일한 정보를 제공





### (5) scatterplot(산점도 그래프), regplot(회귀선이 추가된 산점도그래프)

scatterplot은 두 연속형 변수 간의 관계를 시각화 -> 변수 간의 패턴, 상관관계, 부포 등을 쉽게 파악할 수 있음!



regplot은 두개의 연속 변수 사이의 산점도를 그리고 회귀선을 함께 나타냄



### (6) catplot

catplot은 categoryplot의 준말로, 수치형 데이터와 범주형 데이터의 관계를 볼 때 주로 사용



### (7) pieplot

pieplot은 데이터의 부분과 전체 간의 비율을 표현하는 그래프



### (8) heatmap

heatmap은 변수간 상관계수를 직관적으로 볼 수 있는 그래프  
`corr()` 메서드로 변수간의 상관계수를 구하고 이를 히트맵에 표현 가능

상관계수란 두 수치형 변수 사이의 상관 관계의 정도를 수치적으로 나타낸 계수
- 1에 가까울수록 양의 상관계수가 큰 것, 비례
- -1에 가까울수록 음의 상관계수가 큰 것, 반비례



### (9) violinplot

바이올린 플롯은 박스 플롯과 커널밀도추정 함수 그래프를 합쳐 놓은 그래프

```python
sns.violinplot(x='class', y='age', hue='sex', data=data, split=True)
```
