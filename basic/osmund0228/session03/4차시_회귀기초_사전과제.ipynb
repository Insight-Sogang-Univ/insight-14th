
# 머신러닝과 모델링

## 1. 머신러닝
- 인공지능의 한 분야
- 데이터에서 패턴을 찾아내고, 이 패턴을 바탕으로 새로운 데이터에 대한 예측 가능

💡 **머신러닝의 종류**
|	| 정답 여부 |	방식 |	예시 |
| --- | --- | --- | --- |
|지도 학습 |	정답(label)이 있는 데이터 사용 |	예측값(prediction)을 이미 만들어둔 정답과 같아지도록 기계를 학습시키는 것(=문제와 정답을 모두 알려주고 공부시키는 방법) | 	회귀, 분류 |
|비지도 학습 |	정답(label)이 없는 데이터 사용 |	데이터 속의 패턴 또는 각 데이터 간의 유사도를 기계가 학습하도록 하는 것(=답을 가르쳐주지 않고 공부시키는 방법) | 군집화 |
|강화 학습|   |	 시행착오를 반복하여 정답을 찾는 과정(=보상을 통해 상 최대화, 벌 최소화 방향으로 행위를 강화하는 학습)| 알파고 |

**비지도학습(ex.군집화)**
- 정답이 없는 데이터를 통해 그 속의 패턴 또는 각 데이터 간의 유사도를 기계가 학습하도록 하는 것
- 지도학습과는 다르게 정답이 있지 않고 주어진 데이터로만 학습함(target/label 없이 feature만 있는 데이터)
- 데이터를 스스로 학습하여 그 속의 패턴 또는 각 데이터 간의 유사도를 학습해 분류되지 않은 데이터에서 구조나 패턴을 찾는 데 도움을 줌
- 비슷한 데이터를 묶어 큰 단위로 만드는 군집화 / 데이터 분포를 예측하는 밀도 추정 / 데이터 차원을 간추리는 차원 축소가 비지도 학습에 해당함
- 군집화 : anomaly detection, 고객세분화에 사용 / 클러스터 수 지정, 수정 가능

## (2) 지도 학습: 회귀와 분류
### [지도학습]
💡**지도학습** : 입력 값과 함께 결과 값(정답 레이블)을 같이 주고 학습을 시키는 방법
- 학습의 방향 : 예측값(prediction)과 정답(label)이 최대한 같아지도록 학습
- 구체적인 과정
    1. labeled data(독립변수(feature)와 종속변수(target/label)가 모두 존재하는 데이터, 즉 입력값과 출력값이 있는 데이터)를 training set와 test set로 나눕니다.
    2. training set를 통해 모델을 학습: 독립변수를 통해 추정한 예측값이 주어진 종속변수와 같아지도록(가까워지도록) 모델을 훈련시킵니다.
    3. test set를 통해 기계의 예측이 우리가 의도한 정답에 가까운지 테스트합니다.

### [회귀와 분류]
- 지도학습에는 대표적으로 연속적인 숫자값을 예측하는 회귀(Regression), 입력된 데이터를 주어진 항목들로 나누는 분류(Classification)가 있습니다.

💡**회귀(Regresssion)**
- 주어진 데이터(X)를 기반으로 정답(Y)를 잘 맞추는(fit)하는 함수(function, 아리 그림의 파란선)을 찾는 문제

- e.g. 주택 가격 (타겟) 예측
    - 종속변수(y축) = 주택 가격
    - 독립 변수(x축) = 주택 평 수, 역까지의 거리, 지역의 평균 소득 등

💡**분류(Classification)**
- 기존 데이터가 **어떤 레이블에 속하는지** 패턴을 알고리즘으로 인지한 뒤에 새롭게 관측된 데이터에 대한 **레이블을 판별**하는 것
- e.g. 대출상환여부 (0/1) (타겟) 예측
    - 종속변수 = 대출상환여부(0/1)
    - 독립변수 = 고객의 특성(신용등급, 소득변화 등), 고객이 받은 대출의 특성
- + 이메일이 스팸인지 여부, 종양이 악성인지 여부도 분류에 해당

💡**회귀와 분류의 차이**
- 회귀 : 데이터가 연속형 변수(종속변수)를 예측하기 위해 사용될 때(ex. 배달시간)
- 분류 : 데이터가 범주형 변수(종속변수)를 예측하기 위해 사용될 때(ex. 당뇨병 여부)

# 회귀(Regression)와 모델링

## 1. 회귀(Regression)란?
- 회귀는 데이터에서 패턴을 찾아내어 미래 값을 예측하는 것
- 통계나 머신러닝의 주요 개념
- 집값 예측, 주식 가격 예측 등과 같이 연속적인 숫자가 목표변수(종속변수)일때 사용

## 2. 회귀의 종류
- 선형회귀, 비선형회귀, 릿지회귀, 라쏘회귀, 다항회귀 등 다양한 회귀 기법 존재
- 이중에서 4차시에는 선형회귀를 다룸!

## 3. 선형 회귀

📌선형회귀란?


- 선형 회귀 : 데이터를 가장 잘 설명하는 **회귀선(ex.직선 y=ax+b)**을 찾는 과정
- 회귀선은 현재 데이터 요약 + 미래 값을 예측하는 도구
- 예측하고자 하는 것: 종속변수(y), 예측을 위해 사용하는 변수: 독립변수(x)

📌선형 회귀의 종류
- 단순 선형 회귀: 데이터를 직선 형태(하나의 독립변수)로 표현하는 경우
- 다중 선형 회귀: 데이터를 두 개 이상의 독립변수를 사용하여 표현하는 경우

### (1) 단순선형회귀분석

- 단순 $\rarr$ 하나의 독립변수(x)만 존재하는 선형회귀
- 데이터를 가장 잘 설명하는 **직선 y=ax+b**을 찾는 과정에 해당
- y = ax + b 에서 a는 회귀계수(기울기) b는 y절편을 의미함
- 직선을 결정짓는 적절한 회귀계수(a)와 절편(b)를 찾는 과정
- 어떻게 a와 b를 잘 찾을 수 있을까?
    - 잔차(residual)를 줄여나가면 잘 찾을 수 있을 것이다!
    - 잔차 : 실제값 - 예측값 간의 차이

📌 최소제곱법(Least Squares Method) : 최적의 회귀선 찾는 방식 
- 최소자승법이라고도 부름

- **잔차의 제곱의 합**이 최소가 되는 지점으로 최적의 회귀선을 구하는 방식
- 잔차 : 실제값과 예측값의 차이
- **핵심: 회귀선과 실제 데이터의 차이를 최소화**하기 위한 시도이다!

####  1️⃣ 경사하강법 (Gradient Descent)
- 최소제곱법과 마찬가지로 y=ax+b에서 a와 b를 찾고자 함
    - 잔차제곱을 줄이려는 목적은 그대로!
    - 계산과정이 복잡할 때 최소로 가는 과정에서 경사하강법을 쓴다!

##### (0) 목적함수, 비용함수, 손실함수
- 목적함수 : 우리가 달성하고 싶은 목표
    - 최적화를 위해 오차를 최소화하는 것이 목적인 함수
    - 즉, 최적화하려는 대상을 수학적으로 표현한 함수
        - 시험점수를 최대화
        - 실수를 최소화
        - $\rarr$ case별로 어떤 방향으로 최적화하는지 달라짐
    - 머신러닝에서 모델의 예측이 얼마나 잘 맞는지 평가하는 기준이 됨

- 손실함수 : 개별 데이터 샘플에 대한 오차를 측정하는 함수
    - 한 개의 데이터 포인트에 대해 모델이 얼마나 틀렸는지 평가
    - 예측값 \hat(y)와 실제값 y의 차이를 계산
    - 예시
        - 회귀분석 $\rarr$ 평균제곱오차(MSE), 평균절대오차(MAE)  
        $\text{(MSE for one sample)} = (y - \hat{y})^2$
        - 분류 문제 $\rarr$ 교차 엔트로피(Cross-Entropy Loss)

- 비용함수 : 전체 데이터셋에서 평균적인 손실을 측정하는 함수
    - 손실함수를 모든 데이터 포인트에 대해 계산한 후 평균 or 합산한 값
    - 머신러닝에서는 대부분 모델을 훈련할 때 비용함수를 최적화(최소화)하는 것이 목표
    - 예시 :
        - MSE를 비용함수로 사용할 경우  
        $\text{(MSE)} = \frac{1}{N} \sum_{i=1}^{N} (y_i - \hat{y}_i)^2$
    - 모델을 학습할 때 대부분은 비용함수를 최소화하는 것이 목표!  
    이때는 비용함수 최소화 = 목적함수로 최적화의 방향을 설정함

👻 머신러닝 과정에서는 모델을 평가하고 최적화하는 함수 찾는게 핵심  
🧐 손실함수와 비용함수는 혼용하는 경우가 많음, 최소화를 목적으로 두고 계산하기 위한 함수임!        


##### (1) 손실함수

- t_i : i번째 데이터포인트 실제값
- y_i : i번째 예측값
- 잔차를 제곱한 뒤 합하는 형태!


- y식을 대입해 최종적인 손실함수를 이렇게 씀
- 여기서 우리의 목표는 E를 최소화하는 W와 b를 찾는 것


- 최소지점을 갖는 손실함수(W로 x축을 구성한 경우, b로 x축을 구성할 수도 있음)


- W값(or b값)이 (E값의)최소지점보다 오른쪽이라면 왼쪽으로 이동해야함
    - 미분계수가 양수일 때, "현재 값 - 미분값(양수)"를 하면 W(b)는 왼쪽으로 이동 (감소)
- W값(or b값)이 (E값의)최소지점보다 왼쪽이라면 오른쪽으로 이동해야함
    - 미분계수가 음수일 때, "현재 값 - 미분값(음수)"를 하면 W(b)는 오른쪽으로 이동 (증가)
- 이렇게 W값(or b값)을 업데이트 해나가고, 그러다 (E값의)최솟값에 도달하면 미분값이 0이 되어 ***"현재 값 -0"***으로 더이상 변화하지 않고 멈춤
- 최솟값 왼쪽의 값은 점점 오른쪽으로 이동하고, 최솟값 오른쪽의 값은 왼쪽으로 이동하며 최종적으로 최솟값에 도달하게 되는 원리!
- 정리



##### (2) 학습률(α: learning rate)
- 얼마만큼 이동할 것인가 = 등산에서 '보폭'
- 경사하강법의 '학습한다' = '오차의 최솟값을 찾아나간다'
- 학습률은 적당한 것이 가장 좋다!
    - Learning Rate가 너무 작다면, 모델이 수렴할 때까지 시간이 너무 많이 걸림
    - Learning Rate가 너무 크다면, 모델은 수렴하지 못하고 발산(손실 값이 계속 커짐)
- 학습률은 개발자가 직접 설정하는 하이퍼파라미터
    - 초기에 작은 값부터 경험적으로 시도
    - 실험을 통해 조정 : 학습이 너무 느리면, 늘리고 / 학습이 발산하면, 줄임

##### (3) Local Minima 문제

- 특히 딥러닝의 경우 비용함수가 정확히 어떤 모양인지 예측할 수 없어서 local minima에 빠지는 경우를 주의해야 하며, local minima을 탈출하는 방법에 대해 여러 알고리즘이 존재합니다.

##### (4) 해결법 - 모멘텀
- 기존 경사하강법은 이전 기울기는 잊어버리고 현재 기울기만 기준으로 움직임  
 $\rarr$ 이전의 기울기, 이동해오던 방향을 고려하여 관성을 부여함!
- 관성을 가진 공처럼 이전 이동방향을 참고하면서 더 부드럽고 빠르게 작은 언덕도 넘을 수 있음

- 맨 오른쪽 공 : 경사하강법이 이동하려고 하는 방향(빨강)보다 이전 단계의 이동을 반영한 모멘텀 방향(초록)이 더 커서 실제로는 오른쪽(파랑)으로 이동함
- 모멘텀의 장점
    - 기울기에 관성을 부여하여 작은 기울기는 쉽게 넘어감
    - 지역 최솟값을 탈출할 수 있음



### (2) 다중선형회귀분석

- 독립 변수 X_i가 2개인 경우 아래와 같이 3차원 공간에 표현됨, 회귀선은 평면

- 파악이 힘들 땐 경사하강법 사용

#### 1️⃣ 다중공선성 (Multicollinearity)
✔ 다중공선성은 다중선형회귀 과정에서 중요한 체크 사항!
- ex) 주택 가격 예측 모델에서 '주택의 크기'와 '방의 개수'라는 두 독립 변수가 강한 **상관관계**를 가질 때, 이 두 변수 간에 다중공산성이 발생함
- 변수들이 겹치는 상황, 즉 독립 변수들 간에 상관관계가 큰 경우에 발생함
- ⇒ 다중공선성이 높은 경우 어떤 독립 변수가 종속 변수에 얼마나 영향을 미치는지를 정확하게 구분하기 어렵기 때문에 회귀 모델은 어떤 변수의 영향을 반영해야 할지 불확실해지고, 그 결과 회귀분석의 정확도가 낮아짐.

📌 다중공선성 확인 방법
1. 상관계수
- r이 -1 혹은 1에 가까울 수록 상관성이 높고, 0에 가까울수록 상관성이 없음
- 히트맵으로 확인 가능
```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.heatmap( df.corr(), annot = True)
plt.show()
```
- corr()함수로도 확인 가능
- pairplot과 같은 시각화로 산점도를 찍어 상관성을 직접 눈으로 확인할 수 있음

2. VIF지수(분산 팽창 인수)
- 회귀 모델의 결정계수 R square를 사용하여 계산됨
- VIF가 높으면 다중공선성이 존재한다고 판단함(다음은 일반적인 가이드라인)
    - VIF = 1 : 해당 독립변수는 다른 변수와 상관관계가 전혀 없음
    - VIF < 5 : 일반적으로 다중공선성 문제가 없다고 간주
    - VIF > 5 : 다중공선성의 징후가 있을 수 있음
    - VIF > 10 : 다중공선성이 심각하다고 간주, 변수 제거 or 상관관계 줄이기 위한 조치 필요

📌 다중공선성 대처 방법 

1. 변수 제거(변수 선택법)
- 독립변수로서 사용할 변수를 선택하는 방법

2. 변수 변환
- 변수들을 더하거나 빼서 새로운 변수 생성
- 독립변수를 더하거나 빼더라도 문제가 없는 경우
- 예) 남편 수입과 아내 수입이 서로 상관성이 높으면, 두 개를 더해 가족 수입이라는 하나의 변수로 투입

3. 규제 선형 모델 활용
- 릿지, 라쏘, 엘라스틱넷 등의 방법을 통해 모델의 복잡도를 줄이는 방법 사용

4. PCA(주성분분석)
- 데이터의 차원을 축소하는 데 사용되는 통계적 기법

- (a)를 확인하면 x2차원은 거의 영향이 없는 것을 확인할 수 있음 -> x2차원을 없앰
- 데이터의 차원이 높으면 분석과 시각화하기 어렵기 때문에 이렇게 줄임

**💡 상관 있는 변수들을 묶어서 더 적은 수의 대표 축으로 바꾸는 것이 PCA의 핵심 아이디어**


🌟 데이터의 방향성을 찾아 변환하고 중요한 정보만 남겨 차원을 축소한다!

- PCA 과정?

    
    1. 데이터 전처리 (정규화)
    
    (b) 평균을 0으로 조정 → 데이터 중심을 원점으로 이동
    
    (c) 표준화 (스케일링) → 분산을 1로 맞춰 단위 영향 제거
    
    2. 주성분 찾기
    
    (d) 공분산 행렬 계산 & 고유벡터, 고유값 추출
    
    고유벡터 → 데이터가 가장 많이 퍼진 방향(주성분 축)
    
    3. 차원 축소
    
    (e) 주성분 축에 데이터 투영 → 중요한 정보만 유지
    
    4. 복원 가능 (선택적)
    
    (f) 원래 좌표계로 변환 가능 (표준화 해제)




## 4. 규제선형모델
- underfitting, good fitting, overfitting 중 good fitting이 가장 좋음
- overfitting된 모델은 학습 데이터(train set)에는 완벽히 맞지만 새로운 데이터에(test set)는 성능이 크게 떨어짐

📌 규제선형모델
- 모델이 학습 데이터에 overfitting되지 않도록 규제를 가하고자 등장
- 선형 회귀 모델에서는 특성에 곱해지는 계수(기울기)의 크기를 조정하는 것
- $y = W_1 x_1 + W_2 x_2 + ... + W_n x_n + b$  
이 다중회귀식에서 $W_n$이 회귀계수임, 회귀계수가 커지면 특정 독립변수(특징)에 과하게 의존한다는 뜻이고, overfitting이 발생함
- 비용 함수는 학습 데이터의 잔차 오류 값을 최소화하는 RSS 최소화 방법(최소제곱법)과 과적합을 방지하기 위해 회귀 계수 값이 커지지 않도록 하는 방법이 서로 균형을 유지해야 함
- 비용 함수 목표 = Min(학습데이터 잔차 오류 최소화 + 회귀계수 크기 제어)
- 비용 함수에 alpha값으로 페널티를 부여해, 회귀 계수 크기를 감소시켜 과적합 개선

📌 규제선형모델의 종류
1. L2규제: W(회귀계수)의 제곱에 대해 페널티를 부여하는 방식, 릿지(Ridge)회귀
2. L1규제: W(회귀계수)의 절댓값에 대해 페널티를 부여하는 방식, 라쏘(Lasso) 회귀
3. L2규제 + L1규제 결합: 엘라스틱넷 회귀

**이번 세션에서 규제선형에 대해서 가져갈 내용**
1. 과적합 방지를 위해 회귀계수에 대한 조치가 필요하다!
2. 이에 회귀계수에 페널티를 부여하는 규제선형모델을 사용한다!
3. 그 방식에 따라 3가지의 규제선형모델 (릿지 회귀, 라쏘 회귀, 엘라스틱넷 회귀)가 존재한다!

### L2규제(릿지 회귀)와 L1규제(라쏘 회귀)
#### (1) 릿지 회귀(L2규제)
- 🔨 L2규제는 기둥 크기에 따라 다른 힘으로 누르는 망치
- 큰 기둥은 세게, 작은 기둥은 살살 눌러서 비율이 균등하지 않을 수 있지만 전체적으로 작아짐

$\text{Cost Function} = \text{RSS} + \alpha \sum_{j=1}^{p} w_j^2$

- 회귀계수($w_j$)의 제곱을 패널티로 주기 때문에, 큰 회귀계수일 수록 훨씬 강한 패널티 받음
- 모든 기둥(모든 계수)의 크기는 작아지지만, 완전히 사라지지는 않음(회귀계수가 0이 되지 않음) -> 변수 제거 X
- 모든 변수를 유지하면서도, 계수의 크기를 적절히 줄여 과적합 방지!

#### (2) 라쏘 회귀(L1규제)
- 🔨 L1규제는 모든 기둥을 같은 힘으로 누르는 망치 
- 작은 기둥들은 완전히 사라질 수도 있음

$\text{Cost Function} = \text{RSS} + \alpha \sum_{j=1}^{p} |w_j|$

- 회귀계수($w_j$)의 절댓값을 패널티로 주기 때문에, 회귀계수의 크기와 상관없이 같은 패널티 받음
- 작은 기둥(작은 계수)은 힘을 이기지 못해 0이 되기도 함 → 일부 기둥이 완전히 사라짐(변수 선택 기능)
- 비중이 낮은 변수(작은 계수)는 아예 사라지고, 중요한 변수만 남음!

#### (3) 엘라스틱넷(L1+L2)
- L1 + L2 망치 동시에 사용
    - L1 망치 : 작은 기둥을 아예 부숴서(변수 삭제) 불필요한 변수를 제거
    - L2 망치 : 남은 기둥을 균형 잡히게 눌러 크기를 조절

$\text{Cost Function} = \text{RSS} + \alpha_1 \sum_{j=1}^{p} |w_j| + \alpha_2 \sum_{j=1}^{p} w_j^2$

- L1의 변수 선택 기능(일부 계수를 0으로 만듦)과 L2의 규제 기능(큰 계수를 줄임)을 동시에 활용하는 모델!
- L1은 상관관계가 높은 변수 중 일부만 선택하는 경향이 있음  
    → L2를 함께 적용하면 상관관계가 높은 변수를 함께 고려하면서도 최적의 규제가 가능!
- **안정적인 예측(모든 변수가 균형 있게 줄어듦)과 희소성(변수 제거)을 동시에 달성할 수 있음!**

#### 정리
1. 릿지회귀(L2 규제) = 회귀계수의 제곱 합에 페널티 부여

    (1) 예측 변수가 많을 때 -> 고차원 데이터에서 모든 변수에 작은 가중치를 부여하고자 할 때  
    (2) 다중공선성 존재할 때 -> 변수 선택이 필요하지 않은 상황

2. 라쏘회귀(L1 규제) = 회귀계수의 절댓값 합에 페널티 부여

    (1) 예측 변수 수가 많고, 그 중 일부만이 실제로 중요할 때 -> 계수 중 일부를 정확히 0으로 만들어 변수 선택 효과 부여  
    (2) 모델의 해석을 간단하게 유지하고자 할 때 -> 불필요한 변수를 제거하고 모델을 단순화

***L1 규제를 사용하는 라쏘는 최적화 문제가 비선형적이고 복잡할 수 있으며, 특정 상황에서는 수치적으로 불안정할 수 있습니다. L2 규제를 사용하는 릿지는 수치적으로 더 안정적입니다.***

3. 엘라스틱넷 (L1+L2 규제) = 릿지의 계수 축소 기능과 라쏘의 변수 선택 기능 모두 가짐
    
    (1) 예측 변수 수가 많고, 그 중 중요한 변수를 선택하면서도 다중공선성을 관리할 때 -> 변수 간의 상관관계가 높은 경우 릿지만 사용할 때보다 더 나은 성능

------
모든 계수에 대해 균등한 페널티를 부과하여 모델의 복잡성을 줄이기 위해서 규제는 스케일링이 꼭 필요하며, StandardScaler와 Min-Max scaler을 주로 이용


## 5. 모델평가방법

📌 모델 성능 및 유의성 평가

### (1) 성능 평가지표

1️⃣ 평균 제곱 오차 (Mean Squared Error, MSE)
- MSE는 Mean Squared Error의 약어
- 회귀분석에서 모델의 예측값과 실제 관측값 사이의 오차 제곱의 평균을 의미
- 모델의 예측 성능을 평가하는 지표 중 하나
- $\text{MSE} = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2$

2️⃣ 평균 절대 오차 (Mean Absolute Error, MAE)
- MAE는 Mean Absolute Error의 약어
- 회귀 분석에서 모델의 예측값과 실제 관측값 사이의 절대값 오차의 평균을 의미
- 모델의 예측 성능을 평가하는 지표 중 하나
- $\text{MAE} = \frac{1}{n} \sum_{i=1}^{n} |y_i - \hat{y}_i|$


### (2) 변수 유의성 평가

1️⃣ t 검정

1. 개념
- t-검정은 독립 변수의 회귀계수가 유의미한지 검정하는데 사용됨
    - 귀무가설 ($H_0$): 회귀계수(a)가 0이다 → 독립변수가 종속변수에 영향을 주지 않는다.
    - 대립가설 ($H_1$): 회귀계수(a)가 0이 아니다 → 독립변수가 종속변수에 영향을 준다.

2. 검정 과정

(1) t-값 계산
- $t = \frac{\hat{\beta} - 0}{SE(\hat{\beta})}$
- $\hat{\beta}$ : 추정된 회귀계수
- $SE(\hat{\beta})$ : 표준오차
- t-분포를 따름.

(2) p-value 확인 및 판단
- **p-value < 0.05** → 유의 수준 5%에서 $H_0$ 기각 → **해당 변수는 유의미하다.**
- **p-value ≥ 0.05** → $H_0$ 기각 불가 → **해당 변수는 유의미하지 않다.**

(3) 해석
- 회귀계수가 0이면 해당 독립변수는 의미 없음
- p-value가 작을수록 해당 변수는 종속 변수에 더 큰 영향을 줄 확률이 높아짐
- t-값이 크면 귀무가설을 기각할 가능성이 높아짐

**📌t-검정을 통해 독립 변수가 종속 변수에 유의미한 영향을 주는지 판단할 수 있음!**
