# INSIGHT 교육세션 사전과제
## 09. 데이터로 보는 비즈니스 상황

## <span style="background-color:#F993C0; color:#311d26; padding:4px; border-radius:0px;"> 데이터 기반 의사결정 </span>

#### <span style="background-color:#633a4c; color:#fde9f2; padding:4px; border-radius:0px;"> 무엇을 위한 데이터 분석인가 </span>

### 데이터 분석이 지닌 가장 큰 특징: 목적 달성 프로세스의 질!



<span style="color:#f993c0;"> 액션 </span>

액션은 크게 결정, 계획, 행동으로 나누어짐.  
→ 이 액션들을 위해 데이터를 분석함!

#### <u>객관적으로 과거를 되돌아보고 미래를 전망하는 것</u>

<span style="color:#f993c0;"> 결정 </span>

Ex. 액션: ‘신제품 출시에 관한 의사결정을 내린다’
- ‘지금 상태로 시제품을 출시해야 할지'
- ‘언제 출시해야 할지’
- ‘어떤 프로모션을 할지’
- '얼마나 팔릴 것으로 예상되는지’

→ 결정을 뒷받침하기 위한 데이터 분석을 수행

<span style="color:#f993c0;"> 계획 </span>

Ex. 액션: ‘미디어 광고나 전단진 배포 등의 프로모션을 계획한다’
- ‘매출과 미디어의 관계 분석’
- ‘매출 시뮬레이션’

→ 수치적 근거에 기반한 계획 수립을 위한 데이터 분석을 수행

<span style="color:#f993c0;"> 행동 </span>

Ex. 액션: ‘전단지를 배포한다’
- ‘계절성·기온·지역·매장 특성과 전단지의 관계’
- ‘매출과 전단지 크기의 관계'
- ‘매출과 전단지 배포 요일의 관계'
- ‘비용 대비 효과를 극대화하는 전단지 매수’

→ 일별 계획 실행에 활용할 수 있는 데이터 분석을 수행

#### <span style="background-color:#633a4c; color:#fde9f2; padding:4px; border-radius:0px;"> Data-Driven-Decision-Making(DDDM) </span>

: 비즈니스 의사 결정을 제공하기 위해 직관 대신 <u>데이터와 분석을 사용</u>하는 것을 강조하는 접근 방식

#### <span style="background-color:#633a4c; color:#fde9f2; padding:4px; border-radius:0px;"> 6 Steps of DDDM </span>

<span style="color:#f993c0;"> 1. 목표 정의 </span>

- 조직의 목표를 명확하게 표현하는 것   
- 목표가 정의된 후 회사는 목표를 달성하기 위해 집중적이고 목적이 명확한 노력을 기울임

<span style="color:#f993c0;"> 2. 데이터 식별, 준비 및 수집 </span>

- 명확한 목표를 설정하고 데이터 요구 사항을 판별하며 데이터 소스를 평가 및 준비
- 이후 체계적으로 데이터를 수집 및 검증

<span style="color:#f993c0;"> 3. 구성 및 탐색 </span>

- 새로운 패턴, 추세 및 가치 있는 인사이트를 발견할 수 있도록 데이터 구조화 
- 데이터를 시각화해서 원시 데이터에서 즉시 명확하지 않은 패턴, 이상치 및 추세를 식별

<span style="color:#f993c0;"> 4. 데이터 분석 수행 </span>

- 다양한 기술과 방법론을 사용하여 원시 데이터를 실행 가능한 인사이트로 변환해 비즈니스 전략에 영향을 미치는 패턴, 상관관계 및 추세를 파악
- 데이터 분석을 수행함으로써 조직은 전략적 결정을 내리고 전반적인 성과를 개선

<span style="color:#f993c0;"> 5. 결론 도출 </span>

- 주요 데이터 분석 결과를 검토하고 올바른 비즈니스 상황에 배치하여 실행 가능한 인사이트와 권장 사항을 구성

<span style="color:#f993c0;"> 6. 구현 및 평가 </span>

- 데이터 기반 권장 사항을 기반으로 실행 계획을 개발하고 리소스를 할당하며 이니셔티브의 진행 상황을 지속적으로 모니터링
- 사전 정의된 핵심성과지표(KPI)를 기준으로 결과를 측정하고 결과를 분석하며 피드백을 수집하고 해당 정보를 기반으로 지속적인 모니터링 및 반복적인 개선 주기를 통해 조정을 수행

#### <span style="background-color:#633a4c; color:#fde9f2; padding:4px; border-radius:0px;"> 데이터 분석 유형 </span>

<span style="color:#f993c0;"> 1. 설명 분석 </span>

- 데이터 취합 및 마이닝을 통해 과거 데이터를 설명하고 요약하여 과거 성과에 대한 인사이트를 제공하는 것이 목표
- 월별 판매 보고서, 고객 만족도 설문 조사 및 웹 사이트 트래픽 분석을 작성하는 데 사용

<span style="color:#f993c0;"> 2. 진단 분석 </span>

- 특정 이벤트가 발생한 이유를 파악
- 데이터 디스커버리, 마이닝 및 상관관계 식별이 포함
- 판매 감소 또는 고객 불만 급증과 같은 추세 또는 사고의 근본 원인을 파악

<span style="color:#f993c0;"> 3. 예측 분석 </span>

- 과거 데이터를 기반으로 미래의 추세나 결과를 예측
- 기업은 통계 모델, 머신 러닝 및 예측 기술을 사용하여 판매 및 고객 행동을 더욱 효과적으로 예측하고 위험을 관리

<span style="color:#f993c0;"> 4. 처방 분석 </span>

- 데이터에 기반하여 조치를 권장
- 예측 분석과 최적화 알고리즘을 결합하여 공급망 최적화, 마케팅 캠페인 전략 및 리소스 할당 결정을 지원하는 최상의 행동 방침을 제안

## <span style="background-color:#F993C0; color:#311d26; padding:4px; border-radius:0px;"> 목표 정의 </span>

#### <span style="background-color:#633a4c; color:#fde9f2; padding:4px; border-radius:0px;"> 지표 </span>

#### 방향이나 목적, 기준 따위를 나타내는 표지
→ 추상적인 개념과 관련되는 지수, 척도 등을 말하며 측정하고 싶은 것을 숫자로 표현한 것

#### <span style="background-color:#633a4c; color:#fde9f2; padding:4px; border-radius:0px;"> 지표 정의 </span>

Proxy Metric: 활성화와 인게이지와 같이 추상적이여서 프로덕트마다 나름의 가정과 정의, 논의가 필요한 지표   
→ 절대적인 수치를 구할 수 없을 때 사용!

#### <span style="background-color:#633a4c; color:#fde9f2; padding:4px; border-radius:0px;"> 북극성 지표 </span>

**북극성 지표:** 기업이 가장 중요하게 생각하는 지표.  
Ex. 에어비엔비 - 예약된 박 수, 페이스북 - DAU

**L1 / L2 지표:**
- L1: 전사적인, 고수준의 성과 지표(전략 관점)
- L2: L1 지표의 세부 사항을 측정하는 지표(전술 관점)

#### <span style="background-color:#633a4c; color:#fde9f2; padding:4px; border-radius:0px;"> AARRR </span>

#### 프로덕트가 가치를 제공하는 방식이 다르면, 측정하는 지표도 달라져야 함!



<span style="color:#f993c0;"> Acquisition </span>

- 사용자 획득, 즉 사용자가 여러 채널을 통해 여러분의 앱 혹은 웹사이트에 방문하는 것을 의미
- 고객이 유입되는 채널은 광고, 검색엔진 최적화(SEO), 지인 추천, 오프라인 행사 등 매우 다양함
- ‘신규 고객을 어떻게 획득할지’, ‘어디로부터 획득할지’, ‘주로 어떤 채널을 통해 획득할지’를 고민

→ 어떤 제품인지에 따라 고객 획득을 측정하는 기준이 달라지지만, 대체로 **조회수, 가입 사용자 수 등 지표**를 통해 측정

Ex.
- CAC (Customer Acquisition Cost): 유저 획득 비용
- CTR (Cost Per Click): 클릭 한 건당 얼마 지출
- CPC (Cost Per Install): 인스톨 당 과금되는 상품
- CPL (Cost Per Lead): 유저 행동 당 드는 비용
- ROAS (Return On Ads Spending): 광고로 인한 매출액/광고비

<span style="color:#f993c0;"> Activation </span>

- 사용자 활성화, 사용자가 앱 또는 웹사이트에 첫 방문하여 주요 기능을 직접 사용하는 것을 의미
- 사용자에게 제품의 가치를 알림으로써 활성화 정도를 점차 높여나가야 함

→ 제품의 성격에 따라 다양한 기준으로 측정
    - 회원가입 10일 내 최소 7명의 친구를 가진 신규 사용자
    - 회원가입 3일 내 최소 1건을 주문한 신규 사용자
    - 제품 핵심 기능의 활용도, 사용자의 온라인 접속 시간 등

4단계로 구성
- Sign-up: 유저가 제품에 들어온 시점
- Setup Moment: 유저가 제품의 핵심가치를 경험할 준비 완료
- A-HA! Moment: 유저가 핵심가치를 처음 경험한 순간
- Habit Moment: 유저가 제품 사용에 대한 습관 형성

<span style="color:#f993c0;"> Retention </span>

- 유지 단계에서는 신규 사용자를 장기 사용자로 전환시키는 것에 초점
- 어떻게 장기 고객을 양성할 것인지, 또 어떻게 고객의 지속적인 구매 행위를 증가시킬 수 있을지 고민
- 측정 지표: 익일 유지율, 1주차 유지율, 2주차 유지율, 월 유지율, EDM 메일 개설율 등

Ex.
- DN Retention: N일 안으로 재접속하는 경우
- MAU/DAU/WAU: 일/주/월마다 방문하는 횟수
- Stickness(DAU, MAU): 해당기간 동안 활성화 수 (*점성)

<span style="color:#f993c0;"> Revenue </span>

- 수익 단계는 사용자가 여러분의 제품 혹은 서비스를 유료 구매하기로 결정한 단계
- 사용자의 소비를 유도해 매출을 높이는데 총력을 기울여야 하며, 최대한 사용자를 소비하게끔 만들거나 매출을 늘릴 수 있는 행위를 하도록 유도해야 함
- 측정 지표는 제품 성격에 따라 달라질 수 있음
    - 유료화 상품 판매 중심의 수익 구조라면, 사용자가 구매한 게임 아이템, 스킨의 소비액이 주요 측정 지표
    - 광고 시청료 위주의 수익 구조라면, 광고주가 지불하는 광고비가 주요 지표
- 고객의 평생 가치(LTV)를 지속적으로 높여 나가고, 고객 유치 비용은 계속해서 줄여나가야 함

Ex.
- PU(Paying User): 전체 사용자 중에 결제를 한 사용자 비율
- Average Check: 구매 1회당 평균 금액 (Revenue/Transaction)
- ARPU(Average Revenue Per User): 인당 평균 매출
- ARPPU(Average Revenue Per Paying User): 결제자당 평균 매출
- LTV(Life Time Value): 유저가 진입부터 이탈까지 누적 수익

<span style="color:#f993c0;"> Referral </span>

- 제품의 지속적인 성장을 위한 매우 중요한 단계로, ‘어떻게 충성도 높은 사용자를 보이지 않는 브랜드 홍보자로 만들 것’이며, 또 ‘어떻게 제품을 잠재 고객들에게 추천하도록 만들 것’인가를 고민해봐야 함
- 측정 지표: ‘순 추천고객 지수(NPS)’, ‘바이럴 지수(Viral Coefficient)’
    - 브랜드에 대한 고객의 만족도와 사용자가 평균적으로 추천하는 고객의 수

Ex.
- 바이럴 계수: 기존 고객 한명당 데리고 온 새로운 고객의 수
    - 바이럴 계수가 1이 넘으면 Acquisition 단계를 하지 않아도 서비스가 성장
- 바이럴 주기: 기존 고객 1명이 새로운 고객을 데려오는데 걸리는 시간
