# INSIGHT 교육세션 사전과제
## 08. 추천시스템 및 연관분석 기초

## <span style="background-color:#F993C0; color:#311d26; padding:4px; border-radius:0px;"> 추천 시스템이란? </span>

#### <span style="background-color:#633a4c; color:#fde9f2; padding:4px; border-radius:0px;"> 데이터의 종류와 특징 </span>

<span style="color:#f993c0;"> 명시적 데이터 </span>

: 고객들에 의해 입력된 데이터, 사용자가 선호도를 직접적으로 표현한 데이터   
→ 당시의 사용자의 취향이 정확하게 반영된다는 장점이 있으나, 데이터 자체를 얻기에는 힘들다는 단점 존재

Ex) 평점, 좋아요, 투표, 스케일 점수

<span style="color:#f993c0;"> 암시적 데이터 </span>

: 고객들의 행동을 기반으로 수집된 데이터, 사용자의 선호도가 간접적으로 표현된 데이터   
→  데이터를 얻기 굉장히 쉽지만, 사용자의 취향이나 선호도를 직접적으로 알 수 없다는 단점 존재

Ex) 조회수, 영상시청 기간, 구매 기록, 페이지 체류 기간

#### <span style="background-color:#633a4c; color:#fde9f2; padding:4px; border-radius:0px;"> 추천시스템의 분류 </span>

<span style="color:#f993c0;"> 내용 기반 필터링 </span>

- 콘텐츠의 유사도를 기반으로 사용자에게 추천하는 방법
- 사용자가 이용하거나 선택한 아이템과 유사한 아이템을 추천하는 방식

#### <u> 즉, 한 사용자의 행동 이력을 기준으로 추천!</u>

<span style="color:#f993c0;"> 협업 필터링 </span>

- 서비스 내에 있는 다수의 사용자로부터 얻은 선호도 정보를 기반으로 사용자에게 추천하는 방법
- 두 명의 사용자가 비슷한 관심사를 가지고 있다면, 한 사용자의 데이터를 바탕으로 다른 사용자에게 추천해 주는 방식

1. 사용자 기반 (User-Based Filtering): 특정 사용자와 **유사한 취향의 사용자**가 선호한 아이템 추천
2. 아이템 기반 (Item-Based Filtering): 사용자가 선호한 아이템과 **유사한 아이템**을 추천

#### <u>즉, 여러 사용자의 선호도를 활용하여 추천!</u>

<span style="color:#f993c0;"> 하이브리드 필터링 </span>

- 2가지 이상의 다양한 종류의 추천시스템 알고리즘을 조합하여 만들어진 알고리즘
- 콘텐츠 기반 필터링과 협업 필터링 각각의 단점들을 보완하기 위해 제시된 방식

## <span style="background-color:#F993C0; color:#311d26; padding:4px; border-radius:0px;"> 내용 기반 필터링 </span>

#### <span style="background-color:#633a4c; color:#fde9f2; padding:4px; border-radius:0px;"> 유사도 </span>

1. 영상, 이미지, 텍스트 등의 형태로 존재하는 아이템 정보 데이터를 **벡터화**한 이후, **아이템 벡터간의 유사도**를 측정
2. 측정된 유사도를 바탕으로 **고객이 선호하는 아이템과 비교**하여 **유사도가 높은 아이템을 선별**

1) 인풋 데이터 만들기
    - 아이템 정보 활용
    - 테이블 형태로 데이터 정리
    - 숫자로 변환하기 (원-핫 인코딩)

2) 유저 정보 활용
    - 유저의 취향 파악
    - 유저를 n개의 feature로 표현

3) 벡터의 내적과 유사성
    - **벡터 내적(dot product)**: 두 벡터의 동일 위치 값들끼리 곱한 후 모두 더한 값
    - **유사성(similarity)**: 내적 값이 클수록 두 벡터의 유사성이 높음



- **코사인 유사도(Cosine Similarity)**: 두 점 사이의 각도를 측정한 지표로, 값이 작으면 두 데이터가 가까이 있다 = 유사하다는 의미.
- **유클리드 거리(Euclidean Distance)**: 두 점 사이의 거리를 측정한 지표로, 값이 작으면 두 데이터가 가까이 있다 = 유사하다는 의미.

#### <span style="background-color:#633a4c; color:#fde9f2; padding:4px; border-radius:0px;"> 사용자 프로필 </span>

: **추천 시스템에 사용되는 데이터**

1. 직접 지정형: 사용자에게 자신이 선호하는 아이템의 특징을 명시적으로 지정하도록 하는 방법

    ex) OTT 서비스 가입시, 제시하는 몇 개 영화들에 대한 선호도를 체크해야 하는 경우

2. 간접 지정형: 사용자의 과거 행동 이력에 기반하여 사용자의 프로필을 작성하는 방법

    ex) SF영화를 즐겨본 이력이 존재하는 유저에게 인터스텔라를 추천해주는 경우 

#### 좋은 사용자 프로필을 구성할수록, 추천 시스템의 정확도와 만족도가 높아지고 다양한 방식(명시적 + 암묵적)을 혼합하여 사용하는 것이 일반적!

#### <span style="background-color:#633a4c; color:#fde9f2; padding:4px; border-radius:0px;"> 아이템 특징 </span>

- Items를 벡터 형태로 표현, 도메인에 따라 다른 방법이 적용
- 컨텐츠의 내용을 분석하는 아이템 분석 알고리즘
    - Clustering
    - Machine Learning
    - TF-IDF

<span style="color:#f993c0;"> TF-IDF </span>

: 특정 문서 내에 특정 단어가 얼마나 자주 등장하는 지를 의미하는 단어 빈도(TF) 와 전체 문서에서 특정 단어가 얼마나 자주 등장하는지를 의미하는 역문서 빈도(IDF)를 통해서 "다른 문서에서는 등장하지 않지만 특정 문서에서만 자주 등장하는 단어"를 찾아서 
문서 내 단어의 가중치를 계산하는 방법

$TF(d,t)$ : 특정 문서 d에서의 특정 단어 t의 등장 횟수

$DF(t)$ : 특정 단어 $t$가 등장한 문서의 수

$IDF(d,t)$ : $idf(d,t) = log(\frac{n}{1+df(t)})$

계산 예시: $TF-IDF = TF(d,t) \times IDF(d,t)$

#### <span style="background-color:#633a4c; color:#fde9f2; padding:4px; border-radius:0px;"> 장단점 </span>

| **장점** | **단점** |
| --- | --- |
| 아이템에 대한 **콜드 스타트 문제 해결 가능**(협업 필터링은 다른 사람들의 평점이 필요) | item의 feature를 추출해야 하고 이를 통해서 추천하기 때문에 정확한 feature를 추출하지 못하면 정확도가 떨어짐 |
| item의 feature를 통해서 추천을 하기 때문에 추천이 된 이유를 설명하기 용이함 | 기존의  item과 유사한 item 위주로만 추천하기에 새로운 장르의 item을 추천하기 어려움** |

## <span style="background-color:#F993C0; color:#311d26; padding:4px; border-radius:0px;"> 협업 필터링 </span>

- 기존 사용자들과 아이템의 상관 관계를 기반으로 새로운 사용자와 아이템의 상관 관계를 예측하는 것
- 아이템에 대한 **여러 사용자의 평점을 기반으로 추천**이 이루어짐

#### <span style="background-color:#633a4c; color:#fde9f2; padding:4px; border-radius:0px;"> 메모리 기반 협업 필터링 </span>

<span style="color:#f993c0;"> 메모리 기반 협업 필터링이란? </span>

= **이웃 기반 협업 필터링 (Neighborhood based Collaborative Filtering)**
- 사용자 또는 아이템 간 **유사도**를 기반으로 함
- 비교적 초기에 개발된 알고리즘
- 사용자 A와 **비슷한 취향을 가진 사용자가 선호하는 아이템을 추천**

<span style="color:#f993c0;"> 사용자 기반 접근 </span>

: 유저 간의 유사도가 높을 수록 가중치를 부여하는 방식, 같은 그룹의 다른 유저가 선호하는 아이템을 추천



추천하고자하는 target user와 유사한 사용자들의 평점을 활용

<span style="color:#f993c0;"> 아이템 기반 접근 </span>



추천하고자하는 target item과 유사한 아이템의 평점을 활용

<span style="color:#f993c0;"> 사용자 vs 아이템 </span>

| 비교 기준 | **User-based CF (UBCF)** | **Item-based CF (IBCF)** |
| --- | --- | --- |
| 추천의 정확성 | 아이템 수가 사용자 수보다 많은 경우 적합 | 사용자 수가 아이템 수보다 많은 경우 적합 |
| 시스템 공간 및 효율성 | 아이템 수가 사용자 수보다 많은 경우 적합 | 사용자 수가 아이템 수보다 많은 경우 적합 |
| 안정성 | 아이템이 수시로 바뀌는 경우 적합 | 아이템 리스트의 변화가 적고 고정적인 경우 적합 |
| 다양성 | 특정 사용자가 독특한 아이템을 평가한 경우 독창적 아이템 추천 가능 | 유사한 아이템 위주로 추천되어 다양성 및 참신성 부족 |

<span style="color:#f993c0;"> 메모리 기반 협업 필터링의 한계 </span>

- 아이템/유저가 계속 늘어날 경우, **확장성**이 떨어짐 
- 주어진 평점/선호도 데이터가 적을 경우, **성능**이 저하됨     
- 아이템 i에 대한 평점을 예측하기 위해서는 **아이템 i에 대해 평가를 한 모든 유저와의** 유사도를 구해야 함

#### <span style="background-color:#633a4c; color:#fde9f2; padding:4px; border-radius:0px;"> 모델 기반 협업 필터링 </span>

<span style="color:#f993c0;"> 모델 기반 협업 필터링이란? </span>

= 잠재 요인 협업 필터링
- **사용자-아이템 평점 매트릭스** 속에 숨어 있는 **잠재 요인을 추출**해 **추천 예측**을 할 수 있게 하는 기법

#### 모델 기반 협업필터링은 기계학습 모델을 통해 관측 데이터로부터 숨겨진 선호 요인. 즉, <u>잠재 요인을 수치적으로 학습한 후, 벡터(행렬)의 계산을 통해 평점을 예측</u>하기에 메모리 기반 협업 필터링의 문제점 해결 가능!!

<span style="color:#f993c0;"> 잠재 요인과 행렬 분해 </span>

<span style="color:#9400D3;"> 잠재 요인 </span>: 사용자와 아이템 사이의 관계를 설명해주는, 눈에 보이지 않는 숨겨진 특성(요인)   
- 모델이 **관측된 데이터를 바탕으로 수치적으로 추정**하는 값
-  취향, 성향, 특징 등 **실제 데이터에는 없지만**, **평점 패턴을 분석**하여 **수치로 추정**할 수 있는 값

<span style="color:#9400D3;"> Latent Factor Model </span>: 주어진 데이터로는 알 수 없는 사용자-아이템 간의 숨겨진 특성(feature)을 찾기 위해 주어진 Matrix를 사용자-잠재요인 아이템-잠재요인으로 각각 분해하여 학습하는 모델

<span style="color:#9400D3;"> 행렬 분해 </span>
- User-Item Matrix (유저-아이템 행렬)에는 유저의 취향+아이템의 특성이 녹아 있지만, 그 결과값만 볼 수 있음
- 이 행렬을 분해하여 평점을 유저의 잠재 요인과 아이템의 잠재요인의 내적으로 설명할 수 있다면, 이 방법을 통해 평점을 예측 가능

## <span style="background-color:#F993C0; color:#311d26; padding:4px; border-radius:0px;"> 추천시스템 평가 </span>

#### 추천시스템이 성장하기 위해선 반드시 평가가 실행되어야 함!

<span style="color:#f993c0;"> 오프라인 평가(실시간 연동 X) </span>

- 사용자 행동 이력에서 얻은 **과거의 로그**를 사용하여 모델의 예측 정밀도 등을 평가
- 과거의 사용자 행동을 학습하고 미래의 사용자 행동에 대한 예측을 수행

<span style="color:#f993c0;"> 온라인 평가(실시간 연동 O) </span>

- 추천 모델이나 새로운 사용자 인터페이스를 일부 사용자에게 실제로 표시함으로써 평가를 수행

<span style="color:#f993c0;"> 사용자 스터디 </span>

- 사용자에게 직접 인터뷰나 설문을 통해 데이터를 얻는 것

## <span style="background-color:#F993C0; color:#311d26; padding:4px; border-radius:0px;"> 연관 분석이란? </span>

#### <span style="background-color:#633a4c; color:#fde9f2; padding:4px; border-radius:0px;"> 연관 분석의 정의 </span>

: 상품이나 서비스를 구매하는 등 일련의 거래나 사건 안에 존재하는 항목 간의 일정한 연관 규칙을 발견하는 분석

#### <span style="background-color:#633a4c; color:#fde9f2; padding:4px; border-radius:0px;"> 연관규칙 분석 </span>

> **연관규칙의 형태**

‘A를 구매하였을 때, B 또한 구매할 것이다’와 같이 `if-then` 의 형식을 띄고, `A→B`으로 표현됨

- **조건절(Antecedent)**: `만일 ~라면`에 해당하는 부분
- **결과절(Consequent):** 그 뒷부분에 해당하는 내용
- **아이템 집합(Item set):** 조건절 또는 결과절을 구성하는 아이템들의 집합

> **연관규칙 주의점: 인과관계 파악 불가**
> 
연관규칙으로는 상관관계를 파악할 뿐!

1. 실험이 아닌 **관찰 데이터**이기 때문
    - A를 사게 한 원인이 통제되지 않음.
    - 예: A와 B가 둘 다 여름 제품일 수도 있음 (숨겨진 공통 원인)
2. **방향성은 그냥 수학적 비율일 뿐**
    - A → B는 A가 원인인 것이 아니라 A를 샀을 때 B 산 비율이 높은 것 뿐
3. **상관관계 ≠ 인과관계**
    - 예: 얼음 판매량↑ → 아이스크림 판매량↑
        → 진짜 원인은 더운 날씨 (공변 요인)

> 연관분석 vs 협업필터링

- **협업필터링**은 **개인화된 상품 추천**에 자주 활용됨
- **연관 분석**은 **상품 노출 또는 세트상품 구성**에 많이 활용됨

## <span style="background-color:#F993C0; color:#311d26; padding:4px; border-radius:0px;"> 연관 분석의 주요 개념 </span>

#### <span style="background-color:#633a4c; color:#fde9f2; padding:4px; border-radius:0px;"> 연관규칙의 조건 </span>

1. 두 품목(품목 A와 B)이 함께 구매한 경우의 수가 일정 수준 이상이어야 한다. 
(**일정 이상의 지지도**)
2. 품목 A를 포함하는 거래 중 품목 B를 구입하는 경우의 수가 일정 수준 이상이어야 한다. 
(**일정 이상의 신뢰도**)

<span style="color:#f993c0;"> 지지도 </span>

$지지도 = P(A \cap B)$

: 전체 거래항목 중 품목 A와 품목 B가 동시에 포함하는 거래의 비율

<span style="color:#f993c0;"> 신뢰도 </span>

$신뢰도 = \frac{P(A \cap B)}{P(A)} = \frac{지지도}{P(A)}$

: 품목 A를 포함하는 거래 수 중 품목 A와 품목 B가 동시에 포함하는 거래의 비율

#### <span style="background-color:#633a4c; color:#fde9f2; padding:4px; border-radius:0px;"> 연관 분석의 평가 측도 </span>

<span style="color:#f993c0;"> 향상도 </span>

: 품목 A가 주어지지 않았을 때의 품목 B의 확률 대비, 
품목 A가 주어졌을 때의 품목 B의 확률의 증가 비율
#### = A와 B가 우연히 함께 등장할 확률과 실제 함께 등장할 확률을 비교!

$향상도 = \frac{P(A \cap B)}{P(A)P(B)} = \frac{신뢰도}{P(B)}$

## <span style="background-color:#F993C0; color:#311d26; padding:4px; border-radius:0px;"> 연관 분석 알고리즘 </span>

#### <span style="background-color:#633a4c; color:#fde9f2; padding:4px; border-radius:0px;"> Apriori 알고리즘 </span>

#### 선험적 경험을 이용한다 (A Priori)

<span style="color:#f993c0;"> Apriori 알고리즘이란? </span>

등장 배경 : 무차별 탐색의 비효율

개념: 상위 조합에서부터 차례로 스캔하면서 특정 조합이 자주 발생하지 않는다면 이의 결과물로 탄생한 후속 조합들까지 모두 후보에서 **배제**하는 방식의 알고리즘

장점: Apriori 알고리즘을 활용하면 하나의 조합만 검사하고도 이에서 파생된 다른 조합들까지 후보에서 배제할 수 있게 되므로 시간과 연산량을 효과적으로 줄일 수 있다.



집합 `{A,B}` 의 지지도가 사용자가 정한 **최소 지지도를 넘지 못하면**, 
`{A,B}` 를 포함한 `{A,B,C}`,`{A,B,D}` 등의 규칙들도 **모두 한번에 제거**가 된다! 

→ 연산량 감소

<span style="color:#f993c0;"> Apriori 알고리즘 프로세스 </span>

#### 빈발 항목 집합을 찾고 → 단계적으로 조합을 생성하고 필터링

1. 단일 항목 집합 생성
- 모든 개별 상품에 대해 지지도를 계산
- **최소 지지도**를 넘는 항목들만 다음 단계로!

2. 2개 항목 집합 생성
- 1단계에서 선별된 상품들로 2개 항목의 조합 생성
- 이 조합에 대해 지지도를 계산하고, 최소 지지도를 넘지 않는 조합을 제거

3. 더 큰 항목 집합 생성
- 이전 단계에서 남은 조합들을 바탕으로 3개 이상의 항목으로 구성된 조합 생성
- 지지도를 계산하고, 최소 지지도를 넘지 않는 조합을 제거
- 더 이상 조합을 생성할 수 없을 때 까지 이 과정을 반복

4. 최종 빈발 항목 집합 생성
- 마지막으로 남은 빈발 항목 집합들을 바탕으로 연관 규칙을 생성
- 각 규칙에 대해 지지도, 신뢰도, 향상도를 계산하여 가장 유용한 규칙을 선택

<span style="color:#f993c0;"> Apriori 알고리즘 장단점 </span>

**장점**

- 수많은 상품 연관 구매 패턴 발견
- 다른 연구가설 탐지 가능
- 원리가 간단하고, 이해분석이 용이함

**단점**

- 비즈니스 측면 중요한 현실적 중요 연관 규칙 부족
- 여전히 데이터가 커질수록 속도가 느려짐

#### <span style="background-color:#633a4c; color:#fde9f2; padding:4px; border-radius:0px;"> FP-Growth 알고리즘 </span>

#### Apriori의 연산 속도 문제를 해결할 수 있게 자료구조를 잘 이용한 알고리즘

<span style="color:#f993c0;"> FP-Growth 알고리즘이란? </span>

등장 배경: Apriori 알고리즘은 여전히 데이터가 커질수록 속도가 느려진다는 단점, 이를 해결하기 위한 알고리즘

- FP-Growth (Frequent Pattern Growth) 알고리즘은 빈발 품목 집합을 효율적으로 찾아냄
- **FP-Tree** 라는 구조를 이용하여 Apriori를 효과적으로 구현
- Tree, Array, Linked_List를 합쳐놓은 구조!
- 모든 거래를 확인해 각 아이템마다 지지도를 계산하고 , 최소 지지도 이상의 아이템만 선택



<span style="color:#f993c0;"> FP-Growth 알고리즘 프로세스 </span>

1. 빈발 항목 필터링
- 모든 항목의 빈도를 계산한 후, 최소 지지도 이상인 항목만을 남김

2. 데이터 재정렬
- 남은 항목들을 각 거래별 빈도순으로 재정렬

3. FP-트리 생성
- 재정렬된 데이터를 바탕으로 FP-트리를 생성

4. 빈발 항목 집합 추출
- FP-트리를 바탕으로 특정 항목을 기준으로 부분 트리를 만들어 빈발 항목 집합을 추출

5. 연관 규칙 도출
- FP-트리에서 도출된 빈발 항목 집합을 바탕으로 연관 규칙 생성
- 각 규칙에 대해 지지도(Support), 신뢰도(Confidence), 향상도(Lift) 등의 지표를 계산해 가장 유용한 규칙을 선택

<span style="color:#f993c0;"> FP-Growth 알고리즘 장단점 </span>

**장점**

- 데이터를 두 번만 스캔하는 Tree 구조이기 때문에 Apriori보다 훨씬 **빠르다**.
- 후보 Itemset을 생성할 필요 없이, Tree만 구성하면 끝이다.

**단점**

- 대용량 데이터셋에서 메모리를 효율적으로 사용하지 않는다.
- Apriori에 비해 설계하기 어렵고, 지지도의 계산은 무조건 FP-Tree가 만들어져야 가능하다.
