## 06. 분류

## <span style="background-color:#F993C0; color:#311d26; padding:4px; border-radius:0px;"> 분류란? </span>

지도 학습 중 그동안 **'회귀'** 를 공부했고, 이제부터 다룰 것은 **'분류'**

<span style="color:#f993c0;"> 회귀 vs 분류 </span>
- 회귀: 연속형 데이터를 예측
- 분류: 범주형 데이터를 예측

즉, 분류란 <U>기존 데이터가 어떤 레이블에 속하는지 패턴을 알고리즘으로 인지한 뒤에 새롭게 관측된 데이터에 대한 레이블을 판별하는 것</U>

<span style="color:#f993c0;"> 이진 분류와 다중 분류 </span>

분류에는 두 가지 종류가 있음
- **이진 분류(Binary Classification):** 예측하고자 하는 변수가 어떤 기준에 대하여 참(True) 또는 거짓(False)의 값만을 가질 때(2개)
- **다중 분류(Multiclass Classification):** 예측하고자 하는 변수가 가질 수 있는 값이 3개 이상일 때



## <span style="background-color:#F993C0; color:#311d26; padding:4px; border-radius:0px;"> 분류 모델 </span>

#### <span style="background-color:#633a4c; color:#fde9f2; padding:4px; border-radius:0px;"> 로지스틱 회귀 </span>

로지스틱 회귀: 독립 변수의 선형 조합(선형 회귀)에 로지스틱 함수를 적용하여 출력값을 0에서 1 사이로 변환해주는 것   
→ **이진 분류 문제를 푸는 대표적인 알고리즘!**

<span style="color:#f993c0;"> 로지스틱 회귀의 필요성 </span>



이런 식으로, 이진 분류를 단순 선형 회귀로 풀면 분류 작업이 제대로 동작하지 않는 문제가 발생함

→ 따라서 출력이 0과 1사이의 값을 가지면서 S자 형태로 그려지는 함수를 이용!

<span style="color:#f993c0;"> 시그모이드 함수 </span>

시그모이드 함수 ≈ 로지스틱 함수: 출력이 0과 1사이의 값을 가지면서 S자 형태로 그려지는 함수



출력이 0과 1 사이의 값을 가지면서 S자 형태로 그려지기 때문에, 이진 분류 작업에 적용 가능하다

→ **출력값이 특정값 이상이면 1(True), 특정값 이하면 0(False)로 정해서 이진 분류 문제를 풀 때** 사용하자!

시그모이드 함수가 $H(x)=\frac{1}{1+e^{-(wx+b)}}=sigmoid(wx+b)=\sigma(wx+b)$일 때,

→ 인공지능은 주어진 데이터에 적합한 가중치 $w$와 $b$를 계산해서 찾음

#### <span style="background-color:#633a4c; color:#fde9f2; padding:4px; border-radius:0px;"> 의사결정나무 </span>

의사결정나무: 조건에 따라 데이터를 분류하며, 최종적으로 데이터가 순수한 label의 집합으로 구성될 때까지 분류를 반복하는 분석 방법

<u>like 스무고개 놀이 !!</u>

<span style="color:#f993c0;"> 의사결정나무의 구성 요소 </span>



- **Root Node**: 가장 상위에 위치한 노드
- **Parent Node**: 주어진 노드의 상위에 있는 노드
- **Child Node**: 하나의 노드로부터 분리된 2개 이상의 노드들
- **Binary Tree**(이진 트리): Tree 중에서 children이 최대 2개인 tree
- **Leaf Nodes**: 자식마디가 없는 노드, 모델에서 label에 해당
- **Edge**: Parent node와 child node를 연결하는 연결선
- **Height**: 특정 노드에서 가장 멀리 있는 leaf node까지의 경로에 있는 edge 개수
- **Depth:** Root node에서 특정 노드에 도달하기 위한 edge의 수

#### <span style="background-color:#633a4c; color:#fde9f2; padding:4px; border-radius:0px;"> SVM(Support Vector Machine) </span>

SVM: 클래스를 분류할 수 있는 다양한 경계선 중 최적의 라인을 찾아내는 알고리즘

= 데이터를 분리하는 *초평면(Hyperplane)* 중에서 데이터들과 *가장 거리가 먼 초평면을 선택하여 분리*하는 지도 학습 기반의 이진 선형 분류 모델

→ 명확하게 분류할 수 있는 데이터 집단에서 뛰어난 성능을 보이며, 고차원 공간(다수의 feature)에서도 효과적으로 사용 가능

<span style="color:#f993c0;"> SVM의 구성 </span>



- Support vector (서포트 벡터): 데이터 중에서 결정 경계와 가장 가까이에 있는 데이터들의 집합
- Decision Boundary (결정 경계): 데이터 분류의 기준이 되는 경계
- Margin (마진): 결정 경계에서 서포트 벡터까지의 거리(여유 공간)
- Hyperplane (초평면): $n$차원 공간의 ($n-1$)차원 평면
- Slack Variables (슬랙 변수/여유변수): 완벽한 분리가 불가능할 때 분류를 위해 허용된 오차에 대한 변수

<span style="color:#f993c0;"> SVD의 장단점 </span>


장점
- 서포트 벡터만을 이용해서 결정 경계를 생성하므로 데이터가 적을 때 효과적
- 새로운 데이터가 입력되면 전체 데이터 포인트와의 거리를 계산하지 않고 서포트 벡터와의 거리만 계산하면 되기 때문에 연산량 최소화
- 정확성이 뛰어나며, 비선형 모델 분류 가능
- 다른 모형보다 과대 적합의 가능성이 낮고, 노이즈의 영향이 적음

단점
- 데이터 전처리 과정이 중요
- 데이터가 많아질수록 최적화를 위한 평가 과정이 많아져서 속도가 느림

#### <span style="background-color:#633a4c; color:#fde9f2; padding:4px; border-radius:0px;"> KNN(K-Nearest Neighbor) </span>

말 그대로 가까이 있는다는 점을 이용한 분류 알고리즘! = **유유상종~~**

KNN: 데이터로부터 거리가 가까운 k개의 다른 데이터 레이블을 참조하여 분류하는 알고리즘
- K개의 이웃 설정 ⇒ 가까운 K개의 데이터를 기반으로 분류


**KNN 알고리즘 계산 순서**

1. 데이터 준비
    
    - KNN은 미리 학습하는 과정이 없기 때문에, 데이터를 준비
    - 각 데이터는 특징 벡터와 레이블(어떤 class인지)로 이루어짐
    
2. K 값 설정
    
    - 원하는 K 값을 설정
    - **K**는 **가장 가까운 이웃의 개수**를 나타내며, 보통 **홀수**개로 설정
    
3. 거리 계산
    
    - 새로운 데이터(예측하려는 데이터)가 주어지면, 이 값과 **기존 모든 데이터 간의 거리를 계산**
    - 유클리드 거리, 맨해튼 거리 등을 사용
        
4. 가장 가까운 K개의 이웃 선택
    
    - 계산된 거리 중에서 **가장 작은 거리 값을 가진 K개의 데이터를 선택**
    - 이 데이터 포인트들이 가장 가까운 이웃!
    
5. 분류하기
    
    - K개의 이웃 중 **가장 많이 등장하는 클래스가 예측 결과**


**최적의 K 선택하는 법**

- 일반적으로는 학습용 데이터 개수의 제곱근으로 설정
- K를 너무 크게 설정하면 주변에 있는 점과의 근접성이 떨어져서 분류가 잘 이뤄지지 않고 과소 적합
- K를 너무 작게 설정하면 이상값, 잡음 데이터와 이웃이 될 가능성이 있음

<span style="color:#f993c0;"> K-NN 알고리즘 예시 </span>



- 이렇게 6개의 기존 데이터 A~F와 1개의 신규 데이터 N이 있고, 주변에 있는 이웃의 개수를 K라고 지칭



- K=1이라면, 거리가 첫 번째로 가까운 C를 보고 신규 데이터인 N을 동그라미 그룹으로 분류



- 하지만, 만약 K=3이라면 신규 데이터인 N과 거리가 세 번째로 가까운 C, D, E를 보고, N과 가장 가까운 이웃들이 주로 속해있는 클래스로 N을 분류
- 이때 그룹이 나눠지면 다수결의 원칙에 따른다. 여기서는 1:2가 되어 N은 세모로 분류

IF) 동점이 나오면?
- 동점이 발생한 클래스들 중, 더 가까운 이웃에 해당하는 클래스로 분류
- 동점이 발생한 클래스들 중, 무작위로 하나를 선택

<span style="color:#f993c0;"> KNN의 장단점 </span>


KNN의 중요한 특징은, <U>**어떠한 학습도 필요하지 않다는 것!!!**</U>

장점
- 범주를 나눈 기준을 알지 못해도 데이터를 분류할 수 있음
- 입력 데이터만 주어지면 바로 예측값을 구할 수 있어서 다른 분류 모형과는 달리 학습 과정이 필요 없음
- 이해가 쉽고, 구현이 간단
- 추가된 데이터의 처리가 용이
- 훈련 데이터를 요약하거나 일반화된 모델로 만들지 않고, 원본 데이터 전체를 그대로 저장

단점
- 다른 분류 모형과 달리 테스트 데이터의 개수에 따라 시간이 오래 걸림
- 학습 데이터 모두를 거리 계산에 사용하기 때문에 학습 데이터의 양도 계산 시간에 영향을 미침
- K의 값 결정이 어려움
- 수치형 데이터가 아니면 유사도를 정의하기 어려움
- 데이터 내 이상치가 존재하면 분류 성능에 큰 영향을 받음

## <span style="background-color:#F993C0; color:#311d26; padding:4px; border-radius:0px;"> 분류 평가 지표 </span>

#### <span style="background-color:#633a4c; color:#fde9f2; padding:4px; border-radius:0px;"> 혼동 행렬 </span>

혼동 행렬: 분류 모델의 예측 결과를 정확한 예측과 잘못된 예측으로 구분하여 나타낸 표(행렬)



이미지에서, 
- **파란색 상자** : 예측을 올바르게 한 경우 (**예측과 실제 결과가 모두 일치**)
- **빨간색 상자** : 예측을 잘못한 경우 (**예측과 실제 결과가 불일치**)

→ 예측 성공 여부(T/F) + 예측값(P/N)으로 구분하여 총 4종류의 경우의 수 존재!

| 구분 | 분류 값 | 설명 |
| --- | --- | --- |
| 예측이 정확한 경우 | **True Positive (TP)** | 정답이 참, 참으로 예측 |
|  | **True Negative (TN)** | 정답이 거짓, 거짓으로 예측 |
| 예측이 틀린 경우 | **False Positive (FP)** | 정답이 거짓, 참으로 예측 (예측이 틀림)  |
|  | **False Negative (FN)** | 정답이 참, 거짓으로 예측 (예측이 틀림)  |

<span style="color:#f993c0;"> 혼동 행렬을 이용한 분류 모델 평가 지표 </span>

##### 정확도, 정밀도, 재현도
: 모두 0과 1 사이의 값이다!

**정확도 (Accuracy): 모든 가능한 예측 중 참인 비율**
: 모델이 입력된 데이터에 대해 얼마나 정확하게 예측하는지를 나타냄
- 모든 가능한 예측 중 참인 비율
- (TP+TN) / (TP+TN+FP+FN)
- 단점 : 정답 레이블의 비율이 불균형하면 모델의 정확도를 신뢰할 수 없음

**정밀도 (precision): 참이라고 예측한 경우 실제 참의 비율**

**재현도 (recall): 실제로 참인 경우 중 참으로 예측하는 비율**

<span style="color:#f993c0;"> 정밀도와 재현도 예시 </span>



정밀도 & 재현도 그래프에서 *두 value 값이 만나는 지점을 Threshold*로 정하면, **예측 오류를 최소화**할 수 있다!



요렇게

#### <span style="background-color:#633a4c; color:#fde9f2; padding:4px; border-radius:0px;"> F1-Score </span>

F1-Score: 정밀도(precision)와 재현율(recall)의 조화 평균



<u>머신러닝 모델의 성능을 평가하는 주요지표!</u>

#### <span style="background-color:#633a4c; color:#fde9f2; padding:4px; border-radius:0px;"> ROC/AUC Curve </span>

: 분류 모델의 성능을 평가하는 중요한 도구

<span style="color:#f993c0;"> ROC Curve </span>

ROC Curve: 얼마나 분류가 잘 되었는가를 보여주는 그래프

- True Positive Rate (TPR) : 참인 것들 중에 참이라고 예측한 비율 (=Recall)
- False Positive Rate (FPR) : 거짓인 것들 중에 참이라고 잘못 예측한 비율 

<span style="color:#f993c0;"> AUC Curve </span>

AUC Curve: ROC와 x축 사이의 면적(적분값)

- 모델의 성능을 숫자로 나타냄
- 0.5에서 1사이의 값을 가지며, 1에 가까울수록 분류 성능이 좋은 것

#### <span style="background-color:#633a4c; color:#fde9f2; padding:4px; border-radius:0px;"> 다중 분류 평가 지표 </span>

: 이진 분류 평가 지표를 사용해서 클래스별로 점수를 구한 뒤, 이를 적절히 평균을 내리는 것!

- Macro average: 클래스별로 구한 평가 지표 평균
- Weighted average: 클래스별로 구한 평가 지표 가중 평균
- Micro average: 모든 클래스의 예측 결과를 더하여  전체적인 성능을 평가하는 지표

## <span style="background-color:#F993C0; color:#311d26; padding:4px; border-radius:0px;"> 하이퍼파라미터 최적화 </span>

#### <span style="background-color:#633a4c; color:#fde9f2; padding:4px; border-radius:0px;"> 하이퍼파라미터란? </span>

**Hyperparameter**
    
:모델 학습 과정에 반영되는 값으로, 학습 시작 전에 사용자가 직접 설정하는 변수

**Hyperparameter Optimization**
    
: 하이퍼파라미터 최적화란, Tuning을 거쳐 **적절한 하이퍼파라미터**를 찾음으로써 Model 성능을 향상시키는 것

#### <span style="background-color:#633a4c; color:#fde9f2; padding:4px; border-radius:0px;"> 하이퍼파라미터 최적화 방법 </span>

하이퍼파라미터 최적화 방법
- Grid Search: 정해진 범위에서 Hyperparameter를 모두 순회
- Random Search: 정해진 범위에서 Hyperparameter를 무작위로 탐색
- Bayesian Optimization: 사전 정보를 바탕으로 Hyperparameter 값을 확률적으로 추정하며 탐색

1. **Grid Search**   

    : 정해진 범위에서 Hyperparameter를 **모두** 순회하며 가장 좋은 성능을 내는 값을 찾는 기법

    `GridSearchCV` (scikit-learn)

2. **Random Search**   

    : 정해진 범위에서 Hyperparameter를 무작위로 탐색해 가장 좋은 성능을 내는 값을 찾는 기법   
    `RandomSearchCV` (scikit-learn)

3. **Bayesian Optimization**

    : 사전 정보를 바탕으로 최적 Hyperparameter 값을 확률적으로 추정하며 탐색하는 기법   
    `bayes_opt`
