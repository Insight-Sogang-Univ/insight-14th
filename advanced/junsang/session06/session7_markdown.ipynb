# **시계열 1**

# 1. 시계열 데이터란?

### 1-1. 시계열 데이터의 정의

시계열 데이터 : 순차적인 시간의 흐름에 따라 수집된 데이터
- 연속한 관측치에는 서로 상관관계 존재, 순서가 중요한 시퀀스 데이터로 간주



- 횡단면 데이터 : 비시계열, 몇 가지 주체에 대해 특정 시점에서 수집
- 종단면 데이터 : 시계열, 동일한 개체에 대해 특정 시간 간격으로 수집된 데이터

- 단변량 시계열 : 시간에 따라 연속적으로 측정된 단일 변수로 이루어진 시계열 데이터
- 다변량 시계열 : 같은 시간에 기록된 둘 이상의 변수로 이루어진 시계열 데이터
    - 변수 간 상호 작용 및 영향력 파악

### 1-2. 시계열 분석의 목적

시계열 분석의 주 목표 = 미래 데이터 예측
- 시계열 분석은 시계열 데이터를 분석하고 모델링하는 통계적인 방법과 기법
- 데이터 구조 파악, 예측, 이상치 탐지를 포괄하는 개념

### 1-3. 시계열 데이터의 특징

1. 시간의존성
- 시계열 데이터는 시간에 따라 변화한다
- 시계열 데이터는 시간 순서에 따라 관측되므로 데이터의 순서를 바꾸면 의미가 달라짐

2. 계절성 & 추세
- 계절성 : 일정 주기에 따라 반복적으로 나타나는 데이터의 변동
    - 주기적인 패턴을 고려한 예측 모델링 가능
- 추세 : 시간의 흐름에 따라 나타나는 장기적인 증가/감소 경향 
    - 전반적인 방향성을 파악할 수 있음

3. 자기상관성
- 자기자신끼리 상관관계를 갖는 특성을 의미
- 관측값은 서로 독립적이지 않다

**자기상관성 때문에 회귀분석이 불가능하다!**
1. 통계적 추론의 제약 및 오류
- 회귀분석의 가정중 오차항 간 상관이 없어야함 (오차항의 독립성 가정) -> 시계열은 오차항 간 상관관계가 있음

2. 예측의 어려움 & 모델 복잡성 증가
- 잔차가 시간에 따라 패턴을 가짐 -> 미래값 예측의 왜곡
- 단순 회귀모형은 적절치 않기 때문에, 시계열 모델을 사용해야 함

### 1-4. 시계열 데이터의 분석 요인

1. 추세요인
- 데이터 전반에 걸쳐 시간에 따라 지속적으로 증가/감소하는 경향
- 꾸준히 증가하거나 감소하는 선형 그래프

2. 계절 요인
- 일정한 주기로 반복되는 패턴
- 주기적인 변동을 나타내는 그래프

3. 순환 요인
- 계절성보다 더 길고 불규칙한 주기로 반복되는 변동
- 길게 상승과 하강을 반복하는 파형 그래프

4. 불규칙 요인
- 추세, 계절성, 순환성을 모두 제거한 뒤 남은 예측 불가능한 변동 
- 노이즈, 우연적 사건, 일시적인 변동, 이상치 등을 모두 포함
- 불규칙한 변동을 나타내는 그래프

### 1-5. 시계열 분석의 워크플로우

1. 데이터 전처리 - 결측치 보간, 이상치 처리, 요소분해
2. 정상성 판단 
3. 모델링 및 예측

# 2. 시계열 데이터 전처리

### 2-1. 이상치, 결측치 처리

#### 2-1-1. 결측치 처리

시계열 데이터의 결측치 
- 특정 시점에서 값이 측정되지 않았거나 누락된 상태
- 센서 오루, 수집 노락 등으로 발생
- 결측치가 존재하면 추세/계절성이 왜곡되므로 적절한 대치와 보간이 필요함


결측치 처리 방법
1. 삭제
2. 단순 대치 : 평균, 중앙값, 최빈값, 0 등으로 결측치를채움, 패턴 왜곡 가능성 높음
3. 보간법 : 주어진 데이터 포인트의 값을 사용하여 두 개 이상의 데이터 포인트 사이에서 누락된 값을 추정 -> 합리적으로 매꿔주기
- 선형 보간법
- 다항식 보간법
- 스플라인 보간법
- 조각별 상수 보간법
    - 이전 값 유지
    - 다음 값 활용
    - 최근접 이웃 보간

#### 2-1-2. 이상치 탐지 및 처리

시계열 데이터의 이상치
- 값이 측정되었으나, 측정된 값이 일반적이지 않은 경우
- 단순히 이상한 값이 아니라 신호의 중요한 의미일 가능성도 있으므로 신중한 접근 필요  -> 사전 지식 필요!!

이상치 탐지 방법
- IQR
- Z-score (보통 z>=3)
- STL 분해
    - 시계열을 추세, 계절성, 잔차로 분해, 잔차가 특정 임계값을 초과하면 이상 탐지
    - 계절성이나 추세가 뚜렷한 시계열에 적합
    - 계산 비용이 크다

이상치 탐지 모델
- 분류 및 회귀 트리 (CART)
    - 데이터를 분할하여 예측값을 산출함
    - 예측 오차가 일정 임계값 이상인 지점을 이상치로 간주
- 클러스터링 기반 이상 탐지
    - 이상치는 대부분의 클러스터 중심에서 멀리 떨어져서 나타남
- 오토 인코더
    - 인공신경망 기반 비지도 학습 모델
    - 정상 데이터에 대한 특징을 학습하고, 학습된 모델에 데이터를 입력했을 때 재구성한 결과와 학습된 정상 특징과의 차이점을 비교하여 판단
    - 정상 패턴은 잘 복원되지만, 이상 패턴은 복원오차가 커짐

평활화
- 시계열 데이터에서 노이즈를 줄이고, 전체적인 흐름이나 추세를  더 명확하게 파악하기 위해 데이터를 부드럽게 만드는 기술
- 이동평균 : 시계열의 부분집합으로 평균을 구하여 평활화 효과를 얻는 방법
    - 단순 이동평균
    - 가중 이동평균 (최근 데이터에 더 높은 가중치를 부여)
    - 지수 이동평균 (최근 데이터에 더 많은 가중치, 예전 데이터에 지수적으로 감소하는 가중치 적용)

### 2-2. 요소 분해 (시계열 분해)


$Y_t$ = $f$( $T_t, S_t, C_t, R_t$)

- $T_t$ : 추세성분 (시점 t에서의 추세요소)
- $S_t$ : 계절성분 (시점 t에서의 계절요소)
- $C_t$ : 순환성분 (시점 t에서의 순환요소)  (일반적으로 장기간에만 고려되므로 0으로 설정함)
- $R_t$ : 불규칙성분 (시점 t에서의 불규칙요소)

요소 분해를 하면 각각의 패턴을 명확히 이해 가능, 예측, 이상 탐지 등 분석 작업을 정교하게 수행할 수 있음

 
$$
y_t = T_t + S_t + R_t
$$

가법 모형 
- 시계열 = 추세 + 계절성 + 불규칙성
- 시계열 관측값은 변동요인들의 가법(+)으로 설명될 수 있다 가정
- 각 변동 요인들이 독립적으로 더해지며, 모든 요소의 영향이 일정함
- 여름철 한국에 방문하는 관광객은 매년 평균 10만명씩 증가한다 같은거

---------

$$
y_t = T_t * S_t * R_t
$$

$$
log(y_t) = log(T_t) + log(S_t) +log(R_t)
$$

승법 모형
- 시계열 = 추세 x 계절성 x 불규칙성
- 시계열 관측값은 변동요인들의 승법으로 설명될 수 있다 가정
- 추세가 커질수록 계절성이나 불규칙성의 효과도 비례해서 커짐
- 여름철 한국에 방문하는 관광객은 매년 평균 10%씩 증가한다

# 3. 정상성 판단

### 3-1. 정상시계열

정상성 : 시계열 데이터가 평균, 분산 등의 통계적 특성이 시간에 따라 일정한 수준으로 유지되는 것
- 정상 시계열 : 정상성을 띠는 시계열 데이터
- 비정상 시계열 : 정상성을 띠지 않는 시계열 데이터

강한 정상성 
- 모든 적률(평균, 분산, 왜도, 첨도)이 시간과 무관하게 일정함
- 현실에서 찾아보기 매우 어려움

약한 정상성 (통상적으로 시계열 분석에서 정상성이라고 부름)
1. 평균이 모든 시점에 대해 일정하다
2. 분산이 모든 시점에 대해 일정하다
3. 공분한은 시점에 의존하지 않고, 단지 시차에만 의존한다

시각적으로 정상성 판단하기
- 추세가 있으면 안됨
- 계절성이 있으면 안됨
- 분산이 증가하거나 감소하면 안됨

### 3-2. 검정을 통해 정상성 판단하기

단위근 : 단위 + 근     
시계열 모델에서 특성방적식의 근이 1인 경우, 해당시계열은 단위근을 가짐 -> 이는 비정상 시계열을 의미
1. 평균이 일정하지 않아서 예측이 어렵다
2. 분산이 계속 커진다
3. 과거 충격이 영원히 남는다
- 랜덤워크 : 시계열이 일정한 분포나 상태로 수렴하지 않음 -> 예측불가능

1. ADF 검정 : 딘위근의 존재 여부를 통해 시계열의 정상성 판단

귀무가설 : 시계열에 단위근이 존재한다 (비정상 시계열)     
대립가설 : 시계열에 단위근이 존재하지 않는다 (정상 시계열)


2. KPSS 검정

귀무가설 : 시계열에 단위근이 존재하지 않는다 (정상 시계열)    
대립가설 : 시계열에 단위근이 존재한다 (비정상 시계열)


확정적 추세가 존재하는 경우, 두 검정의 결론이 다르게 나올 수 있음

### 3-3. 정상성 확보를 위한 여러 방법들

1. 추세 제거 - 차분
- 차분 : 연이은 관측값들의 차이를 계산하는 것
- 시계열 데이터에서 한 시점의 값과 그 이전 시점의 값을 차분하여 변화량을 계산
- 대부분의 비정상적 시계열은 값이 누적되는 누적과정이기 때문에 누적된 값을 빼줌으로써 정상적 시계열을 얻어 낼 수 있음

2. 계절성 제거 - 계절 차분
- 시계열이 주기적인 패턴을 가질 때 적용
- t시점의 y값에서 t-m시점의 y값을 빼는 것 (m: 주기의 길이)
-  데이터에 계절성 패턴이 강하게 나타나면, 계절성 차분을 먼저 계산하는 것이 좋음.

3. 분산 안정화 - 로그변환, 박스콕스 변환

- 시계열의 변동 폭이 점점 커지거나 작아질 때 적용
- 로그변환 : 분산의 정상성 확보하기 위한 방법
    - 로그 변환은 음수나 0을 처리할 수 없음
    - 분산 변화가 비선형적이거나 복잡한 패턴을 가질 경우 충분치 않을 수 있음
- 박스 콕스 변환
