# **1. 인공신경망&퍼셉트론**  

- 인공신경망: 우리 뇌의 신경망을 모방하여 만든 모델  
- 뉴런과 인공신경망이 정보를 처리하는 구조  

|뉴런|인공신경망|
|---|---|
|1.가지돌기를 통해 외부 전기신호 받아들임|1.다수의 입력을 받아 가중치 적용|
|2.모인 신호가 일정 이상의 크기를 가지면|2.종합한 값이 일정 기준 넘으면|
|3.축삭돌기 통해 신호 전달|3.output 출력|
 
- 뉴런과 퍼셉트론의 연결 방식 
 


## 퍼셉트론 
: 초기 형태의 인공신경망  

  

- Trainig(훈련 또는 학습): 원하는 출력값을 내보내도록 가중치를 조정해가는 작업  
1. 입력값과 가중치를 곱함 →  ${x_1*w_1}$  
2. 곱한 값들의 총합을 구함 → ${x_1*w_2+x_1*w_2+x_3*w_3 +...+x_n*w_n}$  
3. 총합이 임계값(${\theta}$**)**을 넘으면 $1$, 넘지 않으면 $0$을 출력 (= Step 활성화 함수의 작업)  

- 가중치(weight): 입력값이 출력값에 주는 **영향력(중요도)** 결정, 클수록 해당 입력값이 중요함을 의미  
- 활성화 함수: 신호가 일정 이상의 값을 넘으면 1, 넘지 못하면 0을 출력하는 퍼셉트론의 성질은 활성화 함수로 계단 함수를 사용(step function), Sigmoid/Tanh/ReLU/leaky ReLU 등  
- 임계값(threshold): 가중합이 임계값을 넘으면1, 안 넘으면 0  
  

- 편향(bias): 임계값을 좌변으로 넘기면 편향(b)으로 표현 가능  
  


## 단층 퍼셉트론과 그 한계  
- 단층 퍼셉트론: 두 개의 층(입력, 출력층)으로 이루어진 퍼셉트론  
- 논리 게이트: 디지털 회로에서 기본적인 논리 연산을 수행하는 데 사용  
  - AND 게이트: 두 개의 입력값이 모두 1인 경우에만 출력값: 1  
  - NAND 게이트: 두 개의 입력값이 1인 경우에만 출력값: 0, 나머지: 1  
  - OR 게이트: 두 개의 입력값이 모두 0인 경우 출력값: 0, 나머지: 1  
  - XOR 게이트: 입력값 2개가 서로 다른 값을 갖고 있을 때에만 출력값: 1 -> **단층 퍼셉트론으로 구현 불가능**  

## 다층 퍼셉트론  
- 다층 퍼셉트론 예시 (XOR 게이트)  



- 중간에 층을 더 추가하여 입력층 & 하나 이상의 은닉층 & 출력층으로 구성  
- 두 개 이상의 은닉층을 갖는 MLP가 딥러닝의 시초 모델로 발전  
 - 심층신경망: 은닉층이 2개인 신경망  
 - 딥러닝: 심층신경망을 (가중치를 스스로 찾도록) 학습시킴  
  퍼셉트론 ⊂ 다층 퍼셉트론(MLP) ⊂ 딥러닝 ⊂ 인공신경망(ANN)

# **2. 딥러닝 모델의 학습**

## 순전파와 역전파  
- 학습 알고리즘 구조  
 1. 순전파를 통해 **입력값으로부터 예측**을 수행  
 2. **오차 계산**을 통해 **예측 정확도**를 측정  
 3. **역전파**를 통해 출력층에서 다시 입력층 방향으로 **오차를 전달**  
 4. **가중치 업데이트**  
 5. 1~4를 반복  

- 순전파: 예측을 수행  
- 역전파: 틀린 문제 다시 학습  

- 연쇄 법칙: 서로 얽혀있는 변수 간의 상관관계를 계산하는 방법  
  A->B->C 일 때  
  각 은닉층은 연결돼 있으므로 변수 A가 B에게, B가 C에게 영향을 미침  
  A로 인해 발생한 C의 오차=B가 C에게 준 영향 X A가 B에 준 영향  
  **합성 함수의 미분은 합성 함수를 구성하는 각 함수의 미분의 곱으로 나타낼 수 있음**  

  - **기울기 소실 문제**: 은닉층의 수가 많아질수록 앞쪽 은닉층의 영향력은 거의 사라짐 (계속 곱해지면서)  
  - 기울기 구하는 이유:  
    딥러닝 모델의 성능을 높이려면 모델의 오차를 계산해 순전파와 역전파를 활용해서 오차를 줄이는 것이 필요  
    손실 함수 = 오차값과 실제값을 수치로 표현  
    연쇄법칙을 통해 구하는 기울기 = ‘손실 함수를 줄이기 위해 가중치를 어떻게 바꿀까?’를 알려주는 신호


## 손실 함수  
- 손실함수: 모델이 정확한지 오차를 계산하는 방법, **실제값과 예측값의 차이**를 수치화하는 함수  
- 손실함수의 종류: 
 - 평균 제곱 오차(MSE): 예측값이 연속적, 예측값과 실제값의 차이를 직접적으로 측정하는 **회귀 문제**에 적합  
 - 이진 크로스 엔트로피(BCE): **이진 분류 문제(로지스틱 회귀)** 에서 사용  
   **낮은 확률로 예측해서 맞추거나 높은 확률로 예측해서 틀리는 경우가 loss 큼**
   1. 실제 정답이 1일 때:  
      예측확률이 0에 가까워질수록 손실값 증가, **예측 확률이 0에 가깝다=모델이 정답에 확신없음**  
   2. 실제 정답이 0일 때:  
      예측확률이 1에 가까워질수록 손실값 증가, **예측 확률이 1에 가깝다=모델이 확신을 가지고 틀린 예측을 함**  
 - 크로스 엔트로피 오차(CEE): **다중 클래스 분류문제**에서 사용, (진짜 분포 / 예측 분포)의 유사도 측정  


## 활성화 함수  
- 활성화 함수: 출력값을 다음 단계에서 쓸 수 있는 형태로 적절히 바꿔주는 역할 , 비선형성 만들어줌->복잡한 패턴 학습 가능  
- 활성화 함수 종류:  
  1. 계단함수:  

    

  입력 값과 가중치의 곱(입력 신호)을 모두 더한 뒤, 특정 기준을 넘으면 “1”, 아니면 “0”을 출력하는 간단한 구조   
  주로 **단층 퍼셉트론**에서 사용  

  2. 시그모이드 함수:  

    

  **곡선 형태의 비선형 함수**, 입력 신호를 특정 범위(0~1)로 변환하는 역할  

  3. **ReLU 함수**:  

    

  **비선형 함수**, 입력이 0을 넘으면 그대로/0이하면 0 출력  
   - 계산 간단하고 효율적  
   - 비선형성 제공  
   - 같은 신경망에도 역전파 잘 됨  

  4. Tanh(쌍곡탄젠트) 함수:  

    

  시그모이드 함수의 변형, 출력값의 범위(-1~1) -> 입력값의 중심을 0으로 맞추고 싶을 때 사용  

  5. Softmax 함수:  
  
    

  **다중 클래스 분류 문제의 활성화 함수**, 개별 클래스의 확률을 알 수 있음  
   - **Softmax 함수 + 크로스 엔트로피 손실 함수**  
     **예측의 확률(softmax 함수)과 정답 확률과의 차이(크로스 엔트로피 손실 함수)**를 바로 계산할 수 있어 학습 효율이 높음 -> 가장 **표준적인 다중 클래스의 분류 조합**  


## 경사하강법과 옵티마이저  
- 경사하강법: 손실함수를 최소화하여 모델 학습시키는 방법 -> 오차를 가장 작게 만드는 가중치&편향 찾는 방법  
- 배치(batch): 학습 과정에서 가중치와 같은 매개변수를 조정하기 위해 사용하는 데이터의 묶음  
  - 배치 크기는 사용자가 임의로 설정 가능한 **하이퍼파라미터**  
  - 크기가 너무 크면 학습이 느려지고 메모리가 부족해질수도 있음  
  - 너무 작으면 학습이 불안정해질 수 있음  
- 이포크(epoch): 모든 데이터셋을 학습하는 횟수  
- 이터레이션/스텝: 한 번의 에포크를 끝내기 위해서 필요한 배치의 수  
  - 1 Epoch에서 다루는 데이터의 수=배치의 크기*이터레이션  

1. 배치 경사 하강법:  
(전체 데이터를 활용하기 때문에)
- 가장 기본적인 경사 하강법, 오차를 구할 때 전체 데이터를 한 묶음으로 묶음  
- 업데이트 횟수 적은  
- 메모리 많이 필요  
- 수렴이 안정적  

2. 배치 크기가 1인 확률적 경사 하강법:  
- **랜덤으로 선택된 하나의 데이터**에 대해서만 계산  
- 자원이 적은 컴퓨터에서도 사용 가능  
- 매개변수의 변경폭 불안정, 정확도 낮을 수 있음  

3. 미니 배치 경사 하강법:  
- **배치 크기 적절히 지정**, 해당 데이터 개수만큼에 대해 매개 변수 값 조정  
- 전체 데이터 계산보다는 빠르고, 배치 크기가 1인 경사하강법보다 안정적 -> 가장 많이 사용  

- 옵티마이저: 경사하강법을 더 효율적이고 안정적으로 수행하기 위해 필요  

1. 모멘텀:  
- **경사하강법+관성**  
- 경사 하강법에서 계산된 접선의 기울기에 바로 한 시점 전의 접선의 기울기값을 일정 비율만큼 반영  
- 작은 웅덩이(local minimun)에 빠지더라도 관성의 힘으로 넘어올 수 있음  

2. 알엠에스프롭(RMSProp)  
- 기존 경사하강법에서는 기울기가 큰 방향에서는 너무 많이, 작은 방향에서는 너무 적게 이동  
- 최근 기울기 참고 -> 급한 경사에서는 속도 줄이고 완만한 경사에서는 속도 유지/증가  

3. 아담  
- RMSProp + Momentum  
- 각 파라미터의 1차 모멘트(기울기의 평균)와 2차 모멘트(기울기의 제곱 평균)을 사용하여 학습률을 조정  

## 학습을 향상시키는 기법: 데이터 증강과 전이 학습  

- 데이터 증강: 기존 데이터에서 다양한 변형을 만들어 학습 데이터를 늘림 -> 더 효과적인 학습 가능  
- 데이터 증강의 효과: 
  1. 모델 성능의 향상: 더 일반화된 패턴을 익힐 수 있음  
  2. 데이터 의존성 감소: 과적합 방지  
  3. 과적한 완화: 다양한 입력값 가질 수 있음  
  4. 데이터 프라이버시 보호: 원본 데이터의 노출 최소화할 수 있음  

- 전이 학습: 한 문제에서 학습한 지식을 다른 문제에 활용  
- 전이 학습의 장점:  
  1. 계산 비용 절감  
  2. 데이터셋 크기 문제 완화  
  3. 일반화 가능성 향상  
  4. 성능의 향상  
- 전이 학습 플로우: 
  1. Pre-trained model: 일반적 특징 학습  
     - 이미 큰 데이터셋으로 학습된 모델, 계산 비용 절감 가능
  2. Fine-Tuning: 새로운 데이터에 맞춰 일부/전체 층 재학습  
     - 초기 층은 그대로 두고, 출력층이나 후반 층만 재학습하거나, 모든 층을 새로운 데이터에 맞게 재조정  


# **3. 합성곱 신경망(CNN)**  

- 픽셀: 입력 이미지의 가장 기본적인 데이터 단위  
- 채널: 픽셀의 색을 숫자로 표현하는 방식  

### CNN  
- CNN: 이미지 공간정보를 유지한 상태로 학습이 가능한 방식  

  

- CNN 특징:  
 - Feature learning(위치별 특징 추출): 
   1. concolution layer: 입력 데이터에 필터를 적용 후 활성화 함수를 반영하는 필수 요소  
   2. pooling layer: Convolution Layer 다음에 위치하며, 이미지의 공간 크기를 줄여 중요한 특징을 강조하는 역할, 모델의 계산 효율성을 높이며 과적합을 방지하는 데 도움을 줌  
 - classification(클래스를 분류하는 부분):  
   -  CNN 모델의 마지막에 위치하고 Fully Connected Layer로 구성   

- 이미지 처리에 MLP가 아닌 CNN 사용하는 이유  
1. 이미지는 구조를 유지하는 것이 중요 -> CNN 사용  
2. MLP는 위치, 크기, 회전 등의 변화에 취약  
  - cnn에서는 convolution과 pooling으로 어떤 영역에 특정한 특징이 존재하는가를 봄  
3. 파라미터 폭발  
  - cnn은 이미지의 특정 피쳐에만 집중하여 효율적인 파라미터 사용  
4. 공간 정보 소실 
  - cnn에서는 공간 정보 보존하면서 학습  

## 합성곱(convolution)과 풀링(pooling)  
- convolution layer  
  1. 필터 또는 커널이 이미지 위를 슬라이딩  
  2. 각 위치에서 요소별 곱셈 후 합계 계산  
  3. 결과를 feature map(input data에 filter를 적용하여 얻은 결과물)에 저장  
- padding: 필터를 거치며 정보를 잃어버리는 것을 막기 위해 이미지 바깥을 0으로 둘러싸는 기법  
  
    

- slide(보폭): 이미지를 몇 칸씩 점프하면서 흝을 것인지 결정하는 값, 커질수록 feature map 크기가 작아짐  
- pooling: 피처맵에서 특징을 추출하기 위해 이미지를 격자 형태로 분할한 뒤 해당 영역 안에서 가장 큰 값 또는 평균값을 취하는 것  
  1. max pooling: 최댓값 선택  
  2. average pooling: 평균값 선택  


## 대표적인 CNN 아키텍쳐  
- AlexNet:  
머신러닝에서 벗어나 최초로 GPU를 이용해 학습한 대규모 CNN 모델, ReLU 함수 이용  
- VGGNet:  
이미지 인식에 강한 깊은 층의 CNN 모델, conv+pool을 여러 번 반복, 전이 학습에 많이 사용  
- ResNet  
잔차 연결 개념을 도입해 깊은 신경망의 소실 문제를 해결한 모델  
  - 잔차 연결: 이전 레이어의 출력을 몇 레이어를 건너뛰고 다음 레이어의 입력으로 직접 추가하는 방식  

# **4. 이미지 딥러닝 응용** 

1. 이미지 분류  

| 모델명 | 특징 요약 | 주의사항 |
| --- | --- | --- |
| **AlexNet** | 비교적 단순한 구조, 초기 CNN 모델들 | 작은 데이터셋이나 학습 개념 설명용으로 적절 |
| **VGG** | 더 깊은 구조, 복잡한 블록 구조 적용 | 계산량 증대, 메모리 요구 높음 |
| **ResNet** | Residual 연결 (skip connections) → 매우 깊은 네트워크 가능 | 과적합 주의, batch size / regularization 조절 필요 |
| **Vision Transformer (ViT)** | Transformer 구조 응용 → 패치 기반 입력 처리 | 이미지와 순차 처리 모델의 결합, 대규모 데이터 필요 |  

2. 객체 탐지  
객체 검출(object detection)은 한 이미지에서 객체와 이 객체를 둘러싸는 가장 작은 직사각형으로 정의되는 bounding box를 찾는 작업  
ex) YOLO, Faster R-CNN, SSD  

3. 이미지 캡셔닝  
이미지 내에 있는 객체에 대한 판단뿐만 아니라 객체들 간의 관계를 파악하고 자연어의 형태로 알맞게 표현  
- CNN을 이용한 특징 추출  
  

- Attention Network를 이용하여 단어 생성  
  

- RNN 이용하여 문장 완성
