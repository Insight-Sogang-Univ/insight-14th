# 1 시계열 데이터란? 
## 1.1 시계열 데이터의 정의
* 순차적인 시간의 흐름에 따라 수집된 데이터
* 시간에 따라 상관관계가 존재한다, 즉 과거 데이터가 미래에 어떤 영향을 미치는가? 이것에 대해 알아보는 것이 가장 중요!
* 종단면 데이터와 횡단면 데이터
    1. 종단면 : 동일한 개체에 대해 시간에 따른 변화
        * 개체 또는 단위 추적, 패턴 및 동향 분석, 효과 및 원인 분석
    2. 횡단면 : 몇가지 주제에 대해 특정 시점에서 수집  
        * 다중 변수, 비교분석, 상관관계
* 단변량 시계열과 다변량 시계열
    1. 단변량
        * 시간에 따라 연속적으로 측정된 단일 변수
    2. 다변량
        * 같은 시간에 기록된 둘 이상의 변수

## 1.2 시계열 분석과 그 목적
* 목표는 미레 데이터를 예측!
* 데이터 구조를 파악하고 예측, 이상치 탐지까지

## 1.3 시계열 데이터의 특징
1. 시간의존성
* 시계열 데이터는 시간에 영향을 받는다는 것!
* 시간을 바꾸면 의미가 달라지고, 과거 값이 현재 값에 영향을 미친다!
2. 계절성 & 추세
* 시간에 따라 장기적으로 증가 및 감소가 일어난다
* 일정 주기에 따라 반복이 진행된다
3. 자기상관성
* 이전의 시점 데이터가 이후 시점의 데이터에 영향을 준다
* 즉, 자기 자신이 시점별로 서로서로 상관관계를 가진다!
* 관측값들이 독립 아님
* 자기상관성 때문에, 회귀분석이 불가능!
    1. 통계적 추론의 제약 및 오류
        * 자기상관성에 의해 회귀분석의 추론이 왜곡된다
    2. 예측의 어려움 & 모델 복잡성 증가
        * 잔차가 시간에 따라 패턴 -> 미래 값 예측 왜곡 -> 갈수록 신뢰도 하락
        * 모델 복잡성 증가

## 1.4 시계열 데이터의 분석요인
1. 추세 요인
    * 데이터 전반적인 시간에 따른 증가 및 감소 경향
2. 계절 요인
    * 일정한 주기로 반복되는 패턴
3. 순환 요인
    * 계절성보다 더 길고 불규칙한 주기로 반복되는 변동
    * 계절 요인과의 차이 : 주기의 반복성의 존재 유무
4. 불규칙 요인
    * 위 3개를 다 빼고 남은 예측 불가능 변동
    * 노이즈, 우연, 일시적, 이상치 모두 존재

## 1.5 시계열 분석의 워크플로우
1. 데이터 전처리
2. 정상성 판단
3. 모델링 및 에측


# 2. 시계열 데이터 전처리

## 2.1 이상치/결측치 처리
* 결측치 : 특정 시점이 누락, 추세 및 계절성 왜곡 방지를 위해 적절한 대치와 보간 필요
* 이상치 : 값이 일반적이지 않은 경우 -> 이상한지, 중요 의미인지 사전 지식이 필요
* 시계열 데이터 전처리에서는, 특성을 왜곡되지 않도록 하는 전처리가 필요
* 데이터 확인/시각화 -> 결측치 확인 및 처리 -> 이상치 탐지 및 처리 -> 요소 분해

### 2.1.1 결측값 처리
* 삭제/대치/보간/머신러닝 기법 등이 존재
* 보간법 : 두 개 이상의 데이터 포인트 사이에서 누락된 값을 추정 -> 사라진 값을 합리적으로 메우기
    * 선형보간 : 점끼리 선으로 
    * 다항식보간 : 점끼리 다항식으로
    * 스플라인 보간 : 부분별 다항식으로
    * 조각별 상수: 가까운 상수로
 
### 2.1.2 이상치 탐지 및 처리
* 이상치 탐지 방법
    * IQR Rule-based Anomaly Detection : 사분위수 기준으로 이상치 탐지
    * Z-score : 특정값이 평균에서 얼마나 멀어져 있는지 확인
    * STL 분해 : 계절성 및 추세가 뚜렷할 때 분리해서 생각
* 이상치 탐지 모델
    * 분류 및 회귀 트리 : 데이터를 분할해서 예측 오차가 일정 임계값 이상인 지점을 이상치로
    * 클러스터링 기반 : 정상 데이터는 특정 클러스터에 모이고, 이상치들을 소수 클러스터로 나타남
    * 오토인코더 : 정상 데이터 특징을 학습하고 데이터를 입력했을 때 정상 특징과의 차이점을 비교


* 이상치 처리 방법 : 이상치를 결측치로 간주해서 결측치와 동일하게 처리
1. 평활화 : 노이즈를 줄이고 전체적 흐름 및 추세를 더 명확하게 파악
    * 이동평균 : 부분집합의 평균을 구해서 평활화
        * 단순 이동평균 : 단순 평균 구해서 대치
        * 가중 이동평균 : 가중 평균 구해서 대치
        * 지수 이동평균 : 지수적 감소 가중치 적용

## 2.2 요소분해 (시계열 분해)
* 요소분해 : 시계열 데이터를 개별 성분으로 분해하는 과정
    * 추세성분, 계절성분, 순환성분, 불규칙성분으로!
    * 시계열 데이터는 다양한 변동이 결합되어 있으므로, 각 요소를 분리하면 각각 패턴 명확히 이해 가능

* 가법모형 : 각 요인이 합연사으로 이루어져있다고 생각, 계절성의 진폭이 일정할 때 적합
* 승법모형 : 로그 변환을 통해 가법으로 만들 수 있는 곱해진 모델, 추세가 커질수록 계절성이나 불규칙성의 효과도 비례해서 커질때
* 요약: 각 요소를 독립적으로 분석해서, 특정 시점에 급격한 변화가 있다면 어느 요소에서 비롯된 것인지 분석해서 이상 탐지 및 원인 분석 가능


# 3. 정상성 판단   
## 3.1 정상성이란? 
* 정상성 : 시간에 상관없이 일정한 성질을 띠고 있다! 평균 및 분산 등이 시간에 따라 일정한 수준으로 유지
* 정상성을 띠면 정상시계열, 아니면 비정상 시계열

## 3.2 강한 정상성과 약한 정상성
* 강한 정상성과 약한 정상성
    * 강한 정상성 : 모든 적률이 시간과 무관하게 일정함, 즉 언제 관찰하든 확률정 패턴이 같다
        * 기저를 이루는 확률 분포 자체가 언제나 같아야 한다
        * 현실에서 찾아보기 어려움 : 노이즈와 외부 요인이 존재하므로!
    * 약한 정상성 
        * 조건 : 평균은 모든 시점에 대해 일정하다, 분산은 모든 시점에 대해 일정, 공분산은 시차에만 의존
        * 공분산이 시차에만 의존하고, 시점에는 의존하지 않는다!
## 3.3 정상성을 시각적으로 파악하기

* 추세가 있는 시계열 : 비정상
* 계절성 : 비정상
* 분산이 증가하거나 감소? : 비정상

## 3.4 검정을 통해 정상성 판단하기!
* 단위근 : 양의 정수 N에 대해 X^N = 1 의 근이 되는 복소수
* 시계열적 의미 : 특성방정식에서 근이 1인 경우, 단위근을 가진다 -> 비정상 시계열!
* 정상 시계열은 충격을 받아도 시간이 지나면 돌아옴
* 비정상 시계열을 충격 받으면 그 상태를 유지한다.
* 단위근이 있으면 랜덤워크로 인해 예측의 불확실성이 생긴다
    1. 평균이 비일정
    2. 분산이 커짐
    3. 과거 충격이 영원히 남는다

* 단위근으로 정상 / 비정상 시계열 판단하는 과정
* AR 모델 : 시계열의 관측값과 백색잡음으로 식을 이룬다
    * 알파가 0이면 과거 값의 영향을 받지 않는 완전한 정상성
    * 알파가 1보다 작으면 이전 값 영향을 받기는 하지만 과거 영향이 줄어들어 안정상태를 유지
        * 시간이 지날수록 사라지고 평균수준으로 돌아오는 정상 시계열
    * 알파가 1 : 단위근이 있는 비정상 시계열, 충격을 받으면 돌아가지 못한다

    * 그럼 알파값으로만 판단이 되는가? 아님! AR 모델 외에 또 계수 = 단위근이라고 할 수 없음

* 단위근 검정
    1. ADF 검정 : 단위근 존재 여부를 통해 판단
    2. KPSS 검정 : 단위근 기반의 검정이지만 귀무가설이 반대
    3. 확장적 추세가 존재할 때 KPSS와 ADF의 결론이 서로 다르게 나올 수 있다
        * 확장적 추세 : 추세가 완벽하게 예측 가능
        * 확률적 추세 : 추세선 자체가 시간에 따라 이동
    * KPSS는 평균 주위에서 정상성을 보는 데 중점을 둔다, 확장적 추세가 있으면 ADF와 결과가 다름
    * ADF는 단위근 때문이라고 생극하고, KPSS는 정상이라고 생각한다
    * 추세항을 추가하면 되잖냐? 이론적으로는 그리하지만 현실적으로 ADF가 검정력이 낮다 -> 결과가 상충되면 확장적 추세가 있음을 의심해보자!

## 3.5 정상성 확보를 위한 여러가지 방법들
*  비정상 시계열에서 이것저것 제거
1. 추세 제거
    * 차분 : 연이은 관측값들의 차이를 계산, 누적된 값을 제거
2. 계절성 제거
    * 계절 차분 : 주기의 길이에 해당하는 y값을 뺀다
    * 백색소음과 같은 형태의 완전히 랜덤한 데이터를 뽑아낼 수 있다
3. 분산 안정화 : 시계열의 변동 폭이 커지거나 작아질때
    * 로그변화 : 눈금 변환
    * 박스-콕스 변환 : 로그 변환의 일반화된 형태
        * 람다 값으로 변환 강도를 조절하여 다양한 변환이 가능, 추정법을 통해 적합한 람다를 찾을 수 있다.

* 요약 : 정상성 확보 -> 패턴 파악 -> 적절한 모델링
* 정상/ 비정상 여부 판단으로 시작
* 비정상 유형에 따라 정상성 확보
* 정상성 확보 후 어떤 프로세스를 따르는지 모델링
