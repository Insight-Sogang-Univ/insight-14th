```python
(x_train, _), (x_test, _) = tf.keras.datasets.cifar10.load_data()

# ëª¨ë“  í”½ì…€ê°’ì„ 0~1 ì‚¬ì´ë¡œ ì •ê·œí™”í•©ë‹ˆë‹¤.
x_train = x_train.astype("float32") / 255.0
x_test = x_test.astype("float32") / 255.0

print("í›ˆë ¨ ë°ì´í„° shape:", x_train.shape)
print("í…ŒìŠ¤íŠ¸ ë°ì´í„° shape:", x_test.shape)
```

### 1-2. AutoEncoder ëª¨ë¸ ì •ì˜

- CIFAR-10ì€ 32Ã—32Ã—3 RGB ì´ë¯¸ì§€ì´ë¯€ë¡œ, ë‹¨ìˆœ ì™„ì „ì—°ê²°ì¸µ(Dense) ëŒ€ì‹  **Convolutional AutoEncoder (CNN ê¸°ë°˜) êµ¬ì¡°**ë¥¼ ì‚¬ìš©í•©ë‹ˆë‹¤.

```python
def build_cnn_autoencoder(input_shape=(32, 32, 3)):
    input_img = layers.Input(shape=input_shape)

    # ì¸ì½”ë” (Encoder) # í™œì„±í™” í•¨ìˆ˜ëŠ” ë ë£¨!
    x = layers.Conv2D(32, (3,3), activation="relu", padding="same")(input_img)
    x = layers.MaxPooling2D((2,2), padding="same")(x)
    x = layers.Conv2D(16, (3,3), activation="relu", padding="same")(x)
    encoded = layers.MaxPooling2D((2,2), padding="same")(x)  # (8,8,16)

    # ë””ì½”ë” (Decoder)
    x = layers.Conv2DTranspose(16, (3,3), strides=2, activation="relu", padding="same")(encoded)
    x = layers.Conv2DTranspose(32, (3,3), strides=2, activation="relu", padding="same")(x)
    decoded = layers.Conv2D(3, (3,3), activation="sigmoid", padding="same")(x)

    model = Model(input_img, decoded)
    return model

autoencoder = build_cnn_autoencoder()
autoencoder.summary()
```

### 1-3. AutoEncoder í•™ìŠµ

```python
# ì˜µí‹°ë§ˆì´ì €ëŠ” ì•„ë‹´(adam), lossëŠ” í‰ê· ì œê³±ì˜¤ì°¨(mse)ë¥¼ ì‚¬ìš©í•˜ê² ìŠµë‹ˆë‹¤!
autoencoder.compile(
    optimizer="adam",
    loss="mse"
)
# ì—í­ì€ 30, ë°°ì¹˜ì‚¬ì´ì¦ˆëŠ” 128, ì…”í”Œë„ Trueë¡œ í•˜ê² ìŠµë‹ˆë‹¤!
# ì˜¤í† ì¸ì½”ë”ëŠ” ë¹„ì§€ë„í•™ìŠµì´ë¯€ë¡œ y_trainì´ ì—†ìŠµë‹ˆë‹¤.
history = autoencoder.fit(
    x_train, x_train,  # x_train ìì²´ê°€ ì…ë ¥ì´ì ëª¨ë¸ì´ ë§ì¶°ì•¼ í•  ì •ë‹µ ì—­í• ì„ ë™ì‹œì— ìˆ˜í–‰í•©ë‹ˆë‹¤.
    epochs=30,
    batch_size=128,
    shuffle=True,
    validation_data=(x_test, x_test)  # ê·¸ë˜ì„œ ì—¬ê¸°ë„ x_testê°€ ë‘ë²ˆì“°ì…ë‹ˆë‹¤!
)
```

### 1-4. ì¬ë³µì›

```python
# decoded_imgsì— ì˜¤í† ì¸ì½”ë”ì˜ 'ì˜ˆì¸¡'ì„ í• ë‹¹í•´ì¤ì‹œë‹¤
decoded_imgs = autoencoder.predict(x_test)

n = 10
plt.figure(figsize=(20, 6))

for i in range(1, n + 1):
    # ì›ë³¸ ì´ë¯¸ì§€
    ax = plt.subplot(2, n, i)
    plt.imshow(x_test[i])
    plt.axis("off")
    if i == 1:
        plt.title("Original")

    # ë³µì› ì´ë¯¸ì§€
    ax = plt.subplot(2, n, i + n)
    plt.imshow(decoded_imgs[i])
    plt.axis("off")
    if i == 1:
        plt.title("Reconstructed")

plt.show()
```

## ğŸ” 2. VAE(Variational AutoEncoder)

### 2-1. ë°ì´í„° ì¤€ë¹„

```python
# CIFAR-10 ë°ì´í„° ë¡œë“œ
(x_train, _), (x_test, _) = tf.keras.datasets.cifar10.load_data()

# ì •ê·œí™”
x_train = x_train.astype("float32") / 255.0
x_test = x_test.astype("float32") / 255.0

print("í›ˆë ¨ ë°ì´í„°:", x_train.shape)
print("í…ŒìŠ¤íŠ¸ ë°ì´í„°:", x_test.shape)
```

### 2-2. VAE ëª¨ë¸ ì •ì˜
- **Encoder**: ì´ë¯¸ì§€ë¥¼ ì ì¬ê³µê°„ì˜ í‰ê· ($\mu$)ê³¼ ë¡œê·¸ ë¶„ì‚°($log\sigma^2$)ìœ¼ë¡œ ì••ì¶•
- **Reparametrization Trick**: $z = \mu + \sigma \cdot \epsilon$
- **Decoder**: $z$ë¡œë¶€í„° ì´ë¯¸ì§€ë¥¼ ë³µì›

```python
latent_dim = 64  # ì ì¬ê³µê°„ í¬ê¸°
```

```python
# ----- ì¸ì½”ë” ----- í™œì„±í™”í•¨ìˆ˜ëŠ” ë ë£¨!
encoder_inputs = layers.Input(shape=(32, 32, 3))
x = layers.Conv2D(32, 3, activation="relu", strides=2, padding="same")(encoder_inputs)
x = layers.Conv2D(64, 3, activation="relu", strides=2, padding="same")(x)
x = layers.Flatten()(x)
x = layers.Dense(128, activation="relu")(x)

z_mean = layers.Dense(latent_dim, name="z_mean")(x)
z_log_var = layers.Dense(latent_dim, name="z_log_var")(x)
```

VAEì˜ ëª©ì ì€ **ì…ë ¥** $X$ë¥¼ **í™•ë¥  ë¶„í¬** $P(z|X)$ë¥¼ ë”°ë¥´ëŠ” **ì ì¬ ë²¡í„° $z$ë¡œ ì¸ì½”ë”©**í•˜ëŠ” ê²ƒì…ë‹ˆë‹¤.

ì´ ê³¼ì •ì—ì„œ **$z$ë¥¼ ìƒ˜í”Œë§**í•˜ê²Œ ë˜ëŠ”ë°, \
**ì¼ë°˜ì ì¸ ë¬´ì‘ìœ„ ìƒ˜í”Œë§**ì€ **ì—­ì „íŒŒ(Backpropagation)ë¥¼ í†µí•´ ëª¨ë¸ì˜ ê°€ì¤‘ì¹˜ë¥¼ ì—…ë°ì´íŠ¸í•  ìˆ˜ ì—†ë‹¤**ëŠ” ë¬¸ì œê°€ ìˆìŠµë‹ˆë‹¤.

**ì¬ë§¤ê°œë³€ìˆ˜í™” íŠ¸ë¦­**ì€ ì´ ë¬¸ì œë¥¼ í•´ê²°í•˜ì—¬,\
$z$ë¥¼ ìƒì„±í•˜ëŠ” ê³¼ì •ì— **'ë¬´ì‘ìœ„ì„±(Randomness)'ì„ ìœ ì§€**í•˜ë©´ì„œë„ **'ë¯¸ë¶„ ê°€ëŠ¥ì„±(Differentiability)'ì„ í™•ë³´**í•©ë‹ˆë‹¤.

```python
# ì¬ë§¤ê°œë³€ìˆ˜í™” íŠ¸ë¦­(Reparametrization trick)
def sampling(args):
    z_mean, z_log_var = args        # zì˜ í‰ê· ê³¼ ë¡œê·¸_ë¶„ì‚°ì„ ìƒ˜í”Œë§í•©ì‹œë‹¤!
    eps = tf.random.normal(shape=(tf.shape(z_mean)[0], latent_dim)) #ë¬´ì‘ìœ„ì„±ì„ ì¤ì‹œë‹¤.
    return z_mean + tf.exp(0.5 * z_log_var) * eps

z = layers.Lambda(sampling, name="z")([z_mean, z_log_var])
encoder = Model(encoder_inputs, [z_mean, z_log_var, z], name="encoder")
encoder.summary()
```

```python
# ----- ë””ì½”ë” -----
latent_inputs = layers.Input(shape=(latent_dim,))
x = layers.Dense(8 * 8 * 64, activation="relu")(latent_inputs)
x = layers.Reshape((8, 8, 64))(x)
x = layers.Conv2DTranspose(64, 3, activation="relu", strides=2, padding="same")(x)
x = layers.Conv2DTranspose(32, 3, activation="relu", strides=2, padding="same")(x)
decoder_outputs = layers.Conv2D(3, 3, activation="sigmoid", padding="same")(x)

decoder = Model(latent_inputs, decoder_outputs, name="decoder")
decoder.summary()
```

### 2-3. trainìš© VAE í´ë˜ìŠ¤ ì •ì˜
1. **VAEì˜ ì´ ì†ì‹¤** = **ì¬êµ¬ì„± ì†ì‹¤** + **KL ë°œì‚° ì†ì‹¤**

2. **ì¬êµ¬ì„±(reconstruction) ì†ì‹¤**ì€ ì¼ë°˜ì ì¸ **ì˜¤í† ì¸ì½”ë”ì˜ ì†ì‹¤ê³¼ ë™ì¼**í•˜ë©°, **VAEì˜ ì••ì¶• ë° ë³µì› ëŠ¥ë ¥**ì„ ë‹´ë‹¹í•©ë‹ˆë‹¤.
   - **ëª©í‘œ**ëŠ” ì›ë³¸ ë°ì´í„° ($X$)ì™€ ë””ì½”ë”ê°€ ë³µì›í•œ ì¶œë ¥ ($\hat{X}$) ì‚¬ì´ì˜ **ì°¨ì´ë¥¼ ìµœì†Œí™”**í•˜ëŠ” ê²ƒì…ë‹ˆë‹¤.
   - ëª¨ë¸ì´ **ì…ë ¥ ë°ì´í„°ì˜ íŠ¹ì§•ì„ ì˜ ì¸ì½”ë”©**í•˜ì—¬ **ì •í™•í•˜ê²Œ ë³µì›í•  ìˆ˜ ìˆë„ë¡ ê°•ì œ**í•˜ëŠ” ì—­í• ì…ë‹ˆë‹¤!

3. **KL(Kullback-Leibler) ë°œì‚° ì†ì‹¤**ì€ VAEê°€ ë‹¨ìˆœí•œ ì˜¤í† ì¸ì½”ë”ë¥¼ ë„˜ì–´ **ìƒˆë¡œìš´ ë°ì´í„°ë¥¼ ìƒì„±í•  ìˆ˜ ìˆê²Œ ë§Œë“œëŠ”** í•µì‹¬ ìš”ì†Œì…ë‹ˆë‹¤.
   - **ëª©í‘œ**ëŠ” **ì¸ì½”ë”ê°€ í•™ìŠµí•œ ì ì¬ ê³µê°„ì˜ ë¶„í¬** $P(z|X)$ê°€ **ë¯¸ë¦¬ ì •ì˜ëœ ì´ìƒì ì¸ ë¶„í¬** (Prior Distribution, ë³´í†µ í‘œì¤€ ì •ê·œ ë¶„í¬)ì™€ **ìµœëŒ€í•œ ìœ ì‚¬í•˜ë„ë¡ ê°•ì œ**í•˜ëŠ” ê²ƒì…ë‹ˆë‹¤.
   - ì—­í• ë¡œëŠ” **ì ì¬ ê³µê°„ $Z$ê°€ ê·œì¹™ì ì´ê³  ì—°ì†ì ì¸ êµ¬ì¡°**ë¥¼ ê°–ê²Œ í•©ë‹ˆë‹¤. \
   ì ì¬ ê³µê°„ì˜ **ë¹ˆí‹ˆì´ ì—†ë„ë¡** ë§Œë“¤ì–´, ì„ì˜ì˜ $z$ë¥¼ ìƒ˜í”Œë§í•´ë„ **ì˜ë¯¸ ìˆëŠ” ë°ì´í„°ë¥¼ ìƒì„±**í•  ìˆ˜ ìˆê²Œ í•©ë‹ˆë‹¤ (Generative, ìƒì„± ëŠ¥ë ¥ ë¶€ì—¬).

```python
class VAE(Model):
    def __init__(self, encoder, decoder, **kwargs):
        super(VAE, self).__init__(**kwargs)
        self.encoder = encoder
        self.decoder = decoder

    def compile(self, optimizer, **kwargs):
        super(VAE, self).compile(**kwargs)
        self.optimizer = optimizer

        self.total_loss_tracker = tf.keras.metrics.Mean(name="total_loss")
        self.recon_loss_tracker = tf.keras.metrics.Mean(name="reconstruction_loss")
        self.kl_loss_tracker = tf.keras.metrics.Mean(name="kl_loss")

    @property
    def metrics(self):
        return [self.total_loss_tracker,
                self.recon_loss_tracker,
                self.kl_loss_tracker]

    def call(self, inputs):
        z_mean, z_log_var, z = self.encoder(inputs)
        reconstructed = self.decoder(z)
        return reconstructed

    def train_step(self, data):
        if isinstance(data, tuple):
            data = data[0]
        with tf.GradientTape() as tape:
            z_mean, z_log_var, z = self.encoder(data)
            reconstruction = self.decoder(z)
            # ì¬êµ¬ì„± ì†ì‹¤ (í”½ì…€ë³„ MSE)
            recon_loss = tf.reduce_mean(tf.reduce_sum(tf.square(data - reconstruction), axis=[1,2,3]))
            # KL ì†ì‹¤
            kl_loss = -0.5 * tf.reduce_mean(
                tf.reduce_sum(1 + z_log_var - tf.square(z_mean) - tf.exp(z_log_var), axis=1)
            )
            total_loss = recon_loss + kl_loss

        grads = tape.gradient(total_loss, self.trainable_weights)
        self.optimizer.apply_gradients(zip(grads, self.trainable_weights))
        self.total_loss_tracker.update_state(total_loss)
        self.recon_loss_tracker.update_state(recon_loss)
        self.kl_loss_tracker.update_state(kl_loss)
        return {
            "loss": self.total_loss_tracker.result(),
            "recon_loss": self.recon_loss_tracker.result(),
            "kl_loss": self.kl_loss_tracker.result(),
        }

    def test_step(self, data):
        if isinstance(data, tuple):
            data = data[0]
        z_mean, z_log_var, z = self.encoder(data)
        reconstruction = self.decoder(z)

        recon_loss = tf.reduce_mean(
            tf.reduce_sum(tf.square(data - reconstruction), axis=[1, 2, 3])
        )
        kl_loss = -0.5 * tf.reduce_mean(
            tf.reduce_sum(1 + z_log_var - tf.square(z_mean) - tf.exp(z_log_var), axis=1)
        )
        total_loss = recon_loss + kl_loss

        return {
            "loss": total_loss,
            "recon_loss": recon_loss,
            "kl_loss": kl_loss
        }
```

### 2-4. VAE í•™ìŠµ

```python
# ëª¨ë¸ ìƒì„±
vae = VAE(encoder, decoder)
vae.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-4),
    loss=lambda x, y: 0.0  # placeholder loss (validationìš©)
)
```

```python
# í•™ìŠµ ì—í­ì€ 30, ë°°ì¹˜ì‚¬ì´ì¦ˆëŠ” 128
history = vae.fit(
    x_train,
    epochs=30,
    batch_size=128,
    validation_data=(x_test, None)
)
```

## ğŸ” 3. GAN

### 3-2. ìƒì„±ì(Generator) ëª¨ë¸ ì •ì˜

```python
# Generator ëª¨ë¸ ì •ì˜
class Generator(nn.Module):
    def __init__(self, nz, ngf, nc):
        super(Generator, self).__init__()
        self.main = nn.Sequential(
            nn.ConvTranspose2d(nz, ngf * 8, 4, 1, 0, bias=False),
            nn.BatchNorm2d(ngf * 8),
            nn.ReLU(True),
            nn.ConvTranspose2d(ngf * 8, ngf * 4, 4, 2, 1, bias=False),
            nn.BatchNorm2d(ngf * 4),
            nn.ReLU(True),
            nn.ConvTranspose2d(ngf * 4, ngf * 2, 4, 2, 1, bias=False),
            nn.BatchNorm2d(ngf * 2),
            nn.ReLU(True),
            nn.ConvTranspose2d(ngf * 2, ngf, 4, 2, 1, bias=False),
            nn.BatchNorm2d(ngf),
            nn.ReLU(True),
            nn.ConvTranspose2d(ngf, nc, 4, 2, 1, bias=False),
            nn.Tanh()
        )

    def forward(self, input):
        return self.main(input)
```

### 3-3. íŒë³„ì(Discriminator) ëª¨ë¸ ì •ì˜

```python
# Discriminator ëª¨ë¸ ì •ì˜
class Discriminator(nn.Module):
    def __init__(self, nc, ndf):
        super(Discriminator, self).__init__()
        self.main = nn.Sequential(
            nn.Conv2d(nc, ndf, 4, 2, 1, bias=False),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Conv2d(ndf, ndf * 2, 4, 2, 1, bias=False),
            nn.BatchNorm2d(ndf * 2),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Conv2d(ndf * 2, ndf * 4, 4, 2, 1, bias=False),
            nn.BatchNorm2d(ndf * 4),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Conv2d(ndf * 4, ndf * 8, 4, 2, 1, bias=False),
            nn.BatchNorm2d(ndf * 8),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Conv2d(ndf * 8, 1, 4, 1, 0, bias=False),
            nn.Sigmoid()
        )

    def forward(self, input):
        return self.main(input)
```

ToTensor() ë³€í™˜ í›„ 0ì—ì„œ 1 ì‚¬ì´ì˜€ë˜ í”½ì…€ ê°’ë“¤ì´ ì´ì œ -1ì—ì„œ 1 ì‚¬ì´ë¡œ ë§¤í•‘ë˜ëŠ” ê²ƒì´ì§€ìš”!


```python
# ë°ì´í„°ì…‹ ë¡œë” ì„¤ì •
def load_data(batch_size):
    dataset = ImageFolder(
        root="data/cats",
        transform=transforms.Compose([
            transforms.Resize(64),       #í¬ê¸° ì¬ì„¤ì •
            transforms.CenterCrop(64),
            transforms.ToTensor(),        #í…ì„œí™”
            transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5)),   #ì •ê·œí™”
        ])
    )
    dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=True)
    return dataloader
```

ì•„ë˜ ì½”ë“œì—ì„œ initialize_models í•¨ìˆ˜ëŠ” Generatorì™€ Discriminator ëª¨ë¸ì„ ì´ˆê¸°í™”í•˜ê³ , ê° ëª¨ë¸ì˜ ê°€ì¤‘ì¹˜ë¥¼ weights_init í•¨ìˆ˜ë¡œ ì„¤ì •í•œ ìˆ˜ ë°˜í™˜í•˜ëŠ” ì½”ë“œì…ë‹ˆë‹¤.

weights_init í•¨ìˆ˜ëŠ” Convolution ë° BatchNorm ë ˆì´ì–´ì˜ ê°€ì¤‘ì¹˜ë¥¼ ì´ˆê¸°í™”í•˜ëŠ” í•¨ìˆ˜ë¡œ ê° ëª¨ë¸ì˜ ì´ˆê¸° ê°€ì¤‘ì¹˜ë¥¼ ì„¤ì •í•´ì„œ í•™ìŠµì˜ ì•ˆì „ì„±ì„ ë†’ì…ë‹ˆë‹¤.

Convolution ë ˆì´ì–´ëŠ” í‰ê· ì´ 0, í‘œì¤€í¸ì°¨ê°€ 0.02ì¸ ë¶„í¬ë¡œ ì´ˆê¸°í™”ë˜ê³ \
BatchNorm ë ˆì´ì–´ëŠ” í‰ê· ì´ 1, í‘œì¤€í¸ì°¨ê°€ 0.02ì¸ ì •ê·œë¶„í¬ë¡œ ì´ˆê¸°í™”ë©ë‹ˆë‹¤.

```python
# ëª¨ë¸ ì´ˆê¸°í™”
def initialize_models(nz, ngf, ndf, nc, device):
    netG = Generator(nz, ngf, nc).to(device)
    netD = Discriminator(nc, ndf).to(device)
    netG.apply(weights_init)
    netD.apply(weights_init)
    return netG, netD

# ê°€ì¤‘ì¹˜ ì´ˆê¸°í™” í•¨ìˆ˜
def weights_init(m):
    if isinstance(m, (nn.Conv2d, nn.ConvTranspose2d)):
        nn.init.normal_(m.weight.data, 0.0, 0.02)
    elif isinstance(m, nn.BatchNorm2d):
        nn.init.normal_(m.weight.data, 1.0, 0.02)
        nn.init.constant_(m.bias.data, 0)
```

### 3-5. GAN í•™ìŠµ

ì´ì œ ëª¨ë¸ì„ íŠ¸ë ˆì´ë‹í•˜ëŠ” í•¨ìˆ˜ë¥¼ ë§Œë“­ë‹ˆë‹¤.

fixed_noiseì„ ì„¤ì •í•´ì„œ ê³ ì •ëœ latent vectorì„ ì‚¬ìš©í•´ í•™ìŠµ ê³¼ì • ë™ì•ˆ ìƒì„±ëœ ì´ë¯¸ì§€ê°€ ì–´ë–»ê²Œ ë³€í•˜ëŠ”ì§€ ê´€ì°°í•´ë´…ì‹œë‹¤.

```python
#ì˜µí‹°ë§ˆì´ì €ëŠ” ì•„ë‹´
def train_dcgan(num_epochs, dataloader, netG, netD, device, fixed_noise, save_interval=5):
    criterion = nn.BCELoss()
    optimizerD = optim.Adam(netD.parameters(), lr=0.0002, betas=(0.5, 0.999))
    optimizerG = optim.Adam(netG.parameters(), lr=0.0002, betas=(0.5, 0.999))

    G_losses, D_losses, D_accuracies = [], [], []

    # tqdm epoch ë£¨í”„
    for epoch in tqdm(range(num_epochs), desc="Training Progress"):
        loop = tqdm(enumerate(dataloader), total=len(dataloader), leave=False, desc=f"Epoch {epoch+1}/{num_epochs}")

        for i, (real_images, _) in loop:
            ############################
            # (1) Discriminator í•™ìŠµ
            ############################
            netD.zero_grad()
            real_images = real_images.to(device)
            b_size = real_images.size(0)

            label_real = torch.ones(b_size, device=device)
            output_real = netD(real_images).view(-1)
            errD_real = criterion(output_real, label_real)
            D_x = output_real.mean().item()

            noise = torch.randn(b_size, nz, 1, 1, device=device)
            fake_images = netG(noise)
            label_fake = torch.zeros(b_size, device=device)
            output_fake = netD(fake_images.detach()).view(-1)
            errD_fake = criterion(output_fake, label_fake)
            D_G_z1 = output_fake.mean().item()

            errD = errD_real + errD_fake
            errD.backward()
            optimizerD.step()

            ############################
            # (2) Generator í•™ìŠµ
            ############################
            netG.zero_grad()
            label_gen = torch.ones(b_size, device=device)
            output_gen = netD(fake_images).view(-1)
            errG = criterion(output_gen, label_gen)
            errG.backward()
            D_G_z2 = output_gen.mean().item()
            optimizerG.step()

            # tqdm ì§„í–‰ë¥ ë°”ì— ì •ë³´ ì—…ë°ì´íŠ¸
            loop.set_postfix({
                "Loss_D": f"{errD.item():.3f}",
                "Loss_G": f"{errG.item():.3f}",
                "D(x)": f"{D_x:.3f}",
                "D(G(z))": f"{D_G_z1:.3f}/{D_G_z2:.3f}"
            })

        # --- ë¡œê·¸ ì €ì¥ ---
        G_losses.append(errG.item())
        D_losses.append(errD.item())
        D_accuracies.append(D_x)

        # --- ê²°ê³¼ ì´ë¯¸ì§€ ì €ì¥ ---
        if epoch % save_interval == 0 or epoch == num_epochs - 1:
            os.makedirs("results", exist_ok=True)
            with torch.no_grad():
                fake_images_fixed = netG(fixed_noise).detach().cpu()
                fake_images_fixed = (fake_images_fixed + 1) / 2
            vutils.save_image(fake_images_fixed, f"results/fake_samples_epoch_{epoch}.png")

    return G_losses, D_losses, D_accuracies
```

ì´ì œ í•˜ì´í¼íŒŒë¼ë¯¸í„°ë“¤ì„ ì„¤ì •í•˜ê³  í•™ìŠµì„ ì§„í–‰í•´ë´…ì‹œë‹¤.

latent vectorì˜ ì‚¬ì´ì¦ˆëŠ” 100, ë°°ì¹˜ ì‚¬ì´ì¦ˆëŠ” 128ë¡œ ë‘ê³  50 ì—í¬í¬ ë™ì•ˆ í•™ìŠµì‹œì¼œë³´ê² ìŠµë‹ˆë‹¤.

```python
# ì£¼ìš” í•˜ì´í¼íŒŒë¼ë¯¸í„° ì„¤ì •
nz = 100    # latent vector size
ngf = 64    # size of feature maps in generator
ndf = 64    # size of feature maps in discriminator
nc = 3      # number of color channels
batch_size = 12
num_epochs = 50
save_interval = 5
```

```python
# í•™ìŠµ í”„ë¡œì„¸ìŠ¤ ì‹¤í–‰   T4ê¸°ì¤€ 22ë¶„ ì†Œìš”
dataloader = load_data(batch_size)
netG, netD = initialize_models(nz, ngf, ndf, nc, device)
fixed_noise = torch.randn(64, nz, 1, 1, device=device)

G_losses, D_losses, D_accuracies = train_dcgan(
    num_epochs, dataloader, netG, netD, device, fixed_noise, save_interval
)
```

## ğŸ” 4. Diffusion Model - "Stable Diffusion"

- **Stable Diffusion**ì€ **VAE + Diffusion U-Net êµ¬ì¡°**ë¡œ ë˜ì–´ ìˆìŠµë‹ˆë‹¤.

  - ì´ë•Œì˜ VAEëŠ” ìˆ˜ì–µ ì¥ì˜ ì´ë¯¸ì§€ë¡œ **ì‚¬ì „í•™ìŠµ(pretrained) ëœ ìƒíƒœ**ì…ë‹ˆë‹¤.

- ë”°ë¼ì„œ **ì´ë¯¸ì§€ â†”ï¸ latent vector ë³€í™˜**ì„ ë§¤ìš° ì •ë°€í•˜ê²Œ ìˆ˜í–‰í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

### 4-1. ë°ì´í„° ì¤€ë¹„

```python
from tensorflow.keras.datasets import cifar10

# CIFAR-10 ë°ì´í„° ë¡œë“œ
(x_train, y_train), (_, _) = cifar10.load_data()

# ìë™ì°¨ í´ë˜ìŠ¤ë§Œ í•„í„°ë§ (class 1)
car_idx = np.where(y_train.flatten() == 1)[0]
cars = x_train[car_idx]

# 0~1 ì •ê·œí™”, ê·¸ë ˆì´ìŠ¤ì¼€ì¼ ë³€í™˜
cars = cars.astype('float32') / 255.0
cars_gray = np.dot(cars[...,:3], [0.299, 0.587, 0.114])[..., np.newaxis]  # (N,32,32,1)

# í•™ìŠµìš©ìœ¼ë¡œ 256x256 í¬ê¸°ë¡œ ì—…ìƒ˜í”Œë§ (ì›í•˜ë©´ ê·¸ëŒ€ë¡œ 32x32ë¡œë„ ê°€ëŠ¥)
cars_resized = np.array([cv2.resize(img, (32,32)) for img in cars_gray])
```

### 4-2. ìˆœë°©í–¥ í™•ì‚°(Forward Diffusion) êµ¬í˜„
- 2ì°¨ì› ê°•ì•„ì§€ ì´ë¯¸ì§€ì— ì ì§„ì ìœ¼ë¡œ ë…¸ì´ì¦ˆë¥¼ ì¶”ê°€í•˜ëŠ” ê³¼ì •ì„ êµ¬í˜„í•´ë´…ì‹œë‹¤.

```python
# ì´ë¯¸ì§€ì— ì •ê·œë¶„í¬ë¥¼ ë”°ë¥´ëŠ” ë…¸ì´ì¦ˆë¥¼ ì¶”ê°€í•˜ëŠ” í•¨ìˆ˜ì…ë‹ˆë‹¤.
# ë…¸ì´ì¦ˆ í¬ê¸°ëŠ” noise_levelë¡œ ì¡°ì ˆë©ë‹ˆë‹¤.
# min_value, max_value: í”½ì…€ê°’ì˜ í•˜í•œ/ìƒí•œ ë²”ìœ„
def add_noise(img, noise_level, min_value=0.0, max_value=1.0):
    """
    img: ì…ë ¥ ì´ë¯¸ì§€ (0~1 ì •ê·œí™”)
    noise_level: í‘œì¤€í¸ì°¨ ìˆ˜ì¤€
    """
    noise = np.random.normal(0, noise_level, img.shape)
    noisy_img = img + noise     # imgì— ë…¸ì´ì¦ˆë¥¼ ì¶”ê°€í•´ì¤ì‹œë‹¤
    noisy_img = np.clip(noisy_img, min_value, max_value)
    return noisy_img, noise
```

```python
# ì´ë¯¸ì§€ì— ì ì§„ì ìœ¼ë¡œ ë…¸ì´ì¦ˆë¥¼ ì¶”ê°€í•´ ì—¬ëŸ¬ ë‹¨ê³„ì˜ ë…¸ì´ì¦ˆ ì´ë¯¸ì§€ë¥¼ ìƒì„±í•˜ëŠ” í•¨ìˆ˜ì…ë‹ˆë‹¤.
def forward_diffusion(imgs, beta_start, beta_end, num_steps, min_value=0.0, max_value=1.0):
    """
    imgs: (N,H,W,1)
    beta_start, beta_end: ë² íƒ€ ê°’ ë²”ìœ„ (ë…¸ì´ì¦ˆ ê°•ë„)
    num_steps: í™•ì‚° ë‹¨ê³„ ìˆ˜
    """
    betas = np.linspace(beta_start, beta_end, num_steps)
    alphas = 1 - betas
    alpha_bars = np.cumprod(alphas)

    noisy_images, noise_list = [], []

    for t in range(num_steps):
        noise_level = np.sqrt(1 - alpha_bars[t])     # DDPM ìˆ˜ì‹ì— ë§ì¶˜ ë…¸ì´ì¦ˆ í¬ê¸°
        noisy, noise = add_noise(imgs, noise_level, min_value, max_value)  #ìœ„ì—ì„œ ì •ì˜í•œ ë…¸ì´ì¦ˆ ì¶”ê°€í•¨ìˆ˜!
        noisy_images.append(noisy[0:1][..., np.newaxis])
        noise_list.append(noise[0:1][..., np.newaxis])

    return np.array(noisy_images), np.array(noise_list), betas, alphas, alpha_bars
```

```python
# ìˆœë°©í–¥ í™•ì‚° ìˆ˜í–‰
num_steps = 100
noisy_images, noise_list, betas, alphas, alpha_bars = forward_diffusion(
    cars_resized,        # CIFAR-10 ìë™ì°¨ ì´ë¯¸ì§€ (0~1)
    beta_start=0.0001,
    beta_end=0.02,
    num_steps=num_steps,
    min_value=0.0,
    max_value=1.0
)

# ê²°ê³¼ ì‹œê°í™”
plt.figure(figsize=(15,5))
for i in range(10):
    plt.subplot(2,5,i+1)
    plt.imshow(noisy_images[i*10][0,...,0], cmap='gray')  # ì²« ë²ˆì§¸ ì´ë¯¸ì§€ ê¸°ì¤€ìœ¼ë¡œ í™•ì¸
    plt.title(f"Step {i*10}")
    plt.axis('off')
plt.suptitle("Forward Diffusion (Adding Noise on CIFAR10 car)", fontsize=14)
plt.tight_layout()
plt.show()
```

### 4-3. ì—­ë°©í–¥ í™•ì‚°(Reverse Diffusion) êµ¬í˜„

```python
# U-Net ëª¨ë¸ ì •ì˜
def build_unet(input_shape):
    inputs = Input(input_shape)

    # ì¸ì½”ë”
    c1 = Conv2D(64, (3,3), activation='relu', padding='same')(inputs)
    c1 = Conv2D(64, (3,3), activation='relu', padding='same')(c1)
    p1 = MaxPooling2D((2,2))(c1)

    c2 = Conv2D(128, (3,3), activation='relu', padding='same')(p1)
    c2 = Conv2D(128, (3,3), activation='relu', padding='same')(c2)
    p2 = MaxPooling2D((2,2))(c2)

    # ë””ì½”ë”
    u1 = Conv2DTranspose(128, (2,2), strides=(2,2), padding='same')(p2)
    u1 = Concatenate()([u1, c2])
    c3 = Conv2D(128, (3,3), activation='relu', padding='same')(u1)
    c3 = Conv2D(128, (3,3), activation='relu', padding='same')(c3)

    u2 = Conv2DTranspose(64, (2,2), strides=(2,2), padding='same')(c3)
    u2 = Concatenate()([u2, c1])
    c4 = Conv2D(64, (3,3), activation='relu', padding='same')(u2)
    c4 = Conv2D(64, (3,3), activation='relu', padding='same')(c4)

    outputs = Conv2D(1, (1,1), activation='linear')(c4)  # ë…¸ì´ì¦ˆ ì˜ˆì¸¡ìš©
    return Model(inputs, outputs)
```

```python
input_shape = (64,64,1)
model = build_unet(input_shape)
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001),   #í•™ìŠµë¥ 
              loss='mean_squared_error')
model.summary()
```

```python
# ì…ë ¥: ë…¸ì´ì¦ˆê°€ ì„ì¸ ì´ë¯¸ì§€ / íƒ€ê²Ÿ: ì‹¤ì œ ë…¸ì´ì¦ˆ
noisy_images_tensor = noisy_images.reshape(-1,64,64,1).astype(np.float32)
noise_tensor = noise_list.reshape(-1,64,64,1).astype(np.float32)
```

```python
num_epochs = 50
model.fit(
    noisy_images_tensor,
    noise_tensor,
    epochs=num_epochs,
    batch_size=64,
    verbose=0,
    callbacks=[TqdmCallback(verbose=1)]
)
```

```python
# ì—­ë°©í–¥ í™•ì‚°(Reverse Diffusion)
# DDPM ê³µì‹ì„ ì´ìš©í•´ ëª¨ë¸ì´ ì˜ˆì¸¡í•œ ë…¸ì´ì¦ˆë¥¼ ì œê±°í•˜ë©° ì›ë³¸ìœ¼ë¡œ ë³µì›í•©ë‹ˆë‹¤.

def reverse_diffusion_ddpm(model, num_steps, betas, alphas, alpha_bars, shape):
    x_t = np.random.randn(*shape)  # ì™„ì „í•œ ë…¸ì´ì¦ˆì—ì„œ ì‹œì‘
    denoised_steps = [x_t]

    for t in reversed(range(num_steps)):
        predicted_noise = model.predict(x_t[np.newaxis, ...])[0]     #ì˜ˆì¸¡í•˜ê¸°!
        if t > 0:
            noise = np.random.randn(*x_t.shape)
        else:
            noise = 0

        # DDPM ì—­í™•ì‚° ê³µì‹
        x_t = (
            1 / np.sqrt(alphas[t])
            * (x_t - ((1 - alphas[t]) / np.sqrt(1 - alpha_bars[t])) * predicted_noise)
            + np.sqrt(betas[t]) * noise
        )
        x_t = np.clip(x_t, 0, 1)
        denoised_steps.append(x_t)

    return denoised_steps
```

```python
# DDPM ì—­í™•ì‚° ìˆ˜í–‰
num_steps = 100
reverse_steps = reverse_diffusion_ddpm(
    model, num_steps, betas, alphas, alpha_bars, (64,64,1)
)
```
