# 0. 1차시 간단 복습!
* 시계열 데이터 : 순차적인 시간의 흐름에 따라 수집된 데이터로 관측치 사이의 상관관계가 존재, 순서가 중요한 시쿼스 데이터
* 단변량 시계열은 단일 변수의 연속적인 시간, 다변량 시계열은 둘 이상 변수의 같은 시간
* 시계열 데이터의 특징
    1. 시간 의존성
    2. 계절성 & 추세
    3. 자기 상관성
* 시계열 데이터의 분석 요인
    1. 추세 요인
    2. 순환 요인
    3. 계절 요인
    4. 불규칙 요인
* 시계열 분성의 기본 프로세스
    1. 데이터 전처리 : 보간, 이상치 처리, 요소분해
    2. 정상성 판단 : 시간과 상관없이 일정한가!
        * 단위근 검정 수행으로 정상/비정상 판단
        * 비정상의 경우 추세 있다면 차분, 계절성 있다면 계절 차분, 분산 변환 있다면 로그변환 및 박스콕스 변환
    3. 모델링 및 예측

# 1. 단변량 시계열 분석

## 1.1 안정적 시계열 분석

### 1.1.1 자기 회귀 모델(AR)
* 과거의 자기 자신 데이터가 미래 값을 설명해줄 것이라고 가정
* 자기 회귀 : 과거 자신의 값을 독립변수로 현재 값을 예측 -> 자기 상관성 이용
* p차 자기 회귀 모형 -> 과거 p개 시점 데이터 선형 조합을 통해 예측
* 시점은 특성에 따라 조정 가능!

### 1.1.2 이동 평균 모델(MA)
* 과거의 잔차를 독립변수 삼아 현재 값을 예측하는 모델 -> 과거의 잘못 예측한 것을 현재 시점에 반영
    * 잔차 : 실제 값과 예측 값의 차이
* q차 이동 평균 모형 -> 과거 q개의 예측 오차의 선형 결합으로 예측하는 모델
* 현재 값은 q시점 이전의 오차의 영향을 받음

### 1.1.3 AR, MA 모델의 전제 조건
1. 정상성
    * 두 모형 모두 안정적 시계열 분석이니 확률적 안정성이 필요, 즉 정상성을 만족해야함
    * AR 모형의 정상성 : 특성 방정식애서 모든 해의 절대값이 1보다 크다
    * MA 모형의 정상성 : 잔차가 이미 정상성을 만족하므로 MA도 항상!
    * 단위근 검정? -> 이건 데이터의 정상성! 위는 모델의 정상성!
2. 가역성
    * MA 모형을 AR로 돌릴 수 있는가?
    * MA는 과거를 오차항으로 -> 측 이를 역추적하면 오차항이 과거 관측값으로 표현될 것!
    * AR 조건의 가역? 은 필요없다
    * 조건 : 특성 방정식의 모든 해의 절댓값이 1보다 크면 가역성 확보!

### 1.1.4 ARMA 모델
* AR 모델 + MA 모델
* 과거 p개의 관측값과 과거 q개의 오차를 활용
* 정상성과 가역성 모두 만족해야함

### 1.1.5 차수 결정 방법 : 자기 상관 함수
* 두 그래프를 통해 현재와 과거 관계성 파악하고 p,q를 결정
* 자기 상관 계수 : ARMA 모형에서 사용할 q개의 오차 결정하는 방법 -> 두 시점 사이 상관 관계 파악
* 부분 자기 상관 계수 : p개의 시점 결정 방법 -> 두 시점 사이 직접적 영향만 파악
* ACF 그래프와 PACF 그래프로 차수 및 모형을 결정
    * PACF 그래프의 진동 감소를 통해 p 결정
    * ACF 그래프의 진동 감소를 통해 q 결정
    * 둘 다 만족하는 차수로 ARMA의 p,q 결정

## 1.2 불안정 시계열
* 안정적 시계열 모델에 정상성 확보 과정이 추가

### 1.2.1 ARIMA
* AR 모형의 차수와 MA 모형의 차수에, 차분 회수 d를 추가한 형태
* 만약 p,q가 없고 상수 없으면? 이전 사실이 영원히 영향을, 무방향성
* p,q 없고 상수항까지 있으면? 시간이 흐를수록 상수항이 누적되어 방향성이 생긴다

### 1.2.2 SARIMA
* 시계열 데이터에 계절성을 처리할 수 있는 모델
* ARIMA 모델에 계절성까지!
* 하이퍼파라미터가 비계절 pdq 파라미터와 계절 PDQ 파라미터로 나누어지고, 계절성 주기 길이 m 추가
* 호빵 예시로 보면, 지금을 12월이라고 생각하면 이 날짜 기준 주기로 비계절 ARIMA (1달) 등 진행하고, 이후 계절성 포함한 작년 12월의 계절 ARIMA 도 진행하는 것

* 계절성 데이터에서의 ACF와 PACR
    * 계절 주기 m에서 ACF가 급등 -> 비계절성 추세 파악이 어려움
    * 따라서 계절성 차분을 먼저 수행해야함

* 불안정 시계열에서 하이퍼파라미터 결정 방법
    1. D, d 결정
        * 단위근 검정으로 정상성 확부 여부 판단후 결정
        * 0 아니면 1 대부분
    2. q,p 결정
        * ACR와 PACF로 범위를 추정
        * AIC, BIC와 같은 정보 기준으로 그리드 서치 후 최종 선택

# 2. 다변량 시계열 분석

## 2.1 VAR 
* 자신의 과거 값과 다른 변수들의 과거 값을 함께 사용하자!
* GDP를 예측할 때 소비 과거 데이터도 이용하는 느낌
* 변수간 의존성이 중요할 때 적합
* VAR 모델을 활용한 분석 방법
    1. 충격 반응 함수 : 한 시계열에 특정 시점 충격이 생기면, 다른 시계열에 시간에 따라 어떤 영향을 주는가?
    2. 예측오차 분산 분해 : 각 변수가 미래의 움직임을 예측하는 데에 어느 정도 기여를 하는가?

* 장점 : 양뱡향 영향 분석 가능, 여러 변수 함께 예측, 동시성 문제 없음
* 단점 : 파라미터 폭발 가능, 관계 정보 손실 가능, 상관관계는 인과관계가 아니다.

# 3 딥러닝을 이용한 시계열 분석
## 3.0 왜 나왓나요?
* 처리 속도 느리고, 크기 고정데이터만 다뤄야하고, 비선형 패턴 포착이 어려웠음!

## 3.1 RNN 계열 
1. RNN : 순환 구조 신경망
    * 짧거나 중간 길이의 순차적 의존성 강한 자기상관 큰 데이터에 적합
2. LSTM : 장기기억 존재
    * 장기 의존성이 중요하거나 과거 이벤트가 먼 매래에 영향을 주는 경우
3. GRU : LSTM 경량화
    * 장기의존성과 학습 속도 모두 잡고 싶고, 데이터가 상대적으로 적을 때

## 3.2 Transformer 계열
* 어텐션 매커니즘으로 모든 시점을 동시에 참조!
* 시계열이 길고, 전역 패턴이 중요하고, 대규모 데이터, 다변량 시계열에 적합!
* 짧으면 오버피팅 가능성
* 상세한 모델이 많이 존재 : Vanilla, Informer, Autoformer, TFT 

## 3.3 Foundation Model
* 거대한 데이터로 사전 학습해서 다양한 문제를 해결할 수 있는 범용 모델
* 튼튼한 기초 공사?
* 특정 문제 전용이 아닌, 범용적인 모델!
1. LLM 기반 모델
    * 대규모 언어 모델의 패턴 인식 능력을 시계열에 활용하는 접근법
    * LLM이 이미 시계열 지식이 있다고 생각하고 프롬프트 엔지니어링 집중
    * LLM을 시계열로 Fine - tuning
    * 시계열을 텍스트로 변환 후 LLM 이용
2. 자체 거대 모델
    * 대규모 시계열 데이터로 사전 학습, 다양한 예측 문제에 범용적으로 적용 가능
