# 1. 단변량 시계열 분석

## 1.1 안정적 시계열 분석


## 1 자기 회귀 모델 (Autoregressive Model, AR)
**과거의 자기 자신 데이터가 미래 값을 설명한다**는 가정이 바로 **자기 회귀(AR)** 모델

###  자기 회귀란?

- **과거 자신의 값(yₜ₋₁, yₜ₋₂ …)**을 독립변수로 삼아 현재 값(yₜ)을 예측하는 회귀
- 시계열 데이터의 **자기상관성(autocorrelation)** 활용  
  → 이전 값이 이후 값에 영향을 미침

### AR 모델 식

$$
AR(p):  y_{t}=c +\phi _1y_{t-1}+\phi_2y_{t-2}+...+\phi _py_{t-p}+\varepsilon _t
$$

- **AR(p)** = 과거 *p개 시점*의 데이터를 선형 조합하여 예측하는 모델

#### AR(1) 예시

$$
y_t = c + \phi_1 y_{t-1} + \varepsilon_t
$$

- 올해 크리스마스 기온(yₜ)은  
  **작년 크리스마스 기온(yₜ₋₁)**의 영향을 받는다

####  예시로 풀어보면

- yₜ: 올해 크리스마스 기온  
- yₜ₋₁: 작년 기온  
- ϕ₁: 작년 기온이 올해에 미치는 영향 정도

→ 재작년 기온까지 영향이 있다면 AR(2), AR(3) 가능!


### OLS 회귀 vs AR(p) 모델

####  OLS 회귀
$$
Y = \beta_0 + \beta_1 X_1 + \cdots + \varepsilon
$$

- 목적: **설명력 중심**
- 예측 가능하지만,  
  미래 예측 시 **예측값을 반복해 또 예측해야 해서 부정확**해짐  
  (연쇄적 예측 오류)

####  AR(p) 모델

- 태생이 **시계열 예측용 모델**
- “예측값을 다시 예측에 사용하는 구조”가 설계에 포함되어 있음
- 파라미터 추정: **MLE(최우도추정)**



## 2 이동 평균 모델 (Moving Average, MA)

MA 모델은 **과거의 오차(ε)**를 이용하여 현재 값을 예측하는 모델!


###  MA 모델 식

$$
MA(q): y_t = \mu +\varepsilon _t+ \theta _1\varepsilon_{t-1}+\theta_2\varepsilon_{t-2}+...+\theta_q\varepsilon_{t-q}
$$

- **MA(q)** = 과거 *q개의 예측 오차*로 현재를 설명

####  MA(1) 예시

$$
y_t = \mu + \varepsilon_t + \theta_1 \varepsilon_{t-1}
$$

- 오늘 주량(yₜ)은  
  **어제 밤샘 충격(εₜ₋₁)**의 영향을 받는다



## 3 AR, MA 모델의 전제 조건


### 1 정상성(Stationarity)

- **AR & MA 모델 모두 안정적 시계열 분석을 전제로 함**
- 즉, **정상성을 만족해야 함**


### AR의 정상성 조건

- 특성방정식:

$$
1 - \phi_1 z - \phi_2 z^2 - \cdots - \phi_p z^p = 0
$$

- **해(|z|)의 절댓값 > 1** → AR 모델 정상성 확보

####  자주 쓰는 AR(1), AR(2) 조건만 기억

- AR(1):  $|\phi| < 1$
- AR(2):  특성방정식의 두 근 모두 |λ| < 1

#### 📌 Lag-Operator(L)로 표현하면…

- Lag 연산자:  
  - \(Ly_t = y_{t-1}\)  
  - \(L^ky_t = y_{t-k}\)

AR(p):
$$
y_t = (\phi_1 L + \phi_2 L^2 + \cdots + \phi_p L^p) y_t + \varepsilon_t
$$

→ 이를 **z로 치환**한 것이 특성방정식



###  MA의 정상성 조건

- 잔차 ε는 이미 **백색잡음(white noise)**  
  → **MA는 항상 정상성 만족**
**데이터의 정상성**을 보는 것이고,  
**모델의 정상성(해의 조건)**은 **특성방정식**으로 판단!


### 2 가역성(Invertibility)

- **MA 모델을 AR(∞) 형태로 재표현 가능해야 함**

예: MA(1)

$$
x_t = (1 + \theta L)\varepsilon_t
$$

역변환:

$$
\varepsilon_t = (1 - \theta L - \theta^2 L^2 - \cdots)x_t
$$

→ MA 모델이 안정적으로 거꾸로 표현될 수 있어야 함  
→ 이를 **가역성 조건**이라 함


### 4 ARMA (Autoregressive Moving Average) 모형


$Y_t = \underbrace{\phi_1 Y_{t-1} + \phi_2 Y_{t-2} + \cdots + \phi_p Y_{t-p}}_{\text{AR(p): 자기회귀}} 
+ \underbrace{\varepsilon_t + \theta_1 \varepsilon_{t-1} + \theta_2 \varepsilon_{t-2} + \cdots + \theta_q \varepsilon_{t-q}}_{\text{MA(q): 이동평균}}$

- **AR(p) 모형 + MA(q) 모형을 결합한 모델**
    - 과거 **p개의 관측값** + 과거 **q개의 오차(white noise)** 활용
    - 조건:  
      **AR(p)의 정상성 + MA(q)의 가역성** → 둘 다 만족해야 함

#### ARMA 모형의 특수한 경우

- **ARMA(p, 0) = AR(p)**
- **ARMA(0, q) = MA(q)**
- **ARMA(0, 0) = 백색잡음 모델**  
  → $y_t = \varepsilon_t$


### 5 차수 결정 방법: 자기 상관 함수

두 그래프(ACF, PACF)를 통해 현재 값이 과거 값들과 어떤 관계가 있는지 파악하여 **AR, MA, ARMA의 적절한 차수(p, q)**를 결정

####  자기 상관 함수 (ACF, AutoCorrelation Function)

- **ARMA(p, q)에서 q(오차항 개수) 결정**
- 시점 간의 **직접 + 간접 영향 포함한 전체 상관 구조** 파악

$$
ACF(k) = Corr(y_t, y_{t-k})
$$

- k 시차(lag)에 대해 두 시점 간 상관계수 계산
- **간접적인 영향도 모두 포함**

예시 (k = 3):

- $y_t$ ←→ $y_{t-3}$ **직접** 영향  
- $y_{t-3} → y_{t-1} → y_t$  
- $y_{t-3} → y_{t-2} → y_t$  
→ **간접 경로도 모두 고려**

- 그래프:
  - x축 = lag k  
  - y축 = ACF(k)


#### 부분 자기 상관 함수 (PACF, Partial ACF)

- **ARMA(p, q)에서 p(시점 개수) 결정**
- 두 시점 사이의 **직접적 영향만 파악**


$e_t = y_t - (\beta_1 y_{t-1} + \cdots + \beta_{k-1} y_{t-(k-1)})$ 일 때,

$$
PACF(k) = Corr(e_t, e_{t-k})
$$

- 중간 시점들의 영향(예: $y_{t-1}, y_{t-2}$)은 **모두 제거**
- 순수한 직접 상관성만 계산

예시 (k = 3):

- $y_t$ ←→ $y_{t-3}$ 간 **직접 상관성만** 계산  
- 간접 경로  
  - $y_{t-3} → y_{t-1} → y_t$  
  - $y_{t-3} → y_{t-2} → y_t$  
  → **제거하고 계산**

- 그래프:
  - x축 = lag k  
  - y축 = PACF(k)



#### ACF & PACF로 모형 차수 결정하기

| 구분 | ACF 그래프 | PACF 그래프 |
| --- | --- | --- |
| **AR(p) (ARMA(p, 0))** | 천천히 감소 or 진동하며 감소 | **p+1 시점부터 0에 근접(절단)** |
| **MA(q) (ARMA(0, q))** | **q+1 시점부터 0에 근접(절단)** | 천천히 감소 or 진동 감소 |
| **ARMA(p, q)** | q+1 시점부터 0에 근접 | p+1 시점부터 0에 근접 |

##### 진동 감소란?

- 양수/음수를 오가며 진동  
- 진폭이 점점 줄어듦



## 1.2 불안정 시계열

###  불안정한 데이터의 시계열 분석 프로세스


1. 비정상 시계열 데이터 (추세 또는 계절성 존재)  
2. **정상성 확보** (차분 또는 계절성 차분)  
3. 정상 시계열  
4. ARMA 기반으로 **안정적 시계열 분석**

불안정 시계열 모델은  **안정적 시계열 모델(ARMA)에 정상성 확보 과정(차분, 계절성 차분)이 추가된 형태!**


### 1 ARIMA (Autoregressive Integrated Moving Average)

$$
ARMA(p, q) + \text{차분} \rightarrow ARIMA(p,d,q)
$$

#### ARIMA 모형의 특수한 경우

| 구분 | 표현 |
| --- | --- |
| 백색잡음 | ARIMA(0,0,0) |
| 자기회귀 | ARIMA(p,0,0) ⇔ AR(p) |
| 이동평균 | ARIMA(0,0,q) ⇔ MA(q) |
| 랜덤워크 | ARIMA(0,1,0) (상수 없음) |
| 표류 포함 랜덤워크 | ARIMA(0,1,0) (상수 있음) |


###  랜덤워크 (상수 없는 ARIMA(0,1,0))

**수식**

$$
Y_t = Y_{t-1} + \varepsilon_t
$$

- 이전 값 + 백색잡음 → **방향성 없음**
- 과거 충격이 영원히 누적됨
### 2 SARIMA (Seasonal ARIMA)


$$
ARIMA(p,d,q) + \text{계절성 차분} 
\rightarrow SARIMA(p,d,q)(P,D,Q)_m
$$

- **계절성이 존재하는 시계열 데이터**에 적합  
- ARIMA에 **계절 AR/MA 요소 + 계절 차분** 추가한 형태


###  SARIMA 하이퍼파라미터

- **비계절 파라미터**  
  - p: AR 차수  
  - d: 차분  
  - q: MA 차수  
- **계절 파라미터**  
  - P: 계절 AR 차수  
  - D: 계절 차분  
  - Q: 계절 MA 차수  
- **m: 계절 주기**  
  - 예: 12(월), 7(요일), 4(분기)


####  계절 파라미터 (P, D, Q)\_12

- D = 1  
  → 계절 차분 1회  
- P = 1  
  → 12개월 전 값(작년 판매량)이 현재에 영향
- Q = 1  
  → 작년 오차(\(\varepsilon_{t-12}\))가 현재에 영향

####  비계절 파라미터 (p, d, q)

- d = 1 → 1차 차분  
- p = 1 → 이전 달 판매량 영향  
- q = 1 → 이전 달 오차 영향

###  계절성 데이터에서의 ACF & PACF 특징


- 계절 주기 **m의 배수(예: 12, 24, …)**에서 ACF가 급등  
- 계절성 때문에 비계절성 패턴이 가려질 수 있음  
- **→ 계절 차분 후 비계절 차분 순서로 진행**


###  불안정 시계열에서의 하이퍼파라미터(p, q, d, D) 결정 방법


#### 1 d, D (차분·계절 차분) 결정

- **ADF(단위근 검정)** 수행  
  → 정상성 확보 여부를 기준으로 결정  
- 대부분 **0 또는 1에서 결정**
- 예: d = 1, D = 1, m = 11


#### 2 p, q (AR, MA 차수) 결정

1. **ACF/PACF 그래프에서 범위 추정**
   - 예: `SARIMA(2,1,2)(1,0,1,11)`
2. **AIC, BIC 기반 그리드 서치로 최종 선택**


AIC/BIC는  
모델의 **적합도 + 복잡도**를 함께 평가하는 지표  
→ 값이 낮을수록 좋은 모델




# 2. 다변량 시계열 분석

## 1 VAR (Vector Autoregression)
**자신의 과거 값 + 다른 변수들의 과거 값**을 함께 사용하여 미래를 예측하는 모델

## 1 수식으로 이해하기

###  단변량 AR(1)

→ X는 자기 자신의 과거 정보만 사용

###  다변량 VAR(1)

→ X와 Y가 서로의 과거를 함께 사용


## 2 예시로 이해하기

| AR 모형 | VAR 모형 |
|--------|----------|
| 2025년 GDP는 2024년 GDP로만 예측 | 2025년 GDP는 **2024년 GDP + 2024년 소비**로 예측 |
| **단방향 (자기 자신만 사용)** | **양방향 (변수끼리 상호 영향)** |


## 3 어떤 경우에 추천하나요?

###  적합한 경우
- GDP·소비·투자 등 경제 지표가 서로 영향을 줄 때  
- 금융 자산 가격 사이의 상호작용  
- 변수 간 **상호 의존성**이 중요한 문제

###  부적합한 경우
- 변수가 너무 많아 파라미터가 폭발하는 경우  
- 변수 간 관계가 명확하지 않은 경우  
- 장기 예측이 목표일 때

### 검정 절차

1. **모델 비교**
   - 제한 모형: Y의 과거값만 사용  
   - 비제한 모형: Y의 과거값 + X의 과거값 사용  

2. **귀무가설**
   - H₀: X는 Y 예측에 도움 안 됨  
   - H₁: X는 Y 예측에 도움 됨  

3. **결과 해석**
   - p-value < 0.05 → X는 VAR 포함 **O**  
   - p-value ≥ 0.05 → X 제외 고려 **X**



## VAR 모델을 활용한 분석 방법



## 1 충격 반응 함수 (IRF, Impulse Response Function)

- 특정 변수에 충격(shock)이 발생했을 때,  
  다른 변수가 **시간에 따라 어떻게 반응하는지** 분석
- “소득(Y)에 충격이 생기면 소비(C)는 며칠 뒤 어떤 방향으로 움직일까?”


## 2 예측오차 분산 분해 (Variance Decomposition)

- 특정 변수의 미래 예측 오차가  
  **각 변수의 충격 때문인지 비중을 분해**하는 방법

예:
- 소비(C) 예측오차가  
  - 얼마나 소득(Y) 충격 때문인지  
  - 얼마나 저축(S) 충격 때문인지 찾는 분석



## VAR 모델 장단점

### 장점
- 변수 간 **양방향 상호작용** 분석 가능  
- 도메인 지식 없어도 여러 변수 동시 예측 가능  
- 동시성 문제 없음(과거값만 사용)

### 단점
- 변수가 많으면 **파라미터 수 폭발**  
- 장기적 구조 정보 손실 가능  
- 상관 ≠ 인과 (Granger도 진짜 인과 아님)


# 3. 딥러닝을 활용한 시계열 분석

## 0 등장 배경

### 전통 통계 모델의 한계
1. 데이터가 많으면 처리 속도가 느려짐  
2. 크기가 고정된 데이터만 가능  
3. 비선형 패턴 포착이 어려움  

## 1 RNN 계열

### RNN (Recurrent Neural Network)

**개념:**  
이전 시점 정보를 현재 시점으로 전달하는 ‘순환 구조’ 신경망

####  적합한 경우
- 짧은~중간 길이 시계열 (10~100 시점)  
- 강한 순차적 의존성(주가, 센서)  
- 자기상관이 큰 데이터  

####  부적합한 경우
- 매우 긴 시계열 (장기 의존성 문제)  
- 병렬 처리 필요한 대규모 작업  



### LSTM (Long Short-Term Memory)

**개념:**  
게이트 구조로 **장기 기억 유지**가 가능한 RNN 개선 모델

#### ✔ 적합한 경우
- 장기 의존성이 중요한 시계열 (100~1000)  
- 먼 과거 사건이 미래에 영향을 주는 경우  
  - 예: 분기별 재무제표 → 연간 실적 예측  

###  GRU (Gated Recurrent Unit)

**개념:**  
LSTM을 단순화해 **더 빠르고, 비슷한 성능**을 내는 모델

#### 적합한 경우
- LSTM 수준의 장기 의존 필요 + 빠른 학습 필요  
- 데이터가 적어 과적합 위험이 큰 경우  
- 실시간 예측  


###  모델 비교

| 모델 | 장점 | 단점 | 사용 예시 |
|------|------|------|------------|
| **RNN** | 구조 단순, 빠름 | 장기 의존성 약함 | 짧은 주가 예측 |
| **LSTM** | 장기 의존성 강함 | 느림, 복잡 | 분기 재무제표 예측 |
| **GRU** | LSTM보다 빠름 | 장기 의존성은 LSTM보다 약함 | 실시간 센서 예측 |


## 2 Transformer 계열


###  Transformer

**개념:**  
어텐션 메커니즘으로 모든 시점을 동시에 보고 병렬 처리

####  적합한 경우
- 매우 긴 시계열 (1000+)  
- 전역 패턴이 중요한 경우  
- 대규모 데이터  
- 다변량 시계열  

####  부적합한 경우
- 짧은 시계열  
- 연산 자원 부족한 환경  

#### 사용 예시
- 전력 수요 예측  
- 다종 금융자산 예측  


## Transformer 계열 주요 모델

### Vanilla Transformer
- 기본 Transformer  
- 짧은~중간 길이 시계열에 적합  


###  Informer
- **ProbSparse Attention**으로 계산 효율화  
- 긴 시계열 예측(LSTF)에 강함  
- 예: 1년 전력 데이터 → 다음달 예측  


###  Autoformer
- **시계열 분해(추세+계절) + 자기상관 구조**  
- 주기성 강한 데이터에 적합  
- 예: 월별 계절 판매량  


###  Temporal Fusion Transformer (TFT)
- 외생 변수까지 함께 활용  
- 정적/동적 변수 모두 처리  
- 해석 가능  
- 예: 유통 예측(날씨/이벤트 적용)


###  Transformer 계열 비교

| 모델 | 언제 사용? | 예시 | 장점 | 단점 |
|------|------------|-------|--------|--------|
| **Vanilla Transformer** | 짧은 시계열 | 주가 30일 예측 | 구조 단순 | 긴 시계열 어려움 |
| **Informer** | 긴 시계열 | 전력 수요 | 매우 효율적 | 짧은 시계열 과적합 |
| **Autoformer** | 계절성 뚜렷 | 월 매출 | 주기 포착 우수 | 구조 복잡 |
| **TFT** | 외생 변수 많음 | 유통 수요 | 해석 가능 | 학습 오래 걸림 |



## 3 Foundation Model


###  Foundation Model이란?

> 대규모 데이터로 사전 학습되어 다양한 과제를 해결할 수 있는 ‘범용 모델’

**특징**
- 특정 문제 전용 아님  
- 사전학습 후 Fine-tuning으로 특화  
- 일반적 → 특정 문제로 빠르게 적응  


###  LLM 기반 접근

#### **Without Adaptation**
- LLM이 이미 시계열 패턴을 익혔다고 가정  
- 프롬프트만 잘 설계  

#### **Adapt LLM**
- LLM 일부 레이어를 시계열로 다시 학습(Fine-tuning)

#### **Adapt to LLM**
- 시계열 → 텍스트 변환  
- LLM이 텍스트 패턴으로 이해하게 변환  


### 시계열 전용 대규모 모델
- 처음부터 시계열 전용으로 대규모 학습  
- 예: **TimeGPT**, **TimesFM**, **Moirai**



## 4 딥러닝 모델 선택 가이드


### 핵심 트레이드오프
- **단순함 ↔ 성능**: RNN < LSTM/GRU < Transformer  
- **해석 가능 ↔ 예측 정확도**  
- **학습 속도 ↔ 정확도**  
- **데이터 크기**:  
  - 적으면 RNN/통계모델  
  - 많으면 Transformer  


### 상황별 추천 모델

| 상황 | 추천 모델 | 이유 |
|------|-----------|------|
| 짧은 시계열 (<100) | 통계 모델, RNN | 단순 모델이 과적합 적음 |
| 중간 길이 (100~1000) | LSTM, GRU | 장기 정보 유지 |
| 매우 긴 시계열 | Informer, Autoformer | 긴 시퀀스 효율적 |
| 장기 의존성 중요 | LSTM, Transformer | GRU보다 강함 |
| 명확한 주기성 | Autoformer | 분해 + 자기상관 구조 |
| 외생 변수 많음 | TFT | 해석 가능하며 변수 융합 |
| 대규모 데이터 | Transformer 계열 | 병렬 처리 최적 |
| 범용 예측 | Foundation Model | 사전학습된 지식 활용 |
