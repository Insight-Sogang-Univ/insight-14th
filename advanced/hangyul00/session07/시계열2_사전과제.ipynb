# 0. 1차시 복습

시계열 데이터: 순차적 시간흐름에 따라 수집된 데이터로 연속한 관측치는 서로 상관관계가 존재하며, 순서가 중요한 "시퀀스 데이터"로 간주된다.

단변량 시계열: 시간에 따라 연속적으로 변화하는 '단일 변수'로 이루어진 시계열 데이터

다변량 시계열: 같은 시간에 기록된 '둘 이상의 변수'로 이루어진 시계열 데이터

시계열 데이터의 특성
1) 시간 의존성
2) 계절성과 추세
3) 자기 상관성

시계열 데이터의 분석 요인
1) 추세 요인 데이터 전반에 걸쳐 시간에 따라 지속적으로 증가/감소하는 경향
2) 순환 요인: 계절성보다 더 길고 불규칙한 주기
3) 계절 요인: 일정 주기로 반복되는 패턴
4) 불규칙 요인

시계열 분석의 기본 프로세스
1) 데이터 전처리: 결측치 보간, 이상치 제거, 요소분해
2) 정상성 판단: 시간에 상관없이 일정한 성질을 나타내는 것으로, '단위근 검정'(ADF, KPSS 등)을 통해 판단하고, 차분, 로그변환을 통해 정상성을 확보한다.
3) 모델링 및 예측

# 1. 단변량 시계열 분석

## 1.1 안정적 시계열 분석

### 자기 회귀 모델
자기 회귀 모델: 과거의 자기 자신 데이터가 미래 값을 설명해줄 것이라고 가정하는 것으로, 시계열 데이터의 자기상관성을 이용한다. 따라서, y의 현재 값은 1 시점 이전의 y값의 영향을 받는다.

### 이동 평균 모델
이동 평균 모델: 과거 시점의 잔차를 독립변수로 삼아 현재 값을 예측하는 모델로 과거의 잘못 예측한 것(오차)을 현재 시점에 반영한다.

### AR, MA 모델의 전제 조건
1) 정상성: AR과 MA 모델은 안정적 시계열 분석으로 확률적 안정성을 전제로 추정하기에 '정상성'을 만족해야 한다. 그리고, 모든 해의 절댓값이 1보다 큰 것으로 AR의 정상성을 확보할 수 있다.
2) 가역성: MA 모형을 AR 형태로 재표현할 수 있는지 여부로서 모든 해의 절댓값이 1보다 커야 MA의 가역성을 확보할 수 있다.

### ARMA (Autoregressive Moving Average) 모형
ARMA 모형: AR모형과 MA모형을 결합한 모형으로 즉, 과거 p개의 관측값 + 과거 q개의 오차를 활용해서 예측하는 방법이고, AR 모형의 정상성과 MA 모형의 가역성을 모두 만족해야 한다.

### 지수 결정 방법: 자기 상관 함수
1) 자기 상관 계수: ARMA 모형에서 사용할 q개의 오차 결정 방법으로, 두 시점 사이의 상관 관계를 파악한다.
2) 부분 자기 상관 계수: ARMA 모형에서 사용할 p개의 시점 결정 방법으로, 두 시점 사이의 직접적인 영향만 파악한다.

## 1.2 불안정 시계열
불안정 데이터의 시계열 분석 프로세스
1) 비정상 시계열 데이터
2) 정상성 확보
3) 정상 시계열
4) ARMA를 이용해 안정적 시계열 분석

### ARIMA (Autoregressive Integrated Moving Average)
ARIMA는 차분, 변환을 통해 AR, MA, ARMA로 정상화하는 방법이다. 즉, ARMA(p,q) + 차분 -> ARIMA(p, d, q)이다.

ARIMA 모형의 특수한 경우
1) 백색잡음: ARIMA(0,0,0)
2) 자기회귀: ARIMA(p,0,0)
3) 이동평균: ARIMA(0,0,q)
4) 랜덤워크: 상수가 없는 ARIMA(0,1,0)
5) 표류를 포함하는 랜덤워크: 상수가 없는 ARIMA(0,1,0)

### SARIMA (Seasonal ARIMA)
SARIMA: 시계열 데이터의 계절성을 처리할 수 있는 모델로 ARIMA 모델에 '계절성 변동을 반영'한 모델이다.

ARIMA(p,d,q) + 계절성 차분 -> SARIMA(p,d,q)(P,D,Q)m이다.

하이퍼파라미터?
1) p: AR의 차수
2) d: 차분의 정도
3) q: MA의 차수
4) P: 계절성 AR의 차수
5) D: 계절성 차분의 정도
6) Q: 계절성 MA의 차수
7) m: 계절성 주기의 길이

예시 - SARIMA(1,1,1)(1,1,1)12 일때?
1) 계절성 추세 제거를 위해 계절 차분 1회 수행
2) 1년 전의 판매량이 올해 판매량에 영향을 준다.
3) 1년 전의 예측 오차가 올해 판매량에 영향을 준다.

4) 추세 제거 위해 1차 바누 진행
5) 이전 달의 판매량이 올해 판매량에 영향을 준다.
6) 이전 달의 예측 오차가 올해 판매량에 영향을 준다.

불안정 시계열에서의 하이퍼파라미터 결정 방법?
1) d, D(차분/계절차분) 결정: ADF로 정상성 확보 여부 판단 후 결정하며, 데이터가 정상성을 갖출 때 결정하고, 차분 정도는 대부분 0 or 1에서 결정된다.
2) p, q(AR, MA 차수) 결정: 먼저 ACF, PACF로 범위를 추정한 뒤, AIC와 BIC 같은 정보 기준으로 그리드 서치 후 최종 선택한다.



# 2. 다변량 시계열 분석

### VAR (Vecto Autoregression)

VAR: 자신의 과거 값과 다른 변수들의 과거 값을 함께 사용하여 미래를 예측하는 과정으로, '단변량 AR'의 경우 X는 자기 자신의 과거만 사용하고, '다변량 VAR'의 경우 X와 Y가 서로의 과거를 함께 사용한다.

-> 즉, 여러 지표가 서로 영향을 주고받는 경우, 변수 간 상호 의존성이 중요한 경우에 사용할 수 있다.

VAR 모델을 활용한 분석 방법?
1) 충격 반응 함수: 한 시계열에 특정 시점에서 충격이 발생할 때 다른 시계열에 시간에 따라 어떤 영향을 주는지 분석하는 함수로서, 'A 변수에 외생적 충격이 생기면, B 변수는 며칠 뒤에 얼마나 어떤 영향으로 반응하는가?'를 확인한다.

2) 예측오차 분산 분해: 어떤 시계열이 상대적으로 어떤 영향을 끼치고 있는지 중요도를 산출하기 위해 수행하며 각 변수가 미래 움직임을 예측하는 데에 어느 정도 기여하는지 평가하는 방법이다.

VAR모델의 장단점
1) 장점
- 변수 간 양방향 영향 분석 가능
- 도메인 지식 없어도 여러 변수를 함께 예측
- 과거값만 사용하므로 동시성 문제 없음
2) 단점
- 변수가 많으면 파라미터 폭발
- 장기적 관계 정보 손실 가능
- 상관관계는 인과관계와 같은 것이 아님


# 딥러닝을 활용한 시계열 분석

### 등장 배경
데이터가 많으면 처리 속도가 느려지고, 크기가 고정된 데이터만 가능하며 비선형 패턴 포착이 어려운 전통 통계 모델의 한계를 극복하고자 나오게 되었다.

### RNN 계열
1) RNN: 이전 시점의 정보를 현재로 전달하는 순환 구조의 기본 신경망으로, 짧은-중간 길이의 시계열, 순차적 의존성이 강한 데이터, 자기상관이 큰 데이터에 적합하다.

2) LSTM: 게이트 메커니즘으로 장기 기억을 유지할 수 있도록 개선한 RNN으로 장기 의존성이 중요한 경우, 중요한 과거 이벤트가 먼 미래에 영향을 주는 경우에 적합하다.

3) GRU: LSTM을 단순화하여 더 빠르게 학습하면서 비슷한 성능을 내는 모델로, 장기 의존성이 필요하지만 학습 속도가 중요한 경우, 데이터가 상대적으로 적을 때, 실시간 처리가 필요할 때 적합하다.

### Transformer 계열
1) Transformer: 어텐션 메커니즘으로 모든 시점을 동시에 참조하여 병렬 처리하는 모델로 매우 긴 시계열, 전역 패턴이 중요한 경우, 대규모 데이터, 다변량 시계열일 때 적합하다.

### Foundation 모델
Foundation Model: 거대한 데이터로 사전 학습부터 다양한 문제를 해결할 수 있는 범용모델로, 특정 문제 전용이 아니며, 범용적이고, 사전 학습 후 fine-tuning으로 특화한다.

1) LLM 기반 모델: 대규모 언어 모델의 패턴 인식 능력을 시계열에 활용하는 접근법으로, LLM이 이미 시계열 지식을 가지고 있다고 가정하여 LLM을 시계열 데이터로 fine-tuning한 후에 시계열을 텍스트로 변환 후 LLM을 활용한다.

2) 자체 거대 모델: 대규모 시계열 데이터로 사전 학습하여 다양한 예측 문제에 범용적으로 적용 가능한 모델로, 시계열 전용으로 처음부터 학습하며 대규모 시계열 데이터 수집이 필수다.
