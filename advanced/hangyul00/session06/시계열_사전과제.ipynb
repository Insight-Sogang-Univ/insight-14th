# 시계열1

## 시계열 데이터란?

### 시계열 데이터의 정의
1) 시계열 데이터: 순차적 시간의 흐름에 따라 수집된 데이터 (시퀀스 데이터)
2) 연속된 관측치는 서로 상관관계가 존재한다.
3) 종단면 데이터: 동일한 개체에 대해 특정 시간 간격으로 수집된 데이터로 시간 의존성, 개체 또는 단위 추적, 패턴 및 동향 분석, 효과 및 원인 분석에 이용된다는 특징이 있다.
4) 횡단면 데이터: 몇 가지 주제에 대해 특정 시점에 수집된 데이터로 시간 의존성이 없고, 여러 변수의 동시 측정 결과로 구성되며, 비교 분석에 활용되며, 다른 변수 간의 상관관계를 분석하는 데에 사용된다.
5) 단변량 시계열: 시간에 따라 연속적으로 측정된 단일 변수로 이루어진 시계열 데이터
6) 다변량 시계열: 같은 시간에 기록된 둘 이상의 변수로 이루어진 시계열 데이터

### 시계열 분석과 목적
1) 미래 데이터 예측하기
2) 시계열 분석은 시계열 데이터를 분석하고 모델링하는 통계적 방법과 기법
3) 데이터 구조를 파악하고, 예측하며 이상치 탐지를 포괄하는 개념

### 시계열 데이터의 특징
1. 시간의존성: 시계열 데이터는 시간에 따라 변화한다.
2. 계절성과 추세: 시간에 따라 장기적으로 증가 혹은 감소하는 추세 + 일정 주기로 반복되는 패턴이 있다.
3. 자기상관성: 자기자신끼리 상관관계를 갖는다. (자기상관성 때문에 회귀분석은 불가능하며 예측의 신뢰도를 낮추고, 더 복잡한 시계열 모델을 요구하게 만든다.)


### 시계열 데이터의 분석 요인
1. 추세 요인: 데이터 전반에 걸쳐 시간에 따라 지속적으로 증가 혹은 감소하는 경향
2. 계절 요인: 일정한 주기로 반복되는 패턴
3. 순환 요인: 계절성보다 더 길고 불규칙한 주기로 반복되는 변동
4. 불규칙 요인: 추세, 계절성, 순환성을 모두 제거한 뒤 남은 예측 불가능한 변동

### 시계열 분석의 work flow
1. 데이터 전처리: 결측치 보간, 이상치 제거, 요소분해
2. 정상성 판단
3. 모델링 및 예측

## 2. 시계열 데이터 전처리

### 이상치/결측치 처리
시계열 데이터의 결측치: 결측치가 존재하면 추세나 계절성이 왜곡되므로 적절한 대처와 보간이 필요하다.

시계열 데이터의 이상치: 단순히 이상한 값이 아니라 신호의 중요한 의미일 수도 있다.

-> 전처리 과정
1) 데이터 확인/시각화
2) 결측치 확인 -> 결측치 처리
3) 이상치 탐지 -> 이상치 처리
4) 요소 분해

1. 결측값 처리

결측치 처리 방법
1) 삭제
2) 단순 대치
3) 보간
4) 머신러닝 기법

2. 이상치 탐지 및 처리

이상치 탐지 방법
1) IQR
2) Z-score
3) STL 분해(계절성 + 추세 + 잔차로 나누는 것)

이상치 탐지 모델
1) 분류 및 회귀 트리
2) 클러스터링 기반 이상 탐지
3) 오토 인코더

이상치 처리 방법
1) 삭제
2) 단순 대치
3) 보간

* 평활화 (smoothing)

평활화: 시계열 데이터에서 노이즈를 줄이고, 전체적인 흐름이나 추세를 더 명확하게 파악하기 위해 데이터를 부드럽게 만드는 기술로 '이동 평균'을 사용한다.

이동평균의 유형
1. 단순 이동평균: 기간 내 모든 평균으로 변동이 심하지 않고, 비교적 안정적인 시계열에서 사용
2. 가중 이동평균: 최근 값에 더 큰 가중치를 부여하며 최근 데이터가 더 중요한 상황에서 사용
3. 지수 이동평균: 최근 데이터에 지수적으로 더 많은 가중치를 부여하며 최근 값에 매우 빠르게 반응해야 하는 시계열에 사용

### 요소분해
요소분해: 주어진 시계열 데이터를 개별 성분으로 분해하는 과정으로 추세, 계절성, 순환, 불규칙성분으로 나눈다.

요소분해를 하는 이유: 시계열 데이터는 추세, 계절성, 순환성, 불규칙성 등의 다양한 변동이 결합된 형태로 각 요소를 분리하면 각 패턴을 명확히 이해할 수 있어 예측, 이상 탐지 등의 분석 작업을 정교하게 수행할 수 있다.

1. 가법모형

시계열 = 추세 + 계절성 + 불규칙성

시계열 관측값은 변동요인들의 가법으로 설명될 수 있다고 가정하며, 각 변동 요인들이 독립적으로 더해지며 모든 요소의 영향이 일정하다.

2. 승법모형

시계열 = 추세x계절성x불규칙성

시계열 관측값은 변동요인들의 승법으로 설명될 수 있다고 가정하며, 추세가 커질수록 계절성이나 불규칙성의 효과도 비례해서 커진다.



## 3. 정상성 판단
정상성: 시간에 상관없이 시계열 데이터가 평균, 분산 등의 통계적 특성이 일정한 성질을 띠는 것

강한 정상성

강한 정상성: 모든 적률(평균, 분산, 왜도, 첨도)이 시간과 무관하게 일정하지만, 현실에서는 잡음과 외부 요인의 영향으로 찾아보기 매우 어렵다.

약한 정상성

약한 정상성: 평균은 시점에 대해 일정하고, 분산은 시점에 대해 일정하며, 공분산은 시점에 의존하지 않고, 단지 시차에만 의존한다는 세 가지 가정을 모두 만족하면 약한 정상성을 가진다고 판단할 수 있다.

정상성 검증을 위한 단위근 검정의 대표 방법

1. ADF 검정: 단위근의 존재 여부를 통해 시계열의 정상성을 판단한다. 귀무가설이 '비정상 시계열'임을 가정하고, 대립가설이 '정상 시계열'임을 가정한다.
2. KPSS 검정: 단위근 기반의 검정이지만, 귀무가설이 ADF 검정과 반대이며, 귀무가설이 정상 시계열'임을 가정하고, 대립가설이 '비정상 시계열'임을 가정한다. 특히, KPSS는 "평균 주위"에서 정상성을 보는 데 중점을 둔다.


정상성 확보를 위한 여러가지 방법들

1. 추세 제거: 시계열이 지속적으로 증가하거나 감소할 때 사용한다. 이때, 시계열 데이터에서 한 시점의 값과 그 이전 시점의 값을 차분하여 변화량을 계산하여 비정상적 변동을 제거하거나 시계열 데이터의 정상성을 확보한다.

2. 계절성 제거: 시계열이 주기적인 패턴을 가질 때 적용하며, t시점의 y값에서 t-m시점의 y값을 빼는 것을 통해서 수행한다.

3. 분산 안정화: 시계열의 변동 폭이 점점 커지거나 작아질 때 적용하며 '로그 변환'을 통해 분산의 변화폭을 감소시키거나, 'Box-Cox 변환'을 통해서 분산을 안정화한다.
