# 1. 시계열 데이터란?

- 순차적 시간의 흐름에 따라 수집된 데이터
- 연속된 관측치 간 상관관계 존재
- 순서가 중요한 "시퀀스 데이터"

### 종단면 데이터와 횡단면 데이터

### 단변량 시계열과 다변량 시계열

단변량 시계열
- 시간에 따라 연속적으로 측정된 단일 변수로 이루어진 시계열 데이터
- 시간의 흐름에 따른 패턴과 트렌드 이해



다변량 시계열
- 같은 시간에 기록된 둘 이상의 변수로 이루어진 시계열 데이터
- 변수 간 상호작용 및 영향력 파악

## 시계열 분석과 그 목적


- 시계열 분석의 주 목표: "미래 데이터 예측"
- 시계열 분석: 시계열 데이터를 분석하고 모델링하는 통계적 방법과 기법
- 데이터 구조 파악, 예측, 이상치 탐지 포함

## 시계열 데이터의 특징

1. 시간의존성
    - 시계열 데이터는 시간에 따라 변화함
    - 순서가 달라지면, 의미가 달라짐
2. 계절성 & 추세
    - 계절성: 일정 주기로 반복되는 패턴
    - 추세: 시간의 흐름에 따라 나타나는 장기적인 증가/감소 경향
3. 자기상관성
    - 이전 시점 값이 이후 시점 값에 영향을 미침
    - 관측값들이 서로 독립적이지 않음

## 자기상관성의 한계

1. 통계적 추론의 제약 및 오류
    - 오차항 간 상관관계가 있음\
    -> 회귀분석의 통계적 추론 (검정, 신뢰구간 등)이 왜곡됨
2. 예측의 어려움 & 모델 복잡성 증가
    - 잔차가 시간에 따라 패턴을 가져 예측이 왜곡됨
    - 예측 오차가 누적되면 신뢰도 하락

## 시계열 데이터 분석 요인

1. 추세 요인
    - 데이터 전반에 걸친 지속적 감소/증가 요인
2. 계절 요인
    - 일정한 주기로 반복되는 패턴
3. 순환 요인
    - 계절성보다 더 길고 불규칙한 주기로 반복되는 변동
4. 불규칙 요인
    - 추세, 계절성, 순환성을 모두 제거한 뒤 남은 예측 불가능 변동

## 시계열 분석의 WorkFlow
1. 데이터 전처리
2. 정상성 판단
3. 모델링 및 예측

# 2. 시계열 데이터 전처리

1. 데이터 확인/시각화
2. 결측치 확인 및 처리
3. 이상치 탐지 및 처리
4. 요소분해

## 시계열 데이터의 결측치
- 특정 시점에서 값이 측정되지 않았거나 누락된 상태
- 추세 계절성의 왜곡 유발\
=> 데이터의 특성이 왜곡되지 않도록 전처리 필요!

### 결측치 처리 방법
- 삭제/단순 대치/보간/머신러닝 기법 등 다양한 방법 존재

#### 보간법
- 주어진 데이터 포인트 값을 사용해 두 개 이상의 데이터 포인트 사이에서 누락된 값을 추정하는 수학적 기법
- 사라진 값을 합리적으로 메워주는 과정
- 선형 보간법/다항식 보간법/스플라인 보간법/조각별 상수 보간법

## 시계열 데이터의 이상치
- 값이 측정되었으나, 측정된 값이 일반적이지 않은 경우
- 단순히 이상한 값이 아닌 신호의 중요한 의미일 가능성이 있으므로 신중한 접근 필요

#### 이상치 탐지 방법
- IQR Rule-based Anomaly Detection
- Z-score
- STL 분해
#### 이상치 탐지 모델 
- 분류 및 회귀 트리 (CART)
- 클러스터링 기반 이상 탐지
- 오토 인코더

### 이상치 처리 방법 
- 이상치를 결측치로 간주하고 결측치와 같은 방법으로 처리

### 평활화 (smoothing)
- 노이즈를 줄이고, 전체적인 흐름이나 추세를 더 명확하게 파악하기 위해 데이터를 부드럽게 만드는 기술
- 사용 목적
    - 노이즈 제거
    - 추세 파악
    - 예측의 기초자료
- 이동평균의 유형

| 종류 | 설명 | 사용 |
| --- | --- | --- |
| 단순 이동평균 (SMA) | 단순히 기간 내 모든 값 평균 | 변동이 심하지 않고 비교적 안정적인 시계열 |
| 가중 이동평균(WMA) | 최근 값에 더 큰 가중치 부여 | 최근 데이터가 더 중요한 상황 |
| 지수 이동평균(EMA) | 최근 데이터에 지수적으로 더 많은 가중치 부여  | 최근 값에 매우 빠르게 반응해야 하는 시계열   ex) 주식 |

## 요소분해 (시계열 분해)
- 주어진 시계열 데이터를 개별 성분으로 분해
<aside>
💡 $Y_t$ = $f$( $T_t, S_t, C_t, R_t$)

- $T_t$ : 추세성분 (시점 t에서의 추세요소)
- $S_t$ : 계절성분 (시점 t에서의 계절요소)
- $C_t$ : 순환성분 (시점 t에서의 순환요소)
- $R_t$ : 불규칙성분 (시점 t에서의 불규칙요소)

→ 일반적으로 순환성은 장기간에서만 고려되므로 $C_t$ = 0 으로 설정함

</aside>

### 요소분해를 하는 이유
- 시계열 데이터는 다양한 변동이 결합된 형태
- 각 요소를 분리하면 각각의 패턴을 명확히 이해 가능
- 이후 예측, 이상 탐지 등 분석 작업을 정교하게 수행할 수 있음

### 가법모형
y_t = T_t + S_t + R_t
- 시계열 = 추세 + 계절성 + 불규칙성
- 각 변동 요인들이 독립적으로 더해지며, 모든 요소의 영향이 일정함
- 시계열 전체에서 계절성의 진폭이 일정할 때 적합함

### 승법모형
y_t = T_t * S_t * R_t
- 시계열 = 추세 x 계절성 x 불규칙성
- 추세가 커질수록 계절성이나 불규칙성의 효과도 비례해서 커짐
- 진폭이 시간에 따라 달라질 때 적합하다

### 요소분해 정리
|  | STL | seasonal decompose |
| --- | --- | --- |
| 요소분해 방식 | LOESS | 이동평균 |
| 계절성 처리 | 시간에 따라 **변하는  계절성** 반영 | **고정된 계절성**만 반영 |
| 이상치 영향 | 적게 받음(robust 옵션 가능) | 많이 받음 |
| 데이터 손실 | 발생 ❌ | 발생 (이동평균 계산 구간만큼 앞뒤 데이터 손실) |

# 3. 정상성 판단

## 정상성이란?
- 시계열 데이터가 평균, 분산 등의 통계적 특성이 시간에 따라 일정한 수준으로 유지되는 것
- 정상 시계열은 예측 및 모델링에 적합하며 일반적으로 대부분의 통계 기법 적용 가능

### 강한 정상성
- 모든 적률(평균, 분산, 왜도, 첨도)이 시간과 무관하게 일정함
- 기저를 이루는 확률 분포 자체가 언제가 같아야 한다는 것
- 현실에서 찾아보기 매우 어려움
    - 실무 데이터는 잡음과 외부 요인 영향으로 불안정함

### 약한 정상성 
- 통상적으로 시계열 분석에서 '정상성'이라 부르는 것
- 강한 정상성보다 완화된 개념
- 약 정상성을 띄는 시계열은 시간이 흘러도 '평균', '분산'이 변하지 않으며 자기공분산이 시차에만 의존한다


## 정상성 판단 방법

### 1. 시각적으로 정상성 판단하기
- 그래프 시각화를 통해 눈으로 보고 정상성 판단
- 직관적인 방법

### 2. 검정을 통해 정상성 판단하기

### 단위근?
- 어떤 양의 정수 n에 대해 아래 방정식의 근이 되는 복소수를 뜻함
- 시계열적 의미
    - 시계열 모델의 특성방정식에서 근이 1인 경우, 해당 시계열은 단위근을 가짐\
    -> 비정상 시계열
- 단위근이 있으면 문제가 생기는 이유
    - 랜덤워크로 인해 예측 불확실성이 생기기 때문

## 단위근 검정의 대표 방법

### ADF(Augmented Dicky-Fuller) 검정
- 단위근 존재 여부를 통해 시계열의 정상성 판단

- **귀무가설(H₀):** ‘시계열에 단위근이 존재한다**’ → 비정상 시계열**
- **대립가설(H₁):** ‘시계열에 단위근이 존재하지 않는다’ **→ 정상 시계열**

**p-value가 유의수준보다 작으면 귀무가설 기각 → 정상 시계열**


### KPSS(Kwiatkowski-Phillips-Schmidt-Shin) 검정

- **귀무가설(H₀):** ‘시계열에 단위근이 존재하지 않는다’ **→ 정상 시계열**
- **대립가설(H₁):** ‘시계열에 단위근이 존재한다**’ → 비정상 시계열**

 **p-value가 유의수준보다 작으면 귀무가설 기각 → 비정상 시계열**


### KPSS vs ADF
- 두 검정 방법은 검정의 방향이 반대
- 확정적 추세가 존재하는 경우, KPSS와 ADF 검정의 결론이 서로 다르게 나올 수 있음

- 확정적 추세  
    - 추세가 완벽하게 예측 가능하고 일정하게 존재
- 확률적 추세
    - 추세선 자체가 시간에 따라 이동함 (랜덤 워크 성격)

- KPSS는 "평균 주위"에서 정상성을 보는 데 중점
    - 확정적 추세가 있으면 ADF와 결과가 다르게 나올 수 있음
    - 시각적으로 추세나 분산 등을 확인하는 것 중요

## 정상성 확보를 위한 방법

### 1. 추세 제거: 차분
- 연이은 관측값들의 차이를 계산하는 것
- 평균의 정상성을 확보하기 위한 방법

- 대부분 비정상적 시계열은 값이 누적되는 누적과정이 있끼 때문에, 누적된 값을 빼줌으로써 정상적 시계열을 얻어낼 수 있음
- 1차 차분, 2차 차분, 역차분

### 2. 계절성 제거: 계절 차분
- 시계열이 주기적인 페턴을 가질 때 적용
- t시점의 y값에서 t-m시점의 y값을 빼는 것 (m: 주기의 길이)
- 백색소음과 같은 형태의 데이터로 변환
    - 차분을 통해 완전히 random한 데이터를 뽑아낼 수 있음

### 3. 분산 안정화
- 시계열의 변동 폭이 점점 커지거나 작아질 때 적용
- 로그변환
    - 분산의 정상성을 확보하기 위한 방법
- Box-Cox (박스-콕스 변환)
    - 로그 변환의 일반화된 형태

## 시계열 분석의 프로세스
1. 정상/비정상 여부 판단
    - 단위근 검정 사용
2. 비정상 유형에 따라 정상성 확보
    - 추세 -> 차분
    - 계절성 -> 계절 차분 / Deseasoning
    - 분산 변화 -> 로그변환, Box-Cox 변환
    - 적절한 방법을 선택
3. 정상성 확보 후 어떤 프로세스를 따르는지에 따라 모델링
    - OLS, AR, MA, ARMA, SARIMA 등
