# 기존 추천시스템의 한계
- 데이터 희소성 문제: 사용자는 전체 아이템중 일부에만 평점을 매긴다.
- 확장성 문제: 사용자와 아이템의 숫자가 증가할수록 유사도를 계산량이 기하급수적으로 늘어난다.
- 새로운 아이템 문제 : 새로운 아이템은 상호작용이력이 없다.

# 모델기반 협업 필터링

- 모델기반은 관측된 평점 데이터로부터 사용자와 아이템의 잠재요인을 학습해 비어있는 평점을 예측한다.

- 잠재요인: 사용자-아이템 상호작용 행렬에 숨겨진, 사용자의 취향이나 아이템의 특성을 설명하는 숨겨진 차원.

- 행렬분해: user-item의 평점행렬이 사용자의 취향벡터와 특성벡터의 내적으로 설명될수 있다는 아이디어.

- SVD: 밀집행렬에 적합해 희소행렬에 직접적용 불가. Truncated SVD를 사용해 핵심 패턴 압축

- SGD: 관측된 평점에 대해서만 오차를 최소화하며 학습하여 희소성 문제를 우회적으로 해결

- NMF: 분해된 행렬에 비음수 제약을 추가해 잠재요인에 대한 직관적 해석력 높임

- PMF : 평점 예측에 확률분포를 사용해 불확실성까지 모델링하는 통계적 접근법. 

# 딥러닝 기반 모델
- MF의 선형성 한계와 불확실성 무시 문제를 극보가기 위해 비선형 활성화 함수와 복잡한 신경망을 사용하는 딥러닝 모델이 등장
1. VAE

- AE는 데이터가 부족할때 추정치가 불안정하고 틀리기 쉬워 노이즈에 취약함에 반면 VAE는 불확실성에도 안정성이 높음.

- 사용자의 아이템 상호작용 벡터를 입력으로 받아, 잠재공간z에 재구성하고 디코더가 z로부터 아이템분포를 복원하는 구조이다. 


2. GNN 기반 모델 (Graph Neural Network)

- GNN은 사용자-아이템 상호작용을 거대한 관계망으로 보고, '관계' 자체를 학습하여 고차원 협업 시그널을 포착.

- 메시지 패싱을 통해 각 노드가 이웃 노드의 정보를 수집하고 자신의 임베딩을 업데이트함.

- 그래프는 현실세계의 개체인 노드와 연결선인 엣지로 이루어짐. 관계, 상호작용과 같이 다소 추상적인 개념을 다루기에 적합함. 

- 작동구조

    1. 그래프 구조 생성

    2. 메시지 패싱( 메시지 생성, 집계, 갱신)

    3. 최종출력

- K번의 메시지 패싱을 통해 '나의 이웃의 이웃', 즉 나와 취향이 비슷한 더 멀리 있는 사용자들의 정보까지 임베딩에 반영함

- 대표모델: lightGCN은 기존 GCN에서 복잡한 연산을 제거하고 오직 정규화된 이웃 임베딩의 합만으로 구성하여 효율성과 성능을 높임

- 장점: 고차원 관계 포착, Cold Start문제 완화에 효과적

- 단점: 과평탄화, 과압축 문제, 높은 계산 복잡도

3. Transformer 기반 모델
- CF와 GNN이 해결하지 못한 사용자 행동의 순서와 문맥을 파악하여 동적인 의도를 예측

- 핵심: Self-Attention 메커니즘을 사용해 시퀀스 내 모든 아이템 간의 관계를 거리에 상관없이 동시에 계산가능. 이로써 장기 의존성 문제도 해결 가능

- Query가 Key와 관련성을 계산하여 가중치를 만들고 이를 Value에 곱해 최종 문맥 벡터를 생성

- 주요모델

    - SASRec: 단방향 어텐션을 사용해 다음 아이템을 예측

    - BERT4Rec: 시퀀스 일부를 [Mask] 토큰으로 가리고 양방향 어텐션을 사용해 가려진 아이템을 에측함. 시퀀스 전체 문맥 이해에 더 강점이 있음.

# 최근 트렌드

- LLM의 등장으로 추천을 언어 문제로 전환하기 시작함.

- 단순 점수계산이 아닌 자연어 문장으로 사용자에게 추천을 제공함

- 텍스트 설명만으로 신규 아이템을 추천할 수 있는 zero-shot 추천을 가능케 해 Cold-Start문제를 강력히 해결함.

- 선형모델은 속도가 빠르고 안정적이며 비용효율적임. 선형패턴에 강하고 정확도 상한선도 존재. 상대적으로 해석하기도 용이함.

- 딥러닝 모델은 느리고 무거우며 막대한 자원량을 요구함. 비선형패턴, 순서, 관계 학습에 최적화, 높은 정확도 보유. 하지만 블랙박스 문제로 해석이 어려움. 

- 따라서 같이 사용해야함(2단계 파이프라인)

    1. 후보군생성(선형모델)
        - 수백만개의 전체 아이템중 사용자가 좋아할만한 수백개의 후보군을 매우 빠른 속도로 필터링

    2. 순위 매기기(딥러닝)

        - 1단계에서 추린 수백개의 후보군만을 대상으로, 복잡한 딥러닝 모델을 사용해 최고의 정확도로 정교한 순위를 매김. 

# 현업의 고민

- 모델성능 뿐 아니라 다음 요소들을 고려하는 것이 중요

- 인기편향: 인기 아이템의 과도한 노출을 제어하여 비인기/ 신규 아이템에 노출 기회를 제공

- 설명가능성: 추천근거를 고객과 비지니스 담당자에게 제시할 수 있어야함.
