# RNN / LSTM의 한계

1. 장기 의존성 문제

시퀀스의 길이가 길어질수록, 앞부분의 정보가 소실되는 기울기 소실 문제가 발생함

2. 병렬화 어려움

시퀀스 데이터의 특성상 직렬이기 때문에, 학습 시간이 매우 길어지는 문제점이 있음

# 트랜스포머의 핵심 원리



기존 모델들인 RNN/LSTM의 순환 구조를 완전히 제거하고 '어텐션(Attention)' 메커니즘만으로 시퀀스 내의 관계를 파악

> 순서대로 입력 시퀀스를 처리하는 것이 아닌, 모든 토큰을 동시에 입력 받아서 병렬 처리를 하는 원리!!

# Self - Attention

self-attention은 "문장 내의 한 단어가 같은 문장 내의 다른 단어들과 얼마나 관련이 있는가?"를 측정하는 메커니즘

1. Query, Key, Value (Q, K, V)

> 셀프 어텐션은 각 입력 토큰(의 임베딩 벡터)으로부터 세 가지 다른 벡터를 생성

* Query (Q): 현재 초점을 맞춘 단어의 "질문" 벡터

* Key (K): 문장 내 모든 단어의 "식별자" 또는 "라벨" 벡터

* Value (V): 문장 내 모든 단어의 "실제 의미" 또는 "내용" 벡터

2. Self Attention vs Cross Attention



* Self-Attention (셀프 어텐션):

    - Q, K, V가 모두 같은 출처(소스)에서 나옴

* Cross-Attention (교차 어텐션):

    - Q는 다른 출처에서, K와 V는 같은 출처에서 나옴


3. Scaled Dot-Product Attention



Dot-Product Attention은 Attention Score를 계산하는 가장 기본적인 방법 중 하나
> 쿼리(Query)와 키(Key) 벡터를 내적(Dot Product)하여 유사도를 구함

# 트랜스포머의 전처리 과정

트랜스포머는 텍스트를 입력(Inputs)받고,<br>
그 <U>텍스트를 토큰화</U>하고, <U>토큰들을 임베딩</U>하고, <U>Positional Encoding으로 순서 정보를 추가</U>하여 인코더 레이어에서 실질적인 연산을 수행



토큰화는 입력 텍스트를 모델이 처리할 수 있는 단위(tokens)로 나누는 첫 번째 단계



임베딩은 토큰화된 각 단위(토큰)를 숫자의 벡터로 변환하는 단계



단어들의 순서 정보를 임베딩 벡터에 추가하는 단계

# 트랜스포머의 인코더/디코더 구조



인코더 (Encoder): 입력 시퀀스를 받아 문맥 정보를 풍부하게 압축한 표현을 만듦



디코더 (Decoder): 인코더의 표현과 이전에 생성된 출력 시퀀스를 참조하여 다음 토큰을 생성

# Multi-Head Attention



한 번의 attention이 한 가지 종류의 관계만 파악할 수 있다면, <U>여러 개의 attention을 동시에 수행하여 다양한 종류의 관계를 한 번에 파악</U>

# Encoder-Decoder Attention
