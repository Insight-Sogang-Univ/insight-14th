# 순차 데이터(시퀀스 데이터)란?



## 순차 데이터: 시간적 순서가 의미를 가지는 데이터

예시:

* 문장(단어들의 순서가 의미를 만든다)
* 음성 신호(시간 흐름에 따라 파형이 달라진다)
* 주가, 기온, 센서 신호(과거 흐름과 트렌드가 중요하다)

주의해야하는 점:<br>
<U>순서가 바뀌면 의미가 달라짐!!</U><br>
<U>“과거 맥락”을 이해해야함</U>

# RNN의 구조



RNN은 이전 시점의 숨겨진 상태를 다음 시점으로 전달하여 시퀀스 데이터를 처리함


$h_t = f(Wxh * x_t + Whh * h_{t-1} + b)$

$h_{t-1}$이 기억(맥락) 역할을 하는 것임!


## 가중치 공유

모든 시점 $t$에서 같은 $Wxh$, $Whh$를 사용

장점: 파라미터 수가 시퀀스 길이에 영향을 받지 않음

의미: 모델이 “시간에 따라 같은 패턴이 반복된다”는 가정을 따름

## 장기 의존성 문제

RNN은 시간이 지나면서 정보가 점점 약해지는 문제가 있음

초반 시점의 정보가 후반에 도달하기 어려움 -> <U>“문장 초반 의미를 끝까지 기억하기” 가 어려움...</U>

# LSTM의 구조




RNN의 장기 의존성 문제를 해결하려고 나온 구조<br>
기억 셀(cell state)을 별도로 관리하며, 기억을 <U>저장 / 유지 / 삭제</U>하는 3개의 게이트를 가지고 있음

| 게이트                  | 역할                |
| -------------------- | ----------------- |
| 입력 게이트 (Input gate)  | 현재 정보를 **얼마나 사용**할지 결정    |
| 망각 게이트 (Forget gate) | 정보를 **얼마나 잊어**버릴지 결정   |
| 출력 게이트 (Output gate) | 다음 층으로 **어떤 정보를 전달**할지 결정 (hidden state를 만듦) |

-> LSTM은 기억을 선별적으로 유지함으로써 <U>장기 문맥 기억에 강함</U>


# GRU의 구조



LSTM을 더 간단하게 줄여놓은 구조

| 게이트                    | 역할                     |
| ---------------------- | ---------------------- |
| 업데이트 게이트 (Update gate) | 과거 기억 **얼마나 유지할 지** 결정        |
| 리셋 게이트 (Reset gate)    | 새로운 입력과 과거 기억 **어떻게 섞을 지** 결정 |

-> LSTM보다 파라미터가 적어서 학습이 빠르고 구현도 단순, 성능은 LSTM과 비슷한 경우가 많음

# Seq2Seq의 구조



입력 시퀀스를 받아 다른 형태의 시퀀스로 변환하는 구조, '번역, 요약, 질의응답' 등에 자주 쓰임

구조:

* Encoder: 입력 시퀀스의 의미를 하나의 벡터로 요약

* Decoder: 그 벡터를 기반으로 출력 시퀀스를 생성

$입력 → [Encoder] → context vector → [Decoder] → 출력$ 의 구조임


## 병목현상




Encoder가 모든 정보를 단 하나의 벡터(context vector)에 압축해야 하는데, 길고 복잡한 문장일수록 정보가 많이 손실됨

-> 이 문제를 해결하려고 나온 것이 Attention (Transformer 차시에 배울 예정~~)!!
