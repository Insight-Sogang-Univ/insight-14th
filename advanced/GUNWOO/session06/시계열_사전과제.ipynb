# 시계열 개념

시계열 데이터 : 시간의 흐름에 따라 일정한 간격(초, 분, 일, 월, 연 등)으로 관측된 데이터의 수열

## 종단면 데이터, 횡단면 데이터

image.png


종단면 데이터 : 하나의 개체를 시간의 흐름에 따라 추적하여 관측한 데이터


횡단면 데이터 : 특정 시점에 다양한 개체들을 동시에 관측한 데이터

## 단변량/다변량 시계열

단변량 시계열 : 시간이 지남에 따라 하나의 변수만을 관측한 시계열 데이터

다변량 시계열 :	시간이 지남에 따라 두 개 이상의 변수들을 동시에 관측한 시계열 데이터

## 시간의존성/계절성/추세/자기상관성

* 시간의존성 : 현재 시점의 관측값이 과거 시점의 관측값에 의해 영향을 받는 관계

* 자기상관성 : 특정 시점의 값과 과거 시점의 값 사이에 존재하는 선형 상관관계

* 추세 : 장기간에 걸쳐 데이터가 지속적으로 상승하거나 하락하는 방향성

* 계절성 : 1년, 1주, 1일 등 특정 기간을 주기로 반복되어 나타나는 규칙적인 패턴

## 추세/계절/순환/불규칙 요인

image.png

* 추세 요인 :	장기적인 방향성을 나타내는 요소

* 계절 요인 : 주기적이고 반복적인 패턴을 나타내는 요소

* 순환 요인 : 경기 변동처럼 특정 주기를 갖지만, 계절성보다 긴 비주기적인 변동을 의미

* 불규칙 요인 :	위 세 가지 요인으로 설명되지 않는, 예측 불가능하고 무작위적인 변동

# 시계열 전처리

## 결측치 처리

결측치를 채울 때에는 주변 시간의 데이터를 활용하는 보간법을 주로 사용함

* 선형 보간 : 결측치 이전 값과 이후 값을 직선으로 연결하여 그 사이에 값을 채우는 방법

* 최근린 보간 : 가장 가까운 시점의 관측값으로 결측치를 대체하는 방법

* Spline 보간 : 다항식 함수를 사용하여 데이터의 곡률을 유지하며 부드럽게 보간하는 방법

## 이상치 탐지

다른 데이터들과 동떨어져 있는 비정상적인 값(이상치)을 탐지하고 수정

+++ 시계열에서는 단기적인 급등락이나 갑작스러운 값의 변화 등이 이상치로 간주될 수 있고, 보통 IQR, ARIMA를 사용한 예측 오차를 사용함

## 평활화

image.png


데이터에 포함된 불규칙 요인을 제거하고 추세나 주기를 부드럽게 파악하기 위해 사용

* 단순 이동평균 : 현재 시점을 포함한 과거 N개 시점의 단순 평균을 현재의 값으로 사용

* 가중 이동평균 : 최근 시점에 더 큰 가중치를 부여하여 평균을 계산

* 지수 평활 : 과거의 모든 데이터에 지수적으로 감소하는 가중치를 적용하여 계산

## 요소 분해

보통 시계열 데이터를 추세, 계절성, 순환, 불규칙 요인으로 분해하여 각 요인의 영향을 분석함!

> 이 분해된 요소들을 활용하여 예측, 정상성을 확보하는 데 사용

# 정상성

## 정상성이란?

정상성 : 시계열의 통계적 특성이 시간의 흐름에 관계없이 일정하게 유지되는 성질

## 강한 정상성/약한 정상성

* 강한 정상성 : 모든 시점에 대해 시계열 데이터의 결합 확률 분포가 시간 이동에 따라 변하지 않는 경우

* 약한 정상성 : 기대값(평균)이 시간에 따라 일정하고 자기공분산이 시점에서가 아닌 시차에서만 의존할 때 만족

## 단위근의 시계열적 의미

시계열 데이터에 단위근이 존재한다는 것은 해당 시계열이 비정상적이며, 특히 확률적 추세를 가지고 있음을 의미한다!

## 단위근 검정

* Dickey-Fuller (DF) Test

* Augmented Dickey-Fuller (ADF) Test: DF 검정을 일반화하여 고차의 자기상관성까지 고려하는 가장 널리 사용되는 검정법

  * 귀무가설($H_0$): 시계열에 단위근이 존재한다(비정상이다).

* Phillips-Perron (PP) Test: ADF 검정의 단점을 보완하여 오차항의 자기상관성 문제를 비모수적으로 처리하는 검정법

## 확정적 추세와 확률적 추세

* 확정적 추세 : 시간에 대한 함수로 완벽하게 설명되는 추세

* 확률적 추세 : 단위근에 의해 발생하는 추세 직전 시점의 오차가 다음 시점 이후에도 계속 누적되어 나타남

## 정상성 확보 방법

* 차분 : 현재 시점의 관측값과 이전 시점의 관측값의 차이를 계산<br>확률적 추세를 제거하고 정상성을 확보하는 가장 일반적인 방법

* 추세 제거 : 회귀분석 등을 통해 추세 함수를 추정하고, 원본 데이터에서 해당 추세 성분을 제거

* 변환 : 분산이 일정하지 않은 경우(이분산성) 로그 변환이나 제곱근 변환 등을 통해 분산을 안정화

* 계절 차분 : 계절성이 존재하는 시계열의 경우, 주기만큼 차분하여 계절 요인을 제거
