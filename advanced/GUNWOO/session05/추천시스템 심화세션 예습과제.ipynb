# 행렬 분해 (Matrix Factoriztion)

| **구분** | **접근 방식** | **특징** |
| --- | --- | --- |
| **SVD(Singular Value Decomposition)** | 특이값 분해, 수학적 분해 기법 | 1. 행렬 대수학에 기반하여 행렬을 분해<br>2. 희소 행렬에는 직접 적용 불가하여 결측치 처리가 필요 |
| **SGD(Stochastic Gradient Descent)** | 경사 하강법, 최적화 알고리즘 | 1. 경사 하강법 기반의 알고리즘<br>2.관측된 평점에 대해서만 학습을 진행하여 희소성 문제를 우회적으로 해결 |
| **NMF(Non-negative Matrix Factorization)** | 비음수 행렬 분해, 제약 조건 기반 모델 | 분해된 행렬에 비음수(Non-Negative)라는 제약 조건을 추가하여 잠재 요인의 해석력을 높임 |
| **PMF(Probabilistic Matrix Factorization)** | 확률적 선형 분해 모델 | 평점 예측에 확률 분포를 사용하여 불확실성까지 모델링하는 통계적 접근법 |

# 딥러닝 기반 모델

## VAE 기반 모델

image.png

**V**ariational **A**uto**E**ncoder, 변이형 오토인코더

원리:<br>
입력 데이터를 잠재 공간에 매핑하고 다시 복원하는 생성 모델<br>
추천 시스템에서는 사용자-아이템 상호작용 벡터를 입력으로 받아, 잠재 표현을 학습한 후 원래의 상호작용 벡터를 재구성하도록 훈련

장점:<br>
상호작용 데이터의 노이즈를 줄이고 데이터 분포의 확률적 특성을 모델링하여 단순히 점을 예측하는 것 이상의 풍부한 잠재 공간을 학습할 수 있음


## GNN 기반 모델

image.png

원리:<br>
추천 시스템 데이터를 그래프 구조로 모델링<br>
노드(Node)는 사용자와 상품을, 엣지(Edge)는 상호작용(평점/클릭 등)을 나타냄<br>
GNN은 노드와 엣지 정보를 사용하여 노드의 잠재 표현을 주변 노드의 정보를 집계하며 학습

장점:<br>
사용자-아이템 그래프에서 고차원의 관계(예: 사용자가 상호작용한 아이템, 그 아이템에 상호작용한 다른 사용자)를 자연스럽게 포착하여 MF보다 표현력이 뛰어남<br>
<U>LightGCN</U>과 같은 모델이 많이 사용됩니다.

# 최근 딥러닝 추천시스템 트렌드

## 선형 모델 vs 딥러닝 모델
선형 모델은 선형 변환을 기반으로 사용자-아이템의 잠재 요인을 학습함
반면, 딥러닝 모델은 비선형 활성화 함수를 사용해서 복잡한 비선형 관계를 학습함

## 차이점

이러한 특징을 기반으로 선형 모델은 잠재 요인 간의 곱으로 상호작용을 모델링하기 때문에 상대적으로 적은 데이터로 안정적인 학습을 하고 계산 복잡도가 낮음!

반면, 딥러닝 모델은 고차원적인 특징과 복잡한 패턴을 포착하고 텍스트/이미지 같은 보조 정보도 통합하여 학습할 수 있지만, 대규모 데이터가 필요함.... 그리고 계산 복잡도 또한 매우 높기 때문에 비용이 큰 문제가 있음!

## 혼용하는 이유

서로의 문제점을 보완하기 때문에 함께 사용하거나 하이브리드 형태로 결합되기도 함!

Wide & Deep Learning 모델 같은 경우, 선형 학습과 딥러닝 학습 모두를 활용한다고 함
