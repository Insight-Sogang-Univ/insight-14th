## <span style="background-color:#F993C0; color:#311d26; padding:4px; border-radius:0px;"> 인공신경망의 시작: 퍼셉트론 </span>

#### <span style="background-color:#633a4c; color:#fde9f2; padding:4px; border-radius:0px;"> 인공신경망이란? </span>

<span style="color:#f993c0;"> 인공신경망과 퍼셉트론 </span>

**인공신경망**

: 우리 뇌의 신경망(뇌를 구성하는 신경 세포 뉴런이 서로 연결되어 정보를 처리하는 구조)을 모방한 모델

[생물학적 뉴런이 정보를 처리하는 구조]



[인공 신경망이 정보를 처리하는 구조]



1. 다수의 입력($x_1, x_2, …, x_n$)을 받아 각각에 **가중치**($w_1, w_2, … ,w_n$) 적용
2. 종합한 값이 **일정 기준을 넘으면**(활성화 함수)
3. **output** 출력

**퍼셉트론**

: 초기 형태의 인공신경망



1. 입력값과 가중치를 곱함 →  ${x_1*w_1}$
2. 곱한 값들의 총합을 구함 → ${x_1*w_2+x_1*w_2+x_3*w_3 +...+x_n*w_n}$
3. 총합이 임계값(${\theta}$)을 넘으면 $1$, 넘지 않으면 $0$을 출력 (= Step 활성화 함수의 작업)
    
    ※ 활성화 함수: 임계값(${\theta}$) 초과 여부에 따라 출력값을 결정하는 역할 
    

→ **Training(훈련 또는 학습)**: 원하는 출력값을 내보내도록 **가중치를 조정**해가는 작업

#### <span style="background-color:#633a4c; color:#fde9f2; padding:4px; border-radius:0px;"> 단층 퍼셉트론과 그 한계 </span>

**단층 퍼셉트론**

: 두 개의 층(입력층, 출력층)으로만 이루어진 퍼셉트론

- 단층 퍼셉트론은 값을 보내는 단계와, 값을 받아서 출력하는 두 단계로만 이루어짐   
- 단층 퍼셉트론을 이용하면 AND, NAND, OR 논리 게이트를 쉽게 구현 가능

**논리 게이트의 종류**

1. **AND 게이트**
    : **두 개의 입력 값이 모두 1**인 경우에만 **출력값이 1**이 나오는 구조

2. **NAND 게이트**
    : **두 개의 입력값이 1**인 경우에만 **출력값이 0**, 나머지 입력값의 쌍에 대해서는 **모두 출력값이 1**

3. **OR 게이트**
    : **두 개의 입력이 모두 0**인 경우에 **출력값이 0**이고 나머지 경우에는 **모두 출력값이 1**

**단층 퍼셉트론의 한계**

→ **XOR 게이트 구현 불가능**

    XOR 게이트란? 입력값 두 개가 서로 다른 값을 갖고 있을 때에만 출력값이 1!

퍼셉트론은 입력값과 가중치의 선형 결합으로 이루어진 구조이기 때문에, 단층 퍼셉트론은 직선 하나로 두 영역을 나눌 수 있는 선형 분류 문제에 대해서만 구현이 가능

    XOR 게이트는 직선이 아닌 곡선, 비선형 영역으로 분리해야만 구현이 가능!

#### <span style="background-color:#633a4c; color:#fde9f2; padding:4px; border-radius:0px;"> 다층 퍼셉트론 </span>

XOR 게이트는 기존의 AND, NAND, OR 게이트를 조합하면 만들 수 있음
→ 즉, 퍼셉트론의 층을 더 쌓으면 가능!



- 중간에 층을 더 추가하여 **입력층** & 하나 이상의 **은닉층** & **출력층**으로 구성
- XOR 문제보다 더욱 복잡한 문제를 해결하기 위해서 **중간에 수많은 은닉층 더 추가 가능**!
- 두 개 이상의 은닉층을 갖는 MLP가 딥러닝의 시초 모델로 발전

## <span style="background-color:#F993C0; color:#311d26; padding:4px; border-radius:0px;"> 딥러닝 모델은 어떻게 학습할까? </span>

#### <span style="background-color:#633a4c; color:#fde9f2; padding:4px; border-radius:0px;"> 순전파와 역전파 </span>

**학습 알고리즘의 기본 구조**

(1) 순전파를 통해 **입력값으로부터 예측**을 수행

(2) **오차 계산**을 통해 **예측 정확도**를 측정

(3) **역전파**를 통해 출력층에서 다시 입력층 방향으로 **오차를 전달**

(4) **가중치를 업데이트**

(5) 1~4를 반복…



**연쇄 법칙 (Chain Rule)**   
: 서로 얽혀있는 변수 간의 상관관계를 계산하는 방법

**A(입력값)으로 인해 발생한 C(결과값)의 오차 = B가 C에 준 영향 X A가 B에 준 영향**

⇒ 즉, **A가 C에 미치는 영향은, B가 C에 미치는 영향과 A가 B에 미치는 영향을 곱한 것**

은닉층의 개수가 많아질수록 앞쪽 은닉층의 영향력은 거의 사라짐
→ 기울기 소실 문제

    합성 함수의 미분은 합성 함수를 구성하는 각 함수의 미분의 곱으로 나타낼 수 있다!

#### <span style="background-color:#633a4c; color:#fde9f2; padding:4px; border-radius:0px;"> 손실 함수 </span>

손실 함수는 모델이 정확한지 오차를 계산하는 방법 

**⇒ 실제값과 예측값의 차이를 수치화**하는 함수

→ 결국 출력층의 결과와 정답의 차이를 계산하는 데 사용되는 함수가 손실 함수!

손실 함수의 종류

1. 평균 제곱 오차 (MSE, Mean Squarred Error)

    $-\frac{1}{N}\sum(y-\hat{y})^2$    

    $y : 예측값 / \hat{y} : 실제값$

- 예측 값이 연속적이고, 예측 값과 실제 값 간의 차이를 직접적으로 측정하는 회귀 문제에 적합

2. 이진 크로스 엔트로피 Binary Cross-Entropy Loss (BCE)

    $L= -(y \log \hat{y} + (1-y) \log(1-\hat{y}))$

    $y : 실제값(0,1)이진값 /  \hat{y} : 예측값(확률)$

- BCE는 이진 분류 문제(로지스틱 회귀)에서 사용 

3. 크로스 엔트로피 오차 (CEE, Cross Entropy Error)

    $-\sum y \log \hat{y}$

    $y : 원-핫 인코딩 값 (1,0)/  \hat{y} : 예측값(확률)$

- 다중 클래스 분류 문제에서 예측 확률이 실제 클래스와 얼마나 일치하는 지를 평가하는 데 적합


#### <span style="background-color:#633a4c; color:#fde9f2; padding:4px; border-radius:0px;"> 활성화 함수 </span>

활성화 함수: 출력값을 다음 단계에서 쓸 수 있는 형태로 적절히 바꿔주는 역할
- 활성화 함수는 비선형성을 만들어주며, 이를 통해 복잡한 패턴도 학습이 가능!



1. 계단함수 (단층 퍼셉트론)

    : 입력 값과 가중치의 곱을 모두 더한 뒤, 특정 기준을 넘으면 “1”, 아니면 “0”을 출력하는 간단한 구조로, 주로 단층 퍼셉트론에서 사용    
    - **입력 신호: 입력값과 가중치의 곱**    
    - **출력 신호: 1과 0**

2. 시그모이드 함수

    : 입력 신호를 특정 범위(0과 1 사이)로 변환하는 역할을 수행!

    → 이때, 함수는 직선이 아니라, **곡선 형태의 비선형 함수**

3. ReLU 함수

    : 입력이 0을 넘으면 그 입력값을 그대로 출력하고, 0 이하면 0을 출력하는 **비선형 함수**

    → 간단하지만 딥러닝에서 **가장 많이 사용하는 활성화 함수**

4. Tanh(쌍곡탄젠트)함수

    : **시그모이드 함수**의 변형

    → 출력값의 범위가 다름 (**시그모이드: 0~1**, **쌍곡탄젠트: -1~1**)

5. Softmax 함수

    : 시그모이드 함수와 비슷한 개형을 갖지만, **“다중 클래스 분류 문제의 활성화 함수”**로 사용

    출력 값을 확률로 변환함으로써 **개별 클래스의 확률**을 알 수 있음

    → 출력값들의 합은 **1**

#### <span style="background-color:#633a4c; color:#fde9f2; padding:4px; border-radius:0px;"> 경사 하강법과 옵티마이저 </span>

경사 하강법은 손실 함수를 최소화하여 모델을 학습시키는 방법

즉, **오차를 가장 작게 만드는 가중치와 편향을 찾는 방법!**

딥러닝에서도 경사 하강법으로 **오차를 가장 작게** 만드는데, 보통 데이터가 너무 방대한데 학습 과정에서 모든 데이터를 한 번에 집어넣으면 메모리가 부족하고, 시간도 오래 걸리는 문제 발생

⇒ 그래서 필요한 것이 **“배치(batch)”**

**배치 (Batch)**    
: 학습 과정에서 가중치와 같은 매개변수를 조정하기 위해 사용하는 데이터의 묶음

**이포크(epoch)**   
: 모든 데이터셋을 **학습하는 횟수**를 의미

**이터레이션(Iteration) 또는 스텝(Step)**
: 한 번의 에포크를 끝내기 위해서 필요한 배치의 수

<경사하강법 종류>

1. 배치 경사 하강법
- 전체 데이터를 통해 학습시키기 때문에, **업데이트 횟수가 적음** (1 Epoch 당 1 업데이트)
- 전체 데이터를 모두 한 번에 처리하므로 **메모리가 많이 필요**
- 항상 같은 데이터 (전체)에 대해서 경사를 구하기 때문에 **수렴이 안정적**

2. 배치 크기가 1인 확률적 경사 하강법(Stochastic Gradient Descent, SGD)
- 하나의 데이터에 대해서만 메모리에 저장하면 되므로 **자원이 적은 컴퓨터에서도 쉽게 사용이 가능**
- 매개변수의 **변경폭이 불안정**하고, **정확도가 낮을** 수 있음

3. 미니 배치 경사 하강법(Mini-Batch Gradient Descent)
- **배치 크기를 적절히 지정**하여 해당 데이터 개수만큼에 대해 **매개 변수의 값을 조정**
- 전체 데이터를 계산하는 것보다는 **빠르고**, 배치 크기가 1인 경사 하강법보다 **안정적**
    
    **⇒ 가장 많이 사용**되는 경사 하강법

**옵티마이저**

1. 모멘텀(Momentum)
- **경사 하강법 + 관성**
- 경사 하강법에서 계산된 **접선의 기울기**에 **바로 한 시점 전의 접선의 기울기값**을 **일정 비율**만큼 반영

2. 알엠에스프롭(RMSProp)
- **급한 경사에서는 속도를 줄이고**
- **완만한 경사에서는 속도를 유지하거나 증가**

3. 아담(Adam)
- **RMSProp**과 **Momentum**을 합친, **“방향과 스텝 사이즈를 적절하게 합친 방법”**   
⇒ 즉, **방향**과 **학습률** 두 가지를 모두 잡기 위한 방법!
- 각 파라미터의 1차 모멘트(기울기의 평균)와 2차 모멘트(기울기의 제곱 평균)을 사용하여 학습률을 조정

#### <span style="background-color:#633a4c; color:#fde9f2; padding:4px; border-radius:0px;"> 데이터 증강과 전이 학습 </span>

데이터 증강의 장점

- 성능의 향상

## <span style="background-color:#F993C0; color:#311d26; padding:4px; border-radius:0px;"> 합성곱 신경망(CNN) </span>

#### <span style="background-color:#633a4c; color:#fde9f2; padding:4px; border-radius:0px;"> 픽셀과 채널 </span>

**픽셀(pixel)**
- 입력 이미지의 가장 기본적인 데이터 단위
- 이미지의 각 점에서 밝기/색상을 나타내는 값

**채널(channel)**
- **색공간**
    - 픽셀의 색을 숫자로 표현하는 방식
    - 대표적인 색공간
    
        : 그레이스케일(gray scale), RGB(Red-Green-Blue), HSV(Hue-Saturation-Value / 색,채도,명도)

#### <span style="background-color:#633a4c; color:#fde9f2; padding:4px; border-radius:0px;"> CNN이란? </span>

사람은 이미지 전체의 모든 부분을 하나하나 다 살펴보는 것이 아님

- 물체(강아지, 고양이, …)가 존재하는 영역 위주로 살펴본다(이미지 전체를 보지 않는다).
- 물체의 특징(눈, 코, 입 모양)을 위주로 살펴본다(물체 전체를 보지 않는다).

    이처럼, 이미지 공간정보를 유지한 상태로 학습이 가능한 방식이 바로 CNN !

<span style="color:#f993c0;"> CNN 모델 </span>

#### <span style="background-color:#633a4c; color:#fde9f2; padding:4px; border-radius:0px;"> 이미지 처리에 CNN을 사용하는 이유 </span>

1. 이미지 데이터의 구조
- 이미지는 일반적인 정형 데이터와 다른 구조를 가지고 있기 때문에 구조를 유지래야 함

2. 위치, 크기, 회전 등의 변화에 취약 (Translation Variant)
- **CNN에서는?**
    - **Convolution으로 특징(feature) 추출**: 모델이 ‘특징’을 학습한다.
    - **Pooling으로 차원 축소**: 위치/크기 변화에 덜 민감하게 만든다.
    - 어떤 영역에 **특정한 특징이 존재하는가**를 보기 때문에 위치에 무관하다!

3. 파라미터 폭발 (Parameter Explosion)
- 부분연결 구조 & Convolution 연산으로 이미지의 특정 피쳐에만 집중

4. 공간 정보 손실 (Loss of Spatial Info)
- **CNN에서는?**
    - 다차원 이미지를 그대로 입력받고, 공간 정보를 보존하면서 학습한다.
    - 특징을 어느 정도 추출한 후 flatten 하기 때문에, 그 손실의 정도가 훨씬 낮다.

#### <span style="background-color:#633a4c; color:#fde9f2; padding:4px; border-radius:0px;"> 합성곱과 풀링 </span>

**CNN 모델의 구조**

CNN 모델은 크게 두 부분으로 나뉨:

<aside>

1. **특징 추출 (Feature Learning)** - **Convolution,** **Pooling**이 담당
2. **클래스 분류 (Classification)** - **Fully Connected Layer**가 담당
</aside>

1. Convolution Layer

    : 전체 이미지를 일정 크기의 필터로 훑으면서 이미지의 특징을 학습
 
    (1) 필터(Filter) 또는 커널(Kernel)이 이미지 위를 슬라이딩   
        - CNN의 필터 값은 MLP의 가중치(weight)와 같은 역할을 함   
    (2) 각 위치에서 **요소별 곱셈** 후 **합계** 계산   
    (3) 결과를 **Feature Map**에 저장   
        - **feature map**: input data에 filter를 적용하여 얻은 결과물   
        - feature map의 차원은 filter의 수에 따라 결정됨.   
        
2.  Padding

    : filter를 거치며 정보를 잃어버리는 것을 막기 위해 이미지 바깥을 0으로 둘러싸는 기법

3. Stride (보폭)

    : 이미지를 몇 칸씩 점프하면서 훑을 것인지 결정하는 값

    - **Stride = 1**: 한 칸씩 이동 (세밀한 특징 추출)
    - **Stride = 2**: 두 칸씩 이동 (빠른 처리, 크기 축소)

    → 계산량 감소, 겹치는 영역을 줄여 효율적으로 요약

    **⇒ stride가 커질 수록 feature map의 크기가 작아진다**

4. Pooling

    : 피처맵에서 특징을 추출하기 위해 이미지를 격자 형태로 분할한 뒤, 해당 영역 안에서 가장 큰 값 또는 평균값을 취하는 것
