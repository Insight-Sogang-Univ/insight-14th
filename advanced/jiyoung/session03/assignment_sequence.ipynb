### LSTM (Long Short-Term Memory)

LSTMì˜ í•µì‹¬ ì•„ì´ë””ì–´ëŠ” ì…€ ìƒíƒœ(Cell State)ë¥¼ ë”°ë¡œ ë‘ëŠ” ê²ƒì…ë‹ˆë‹¤.

ë™ì‘: ì´ì „ ì‹œì ì˜ ì¶œë ¥($h_{t-1}$)ê³¼ í˜„ì¬ ì‹œì ì˜ ì…ë ¥($x_t$)ì„ ë°›ì•„ 0ì—ì„œ 1 ì‚¬ì´ì˜ ê°’ì„ ì¶œë ¥í•©ë‹ˆë‹¤. (1ì´ë©´ ì™„ì „íˆ ê¸°ì–µ, 0ì´ë©´ ì™„ì „íˆ ìŠìŒ)  
> ì˜ˆ: ë¬¸ì¥ì—ì„œ ì£¼ì–´ê°€ ë°”ë€Œë©´, ì´ì „ ì£¼ì–´ì— ëŒ€í•œ ì •ë³´ë¥¼ ìŠë„ë¡(0ì— ê°€ê¹ê²Œ) ì‹ í˜¸ë¥¼ ë³´ëƒ…ë‹ˆë‹¤

> ì˜ˆ: ìƒˆë¡œìš´ ì£¼ì–´ê°€ ë“±ì¥í•˜ë©´, ì´ ì •ë³´ë¥¼ ì…€ ìƒíƒœì— ì¶”ê°€í•˜ë„ë¡(1ì— ê°€ê¹ê²Œ) ì‹ í˜¸ë¥¼ ë³´ëƒ…ë‹ˆë‹¤.  

    > Final memory cellì˜ ê³„ì‚°: $C_t$ = ( Forget Gateì˜ ì¶œë ¥ $\times$ $C_{t-1}$(ì´ì „ ì…€ ìƒíƒœ) ) + ( Input Gateì˜ ì¶œë ¥ $\times$ $\tilde{C}_t$(ìƒˆë¡œìš´ í›„ë³´ ì •ë³´) )

ğŸ“¤ Output Gate (ì¶œë ¥ ê²Œì´íŠ¸)

ì—­í• : ì—…ë°ì´íŠ¸ëœ ì…€ ìƒíƒœ($C_t$)ì˜ ì •ë³´ ì¤‘ì—ì„œ ë¬´ì—‡ì„ í˜„ì¬ ì‹œì ì˜ ì¶œë ¥($h_t$)ìœ¼ë¡œ ë‚´ë³´ë‚¼ì§€ ê²°ì •í•©ë‹ˆë‹¤.  
ë™ì‘: ì…€ ìƒíƒœì˜ ì •ë³´ê°€ ëª¨ë‘ ë‹¤ìŒ ì‹œì ìœ¼ë¡œ ì „ë‹¬ë  í•„ìš”ëŠ” ì—†ìŠµë‹ˆë‹¤. í˜„ì¬ ì‹œì ì˜ ì˜ˆì¸¡ì— í•„ìš”í•œ ì •ë³´ë§Œ í•„í„°ë§í•˜ì—¬ ì¶œë ¥í•©ë‹ˆë‹¤.  


> LSTMì€ ë³„ë„ì˜ 'ì…€ ìƒíƒœ'ë¥¼ í†µí•´ ì¥ê¸° ê¸°ì–µì„ ë³´ì¡´í•˜ê³ , 3ê°œì˜ ê²Œì´íŠ¸ë¥¼ í†µí•´ ì´ ê¸°ì–µì„ ì„ íƒì ìœ¼ë¡œ ìŠê±°ë‚˜, ì¶”ê°€í•˜ê±°ë‚˜, ì¶œë ¥í•©ë‹ˆë‹¤.

### KerasëŠ” PyTorchì— ë¹„í•´ ì‚¬ìš©ë²•ì´ ë§¤ìš° ì§ê´€ì ì…ë‹ˆë‹¤.

- Keras(ë³¸ ê³¼ì œì—ì„œ ì‚¬ìš©): model.fit(), model.compile()ì²˜ëŸ¼ ìì£¼ ì“°ì´ëŠ” ê¸°ëŠ¥ë“¤ì´ ë¯¸ë¦¬ ì •ì˜ë˜ì–´ ìˆìŠµë‹ˆë‹¤. ëª¨ë¸ì„ ë¹ ë¥´ê²Œ í•™ìŠµì‹œê³ , ë°ì´í„°ì˜ ì…ë ¥ë¶€í„° ì¶œë ¥ê¹Œì§€ì˜ ì „ì²´ íë¦„ì„ íŒŒì•…í•˜ëŠ” ë° ìœ ë¦¬í•©ë‹ˆë‹¤.

- PyTorch(ì‹¤ìŠµ ì„¸ì…˜ì—ì„œ ì‚¬ìš©í–ˆì—ˆì£ ): ëª¨ë¸ì˜ ì•„ì£¼ ì„¸ë°€í•œ ë™ì‘(ë£¨í”„, ì»¤ìŠ¤í„°ë§ˆì´ì§•, ê°€ì¤‘ì¹˜ ì œì–´ ë“±)ì„ ì§ì ‘ êµ¬í˜„í•´ì•¼ í•  ë•Œê°€ ë§ìŠµë‹ˆë‹¤. ì´ëŠ” ë¯¸ì„¸í•œ íŠœë‹ì´ë‚˜ ì—°êµ¬ ë¶„ì•¼ì—ì„œ ë” ê°•ë ¥í•œ ìœ ì—°ì„±ì„ ë³´ì—¬ì¤ë‹ˆë‹¤.<br><br>


ì´ë²ˆ ê³¼ì œì˜ ëª©í‘œëŠ” **Kerasë¥¼ í†µí•´ RNNê³¼ LSTMì˜ í•µì‹¬ ê°œë…ì„ ë¹ ë¥´ê²Œ êµ¬í˜„í•˜ê³  ê²½í—˜í•˜ëŠ” ê²ƒ**ì…ë‹ˆë‹¤.

KerasëŠ” `layers.SimpleRNN`, `layers.LSTM`, `layers.GRU`ì²˜ëŸ¼ ê°•ë ¥í•œ ìˆœí™˜ ë ˆì´ì–´ë¥¼ ë‹¨ í•œ ì¤„ì˜ ì½”ë“œë¡œ ì œê³µí•©ë‹ˆë‹¤.
- ì˜ˆë¥¼ ë“¤ì–´, LSTM ë ˆì´ì–´ í•˜ë‚˜ë¥¼ ì¶”ê°€í•˜ëŠ” ê²ƒì€ PyTorchì—ì„œëŠ” ì—¬ëŸ¬ ì¤„ì˜ ì„¤ì •ì´ í•„ìš”í•  ìˆ˜ ìˆì§€ë§Œ, Kerasì—ì„œëŠ” ë‹¨ì§€ `layers.LSTM(64)` í•œ ì¤„ì´ë©´ ëë‚©ë‹ˆë‹¤.

ìš°ë¦¬ëŠ” ì´ 'ë¸”ë¡'ë“¤ì„ ê°€ì ¸ì™€ì„œ ì‹¤ì œ ì‹œí€€ìŠ¤ ë°ì´í„°(í…ìŠ¤íŠ¸, ì‹œê³„ì—´ ë“±)ì— ì–´ë–»ê²Œ ì ìš©í•˜ëŠ”ì§€ ê·¸ íë¦„ì„ íŒŒì•…í•˜ëŠ” ë° ì§‘ì¤‘í•  ê²ƒì…ë‹ˆë‹¤.

'Kerasë¡œ RNN/LSTMì„ ì‚¬ìš©í•˜ë‹ˆ (ë¹„êµì ) ê°„ë‹¨í•˜ê²Œ ì‹œí€€ìŠ¤ ëª¨ë¸ì´ ë§Œë“¤ì–´ì§€ëŠ”êµ¬ë‚˜!' í•˜ëŠ” ê²½í—˜ê³¼ ì§ê´€ì„ ì–»ì–´ê°€ëŠ” ì‹œê°„ìœ¼ë¡œ ìƒê°í•˜ê³ , ì¦ê²ê²Œ ì½”ë“œë¥¼ ì±„ì›Œì£¼ì„¸ìš”!

> ê¸°ë³¸ Keras RNN ë ˆì´ì–´ ì‚¬ìš©ë²•

```python
# Kerasë¡œ ë§Œë“œëŠ” LSTM ë ˆì´ì–´ ì˜ˆì‹œ
# Kerasì˜ RNN ë ˆì´ì–´ëŠ” (batch_size, timesteps, features) í˜•íƒœì˜ ì…ë ¥ì„ ê¸°ëŒ€í•©ë‹ˆë‹¤.

# ê°€ìƒì˜ ì‹œí€€ìŠ¤ ë°ì´í„° ìƒì„±
# ì˜ˆ: 32ê°œì˜ ë¬¸ì¥(batch_size), ê° ë¬¸ì¥ì€ 10ê°œì˜ ë‹¨ì–´(timesteps), ê° ë‹¨ì–´ëŠ” 8ì°¨ì›ì˜ ë²¡í„°(features)
input_shape = (10, 8)
dummy_input = np.random.rand(32, *input_shape)

# LSTM ë ˆì´ì–´ ìƒì„± (64ê°œì˜ ìœ ë‹›)
lstm_layer = layers.LSTM(64)
output = lstm_layer(dummy_input)

print(f"ì…ë ¥ í¬ê¸°: {dummy_input.shape}")
print(f"LSTM ì¶œë ¥ í¬ê¸°: {output.shape}") # ë§ˆì§€ë§‰ íƒ€ì„ìŠ¤í…ì˜ ì¶œë ¥ë§Œ ë°˜í™˜

# return_sequences=True: ëª¨ë“  íƒ€ì„ìŠ¤í…ì˜ ì¶œë ¥ì„ ë°˜í™˜ (RNNì„ ì¸µì¸µì´ ìŒ“ì„ ë•Œ í•„ìš”)
lstm_layer_seq = layers.LSTM(64, return_sequences=True)
output_seq = lstm_layer_seq(dummy_input)
print(f"LSTM (seq=True) ì¶œë ¥ í¬ê¸°: {output_seq.shape}")
```

ê°€ì¥ ê³ ì „ì ì¸ NLP(ìì—°ì–´ ì²˜ë¦¬) ê³¼ì œ ì¤‘ í•˜ë‚˜ì…ë‹ˆë‹¤. ì˜í™” ë¦¬ë·° í…ìŠ¤íŠ¸ë¥¼ ë³´ê³  ê¸ì •/ë¶€ì •ì„ ë¶„ë¥˜í•©ë‹ˆë‹¤.

**1. Embedding (ì„ë² ë”©) ë ˆì´ì–´**  
- ì™œ í•„ìš”í• ê¹Œìš”?  
ì»´í“¨í„°ëŠ” 'good', 'bad' ê°™ì€ í…ìŠ¤íŠ¸ ë‹¨ì–´ë¥¼ ì§ì ‘ ì´í•´í•  ìˆ˜ ì—†ìŠµë‹ˆë‹¤. ëª¨ë“  ê²ƒì„ ìˆ«ìë¡œ ë°”ê¿”ì£¼ì–´ì•¼ í•©ë‹ˆë‹¤.

  > good (ë‹¨ì–´) â” 5 (ì¸ë±ìŠ¤) â” [0.1, 0.5, -0.2, ... , 0.9] (128ì°¨ì› ë²¡í„°)

**2. Bidirectional LSTM (ì–‘ë°©í–¥ LSTM)**  
- ì™œ ê·¸ëƒ¥ LSTMì´ ì•„ë‹ˆë¼ 'ì–‘ë°©í–¥'ì„ ì“¸ê¹Œìš”?  
RNNê³¼ LSTMì€ í…ìŠ¤íŠ¸ì™€ ê°™ì€ ì‹œí€€ìŠ¤ ë°ì´í„°ë¥¼, ì•ì—ì„œë¶€í„°(ì™¼ìª½ì—ì„œ ì˜¤ë¥¸ìª½ìœ¼ë¡œ) ì°¨ë¡€ëŒ€ë¡œ ì½ìœ¼ë©° 'ê¸°ì–µ'(Context)ì„ ìŒ“ì•„ê°‘ë‹ˆë‹¤.

  **í•˜ì§€ë§Œ ë¬¸ì¥ì˜ ì§„ì§œ ì˜ë¯¸ëŠ” ë’¤ì— ë‚˜ì˜¤ëŠ” ë‹¨ì–´ì— ì˜í•´ ê²°ì •ë  ë•Œê°€ ë§ìŠµë‹ˆë‹¤.**

  ì˜ˆì‹œ 1: "This movie was not bad at all."

  "bad"ë§Œ ë³´ë©´ ë¶€ì •ì ì´ì§€ë§Œ, ì•ì— "not"ì´ ìˆìŠµë‹ˆë‹¤.

  ì˜ˆì‹œ 2: "The visual effects were stunning, but the story was terrible."

  "stunning" ë•Œë¬¸ì— ê¸ì •ìœ¼ë¡œ íŒë‹¨í•˜ë‹¤ê°€, ë§ˆì§€ë§‰ "terrible"ì„ ë³´ê³  ë¶€ì •ìœ¼ë¡œ ê²°ë¡  ë‚´ë ¤ì•¼ í•©ë‹ˆë‹¤.

- ì–´ë–»ê²Œ ì‘ë™í•˜ë‚˜ìš”?  
Bidirectional ë˜í¼(Wrapper)ëŠ” LSTM ë ˆì´ì–´ë¥¼ ë‘ ê°œ ë§Œë“­ë‹ˆë‹¤.

  - ì •ë°©í–¥ (Forward) LSTM: í…ìŠ¤íŠ¸ë¥¼ "I loved this movie" ìˆœì„œ (ì• -> ë’¤)ë¡œ ì½ìŠµë‹ˆë‹¤.

  - ì—­ë°©í–¥ (Backward) LSTM: í…ìŠ¤íŠ¸ë¥¼ "movie this loved I" ìˆœì„œ (ë’¤ -> ì•)ë¡œ ì½ìŠµë‹ˆë‹¤.

  ëª¨ë¸ì€ ì´ ë‘ ë°©í–¥ì˜ 'ê¸°ì–µ(ì¶œë ¥)'ì„ ëª¨ë‘ í•©ì³ì„œ ìµœì¢… ê²°ì •ì„ ë‚´ë¦½ë‹ˆë‹¤.

> ê·¸ ê²°ê³¼, ëª¨ë¸ì€ ê° ë‹¨ì–´ì˜ 'ì•ë’¤ ë¬¸ë§¥'ì„ ì™„ë²½í•˜ê²Œ íŒŒì•…í•  ìˆ˜ ìˆê²Œ ë˜ì–´, í›¨ì”¬ ë” ì •í™•í•˜ê²Œ ê¸ì •/ë¶€ì •ì„ ë¶„ë¥˜í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

```python
# 1. ë°ì´í„° ë¡œë“œ ë° ì „ì²˜ë¦¬
max_features = 20000  # [Vocabulary Size] ì‚¬ìš©í•  ë¹ˆë„ìˆ˜ ìƒìœ„ ë‹¨ì–´ì˜ ê°œìˆ˜. ë„ˆë¬´ í¬ë©´ í¬ì†Œì„± ë¬¸ì œê°€, ë„ˆë¬´ ì‘ìœ¼ë©´ ì •ë³´ ì†ì‹¤ì´ ë°œìƒí•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.
maxlen = 200          # [Sequence Length] ê° ë¦¬ë·°ì˜ ìµœëŒ€ ê¸¸ì´. ì´ë³´ë‹¤ ê¸¸ë©´ ìë¥´ê³ (truncating), ì§§ìœ¼ë©´ 0ìœ¼ë¡œ ì±„ì›ë‹ˆë‹¤(padding).

(X_train, y_train), (X_val, y_val) = keras.datasets.imdb.load_data(
    num_words=max_features
)
print(f"{len(X_train)} í›ˆë ¨ ì‹œí€€ìŠ¤, {len(X_val)} ê²€ì¦ ì‹œí€€ìŠ¤")

# ì‹œí€€ìŠ¤ ê¸¸ì´ë¥¼ ë§ì¶”ê¸° (Padding)
X_train = keras.preprocessing.sequence.pad_sequences(X_train, maxlen=maxlen)
X_val = keras.preprocessing.sequence.pad_sequences(X_val, maxlen=maxlen)
```

```python
# 2. ëª¨ë¸ êµ¬ì¶• (Bidirectional LSTM)
embedding_dim = 128  # [Vector Size] ì„ë² ë”© í•  ë•Œ ê° ë‹¨ì–´ë¥¼ ëª‡ ì°¨ì›ì˜ ë²¡í„°ë¡œ í‘œí˜„í• ì§€ ê²°ì •í•©ë‹ˆë‹¤. ë³´í†µ 50~300 ì‚¬ì´ë¥¼ ë§ì´ ì‚¬ìš©í•©ë‹ˆë‹¤.
lstm_units = 64      # [Model Capacity] LSTM ì…€ ë‚´ë¶€ì˜ ì€ë‹‰ ìƒíƒœ ë²¡í„° í¬ê¸°ì…ë‹ˆë‹¤. í´ìˆ˜ë¡ ë” ë³µì¡í•œ íŒ¨í„´ì„ í•™ìŠµí•  ìˆ˜ ìˆì§€ë§Œ, ê³¼ì í•© ìœ„í—˜ë„ ì»¤ì§‘ë‹ˆë‹¤.

# Input ë ˆì´ì–´
inputs = keras.Input(shape=(None,), dtype="int32")

 # Embedding ë ˆì´ì–´: ë‹¨ì–´ë¥¼ ê³ ì°¨ì›ì˜ ë²¡í„°ë¡œ ë³€í™˜í•©ë‹ˆë‹¤. (ì˜ˆ: 'good' -> [0.1, 0.5, -0.2, ...])
x = layers.Embedding(max_features, embedding_dim)(inputs)

# Bidirectional ì ìš©: RNNì´ ì •ë°©í–¥ê³¼ ì—­ë°©í–¥ì˜ í…ìŠ¤íŠ¸ë¥¼ ëª¨ë‘ ì½ìŠµë‹ˆë‹¤.
x = layers.Bidirectional(layers.LSTM(lstm_units, return_sequences=True))(x) # return_sequences=True: ëª¨ë“  íƒ€ì„ìŠ¤í…ì˜ ì¶œë ¥ì„ ë°˜í™˜í•˜ëŠ” ì„¤ì •ì…ë‹ˆë‹¤.
x = layers.Bidirectional(layers.LSTM(lstm_units))(x)

# Dense: ì¶œë ¥ ë ˆì´ì–´
outputs = layers.Dense(1, activation='sigmoid')(x) # ê¸ì •ì  ë¦¬ë·°(1)/ë¶€ì •ì  ë¦¬ë·°(0) ì´ì§„ ë¶„ë¥˜ë¥¼ í•˜ê¸° ë•Œë¬¸ì— í•„ìš”í•©ë‹ˆë‹¤.
                                             # íŒíŠ¸: êµìœ¡ì„¸ì…˜ì˜ [ë¡œì§€ìŠ¤í‹± íšŒê·€]ì—ì„œ ë°°ì› ë˜ í™œì„±í™” í•¨ìˆ˜ì…ë‹ˆë‹¤!

# my_model = keras.Model(inputs(ì¸í’‹ ë ˆì´ì–´), outputs(ì¶œë ¥ ë ˆì´ì–´))
model_imdb = keras.Model(inputs, outputs)
model_imdb.summary()
```

```python
# 3. ëª¨ë¸ ì»´íŒŒì¼ ë° í•™ìŠµ (ì´ ë¶€ë¶„ì€ ì¢€ ì˜¤ë˜ ê±¸ë¦½ë‹ˆë‹¤)
model_imdb.compile(optimizer='adam', # íŒíŠ¸: RMSPropì— ëª¨ë©˜í…€ì´ í•©ì³ì§„ ì˜µí‹°ë§ˆì´ì €!
                   loss="binary_crossentropy", # ì´ì§„ ë¶„ë¥˜ ë¬¸ì œì—ì„œëŠ” ì†ì‹¤ í•¨ìˆ˜ë¡œ "binary_crossentropy"ë¥¼ ì‚¬ìš©í•©ë‹ˆë‹¤.
                   metrics=["accuracy"])       # ëª¨ë¸ì˜ ì„±ëŠ¥ì„ í‰ê°€í•˜ëŠ” ë° ì‚¬ìš©ë˜ëŠ” ì²™ë„ë¥¼ metrics íŒŒë¼ë¯¸í„°ì— ì§€ì •í•©ë‹ˆë‹¤.

# íŠ¸ë ˆì¸ ì…‹ì„ í•™ìŠµì‹œì¼œ ë´…ì‹œë‹¤.
imdb_fit = model_imdb.fit(X_train, y_train, batch_size=128, epochs=10, validation_data=(X_val, y_val))
# íŒíŠ¸: ë¬¸ì œì§‘ì„ 'ëª‡ ë²ˆ' í’€ ì§€ ì •í•˜ëŠ” íŒŒë¼ë¯¸í„°
```

```python
# ëª¨ë¸ì˜ í›ˆë ¨ ê³¼ì •(loss, accuracy)ì„ 'history' ê°ì²´ë¥¼ ì‚¬ìš©í•˜ì—¬ ì‹œê°í™”í•´ë³´ì„¸ìš”.
plt.plot(imdb_fit.history['loss'], label='loss')
plt.plot(imdb_fit.history['accuracy'], label='accuracy')
plt.legend()

plt.show()
```

Bidirectional LSTMì„ í†µí•´ ê°ì„± ë¶„ì„ ëª¨ë¸ì„ í•™ìŠµì‹œì¼°ìŠµë‹ˆë‹¤.  
ê·¸ëŸ¼ ì´ ëª¨ë¸ë¡œ ì§ì ‘ ì˜ì–´ë¡œ ì˜í™”ë¦¬ë·°ë¥¼ ì‘ì„±í•œ ë’¤ ê¸ì •/ë¶€ì •ì„ ì˜ˆì¸¡í•´ ë³¼ê¹Œìš”?

```python
# 1. IMDB ë°ì´í„°ì…‹ì˜ 'ë‹¨ì–´-ì¸ë±ìŠ¤' ë”•ì…”ë„ˆë¦¬ ë¡œë“œ
# (ê° ë‹¨ì–´ê°€ ì–´ë–¤ ìˆ«ìì— ë§¤í•‘ë˜ì—ˆëŠ”ì§€ ì •ë³´ê°€ í•„ìš”í•©ë‹ˆë‹¤)
word_index = keras.datasets.imdb.get_word_index()

# 2. Kerasì˜ ì˜ˆì•½ ì¸ë±ìŠ¤ë¥¼ ë§ì¶°ì£¼ê¸° ìœ„í•œ offset
# (0: íŒ¨ë”©, 1: ë¬¸ì¥ ì‹œì‘, 2: ëª¨ë¥´ëŠ” ë‹¨ì–´)
offset = 3
word_index = {k:(v+offset) for k,v in word_index.items()}
word_index["<pad>"] = 0
word_index["<start>"] = 1
word_index["<unk>"] = 2

# 3. ë¬¸ì¥ì„ -> ìˆ«ì ì‹œí€€ìŠ¤ë¡œ ë³€í™˜í•˜ëŠ” í•¨ìˆ˜
def preprocess_text(text):
    # í…ìŠ¤íŠ¸ë¥¼ ì†Œë¬¸ìë¡œ ë°”ê¾¸ê³  ë‹¨ì–´ ë‹¨ìœ„ë¡œ ë¶„ë¦¬ (ê°„ë‹¨í•œ í† í°í™”)
    tokens = text.lower().split()

    # ë‹¨ì–´ë¥¼ ì¸ë±ìŠ¤ë¡œ ë³€í™˜ (word_indexì— ì—†ìœ¼ë©´ 2 ('<unk>')ë¡œ ì²˜ë¦¬)
    sequence = [word_index.get(word, 2) for word in tokens]

    # 20000 ('max_features')ë³´ë‹¤ í° ì¸ë±ìŠ¤ëŠ” ëª¨ë¥´ëŠ” ë‹¨ì–´ë¡œ ì²˜ë¦¬ (2)
    sequence = [idx if idx < max_features else 2 for idx in sequence]
    return sequence
```

```python
# 4. ì˜ˆì¸¡í•´ ë³¼ ìƒˆë¡œìš´ í…ŒìŠ¤íŠ¸ ë¬¸ì¥
test_sentence = "I loved this movie, it was fantastic and really great!" # ì˜ì–´ë¡œ ëœ ë¦¬ë·°ë¥¼ ì‘ì„±í•´ ë³´ì„¸ìš”!
# ì˜ˆ: "I loved this movie, it was fantastic and really great!""
# test_sentence = "This movie was terrible, I hated it." # ì‹œê°„ ë‚¨ìœ¼ë©´ ë¶€ì •ì ì¸ ë¬¸ì¥ë„ í…ŒìŠ¤íŠ¸í•´ë³´ì„¸ìš”
```

```python
# 5. ë¬¸ì¥ ì „ì²˜ë¦¬ ë° íŒ¨ë”©
tokenized_sentence = preprocess_text(test_sentence)
# (1, maxlen=200) í˜•íƒœë¡œ ë§Œë“¤ê¸° (1ê°œì˜ ìƒ˜í”Œ)
padded_sentence = keras.preprocessing.sequence.pad_sequences([tokenized_sentence], maxlen=maxlen)

print(f"ì›ë³¸ ë¬¸ì¥: {test_sentence}")
print(f"ë³€í™˜ëœ ì‹œí€€ìŠ¤ (ì¼ë¶€): {tokenized_sentence[:10]}...")
print(f"íŒ¨ë”©ëœ ì‹œí€€ìŠ¤ shape: {padded_sentence.shape}")

# 6. ëª¨ë¸ë¡œ ì˜ˆì¸¡!
prediction = model_imdb.predict(padded_sentence)
score = prediction[0][0]

print(f"\n--- ì˜ˆì¸¡ ê²°ê³¼ ---")
print(f"ì˜ˆì¸¡ ì ìˆ˜ (Sigmoid): {score:.4f}")

if score > 0.5:
    print(f"ê²°ê³¼: ê¸ì •ì ì¸ ë¦¬ë·°ì…ë‹ˆë‹¤. ({score*100:.2f}%)")
else:
    print(f"ê²°ê³¼: ë¶€ì •ì ì¸ ë¦¬ë·°ì…ë‹ˆë‹¤. ({(1-score)*100:.2f}%)")
```

ê¸°ì¡´ì— ì“°ë˜ Cross-Sectional ë°ì´í„°ì²˜ëŸ¼ ëœë¤ ë‚œìˆ˜ë¡œ ë¶„ë¦¬í•˜ì§€ ì•ŠìŠµë‹ˆë‹¤.

```python
# 1. ì‚¬ìš©í•  í”¼ì³ì™€ ì¢…ì†ë³€ìˆ˜(íƒ€ê²Ÿ) ì •ì˜
# 14ê°œì˜ íŠ¹ì§• ì¤‘ ì¼ë¶€ë§Œ ì‚¬ìš©í•  ìˆ˜ë„ ìˆìŠµë‹ˆë‹¤. (ì´ë²ˆì—ëŠ” ëª¨ë‘ ì‚¬ìš©)
X = df.drop('T (degC)', axis=1) # ì˜¨ë„ columnì„ ì œì™¸í•œ ë‚˜ë¨¸ì§€
y = df['T (degC)'] # ì˜¨ë„ë¥¼ ì˜ˆì¸¡
```

```python
# 2. ë°ì´í„° ë¶„ë¦¬ (ìˆœì„œëŒ€ë¡œ ìë¥´ê¸°)
# 70% í›ˆë ¨, 20% ê²€ì¦, 10% í…ŒìŠ¤íŠ¸ ë§Œí¼ ì‚¬ìš©í•  ê²ƒì…ë‹ˆë‹¤.
# [í›ˆë ¨ ë°ì´í„°]ì˜ ë¶„ëŸ‰ì…ë‹ˆë‹¤.
split_fraction = 0.7
# [í›ˆë ¨ ë°ì´í„°+ê²€ì¦ ë°ì´í„°]ì˜ ë¶„ëŸ‰ì…ë‹ˆë‹¤.
val_split_fraction = 0.9 # íŒíŠ¸: 70% + 20%

# ì´í›„ì— ìŠ¬ë¼ì´ì‹±ì„ ìœ„í•´ (ì „ì²´ ê¸¸ì´ ì •ìˆ˜)X(í¼ì„¼íŠ¸)ì˜ ì •ìˆ˜í˜•ìœ¼ë¡œ ë°”ê¿”ì¤„ê²Œìš”.
n = len(df)
train_split = int(n * split_fraction)
val_split = int(n * val_split_fraction)
```

```python
# 3. ë°ì´í„° ì •ê·œí™”
# í›ˆë ¨ ì…‹(train_data)ì˜ í‰ê· ê³¼ í‘œì¤€í¸ì°¨ë¡œ ëª¨ë“  ë°ì´í„°ë¥¼ ì •ê·œí™”í•©ë‹ˆë‹¤.
# (ê²€ì¦/í…ŒìŠ¤íŠ¸ ë°ì´í„°ì˜ ì •ë³´ê°€ í›ˆë ¨ ê³¼ì •ì— ëˆ„ìˆ˜ë˜ëŠ” ê²ƒì„ ë§‰ê¸° ìœ„í•¨)
train_mean = X.iloc[:train_split].mean(axis=0)
train_std = X.iloc[:train_split].std(axis=0)

# ì •ê·œí™”ëœ í”¼ì³ ìƒì„±
X_features_normalized = (X - train_mean) / train_std

# DataFrame -> Numpy arrayë¡œ ë³€í™˜
X_features_data = X_features_normalized.values
y_data = y.values
```

```python
# 4. ìµœì¢… ë°ì´í„° ë¶„ë¦¬ (Numpy array)
X_train = X_features_data[:train_split]
y_train = y_data[:train_split]

X_val = X_features_data[train_split:val_split]
y_val = y_data[train_split:val_split]

X_test = X_features_data[val_split:]
y_test = y_data[val_split:]

print(f"x_train shape: {X_train.shape}") # (ìƒ˜í”Œ ìˆ˜, 13ê°œ í”¼ì³)
print(f"y_train shape: {y_train.shape}")   # (ìƒ˜í”Œ ìˆ˜,)
```

3-3. ë°ì´í„° ìœˆë„ì‰ (Windowing) - Keras ìœ í‹¸ë¦¬í‹° ì‚¬ìš©  
KerasëŠ” ì‹œê³„ì—´ ìœˆë„ì‰ì„ ìœ„í•œ `timeseries_dataset_from_array`ë¼ëŠ” ê°•ë ¥í•œ ìœ í‹¸ë¦¬í‹°ë¥¼ ì œê³µí•©ë‹ˆë‹¤.

> ìš°ë¦¬ì˜ ëª©í‘œ<br>  
ì…ë ¥ (X): ê³¼ê±° 5ì¼ (120ì‹œê°„) ë™ì•ˆì˜ ë°ì´í„° (sequence_length=120)  
íƒ€ê²Ÿ (Y): ì…ë ¥ ì‹œí€€ìŠ¤ê°€ ëë‚œ ì‹œì ìœ¼ë¡œë¶€í„° 6ì‹œê°„ ë’¤ì˜ ê¸°ì˜¨ (delay=6)

```python
# 1. ìœˆë„ìš° í•˜ì´í¼íŒŒë¼ë¯¸í„° ì •ì˜
# (1) ì…ë ¥ ì‹œí€€ìŠ¤ ê¸¸ì´ (ê³¼ê±°ë¥¼ ì–¼ë§ˆë‚˜ ë³¼ ê²ƒì¸ê°€?)
sequence_length = 120 # 120ì‹œê°„ = 5ì¼ì¹˜ ë°ì´í„°ë¥¼ ë³´ê³  ë¯¸ë˜ë¥¼ ì˜ˆì¸¡í•˜ê² ë‹¤ëŠ” ì˜ë¯¸ì…ë‹ˆë‹¤.

# (2) ì˜ˆì¸¡ ì§€ì  (ë¯¸ë˜ ì–¼ë§ˆ ë’¤ë¥¼ ì˜ˆì¸¡í•  ê²ƒì¸ê°€?)
# í˜„ì¬ ì‹œì ìœ¼ë¡œë¶€í„° 6ìŠ¤í…(ì•½ 6ì‹œê°„) ë’¤ì˜ ì˜¨ë„ë¥¼ (ì˜ˆì¸¡ì˜) íƒ€ê²Ÿìœ¼ë¡œ ì„¤ì •í•©ë‹ˆë‹¤.
delay = 6

# (3) ë°°ì¹˜ í¬ê¸°
# ì‹œê³„ì—´ ë°ì´í„°ëŠ” ë§ê¸° ë•Œë¬¸ì—(LSTM ì‹œê³„ì—´ì„ í•  ë•ŒëŠ” ë°ì´í„° ì–‘ ìì²´ê°€ ë§ìœ¼ë©´ ì¢‹ì•„ìš”) ë³´í†µ 256, 512 ë“± í¬ê²Œ ì¡ìŠµë‹ˆë‹¤.
batch_size = 256

#ìŠ¬ë¼ì´ì‹±ìœ¼ë¡œ 6ì‹œê°„ì˜ ë”œë ˆì´ë¥¼ ì§ì ‘ ë§Œë“¤ì–´ì£¼ëŠ” ì‘ì—…
X_train_delayed = X_train[:-delay]
y_train_delayed = y_train[delay:]
# val setì—ë„ ë˜‘ê°™ì´ ìŠ¬ë¼ì´ì‹±ì„ í•´ì¤ë‹ˆë‹¤.
X_val_delayed = X_val[:-delay]
y_val_delayed = y_val[delay:]
```

```python
# 2. Keras ìœˆë„ìš° ìƒì„± ìœ í‹¸ë¦¬í‹° ì‚¬ìš©
train_dataset = keras.utils.timeseries_dataset_from_array(
    data=X_train_delayed,    # ë”œë ˆì´ ì ìš©ëœ ì…‹ì„ ì¨ì•¼ í•©ë‹ˆë‹¤.
    targets=y_train_delayed, # íƒ€ê²Ÿì—ë„ ë”œë ˆì´ ì ìš©ëœ ì…‹ì„ ì¨ì•¼ í•©ë‹ˆë‹¤.
    sequence_length=sequence_length,
    sampling_rate=1,
    sequence_stride=1, # ìƒ˜í”Œë§ ê°„ê²© (ì–¼ë§ˆë‚˜ ì´˜ì´˜íˆ ë³¼ ê²ƒì¸ê°€?) # sequence_stride=1: 1ì‹œê°„ ê°„ê²©ì„ ëª¨ë‘ ë´…ë‹ˆë‹¤.  sequence_stride=6: 6ì‹œê°„ ê°„ê²©ìœ¼ë¡œ ë´…ë‹ˆë‹¤.
    batch_size=batch_size,
    shuffle=True # íŠ¸ë ˆì¸ ì…‹ì´ë¯€ë¡œ ìœˆë„ìš° ìˆœì„œë¥¼ ì„ìŒ
)

# 3. ê²€ì¦ ë°ì´í„°ì…‹ (shuffle=False)
val_dataset = keras.utils.timeseries_dataset_from_array(
    data=X_val_delayed,    # ë˜‘ê°™ì´ ë”œë ˆì´ ì ìš©ëœ ì…‹ì„ ì¨ì•¼ í•©ë‹ˆë‹¤.
    targets=y_val_delayed, # ë˜‘ê°™ì´ ë”œë ˆì´ ì ìš©ëœ ì…‹ì„ ì¨ì•¼ í•©ë‹ˆë‹¤.
    sequence_length=sequence_length,
    sampling_rate=1,
    sequence_stride=1,
    batch_size=batch_size,
    shuffle=False # ê²€ì¦ ì…‹ì€ ìœˆë„ìš° ìˆœì„œë¥¼ ì„ì§€ ì•ŠìŒ
)
```

```python
# 4. ìƒì„±ëœ ìœˆë„ìš°ì˜ Shape í™•ì¸
for batch in train_dataset.take(1):
    inputs, targets = batch
    print(f"Input shape (X): {inputs.shape}")   # (Batch, 120, 13) -> (ë°°ì¹˜, ì‹œí€€ìŠ¤ ê¸¸ì´, í”¼ì³ ìˆ˜)
    print(f"Target shape (y): {targets.shape}") # (Batch,) -> (ë°°ì¹˜,)
```


3-4. LSTM ëª¨ë¸ êµ¬ì¶• ë° í•™ìŠµ  
ì´ì œ (120, 13) í˜•íƒœì˜ ì…ë ¥ì„ ë°›ì•„ (1,) í˜•íƒœì˜ ì¶œë ¥ì„ ë‚´ë³´ë‚´ëŠ” ëª¨ë¸ì„ ë§Œë“­ë‹ˆë‹¤.  
ì—¬ê¸°ì„œëŠ” ì¶”ê°€ë¡œ EarlyStoppingì´ë¼ëŠ” íˆ´ì„ ì¨ ë³¼ê²Œìš”.

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
# 1. ëª¨ë¸ êµ¬ì¶• (LSTM ì‚¬ìš©)
inputs = keras.Input(shape=(sequence_length, X_train.shape[-1])) # (120, 13)

# GRU(32) ë˜ëŠ” LSTM(32) ë ˆì´ì–´ë¥¼ ì‚¬ìš©í•˜ì„¸ìš”.
# ì‹œí€€ìŠ¤ì˜ ë§ˆì§€ë§‰ ì¶œë ¥ë§Œ í•„ìš”í•˜ë¯€ë¡œ, [return_sequences=False]ë¡œ ì„¤ì •í•˜ê² ìŠµë‹ˆë‹¤. (ê¸°ë³¸ê°’ì´ False ì…ë‹ˆë‹¤.)
x = layers.LSTM(32, activation='tanh', return_sequences=False)(inputs)

# Dropout ë ˆì´ì–´: ê³¼ì í•© ë°©ì§€
x = layers.Dropout(0.2)(x)
# 0.2ëŠ” Dropout rateë¡œ, í•™ìŠµ ì‹œ ëœë¤í•˜ê²Œ ì „ì²´ ë‰´ëŸ°ì˜ 20%ë¥¼ êº¼ì„œ ê³¼ì í•©ì„ ë°©ì§€í•©ë‹ˆë‹¤.

# ì¶œë ¥ì¸µ (Dense)
# 6ì‹œê°„ ë’¤ì˜ 'ì˜¨ë„' 1ê°œë§Œ ì˜ˆì¸¡í•˜ë©´ ë˜ë¯€ë¡œ units=1
# ì˜¨ë„ ê·¸ ìì²´ë¥¼ ì˜ˆì¸¡í•˜ëŠ” íšŒê·€ ë¬¸ì œì´ê¸° ë•Œë¬¸ì—, í™œì„±í™” í•¨ìˆ˜ë¥¼ ë„£ì§€ ì•ŠìŠµë‹ˆë‹¤
outputs = layers.Dense(1)(x)

model_lstm = keras.Model(inputs, outputs)
```

```python
# 2. ëª¨ë¸ ì»´íŒŒì¼ ('íšŒê·€' ë¬¸ì œ)
# ì†ì‹¤(Loss): Mean Squared Error (í‰ê·  ì œê³± ì˜¤ì°¨)
# ì§€í‘œ(Metric): Mean Absolute Error (í‰ê·  ì ˆëŒ€ ì˜¤ì°¨)
model_lstm.compile(optimizer='adam',
                   loss='mean_squared_error',        # íŒíŠ¸: ì†ì‹¤í•¨ìˆ˜ë¥¼ ì§€ì •í•˜ëŠ” íŒŒë¼ë¯¸í„°
                   metrics=['mean_absolute_error']) # ëª¨ë¸ì˜ ì„±ëŠ¥ í‰ê°€ ì§€í‘œë¥¼ ì§€ì •í•˜ëŠ” íŒŒë¼ë¯¸í„°

model_lstm.summary()
```

```python
# ê³¼ì í•©ì„ ë§‰ê¸° ìœ„í•œ EarlyStopping ê°ì²´
# val_lossê°€ ë” ì´ìƒ ì¢‹ì•„ì§€ì§€ ì•Šê³  ëª‡ ë²ˆ ì—°ì†ìœ¼ë¡œ ë‚˜ë¹ ì§€ë©´, 100ë²ˆì„ ë‹¤ ì±„ìš°ê¸° ì „ì— í›ˆë ¨ì„ "ì¡°ê¸° ì¢…ë£Œ" í•˜ëŠ” ê³¼ì í•© ë°©ì§€ ê¸°ë²•ì…ë‹ˆë‹¤.

early_stopping_callback = EarlyStopping(
    monitor='val_loss', # 'monitor' : ì–´ë–¤ ì§€í‘œë¥¼ ê°ì‹œí•˜ë©´ì„œ ë©ˆì¶œì§€?
    patience=10,        # 'patience' : ëª‡ epoch ë™ì•ˆ ì ìˆ˜ê°€ ë‚˜ì•„ì§€ì§€ ì•Šìœ¼ë©´ ë©ˆì¶œì§€?
    restore_best_weights=True # 'restore_best_weights'=True : í›ˆë ¨ì´ ë©ˆì·„ì„ ë•Œ, ê°€ì¥ ì¢‹ì•˜ë˜ ì‹œì ì˜ ê°€ì¤‘ì¹˜ë¡œ ë³µì›í•œë‹¤ëŠ” ëœ»ì…ë‹ˆë‹¤.
)
```

```python
# 3. ëª¨ë¸ í•™ìŠµ (ì¢€ ë” ì˜¤ë˜ ê±¸ë¦½ë‹ˆë‹¤)
history = model_lstm.fit(
    train_dataset,        # 2. Keras ìœˆë„ìš° ìƒì„± ìœ í‹¸ë¦¬í‹° ì‚¬ìš©ì—ì„œ ë§Œë“  train_datasetì„ ê·¸ëŒ€ë¡œ ë„£ì„ ìˆ˜ ìˆìŠµë‹ˆë‹¤.
    epochs=100,           # ì¼ë‹¨ 100 epochë¥¼ ëŒë¦½ë‹ˆë‹¤. í•˜ì§€ë§Œ earlystopping ê¸°ì¤€ì— ê±¸ë¦¬ë©´ 100ë²ˆ ì „ì— ë©ˆì¶œ ê²ƒì…ë‹ˆë‹¤.
    validation_data=val_dataset, # 2. Keras ìœˆë„ìš° ìƒì„± ìœ í‹¸ë¦¬í‹° ì‚¬ìš©ì—ì„œ ë§Œë“  val_dataset
    callbacks=[early_stopping_callback]
)
```

í•™ìŠµëœ ëª¨ë¸ì˜ val_mae (ê²€ì¦ ë°ì´í„°ì˜ MAE)ë¥¼ í™•ì¸í•˜ê³ , í•™ìŠµ ê³¡ì„ ì„ ê·¸ë¦½ë‹ˆë‹¤.

```python
# 1. í•™ìŠµ ê³¼ì • ì‹œê°í™” (MAE)
plt.plot(history.history['mean_absolute_error'], label='Train MAE')
plt.plot(history.history['val_mean_absolute_error'], label='Validation MAE')
plt.title('Model MAE (í‰ê·  ì ˆëŒ€ ì˜¤ì°¨)')
plt.legend()
plt.show()

# 2. ê²°ê³¼ ë¶„ì„
print(f"ìµœì¢… ê²€ì¦ MAE: {history.history['val_mean_absolute_error'][-1]:.4f} (degC)")
```

```python
# í…ŒìŠ¤íŠ¸ ì…‹ì—ë„ ì ìš©í•´ ë´…ì‹œë‹¤!

# í…ŒìŠ¤íŠ¸ ì…‹ì—ë„ ë”œë ˆì´(6ì‹œê°„)ë¥¼ ìŠ¬ë¼ì´ì‹±ìœ¼ë¡œ ì ìš©í•©ë‹ˆë‹¤.
X_test_delayed = X_test[:-delay]
y_test_delayed = y_test[delay:]

# --- 'test_dataset'ìƒì„± ---
# val_datasetê³¼ ë™ì¼í•œ ë°©ì‹ìœ¼ë¡œ ë§Œë“¤ë˜, 'shuffle=False'ê°€ ë§¤ìš° ì¤‘ìš”í•©ë‹ˆë‹¤!
# (ìˆœì„œê°€ ì„ì´ë©´ ì‹¤ì œê°’ê³¼ ì˜ˆì¸¡ê°’ì„ ê·¸ë˜í”„ë¡œ ë¹„êµí•  ìˆ˜ ì—†ìŠµë‹ˆë‹¤)
test_dataset = keras.utils.timeseries_dataset_from_array(
    data=X_test_delayed, # ë”œë ˆì´ ì ìš©ëœ X_test
    targets=y_test_delayed, # ë”œë ˆì´ ì ìš©ëœ y_test
    sequence_length=sequence_length,
    sampling_rate=1,
    sequence_stride=1,
    batch_size=batch_size,   # (val_datasetê³¼ ë™ì¼í•œ ë°°ì¹˜ í¬ê¸°)
    shuffle=False
)
```

```python
# í…ŒìŠ¤íŠ¸ ì…‹ì„ í†µí•´ ì˜ˆì¸¡í•œ y_predë¥¼ ì„¤ì •í•©ë‹ˆë‹¤.
y_pred = model_lstm.predict(test_dataset)
```

# 4. ê°„ë‹¨í•œ Character-Level Seq2Seq ëª¨ë¸ ì˜ˆì œ

ìš°ë¦¬ê°€ í•˜ë ¤ëŠ” "hello" â” "HELLO" ë³€í™˜ë„, "ì˜ì–´ ë¬¸ì¥" â” "í”„ë‘ìŠ¤ì–´ ë¬¸ì¥" ê¸°ê³„ ë²ˆì—­ë„ ëª¨ë‘ Seq2Seq ëª¨ë¸ì…ë‹ˆë‹¤.

ì´ ê³¼ì œì—ì„œëŠ” ë¬¸ì ë‹¨ìœ„(Character-Level)ë¡œ ì´ êµ¬ì¡°ë¥¼ ê²½í—˜í•´ ë´…ë‹ˆë‹¤.

**1. ì¸ì½”ë” (Encoder): ë¬¸ì¥ ì••ì¶•ê¸° ğŸ§ **  
ì¸ì½”ë”ì˜ ì—­í• ì€ ì…ë ¥ ë¬¸ì¥("hello")ì˜ ëª¨ë“  ì •ë³´ë¥¼ í•˜ë‚˜ì˜ ë²¡í„°ë¡œ ì••ì¶•í•˜ëŠ” ê²ƒì…ë‹ˆë‹¤.

- ì–´ë–»ê²Œ ì‘ë™í• ê¹Œìš”?  
  a. LSTM ë ˆì´ì–´ê°€ ì…ë ¥ ë¬¸ì¥ì„ í•œ ê¸€ìì”© ìˆœì„œëŒ€ë¡œ ì½ìŠµë‹ˆë‹¤. ('h' â” 'e' â” 'l' â” 'l' â” 'o')

> ì´ ìµœì¢… ìƒíƒœë¥¼ **ì»¨í…ìŠ¤íŠ¸ ë²¡í„°**(Context Vector)ë¼ê³  ë¶€ë¦…ë‹ˆë‹¤. "hello"ë¼ëŠ” ë‹¨ì–´ì˜ ì˜ë¯¸ê°€ ì´ ë²¡í„° ì•ˆì— ì••ì¶•ë˜ì–´ ë‹´ê²¨ìˆìŠµë‹ˆë‹¤.

> **ì¸ì½”ë”ëŠ” ì¤‘ê°„ ì¶œë ¥ë¬¼ì€ ë²„ë¦¬ê³ , ì˜¤ì§ ì´ 'ì»¨í…ìŠ¤íŠ¸ ë²¡í„°'ë§Œ ë””ì½”ë”ì—ê²Œ ì „ë‹¬í•©ë‹ˆë‹¤.**

**2. ë””ì½”ë” (Decoder): ë¬¸ì¥ ìƒì„±ê¸° âœï¸**  
ë””ì½”ë”ì˜ ì—­í• ì€ ì¸ì½”ë”ê°€ ì••ì¶•í•´ì¤€ 'ì»¨í…ìŠ¤íŠ¸ ë²¡í„°'ë¥¼ ë°›ì•„ì„œ, ìš°ë¦¬ê°€ ì›í•˜ëŠ” ì¶œë ¥ ë¬¸ì¥("HELLO")ì„ í•œ ê¸€ìì”© ìƒì„±í•˜ëŠ” ê²ƒì…ë‹ˆë‹¤.

- ì–´ë–»ê²Œ ì‘ë™í• ê¹Œìš”?  
  a. ë””ì½”ë” LSTMì€ ì¸ì½”ë”ê°€ ë§Œë“  'ì»¨í…ìŠ¤íŠ¸ ë²¡í„°'ë¥¼ **ì´ˆê¸° ìƒíƒœ**(Initial State)ë¡œ ë°›ìŠµë‹ˆë‹¤. (ì´ê²ƒì´ ì¸ì½”ë”ì™€ ë””ì½”ë”ë¥¼ ì‡ëŠ” ìœ ì¼í•œ ì—°ê²°ê³ ë¦¬ì…ë‹ˆë‹¤!)  
  b. ë””ì½”ë”ëŠ” "ë²ˆì—­ì„ ì‹œì‘í•˜ë¼"ëŠ” ì‹ í˜¸ì¸ **ì‹œì‘ í† í°**(\t)ì„ ì²« ì…ë ¥ìœ¼ë¡œ ë°›ìŠµë‹ˆë‹¤.  
  c. ì»¨í…ìŠ¤íŠ¸ì™€ \të¥¼ ë³´ê³  ë‹¤ìŒ ê¸€ìì¸ "H"ë¥¼ ì˜ˆì¸¡í•©ë‹ˆë‹¤.  
  d. (ì¶”ë¡  ì‹œ) ë°©ê¸ˆ ì˜ˆì¸¡í•œ "H"ë¥¼ ë‹¤ì‹œ ì…ë ¥ìœ¼ë¡œ ë°›ì•„ "E"ë¥¼ ì˜ˆì¸¡í•©ë‹ˆë‹¤.  
  e. ì´ ê³¼ì •ì„ **ì¢…ë£Œ í† í°**(\n)ì´ ë‚˜ì˜¬ ë•Œê¹Œì§€ ë°˜ë³µí•©ë‹ˆë‹¤.  

4-1. ì˜ë‹¨ì–´ ë°ì´í„° ë¡œë“œ ë° ì „ì²˜ë¦¬

```python
# 1. ì›ë³¸ ë°ì´í„°: 20000ê°œì˜ NLTK ëœë¤ ì˜ë‹¨ì–´
# (ì˜ˆì‹œ ì…ë ¥: "hello", ì˜ˆì‹œ íƒ€ê²Ÿ: "\tHELLO\n")
# \t: ì‹œì‘ í† í°(Start of Sequence)
# \n: ì¢…ë£Œ í† í°(End of Sequence)

file_path = 'words.txt' # ê²½ë¡œ ì…ë ¥

raw_data = []

try:
    with open(file_path, 'r', encoding='utf-8') as f:
        for line in f:
            # ì¤„ë°”ê¿ˆ ë¬¸ì ì œê±° í›„, ì‰¼í‘œ(,)ë¥¼ ê¸°ì¤€ìœ¼ë¡œ ë‚˜ëˆ”
            parts = line.strip().split(',')
            # ë°ì´í„°ê°€ ì •ìƒì ìœ¼ë¡œ 2ê°œ(ì†Œë¬¸ì, ëŒ€ë¬¸ì)ì¸ ê²½ìš°ë§Œ ì¶”ê°€
            if len(parts) == 2:
                raw_data.append((parts[0], parts[1]))

    print(f"íŒŒì¼ì—ì„œ {len(raw_data)}ê°œì˜ ë‹¨ì–´ë¥¼ ì„±ê³µì ìœ¼ë¡œ ë¶ˆëŸ¬ì™”ìŠµë‹ˆë‹¤!")
    # í™•ì¸ìš© ì¶œë ¥
    print(raw_data[:5])

except FileNotFoundError:
    print(f"ì˜¤ë¥˜: '{file_path}' íŒŒì¼ì„ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")
```

```python
# 2. ë¬¸ì ì‚¬ì „(Vocabulary) êµ¬ì¶•
input_chars = set()
target_chars = set()

for input_text, target_text in raw_data:
    target_text = "\t" + target_text + "\n" # íƒ€ê²Ÿì— ì‹œì‘ ë¬¸ì(\t)/ì¢…ë£Œ ë¬¸ì(\n) ì¶”ê°€
    for char in input_text:
        input_chars.add(char)
    for char in target_text:
        target_chars.add(char)

input_chars = sorted(list(input_chars))
target_chars = sorted(list(target_chars))
num_encoder_tokens = len(input_chars)
num_decoder_tokens = len(target_chars)
```

```python
# 3. ë¬¸ì <-> ì¸ë±ìŠ¤ ë§¤í•‘ (Dictionary)
input_token_index = dict([(char, i) for i, char in enumerate(input_chars)])
target_token_index = dict([(char, i) for i, char in enumerate(target_chars)])

print(f"ì…ë ¥ ë¬¸ì ì‚¬ì „ (í¬ê¸°: {num_encoder_tokens}): {input_chars}")
print(f"íƒ€ê²Ÿ ë¬¸ì ì‚¬ì „ (í¬ê¸°: {num_decoder_tokens}): {target_chars}")
```

```python
# 4. ì‹œí€€ìŠ¤ ê¸¸ì´ ê³„ì‚°
max_encoder_seq_length = max([len(txt) for txt, _ in raw_data])
max_decoder_seq_length = max([len(txt) + 2 for _, txt in raw_data]) # \t(ì‹œì‘ ì‹ í˜¸ ë¬¸ì), \n(ì¢…ë£Œ ì‹ í˜¸ ë¬¸ì) í¬í•¨
```

```python
# 5. ë°ì´í„° ë²¡í„°í™” (One-Hot Encoding)
# Keras ëª¨ë¸ì— ì…ë ¥í•  3D Numpy ë°°ì—´ (samples, timesteps, features)
# features = ì‚¬ì „ í¬ê¸° (One-hot)

encoder_input_data = np.zeros(
    (len(raw_data), max_encoder_seq_length, num_encoder_tokens), dtype="float32"
)
decoder_input_data = np.zeros(
    (len(raw_data), max_decoder_seq_length, num_decoder_tokens), dtype="float32"
)
decoder_target_data = np.zeros(
    (len(raw_data), max_decoder_seq_length, num_decoder_tokens), dtype="float32"
)
```

```python
# 6. ë°ì´í„° ì±„ìš°ê¸°
for i, (input_text, target_text) in enumerate(raw_data):
    target_text = "\t" + target_text + "\n"

    # 6-1. ì¸ì½”ë” ì…ë ¥ ë°ì´í„°
    for t, char in enumerate(input_text):
        encoder_input_data[i, t, input_token_index[char]] = 1.0

    # 6-2. ë””ì½”ë” ì…ë ¥/íƒ€ê²Ÿ ë°ì´í„° (Teacher Forcing)
    for t, char in enumerate(target_text):
        # ë””ì½”ë” ì…ë ¥: "\tHELLO"
        decoder_input_data[i, t, target_token_index[char]] = 1.0

        # ë””ì½”ë” íƒ€ê²Ÿ: "HELLO\n" (ì…ë ¥ë³´ë‹¤ 1ìŠ¤í… ë°€ë¦¼)
        if t > 0:
            decoder_target_data[i, t - 1, target_token_index[char]] = 1.0
```

```python
print("--- ë°ì´í„° ë²¡í„°í™” (ì˜ˆì‹œ: 'hello' -> '\tHELLO\n') ---")
print(f"encoder_input_data shape: {encoder_input_data.shape}")
print(f"decoder_input_data shape: {decoder_input_data.shape}")
print(f"decoder_target_data shape: {decoder_target_data.shape}")
```

4-2. ëª¨ë¸ êµ¬ì¶•(LSTM Encoder/Decoder) ë° ì¶”ë¡ (Inference) ëª¨ë¸ ë¶„ë¦¬ ê³¼ì •

```python
# latent_dim: ì¸ì½”ë”ê°€ ì…ë ¥ ë¬¸ì¥ì˜ ì •ë³´ë¥¼ ì••ì¶•í•˜ì—¬ ì €ì¥í•˜ëŠ” ë²¡í„° ê³µê°„ì˜ í¬ê¸°, ì¼ë°˜ì ìœ¼ë¡œ 128, 256, 512 ë“±ì˜ ê°’ì„ ì‚¬ìš©í•©ë‹ˆë‹¤
# ë„ˆë¬´ í¬ë©´ ê³¼ì í•©ì´ ë°œìƒí•  ìˆ˜ ìˆê³ , ë„ˆë¬´ ì‘ìœ¼ë©´ ë¬¸ì¥ì˜ ì •ë³´ë¥¼ ì¶©ë¶„íˆ ë‹´ì§€ ëª»í•´ ì„±ëŠ¥ ì €í•˜ê°€ ìƒê¸¸ ìˆ˜ ìˆìŠµë‹ˆë‹¤.

# ì´ë²ˆì—ëŠ” 256ìœ¼ë¡œ ì„¤ì •í•´ ë´…ì‹œë‹¤.
latent_dim = 256

# ì´ ëª¨ë¸ì€ 'Teacher Forcing'ì„ ì‚¬ìš©í•˜ì—¬ í•™ìŠµí•©ë‹ˆë‹¤.
# ìš°ë¦¬ê°€ ë§Œë“  encoder_input_data, decoder_input_data, decoder_target_data 3ì¢…ë¥˜ë¥¼ ì‚¬ìš©í•©ë‹ˆë‹¤.
```

### **decoder_input_dataì™€ decoder_target_dataë¥¼ ì™œ ë¶„ë¦¬í• ê¹Œìš”??**

ğŸ’¡ í•™ìŠµì˜ í•µì‹¬: "Teacher Forcing"

ëª¨ë¸ì„ 'í•™ìŠµ'ì‹œí‚¬ ë•Œ, ìœ„ì˜ d ë‹¨ê³„ì—ì„œ "ìì‹ ì´ ì˜ˆì¸¡í•œ ê°’"ì„ ë‹¤ì‹œ ì…ë ¥ìœ¼ë¡œ ë„£ëŠ” í•™ìŠµì„ ì‹œí‚¨ë‹¤ê³  ìƒê°í•´ ë´…ì‹œë‹¤.  
ì´ ë•ŒëŠ” ì´ˆë°˜ì— ëª¨ë¸ì´ ì—‰ëš±í•œ ê°’ì„ ì˜ˆì¸¡í–ˆë‹¤ë©´, ê·¸ ì´í›„ì˜ í•™ìŠµì´ ë§¤ìš° ë¶ˆì•ˆì •í•´ì§‘ë‹ˆë‹¤. (ì—‰ëš±í•œ ì…ë ¥ì„ ë°›ê³  ê³„ì† ì—‰ëš±í•œ ì˜ˆì¸¡ì„ í•˜ê²Œ ë¨)

> ê·¸ë˜ì„œ ìš°ë¦¬ëŠ” **í•™ìŠµ ì‹œì— "ì •ë‹µ"ì„ ê°•ì œë¡œ ì•Œë ¤ì£¼ë©° í›ˆë ¨ì‹œí‚µë‹ˆë‹¤. ì´ê²ƒì´ Teacher Forcingì…ë‹ˆë‹¤.**

> ê·¸ë¦¬ê³ 
ì´ê²ƒì´ ë°”ë¡œ ì½”ë“œì—ì„œ decoder_input_dataì™€ decoder_target_dataë¥¼ ë¶„ë¦¬í•˜ëŠ” ì´ìœ ì…ë‹ˆë‹¤.

"hello" â” "HELLO" ì˜ˆì‹œ:

Encoder Input: h, e, l, l, o â¡ï¸ (ì¸ì½”ë”ê°€ ì½ê³  ì»¨í…ìŠ¤íŠ¸ ë²¡í„° ìƒì„±)

Decoder Input (Teacher Forcingìš©): \t, H, E, L, L, O (ì‹œì‘ í† í° \tìœ¼ë¡œ ì‹œì‘) â¡ï¸ (ë””ì½”ë”ëŠ” ì´ 'ì •ë‹µ' ì…ë ¥ì„ í•œ ê¸€ìì”© ë°›ìœ¼ë©´ì„œ í•™ìŠµ)

Decoder Target (ì‹¤ì œ ì •ë‹µ): H, E, L, L, O, \n (1 ìŠ¤í…ì”© ë°€ë ¤ë‚¨) â¡ï¸ (ëª¨ë¸ì€ \të¥¼ ë³´ê³  Hë¥¼, Hë¥¼ ë³´ê³  Eë¥¼... ì˜ˆì¸¡í•˜ë„ë¡ í•™ìŠµë¨)

> ì´ëŸ¬í•œ êµ¬ì¡° ë•Œë¬¸ì—, Seq2Seq ëª¨ë¸ì€ **í•™ìŠµìš© ëª¨ë¸**ê³¼ (ë‚˜ì¤‘ì— ê²°ê³¼ë¥¼ ë³¼ ë•Œ ì‚¬ìš©í•˜ëŠ”) **ì¶”ë¡ ìš©(Inference) ëª¨ë¸**ì„ ë”°ë¡œ ì •ì˜í•´ì•¼ í•©ë‹ˆë‹¤.

4-2-1. **í•™ìŠµìš©** ì¸ì½”ë”ì™€ ë””ì½”ë” ì •ì˜, ì»´íŒŒì¼, í•™ìŠµ

```python
## 1. ì¸ì½”ë”(Encoder) ì •ì˜
# (samples, max_encoder_seq_length, num_encoder_tokens) í¬ê¸°ì˜ ì…ë ¥ì„ ë°›ìŒ

# ì¸ì½”ë” ì…ë ¥ ì •ì˜
# Keras ëª¨ë¸ì˜ ì‹œì‘ì ì¸ Input ë ˆì´ì–´ë¥¼ ì‚¬ìš©í•˜ì„¸ìš”.
encoder_inputs = keras.Input(shape=(None, num_encoder_tokens))

# ì¸ì½”ë” LSTM ì •ì˜
# ì‹œí€€ìŠ¤ ë°ì´í„°ë¥¼ ì²˜ë¦¬í•˜ê¸° ìœ„í•œ LSTM ë ˆì´ì–´ë¥¼ ì‚¬ìš©í•˜ì„¸ìš”.
# ë””ì½”ë”ì— ì´ˆê¸° ìƒíƒœë¥¼ ì „ë‹¬í•˜ê¸° ìœ„í•´ return_state=Trueë¡œ ì„¤ì •í•´ì•¼ í•©ë‹ˆë‹¤.
encoder_lstm = layers.LSTM(latent_dim, return_state=True)

# encoder_outputsëŠ” ì‚¬ìš©í•˜ì§€ ì•Šì§€ë§Œ, 'state_h'ì™€ 'state_c' (ì»¨í…ìŠ¤íŠ¸ ë²¡í„°)ë¥¼ ì–»ìŒ
# LSTM ë ˆì´ì–´ëŠ” ì¶œë ¥(outputs), ì€ë‹‰ ìƒíƒœ(state_h), ì…€ ìƒíƒœ(state_c)ë¥¼ ë°˜í™˜í•©ë‹ˆë‹¤.
encoder_outputs, state_h, state_c = encoder_lstm(encoder_inputs)

# 'state_h'ì™€ 'state_c'ë¥¼ ë¬¶ì–´ì„œ 'encoder_states'ë¼ê³  ë¶€ë¦„
# ì´ ìƒíƒœë“¤ì´ ë””ì½”ë”ì˜ ì´ˆê¸° ìƒíƒœë¡œ ì‚¬ìš©ë˜ì–´ ì¸ì½”ë”ì˜ ì •ë³´ë¥¼ ì „ë‹¬í•©ë‹ˆë‹¤.
encoder_states = [state_h, state_c] # ì´ê²ƒì´ 'ì»¨í…ìŠ¤íŠ¸ ë²¡í„°'ì…ë‹ˆë‹¤.
```

```python
## 2. ë””ì½”ë”(Decoder) ì •ì˜
# (samples, max_decoder_seq_length, num_decoder_tokens) í¬ê¸°ì˜ ì…ë ¥ì„ ë°›ìŒ

# ë””ì½”ë” ì…ë ¥ ì •ì˜
decoder_inputs = keras.Input(shape=(None, num_decoder_tokens))

# ë””ì½”ë” LSTM ì •ì˜
# return_sequences=Trueë¡œ ì„¤ì •í•˜ì—¬ ëª¨ë“  íƒ€ì„ìŠ¤í…ì˜ ì¶œë ¥ì„ ë°˜í™˜ë°›ìŠµë‹ˆë‹¤.
# return_state=Trueë¡œ ì„¤ì •í•˜ì—¬ ë””ì½”ë”ì˜ ìƒíƒœë„ ë°˜í™˜ë°›ìŠµë‹ˆë‹¤(ì¶”ë¡  ì‹œ í•„ìš”).
decoder_lstm = layers.LSTM(latent_dim, return_sequences=True, return_state=True)


# â˜† ë””ì½”ë”ëŠ” 'encoder_states'ë¥¼ ì´ˆê¸° ìƒíƒœ(initial_state)ë¡œ ë°›ìŠµë‹ˆë‹¤.â˜†
# ì´ê²ƒì´ ì¸ì½”ë”ì™€ ë””ì½”ë”ë¥¼ ì—°ê²°í•˜ëŠ” ê³ ë¦¬ì…ë‹ˆë‹¤.
# ì¸ì½”ë”ê°€ ìƒì„±í•œ ì»¨í…ìŠ¤íŠ¸ ë²¡í„°ë¥¼ ë””ì½”ë”ì˜ ì‹œì‘ì ìœ¼ë¡œ ì‚¬ìš©í•˜ì—¬ ë²ˆì—­ ë“±ì„ ìˆ˜í–‰í•©ë‹ˆë‹¤.
decoder_outputs, _, _ = decoder_lstm(decoder_inputs, initial_state=encoder_states)
```

```python
# 3. ë””ì½”ë” ì¶œë ¥ì¸µ (Dense)
# ë””ì½”ë”ì˜ ê° íƒ€ì„ìŠ¤í… ì¶œë ¥(latent_dim)ì„ íƒ€ê²Ÿ ë¬¸ì ì‚¬ì „ í¬ê¸°(num_decoder_tokens)ë¡œ ë³€í™˜
decoder_dense = layers.Dense(num_decoder_tokens, activation="softmax")
decoder_outputs = decoder_dense(decoder_outputs)
```

```python
## 4. ìµœì¢… 'í•™ìŠµìš©' ëª¨ë¸ ì •ì˜
# inputs: [ì¸ì½”ë” ì…ë ¥, ë””ì½”ë” ì…ë ¥]
# outputs: [ë””ì½”ë” íƒ€ê²Ÿ]
model = keras.Model([encoder_inputs, decoder_inputs], decoder_outputs)
```

```python
# 5. ëª¨ë¸ ì»´íŒŒì¼ ë° í•™ìŠµ
model.compile(
    optimizer="rmsprop", loss="categorical_crossentropy", metrics=["accuracy"]
)
# ì´ë²ˆì—ëŠ” RMSProp ì˜µí‹°ë§ˆì´ì €ë¥¼ ì”ë‹ˆë‹¤.

print("--- í•™ìŠµìš© ëª¨ë¸(Training Model) ìš”ì•½ ---")
model.summary()
```

```python
from tensorflow.keras.callbacks import EarlyStopping

# EarlyStopping ê°ì²´ ì„¤ì •
early_stopping_callback = EarlyStopping(monitor='val_loss', # íŒíŠ¸: EarlyStoppingì—ì„œ ì–´ë–¤ ì ìˆ˜ë¥¼ ë³´ë©´ì„œ ê°ì‹œí• ì§€ ì •í•˜ëŠ” íŒŒë¼ë¯¸í„°ëŠ”?
                                        patience=20,
                                        restore_best_weights=True)

# ëª¨ë¸ í•™ìŠµ
model.fit(
     [encoder_input_data, decoder_input_data], # 2ê°œì˜ ì…ë ¥[ì¸ì½”ë” ì¸í’‹, ë””ì½”ë” ì¸í’‹]
     decoder_target_data,                      # 1ê°œì˜ ì¶œë ¥[ë””ì½”ë” íƒ€ê²Ÿ]
     batch_size=128,
     epochs=100,  # epochë¥¼ ëŠ˜ë¦¬ê³ , EarlyStoppingì„ ì”ë‹ˆë‹¤.
     validation_split=0.2, # ì „ì²´ í•™ìŠµ ë°ì´í„°ì˜ 20%ë¥¼ ìë™ìœ¼ë¡œ ë¶„ë¦¬í•˜ì—¬ ê²€ì¦ ì…‹ì˜ ìš©ë„ë¡œ ì‚¬ìš©í•©ë‹ˆë‹¤.
     callbacks=[early_stopping_callback]
    )
```

4-2-2. **ì¶”ë¡ ìš©** ì¸ì½”ë”ì™€ ë””ì½”ë” ì •ì˜, ì»´íŒŒì¼, í•™ìŠµ

```python
## 1. ì¸ì½”ë” ëª¨ë¸ (Inference)
# ê¸°ëŠ¥: ì…ë ¥ ë¬¸ì¥ -> ì»¨í…ìŠ¤íŠ¸ ë²¡í„° [state_h, state_c]
encoder_model = keras.Model(encoder_inputs, encoder_states)
```

```python
## 2. ë””ì½”ë” ëª¨ë¸ (Inference)
# ì¶”ë¡ ì„ ìœ„í•œ ë””ì½”ë” ëª¨ë¸ì„ ì •ì˜í•©ë‹ˆë‹¤.
# í•™ìŠµ ì‹œì™€ ë‹¬ë¦¬, ì¶”ë¡  ì‹œì—ëŠ” ì´ì „ íƒ€ì„ìŠ¤í…ì˜ ì¶œë ¥ì„ í˜„ì¬ íƒ€ì„ìŠ¤í…ì˜ ì…ë ¥ìœ¼ë¡œ ì‚¬ìš©í•´ì•¼ í•˜ë¯€ë¡œ
# ì¸ì½”ë”ì˜ ì»¨í…ìŠ¤íŠ¸ ë²¡í„°ë¿ë§Œ ì•„ë‹ˆë¼ ì´ì „ íƒ€ì„ìŠ¤í…ì˜ ì˜ˆì¸¡ ë¬¸ìë„ ì…ë ¥ìœ¼ë¡œ ë°›ì•„ì•¼ í•©ë‹ˆë‹¤.

# ë””ì½”ë”ì˜ ì´ˆê¸° ìƒíƒœ ì…ë ¥ ì •ì˜
# ì¸ì½”ë”ë¡œë¶€í„° ì „ë‹¬ë°›ì€ ì»¨í…ìŠ¤íŠ¸ ë²¡í„°(ì€ë‹‰ ìƒíƒœì™€ ì…€ ìƒíƒœ)ë¥¼ ì…ë ¥ìœ¼ë¡œ ë°›ìŠµë‹ˆë‹¤.
decoder_state_input_h = keras.Input(shape=(latent_dim,))
decoder_state_input_c = keras.Input(shape=(latent_dim,))
decoder_states_inputs = [decoder_state_input_h, decoder_state_input_c]

# ë””ì½”ë” LSTM ë ˆì´ì–´ ì¬ì‚¬ìš©
# í•™ìŠµ ì‹œ ì‚¬ìš©í–ˆë˜ decoder_lstm ë ˆì´ì–´ë¥¼ ê·¸ëŒ€ë¡œ ì‚¬ìš©í•˜ì—¬,
# ì´ì „ íƒ€ì„ìŠ¤í…ì˜ ì˜ˆì¸¡ ë¬¸ìì™€ í˜„ì¬ ìƒíƒœë¥¼ ì…ë ¥ìœ¼ë¡œ ë°›ì•„ ë‹¤ìŒ ë¬¸ìë¥¼ ì˜ˆì¸¡í•©ë‹ˆë‹¤.
# initial_stateì— decoder_states_inputsë¥¼ ì „ë‹¬í•˜ì—¬ ì´ˆê¸° ìƒíƒœë¥¼ ì„¤ì •í•©ë‹ˆë‹¤.
decoder_outputs, state_h, state_c = decoder_lstm(
    decoder_inputs, initial_state=decoder_states_inputs
)

# ë‹¤ìŒ íƒ€ì„ìŠ¤í…ì„ ìœ„í•´ ê°±ì‹ ëœ ìƒíƒœë¥¼ ì €ì¥í•©ë‹ˆë‹¤.
decoder_states = [state_h, state_c]

# ë””ì½”ë” ì¶œë ¥ì¸µ ì¬ì‚¬ìš©
# í•™ìŠµ ì‹œ ì‚¬ìš©í–ˆë˜ decoder_dense ë ˆì´ì–´ë¥¼ ê·¸ëŒ€ë¡œ ì‚¬ìš©í•˜ì—¬,
# LSTMì˜ ì¶œë ¥ì„ ë°”íƒ•ìœ¼ë¡œ ë‹¤ìŒ ë¬¸ìì˜ í™•ë¥  ë¶„í¬ë¥¼ ì˜ˆì¸¡í•©ë‹ˆë‹¤.
decoder_outputs = decoder_dense(decoder_outputs)

# ìµœì¢… 'ë””ì½”ë” ì¶”ë¡ ìš©' ëª¨ë¸
# inputs: [ì´ì „ ì˜ˆì¸¡ ë¬¸ì 1ê°œ, ì´ì „ ìƒíƒœ h, ì´ì „ ìƒíƒœ c]
# outputs: [ë‹¤ìŒ ë¬¸ì ì˜ˆì¸¡(softmax), ê°±ì‹ ëœ ìƒíƒœ h, ê°±ì‹ ëœ ìƒíƒœ c]
# ì´ ëª¨ë¸ì€ ë£¨í”„ ë‚´ì—ì„œ ë§¤ íƒ€ì„ìŠ¤í…ë§ˆë‹¤ í˜¸ì¶œë˜ì–´ ë‹¤ìŒ ë¬¸ìë¥¼ í•˜ë‚˜ì”© ìƒì„±í•©ë‹ˆë‹¤.
decoder_model = keras.Model(
    [decoder_inputs] + decoder_states_inputs, [decoder_outputs] + decoder_states
)
```

```python
print("--- ì¶”ë¡ ìš© ì¸ì½”ë” ëª¨ë¸(Inference Encoder) ìš”ì•½ ---")
encoder_model.summary()
print("\n--- ì¶”ë¡ ìš© ë””ì½”ë” ëª¨ë¸(Inference Decoder) ìš”ì•½ ---")
decoder_model.summary()
```

```python
# 1. ì¸ë±ìŠ¤ -> ë¬¸ì ë³€í™˜ ë”•ì…”ë„ˆë¦¬ (ê²°ê³¼ í™•ì¸ìš©)
reverse_input_char_index = dict((i, char) for char, i in input_token_index.items())
reverse_target_char_index = dict((i, char) for char, i in target_token_index.items())
```

```python
# 2. ì˜ˆì¸¡(ë²ˆì—­)ì„ ìˆ˜í–‰í•˜ëŠ” 'decode_sequence' í•¨ìˆ˜ ì •ì˜
def decode_sequence(input_text):
    # (a) ì…ë ¥ ë¬¸ì¥ì„ ì›-í•«(one-hot) ë²¡í„°ë¡œ ë³€í™˜
    input_seq = np.zeros(
        (1, max_encoder_seq_length, num_encoder_tokens), dtype="float32"
    )
    for t, char in enumerate(input_text):
        if char in input_token_index:
            input_seq[0, t, input_token_index[char]] = 1.0

    # (b) 'encoder_model'ë¡œ ì…ë ¥ ë¬¸ì¥ì˜ ì»¨í…ìŠ¤íŠ¸ ë²¡í„°(states) ì¶”ì¶œ
    states_value = encoder_model.predict(input_seq)

    # (c) ë””ì½”ë”ì˜ ì²« ì…ë ¥: '\t' (ì‹œì‘ í† í°)
    target_seq = np.zeros((1, 1, num_decoder_tokens), dtype="float32")
    target_seq[0, 0, target_token_index["\t"]] = 1.0

    stop_condition = False
    decoded_sentence = ""

    # (d) ë£¨í”„ ì‹œì‘: í•œ ê¸€ìì”© ì˜ˆì¸¡
    while not stop_condition:
        # 'decoder_model'ë¡œ ë‹¤ìŒ ë¬¸ì ì˜ˆì¸¡
        # ì…ë ¥: (ì´ì „ ë¬¸ì, ì´ì „ states_value)
        # ì¶œë ¥: (ì˜ˆì¸¡ ë¬¸ì, ìƒˆë¡œìš´ states_value_h, ìƒˆë¡œìš´ states_value_c)
        output_tokens, h, c = decoder_model.predict([target_seq] + states_value)

        # ì˜ˆì¸¡ëœ ë¬¸ìì˜ ì¸ë±ìŠ¤ ì°¾ê¸° (ê°€ì¥ í™•ë¥ ì´ ë†’ì€ ê²ƒ)
        sampled_token_index = np.argmax(output_tokens[0, -1, :])
        # ì¸ë±ìŠ¤ë¥¼ ë¬¸ìë¡œ ë³€í™˜
        sampled_char = reverse_target_char_index[sampled_token_index]

        # ì˜ˆì¸¡ëœ ë¬¸ìë¥¼ ê²°ê³¼ ë¬¸ì¥ì— ì¶”ê°€
        decoded_sentence += sampled_char

        # (e) ì¢…ë£Œ ì¡°ê±´ í™•ì¸
        # '\n' (ì¢…ë£Œ í† í°)ì´ ì˜ˆì¸¡ë˜ê±°ë‚˜ ìµœëŒ€ ê¸¸ì´ë¥¼ ì´ˆê³¼í•˜ë©´ ì¤‘ì§€
        if sampled_char == "\n" or len(decoded_sentence) > max_decoder_seq_length:
            stop_condition = True

        # (f) ë‹¤ìŒ ë£¨í”„ë¥¼ ìœ„í•´ ì…ë ¥ ì—…ë°ì´íŠ¸
        # target_seq: ë‹¤ìŒ ì…ë ¥ìœ¼ë¡œ ë°©ê¸ˆ ì˜ˆì¸¡í•œ ë¬¸ìë¥¼ ì‚¬ìš©
        target_seq = np.zeros((1, 1, num_decoder_tokens), dtype="float32")
        target_seq[0, 0, sampled_token_index] = 1.0
        # states_value: ë‹¤ìŒ ì…ë ¥ìœ¼ë¡œ ë°©ê¸ˆ ë°›ì€ ìƒˆë¡œìš´ stateë¥¼ ì‚¬ìš©
        states_value = [h, c]

    return decoded_sentence
```
