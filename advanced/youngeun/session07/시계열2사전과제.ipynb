## <span style="background-color:#E0F7FA; color:#1565C0; padding:4px; border-radius:5px;"> 안정적 시계열 분석 </span>
### <span style="background-color:#FFFFE0; color:#CD5C5C; padding:4px; border-radius:5px;"> 자기 회귀 모델(AR) </span>
> 과거 자신의 값들을 독립변수로 삼아 현재의 값을 예측하는 회귀 모델

시계열 데이터의 `자기 상관성` 활용

$AR(p):  y_{t}=c +\phi _1y_{t-1}+\phi_2y_{t-2}+...+\phi _py_{t-p}+\varepsilon _t$

$p$차 자기 회귀 모형: 과거 $p$개 시점의 데이터의 선형 조합을 이용해 예측하는 모델

### <span style="background-color:#FFFFE0; color:#CD5C5C; padding:4px; border-radius:5px;"> 이동 평균 모델(MA) </span>
> 과거 시점의 잔차($\varepsilon$)를 독립변수로 삼아 현재 값을 예측하는 모델

$MA(q): y_t = \mu +\varepsilon _t+ \theta _1\varepsilon_{t-1}+\theta_2\varepsilon_{t-2}+...+\theta_q\varepsilon_{t-q}$

$q$차 이동 평균 모형: 과거 q개의 예측 오차의 선형 결합으로 예측하는 모델

### <span style="background-color:#FFFFE0; color:#CD5C5C; padding:4px; border-radius:5px;"> AR, MA 모델의 전제 조건 </span>
##### <span style="background-color:#DCDCDC; color:#000000; padding:4px; border-radius:5px;"> 정상성 </span>
> AR과 MA 모형은 안정적 시계열 분석으로 확률적 안정성을 전제로 추정


##### **AR 모형**의 정상성 조건
> 특성 방정식: $1 - \phi_1 z - \phi_2 z^2 - \cdots - \phi_p z^p = 0$

: 모든 해의 절댓값 > 1 -> AR의 정상성 확보


##### **MA 모형**의 정상성 조건
> 잔차는 정상성 이미 만족 -> MA는 항상 만족


##### <span style="background-color:#DCDCDC; color:#000000; padding:4px; border-radius:5px;"> 가역성 </span>
> MA 모형을 AR(∞) 형태로 재표현할 수 있는지 여부

##### **MA 모형**의 가역성 만족 조건
> 특성 방정식: $1 + \theta_1 z + \cdots + \theta_q z^q = 0$

: 모든 해의 절댓값 > 1 ⇒ MA의 가역성 확보

### <span style="background-color:#FFFFE0; color:#CD5C5C; padding:4px; border-radius:5px;"> ARMA 모형 </span>


$Y_t = \underbrace{\phi_1 Y_{t-1} + \phi_2 Y_{t-2} + \cdots + \phi_p Y_{t-p}}_{\text{AR(p): 자기회귀}} + \underbrace{\varepsilon_t + \theta_1 \varepsilon_{t-1} + \theta_2 \varepsilon_{t-2} + \cdots + \theta_q \varepsilon_{t-q}}_{\text{MA(q): 이동평균}}$

> $AR(p)$ 모형과 $MA(q)$ 모형을 결합한 모형

: 과거 $p$개의 관측값 + 과거 $q$개의 오차를  활용해서 예측하는 방법

### <span style="background-color:#FFFFE0; color:#CD5C5C; padding:4px; border-radius:5px;"> 차수 결정법: 자기 상관 함수 </span>
> 두 그래프를 통해 현재 값이 과거 값과 어떤 관계가 있는지 파악하고 이를 통해 AR, MA, ARMA의 적절한 차수(p, q)를 결정
##### <span style="background-color:#DCDCDC; color:#000000; padding:4px; border-radius:5px;"> 자기 상관 계수 (ACF) </span>
> ARMA(p, q)모형에서 사용할 q개의 오차 결정 방법

: 두 시점 사이의 **상관 관계** (간접 영향) 파악

##### <span style="background-color:#DCDCDC; color:#000000; padding:4px; border-radius:5px;"> 자기 상관 계수 (ACF) </span>
> ARMA(p, q) 모형에서 사용할 p개의 시점 결정 방법

:두 시점 사이의 **직접적인 영향**만 파악

## <span style="background-color:#E0F7FA; color:#1565C0; padding:4px; border-radius:5px;"> 불안정 시계열 분석 </span>
### <span style="background-color:#FFFFE0; color:#CD5C5C; padding:4px; border-radius:5px;"> ARIMA </span>

$ARMA(p, q) + \text{차분} \rightarrow ARIMA(p,d,q)$


### <span style="background-color:#FFFFE0; color:#CD5C5C; padding:4px; border-radius:5px;"> SARIMA </span>




$ARIMA(p,d,q) + \text{계절성(Seasonal) 차분} \\ \rightarrow SARIMA(p,d,q)(P, D, Q)_m$
> ARIMA 모델에 계절성 변동을 반영한 모델

## <span style="background-color:#E0F7FA; color:#1565C0; padding:4px; border-radius:5px;"> 다변량 시계열 분석 </span>
### <span style="background-color:#FFFFE0; color:#CD5C5C; padding:4px; border-radius:5px;"> VAR </span>
> 자신의 과거 값과 다른 변수들의 과거 값을 함께 사용하여 미래를 예측하는 모델


$X_t = a_{11} X_{t-1} + a_{12} Y_{t-1} + ε_{x,t}$

-> X와 Y가 서로의 과거를 함께 사용

✅
- 여러 경제 지표가 서로 영향을 주고받는 경우 (GDP, 소비, 투자 등)
- 금융 시장에서 여러 자산의 가격이 상호작용하는 경우
- 변수 간 상호 의존성이 중요한 경우

❌
- 변수가 너무 많은 경우 (파라미터 폭발)
- 변수 간 관계가 명확하지 않은 경우
- 장기 예측이 필요한 경우

##### <span style="background-color:#DCDCDC; color:#000000; padding:4px; border-radius:5px;"> 충격 반응 함수 (IRF) </span>
> 한 시계열에 특정 시점에서 충격이 발생할 때 다른 시계열에 시간에 따라 어떤 영향을 주는지 분석

##### <span style="background-color:#DCDCDC; color:#000000; padding:4px; border-radius:5px;"> 예측오차 분산 분해 </span>
> 각 변수가 미래의 움직임을 예측하는 데에 어느 정도의 기여를 하는지 평가하는 방법

## <span style="background-color:#E0F7FA; color:#1565C0; padding:4px; border-radius:5px;"> 딥러닝 시계열 분석 </span>
### <span style="background-color:#FFFFE0; color:#CD5C5C; padding:4px; border-radius:5px;"> RNN 계열

| 모델 | 장점 | 단점 | 사용 예시 |
| --- | --- | --- | --- |
| **RNN** | 간단, 빠름 | 장기 의존성 학습 불가 | 짧은 시계열 (일주일 주가) |
| **LSTM** | 장기 의존성 학습 가능 | 복잡, 느림 | 분기별 재무제표 예측 |
| **GRU** | LSTM보다 빠름, 적당한 성능 | LSTM보다 장기 의존성 약함 | 실시간 센서 데이터 예측 |


### <span style="background-color:#FFFFE0; color:#CD5C5C; padding:4px; border-radius:5px;"> Transformer </span>

| 모델 | 언제 사용? | 사용 예시 | 장점 | 단점 |
| --- | --- | --- | --- | --- |
| **Vanilla Transformer** | 짧은 시계열 | 30일 주가 예측 | 구조 단순 | 긴 시계열 처리 어려움 |
| **Informer** | 긴 시계열 | 전력 수요 예측 | 계산 효율적 | 짧은 시계열 과적합 |
| **Autoformer** | 주기성 뚜렷 | 월별 판매량 예측 | 주기 포착 우수 | 구조 복잡 |
| **TFT** | 외생 변수 많음 | 유통 수요 예측 | 해석 가능 | 학습 시간 오래 걸림 |
### <span style="background-color:#FFFFE0; color:#CD5C5C; padding:4px; border-radius:5px;"> Foundation Model </span>
> 거대한 데이터로 사전 학습하여 다양한 문제를 해결할 수 있는 범용 모델

##### <span style="background-color:#DCDCDC; color:#000000; padding:4px; border-radius:5px;"> LLM 기반 </span>
> 대규모 언어 모델의 패턴 인식 능력을 시계열에 활용하는 접근법

##### <span style="background-color:#DCDCDC; color:#000000; padding:4px; border-radius:5px;"> 예측오차 분산 분해 </span>
> 대규모 시계열 데이터로 사전 학습하여 다양한 예측 문제에 범용적으로 적용 가능한 모델
