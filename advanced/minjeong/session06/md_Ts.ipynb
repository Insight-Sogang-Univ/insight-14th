# 시계열 데이터
## 시계열 데이터 정의<br>
#### <span style='background-color: #fff5b1'> 1. 시계열 데이터란? </span>

= 순차적 시간 흐름에 따라 수집한 데이터 => 연속한 관측치 간 상관관계가 존재하며, <u>시퀀스 데이터</u>로 간주 


#### <span style='background-color: #fff5b1'> 2. 종단면 / 횡단면 데이터 </span>

- 종단면 데이터 (시계열) = 동일 개체에 대해 특정 시간 간격으로 수집한 데이터    
    - 시간 의존성    
    - 개체/단위 추적     
    - 패턴 및 동향 분석     
    - 효과 및 원인 분석: 특정 요인 or 처리 효과 분석    

- 횡단면 데이터 (비시계열) = 몇 가지 주제에 대해 특정 시점에서 수집한 데이터    
    - 시간 의존성 없음 = 데이터 포인트는 동일 시점에서 독립적으로 수집    
    - 다중 변수: 여러 변수의 동시 측정 결과로 구성          
    - 비교 분석: 다른 개체/단위 간 비교 분석      
    - 상관관계 분석    

#### <span style='background-color: #fff5b1'> 단변량 / 다변량 시계열 </span>

- 단변량 시계열 = 시간에 따라 연속 측정된 <u>단일 변수</u>   
- 다변량 시계열 = 같은 시간에 기록된 둘 이상의 변수    


## 시계열 분석과 목적<br>
#### <span style='background-color: #fff5b1'> 1. 시계열 분석과 목적 </span>

- 시계열 데이터를 분석 & 모델링하는 통계적 방법과 기법    
- 데이터 구조 파악, 예측, 이상치 탐지     
=> 미래 데이터 예측    

#### <span style='background-color: #fff5b1'> 2. 활용 방안 </span>

- 웹 트래픽 예측해 서버 최적화    
- 주가, 환율, 금리 변동 예측   
- 환자 생체 신호 분석을 통한 이상 탐지    
- 앱 사용 로그 분석을 통한 이탈 예측    

## 시계열 데이터 특징<br>
#### <span style='background-color: #fff5b1'> 1. 시간의존성 </span>

= 시간에 따라 변화    
- 과거 값이 현재 값에 영향    
- 시점 간 상관관계 고려   

#### <span style='background-color: #fff5b1'> 2. 계절성 & 추세 </span>

= 시간에 따라 장기적으로 증가/감소하는 추세 + 일정 주기로 반복되는 패턴    
- 계절성: 주기에 따라 반복되는 데이터의 변동 => 패턴 고려한 예측 모델링       
- 추세: 시간 흐름에 따라 나타나는 장기적 증가/감소 영향 => 전반적 방향성      

#### <span style='background-color: #fff5b1'> 3. 자기상관성 </span>

= 이전 시점 값이 이후 시점 값에 영향    
= 자기 자신끼리 상관 관계를 갖는 특성    
- 시간이 지남에 따라 값 간 패턴/반복 존재   
- 관측값은 서로 독립적이지 않음   

<br>

**자기상관성 때문에 회귀 분석 불가능**<br>
1. 통계적 추론 제약 및 오류     
- 오차항 간 상관관계가 있어, 회귀 분석의 통계 추론 왜곡 가능성    

2. 예측 어려움 & 모델 복잡성 증가    
- 잔차가 시간에 따라 패턴을 가져 미래값 예측이 왜곡되고, 오차가 누적되며 신뢰도 하락    
- 단순 회귀 모형이 적절하지 않아 AR, ARIMA 등 시계열 모델 대체 필요

## 시계열 데이터 분석요인<br>
#### <span style='background-color: #fff5b1'> 1. 추세 요인 </span>

- 데이터 전반에 걸쳐 시간에 따라 지속적으로 증가/감소하는 경향    
- 꾸준히 증가/감소하는 선형 그래프     

#### <span style='background-color: #fff5b1'> 2. 계절 요인 </span>

- 일정한 주기로 반복되는 패턴    
- 주기적 변동을 나타내는 그래프     
- 주기가 고정적으로 반복    

#### <span style='background-color: #fff5b1'> 3. 순환 요인 </span>

- 계절성보다 길고 불규칙한 주기로 반복되는 변동     
- 길게 상승&하강 반복하는 파형 그래프    
- 주기가 불규칙 or 장기적 & 외부 요인에 의해 발생     

#### <span style='background-color: #fff5b1'> 4. 불규칙 요인 </span>

- 위 요인 모두 제거 뒤 남은 예측 불가능한 변동     
- 노이즈, 우연, 일시 변동, 이상치 등    
- 불규칙한 변동을 나타내는 그래프   

## 시계열 데이터 분석 work flow

(1) 데이터 전처리    
    - 결측치 보간, 이상치 제거, 요소분해      

(2) 정상성 판단       
 
(3) 모델링 및 예측

## 시계열 데이터 전처리
### 이상치/결측치 처리<br>

- 결측치: 특정 시점에서 값 측정 or 누락된 상태 -> 추세, 계절성 왜곡 방지를 위해 적절한 대치와 보간 필요     
- 이상치: 측정된 값이 일반적이지 않은 경우 -> 신호의 중요한 의미일 가능성 있으므로 신중하게 접근        
=> 시계열 특성을 반영한 전처리 필요      

#### <span style='background-color: #fff5b1'> 1. 결측값 처리 </span>

**결측치 처리 방법 - 보간법**<br>
= 주어진 데이터 포인트 값을 사용해 두 데이터 포인트 사이에서 누락된 값 추정    
= 사라진 값 합리적으로 메워주는 과정     
- 결측치/이상치 구간이 짧거나 신호의 패턴이 단순한 경우, 보간 정확도 높음   
- 선형 보간, 다항식 보간, 스플라인 보간, 조각별 상수 보간 등    
- 회귀와 달리, 관측값을 완전히 신뢰하며 직접 연결함    

#### <span style='background-color: #fff5b1'> 2. 이상치 탐지 및 처리 </span>

**이상치 탐지 방법 - IQR Rule based Anomaly Detection**<br>

- 주기성 없거나 안정적 구간의 값에서 이상치 탐지할 때    
(+) 단순하고 명확한 해석, 통계 기반이라 빠른 속도    
(-) 추세나 계절성이 강한 시계열에 부적합, 분포가 비정상적일 경우 오탐률 높음     

**이상치 탐지 방법 - Z-score**<br>

- 특정 값이 평균으로부터 몇 표준편차 떨어진지    
- z-score가 임계값 넘어서면 이상치     
- 계절성/추세가 약하거나 값의 변동폭이 일정할 때     
(+) 간단하고 빠른 계산, 직관적 해석      
(-) 정규분포 가정에 민감, 비정상 시계열에서 안정적 기준 설정 어려움     

**이상치 탐지 방법 - STL 분해**<br>

- 잔차가 특정 임계값을 초과하면 이상치 탐지    
- 계절성/추세가 뚜렷한 시계열에 적합     
(+) 계절성/추세 제거 후 진짜 이상한 값만 탐지 가능, 비정상 시계열 적용 가능     
(-) 큰 계산 비용, 매우 긴 시계열에 파라미터 튜닝 필요    

**이상치 탐지 모델 - 분류 및 회귀 트리(CART)**<br>

- 데이터 분할해 예측값 산출     
- 예측 오차가 임계값 이상인 지점이 이상치    
(+) 비선형 관계 포착, 변수 간 상호작용이 있는 시계열 적용 가능    
(-) 노이즈가 많은 데이터에서 과적합 위험, 트리 깊이 에 따라 다른 탐지 민감도    

**이상치 탐지 모델 - 클러스터링 기반 이상 탐지**<br>

- 정상 구간의 데이터는 특정 클러스터에 잘 모임    
- kmeans, DBSCAN, isolation forest 등 주로 사용    
(+) 데이터 패턴이 복잡하거나 다변량일 때 유용, 사전 분포 가정 필요 없음    
(-) 하이퍼파라미터에 민감, 시계열 순서 정보 직접 반영되지 않음    

**이상치 탐지 모델 - 오토 인코더**<br>

- 인공신경망 기반 비지도 학습 모델     
- 정상 데이터 특징 학습 후, 모델에 데이터 입력 시 재구성한 결과와 학습된 정상 특징과 차이점을 비교해 이상 여부 판단   
(+) 복잡한 비선형 패턴 학습, 다변량 시계열 이상 탐지에 강함    
(-) 많은 학습 데이터 필요, 모델 해석이 어렵고 파라미터 튜닝 까다로움     

**평활화 - 이동평균**<br>

평활화 = 노이즈를 줄이고, 전체적 흐름/추세를 명확하게 파악하기 위해 데이터를 부드럽게 만드는 기술    
이동 평균 = 시계열의 부분 집합으로 평균을 구해 평활화 효과 얻는 방법    

(1) 단순 이동평균 = 기간 내 모든 값 평균   

(2) 가중 이동평균 = 최근 값에 더 큰 가중치     

(3) 지수 이동평균 = 최근 데이터에 지수적으로 가중치      

## 요소 분해(시계열 분해)<br>
#### <span style='background-color: #fff5b1'> 1. 요소 분해 </span>

= 주어진 시계열 데이터를 개별 성분으로 분해하는 과정           
- 다양한 변동이 결합된 시계열 데이터에서 각 요소를 분리해 패턴 이해    
=> 예측, 이상 탐지 등 정교한 분석 작업 수행     

#### <span style='background-color: #fff5b1'> 2. 가법 모형 </span>

= 관측값은 변동 요인들의 가법으로 설명될 수 있다고 가정    
- 각 변동 요인들이 독립적으로 더해지며, 모든 요소의 영향이 일정함     
- 시계열 전체에서 계절성의 진폭이 일정할 때 적합      

#### <span style='background-color: #fff5b1'> 3. 승법 모형 </span>

= 관측값이 변동 요인들의 승법으로 설명될 수 있다고 가정     
- 추세가 커질수록 계적성/불규칙성 효과도 비례해서 커짐     
- 진폭이 시간에 따라 달라질 때 적합    

## 정상성 판단
### 정상성이란?<br>

= 시계열 데이터가 평균, 분산 등 통계적 특성이 시간에 따라 일정한 수준으로 유지되는 것    
- 정상 시계열: 평균 및 분산 일정 => 예측 & 모델링에 적합     

### 강한 정상성과 약한 정상성<br>

**강한 정상성**<br>

= 모든 적률(평균, 분산, 왜도, 첨도 등 확률분포의 모양/특성을 수치로 요약한 것) 시간과 무관하게 일정     


**약한 정상성**<br>

(1) 평균이 모든 시점에 대해 일정하고     
(2) 분산이 모든 시점에 대해 일정하고    
(3) 공분산은 시점에 의존하지 않고, 시차에만 의존함    

### 검정으로 정상성 판단하기<br>

- 그래프만으로 정확히 판단하기 어려움 -> 수치적, 객관적 판단 판단 필요    

**단위근 검정**<br>

단위근 = 양의 정수에 대해 해당 방정식의 근이 되는 복소수    
$x^n = 1$     

- 시계열 모델의 특성 방정식의 근이 1인 경우, 해당 시계열이 단위근을 가짐 = 비정상 시계열    
=> 단위근이 있으면 랜덤워크로 예측의 불확실성이 높아짐. 즉, 단위근이 있는 시계열은 비정상 시계열.    

**AR(1) 모델을 통해 본 단위근**<br>

$y_t = \alpha y_{t-1} + e_t$      

(1) $\alpha = 0$     
- 완전한 백색 잡음 = 과거 값의 영향을 젆 ㅕ받지 않음     
- 외부 충격만으로 시계열 생성      

(2) $|\alpha| < 1$       
- 현재 값이 이전 값의 영향을 받지만 (자기 상관) 시간 흐름에 따라 과거 충격 영향이 희석됨     
- 평균 수준으로 돌아오며, 계속해서 안정 상태 유지    
=> 정상 시계열로 간주     

(3) $\alpha = 1$       
- 비정상 시계열 = 단위금 있음     
- 외부 충격으로 인한 변동이 유지되어 원상태로 돌아가지 못함   
=> 예측 불가능    

<br>

(1) ADF 검정        

= 단위근 존재 여부를 통해 시계열의 정상성 판단    
- $H_0$ : 시계열에 단위근 존재 = 비정상 시계열       
- &H_1& : 시계열에 단위근 존재 X = 정상 시계열      

(2) KPSS 검정    

= 단위근 기반이 검정이지만 귀무가설이 반대    
- $H_0$ : 시계열이 단위근 존재 X = 정상 시계열     
- $H_1$ : 시계열이 단위근 존재 = 비정상 시계열   

**확정적 추세**<br>

= 추세가 완벽하게 예측 가능하고 일정하게 존재     
- 잡음으로 흔들림은 있지만, 시간에 따라 선형적으로 증가   
- 확정적 추세가 있을 때, ADF 모델     

**확률적 추세**<br>

= 추세선 자체가 시간에 따라 이동함    
- 단위근 과정으로, 시간에 따라 우연히 위아래로 이동하며 영구적 충격이 누적됨    
- 확정적 추세가 있을 때, KPSS 모델    


### 정상성 확보를 위한 방법들<br>

1. 추세 제거     
- 시계열이 지속적으로 증가/감소할 떄 적용   
- 차분: 평균의 정상성을 확보하기 위해 연이은 관측값의 차이 계산     
= 시계열 데이터에서 한 시점과 그 이전 시점 값을 차분해 변화량 계산     

2. 계절성 제거    
- 시계열이 주기적인 패턴을 가질 때 적용     
- 계절 차분     

3. 분산 안정화     
- 시계열 변동 폭이 점점 커지거나 작아질 때 적용     
- 로그 변환: 분산의 정상성 확보할 수 있지만, 음수나 0을 처리할 수 없음    
- Box-Cox 변환: 로그 변환의 일반화된 방법

**정상/비정상 여부 판단**<br>
- 단위근 검정으로 정상 시계열 여부 판단 후, 비정상이면 다음 단계 이동    

**비정상 유형에 따른 정상성 확보 방법**<br>
- 추세가 있다면 차분, 추세 제거    
- 계절성이 있다면 계절 차분 / deseasoning    
- 분산 변화가 있다면 로그 변환, box-cox 변환    

**정상성 확보 후 모델링**<br>
- 시계열 구조에 맞는 모델 적용
