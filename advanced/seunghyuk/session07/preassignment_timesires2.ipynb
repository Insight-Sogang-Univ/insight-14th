## 단변량 시계열 분석

1. 안정적 시계열 분석
    1. 자기 회귀 모델 (AR)
    - 과거 p개 시점의 데이터들을 통해 현재의 값을 예측 (선형 조합)
    - 자기 상관성을 활용
    2. 이동 평균 모델 (MA)
    - 과거 q개의 예측 오차의 선형 결합을 통해 예측
    3. AR, MA의 전제 조건
    - 정상성
        - "모델"의 정상성을 확보해야 함.
        - 특성방정식에서 모든 해의 절댓값이 1 초과 => 정상성 확보
    - 가역성
        - MA 모델을 AR형태로 재표현할 수 있는지?
            - 특성 방정식의 모든 해의 절댓값 >1
        - AR은 가역성 체크 필요 x
    4. ARMA
    - AR 모형과 MA 모형의 결합
        - 과거 p개의 관측값 + 과거 q개의 오차
        - AR모형의 정상성과 MA모형의 가역성을 모두 확보해야함
    5. 차수 결정 방법: 자기 상관 함수
    - ACF: ARMA 모형에서 사용할 q개의 오차를 결정
        - 두 시점 사이의 상관 관계 파악 (간접 영향까지)
    - PACF: ARMA 모형에서 사용할 p개의 시점을 결정
        - 두 시점 사이의 직접적인 영향만 파악
        

2. 불안정 시계열
    - 분석 프로세스
        1. 비정상 시계열 데이터
        2. 정상성 확보
        3. 정상 시계열
        4. ARMA를 통한 안정적 시계열 분석
    - 안정적 시계열 모델 + 정상성 확보 과정이 추가
    
    1. ARIMA
    - ARMA + 차분
        
        - 4번 -> 백색 노이즈만 존재해서 예측 불가
        - 5번 -> 랜덤워크 + 상수항 (방향성 존재)
    2. SARIMA
    - ARIMA + 계절성 변동
    - 하이퍼파라미터
        - 비계절성 파라미터: 데이터의 일반적인 추세와 단기적인 자기상관성을 처리
        - 계절성 파라미터: 계절성 주기에 따른 패턴과 장기적인 자기 상관성 처리
    - 하이퍼파라미터 결정
        - 차분 차수 결정: ADF를 통해 정상성 판단. 계절 차분은 먼저 수행하는 것이 일반적
        - 차수 결정: PACF 그래프를 분석해 차수의 초기 범위 추정. 추정된 범위를 바탕을 그리드 서치
            - AIC, BIC와 같은 성능 평가 지표도 활용
    


## 다변량 시계열

1. VAR
- 자신의 과거 값 + 다른 변수 => 미래 예측
- 분석 방법
    1. 충격 반응 함수 (IRF)
    - 특정 시점에서 발생하는 충격이 다른 시계열에 시간에 따라 어떻게 영향을 주는지?
    2. 예측 오차 분산 분해
    - 어떤 시계열이 상대적으로 어떤 영향을 끼치고 있는지?
- 장점
    - 변수 간 양방향 분석 가능
    - 여러 변수를 함께 예측
    - 동시성 문제 없음
- 단점
    - 변수 많음 -> 파라미터 폭발
    - 장기적 관계 정보 손실
    - 상관관계가 인과관계는 아님

## 딥러닝을 통한 시계열 분석

1. RNN 계열
    1. RNN
    - 10 ~ 100시점, 순차적 의존성이 강하고 자기상관이 큰 데이터에 적합
    - 매우 긴 시계열이나 병렬 처리를 해야하는 경우 부적함
    2. LSTM
    - 100 ~ 1000시점, 중요한 과거 이벤트가 먼 미래에 영향을 주는 경우에 적합
    3. GRU
    - 장기 의존성 + 빠른 성능이 필요하거나 실시간 처리가 필요한 경우

2. Transformer 계열
- 매우 긴 시계열(대규모 데이터), 전역 패턴이 중요한 경우, 다변량 시계열인 경우에 적합
- 짧고 자원이 제한된 경우에 부적합

3. Foundation 모델
    1. LLM 기반
    - Without Adaptation
        - 프롬프트 엔지니어링에 집중
    - Adapt LLM
        - 시계열 데이터로 파인튜닝
    - Adapt to LLM
        - 시계열 -> 텍스트 변환 후 LLM 활용
    2. 자체 거대 모델
    - 아예 대규모 시계열 데이터로 pre-training
    - TimeGPT, TimesFM, Moirai
