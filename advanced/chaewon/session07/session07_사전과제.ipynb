## SESSION07_시계열1

### 1. 시계열 개념


##### <span style="background-color:#FFF5F7; color:#000000; padding:4px; border-radius:5px;"> 1-1. 종단면 데이터, 횡단면 데이터


#### 종단면(시계열) 데이터란?
: 동일한 개체에 대해 특정 시간 간격으로 수집된 데이터 
#### 종단면 데이터의 특징은?   
1. 시간 의존성
2. 개체 또는 단위 추적
3. 패턴 및 동향 분석
4. 효과 및 원인 분석
#### 횡단면(비시계열) 데이터란?
: 몇가지 주제에 대해 특정 시점에서 수집된 데이터 

#### 횡단면 데이터의 특징은?     
1. 시간 의존성
2. 다중 변수
3. 비교 분석
4. 상관관계 분석

##### <span style="background-color:#FFF5F7; color:#000000; padding:4px; border-radius:5px;"> 1-2. 단변량 시계열과 다변량 시계열
#### 단변량 시계열이란?
- 시간에 따라 연속적으로 측정된 단일 변수로 이루어진 시계열 데이터
- 시간의 흐름에 따라 패턴과 트랜드 이해
#### 적용 사례는?
: 주식 가격, 온도 기록, 매출 데이터...
#### 다변량 시계열이란?
- 같은 시간에 기록된 둘 이상의 변수로 이루어진 시게열 데이터
- 변수 간 상호 작용 및 영향력 파악
#### 적용 사례는?
: 경제 지표, 기후 모댈 

##### <span style="background-color:#FFF5F7; color:#000000; padding:4px; border-radius:5px;"> 1-3. 시계얄 데이터의 특징

#### 시간의존성이란?
: 시계열 데이터는 시간에 따라 변화한다!!
- 시간 순서에 따라 관측되므로 데이터의 순서를 바꾸면 의미 달라짐
- 과거의 값이 현재의 값에 영향을 미침
- 시점 간의 상관관계 고려해야 함 
#### 계절성과 추세란?
           
: 시간에 따라 장기적으로 증가/감소하는 추세 + 일정 주기로 반복되는 패턴 
- 계절성: 일정 주기에 따라 반복적으로 나타나는 데이터의 변동 -> 주기적인 패턴을 고려한 예측 모델링 가능
- 추세: 시간의 흐름에 따라 나타나는 장기적인 증가/감소 경향 -> 전반적인 방향성 파악 가능 
#### 자기상관성이란?
          
: 이전 시저의 값이 이후 시점의 값에 영향을 미침, 말 그대로 자기자신끼리 상관관계를 갖는 특성!
- 한 시점의 값이 과거의 값과 통계적으로 관련 있는 성질
- 시간이 지남에 따라 값들 간에 일정한 패턴이나 반복이 존재!
- 관측값들은 서로 독립적이지 않음 
#### 자기상관성 떄문에 회귀분석이 불가능하다?!?!
1. 통계적 추론의 제약 및 오류
2. 예측의 어려움과 모델 복잡성 증가 

##### <span style="background-color:#FFF5F7; color:#000000; padding:4px; border-radius:5px;"> 1-4. 시계열 데이터의 분석요인 
#### 추세 요인이란?
: 데이터 전반에 걸쳐 시간에 따라 지속적으로 증가/감소하는 경향 
              
-> 꾸준히 증가 or 감소하는 선형 그래프            
-> ex) 연도별 매출 상승, 출산률 감소...
#### 계절 요인이란?
: 일정한 주기로 반복되는 패턴
     
-> 주기적인 변동을 나타내는 그래프         
-> ex) 크리스마스에 매출 상승     
#### 순환 요인이란?
: 계절성보다 더 길고 불규칙한 주기로 반복되는 변동 
      
-> 길게 상승과 하강을 반복하는 파형 그래프             
-> ex) 경기 호황과 불황기 
#### 불규칙 요인이란?
: 추세, 계절성, 순환성 모두 제거한 뒤 남은 예측 불가능한 변동          
: 노이즈, 우연적 사건, 일시적 변동, 이상치 등 
          
-> 불규칙한 변동을 나타내는 그라프           
-> ex) 자연재해, 팬데믹 
#### 총정리!


### 2. 시계열 전처리

##### <span style="background-color:#FFDDEB; color:#000000; padding:4px; border-radius:5px;"> 2-1. 결측치 처리-보간법 </span> 
#### 시계열 데이터의 결측치란?
- 특정 시점에서 값이 측정 X or 누락 상태
- 센서 오류, 수집 누락, 전송 지연 등
- 결측치가 존재하면 추세, 계절성 왜곡 => 적절한 대치와 보간을 하자
#### 시계열 데이터의 이상치란?
- 값이 측정되었으나, 측정된 값이 일반적 X
- 측정 오류 or 실제로 특이한 이벤트일 수도 있는 극단 값
- 단순히 이상한 값이 아니라 신호의 중요한 의미일 가능성도 있으므로 신중한 접근 하자 -> 신호에 대한 사전 지식!!!!
#### 전처리 프로세스
1. 데이터 확인 및 시각화
2. 결측치 확인 -> 처리
3. 이상치 탐지 -> 처리
4. 요소 분해 
#### 결측치 처리는 어떻게?
: 삭제/단순 대치/보간/머신러닝 기법...
#### 보간법이란? 
- 주어진 데이터 포인트의 값을 사용하여 두 개 이상의 데이터 포인트 사이의 누락된 값 추정 -> 사라진 값을 합리적으로 메워줌
- 결측치/이상치 구간이 짧거나 신호 패턴이 단순한 경우 보간 정확도 높!

##### <span style="background-color:#FFDDEB; color:#000000; padding:4px; border-radius:5px;"> 2-2. 이상치 탐지 및 처리는?
#### 이상치 탐지 방법
- IQR Rule-based Anomaly Detection
- Z-score
- STL 분해
#### 이상치 탐지 모델은?
- 분류 및 회귀 트리 (CART)
- 클러스터링 기반 이상 탐지
- 오토 인코더
#### 이상치 처리 방법은?
: 이상치를 결측치로 간주 후 결측치와 같은 방법으로 처리
- 삭제
- 단순 대치
- 보간 

##### <span style="background-color:#FFDDEB; color:#000000; padding:4px; border-radius:5px;"> 2-3. 평월화란?
: 시계열 데이터에서 노이즈 줄이고 전체 흐름이나 추세를 더 명확히 파악하기 위해 데이터를 부드럽게 만드는 기술 
#### 이동평균??
     
: N이 클수록 긴 기간의 데이터를 평균에 포함 -> 노이즈가 더 많이 제거되어 매끄러운 곡선 생성
#### 여기서 N이란?
: Moving Window의 길이
- 시계열의 부분집합으로 평균을 구하여 평월화 효과를 얻는 방법
- 윈도우 내의 평균을 구하는 방법 
#### 왜 사용하지?
- 노이즈 제거하자
- 추세 파악하자
- 예측의 기초자료 
#### 유형은?
- 단순 이동평균 (SMA)
- 가중 이동평균(WMA)
- 지수 이동평균(EMA)

##### <span style="background-color:#FFDDEB; color:#000000; padding:4px; border-radius:5px;"> 2-4. 요소분해
: 주어진 시계열 데이터를 개별 성분으로 분해하는 과정 
#### 왜 해요?
- 시계열 데이터는 다양한 변동이 결합된 형태
- 각 요소를 분리하면 각 패턴을 명확히 이해할 수 있다!
- 이후 예측, 이상 탐지 등 분석 작업을 정교하게 수행할 수 있다~
#### 가법모형
: 시계열 = 추세 + 계절성 + 불규칙성
- 각 변동 요인들 독립적으로 더해짐 -> 모든 요소의 영향 일정 
#### 승법모형
: 시계열 = 추세 x 계절성 x 불규칙성 (로그 변환을 통해 가법 형태로 만들 수 있다)
- 추세가 커질수록 계절성이나 불규칙성의 효과 비례


### 3. 정상성 

##### <span style="background-color:#FFC0CB; color:#000000; padding:4px; border-radius:5px;"> 3-1. 정상성의 정의
: 시계열 데이터가 평균, 분산 등 통계적 시간에 따라 일정한 수준으로 유지되는 것          
=> 어제나 오늘이나 데이터의 통계적 특성이 일정!
#### 핵심은?
: 시간에 상관없이 일정한 성질~~
- 정상성을 띠는 시계열 데이터 -> 정상성 시계열 
- 정상성 띠지 않는 시계열 데이터 -> 비정상 시계열 
##### 정상 시계열은 예측 및 모델링에 적합, 일반적으로 대부분의 통계 기법 적용 가능!

##### <span style="background-color:#FFC0CB; color:#000000; padding:4px; border-radius:5px;"> 3-2.강한 정상성과 약한 정상성 
#### 강한 정상성이란?
- 강정상성을 띠는 시계열은 모든 적률이 시간과 무관하게 일정 
- 기저를 이루는 확률 분포 자체가 언제나 같아야 함
- 현실에서 찾아보기 매우 어려움 
#### 약한 정상성이란?
- 강항 정상성보다 완화된 개념 
    
=> 시간이 흘러도 평균, 분산이 변하지 않으며 자기공분산이 시차에만 의존!!

##### <span style="background-color:#FFC0CB; color:#000000; padding:4px; border-radius:5px;"> 3-3. 단위근의 시계열적 의미 
#### 단위근이란?
: 단위 + 근을 합친 말 
- 시계열 모델의 특정방정식에서 근이 1인 경우! -> 단위근을 가짐 -> 비정상 시계열 
#### 단위근이 문제가 되는 이유
: 랜덤워크로 인해 예측의 불확실성이 생김
1. 평균이 일정하지 않아서 예측 어려움
2. 분산이 계속 커짐
3. 과거 충격이 영원히 남음 

##### <span style="background-color:#FFC0CB; color:#000000; padding:4px; border-radius:5px;"> 3-4. 단위근 검정 
#### 1. ADF(Augmented Dicky-Fuller) 검정
: 단위근의 존재 여부를 통해 시계열의 정상성 판단!       
=> 단위근의 존재를 의심 
#### 2. KPSS(Kwiatkowski-Phillips-Schmidt-Shin) 검정
: 역시 단위근 기반의 검정이지만 귀무가설 반대!    
=> 정상성을 의심 
#### 확정적 추세가 존재하는 경우 두 검정 결론이 서로 다르게 나올 수 있다 

##### <span style="background-color:#FFC0CB; color:#000000; padding:4px; border-radius:5px;"> 3-5. 확정적 추세와 확률적 추세
#### 확정적 추세란??
: 추세가 완벽하게 예측 가능, 일정하게 존재
- 잡음으로 약간의 흔들림은 있음
- 시간에 따라 일정한 방향으로 움직임 
#### 확률적 추세란??
: 추세선 자체가 시간에 따라 이동 
- 단위근 과정으로 시간에 따라 우연히 위아래로 이동
- 영구적인 충격이 누적 


##### <span style="background-color:#FFC0CB; color:#000000; padding:4px; border-radius:5px;"> 3-6. 정상성 확보 방법 
#### 1. 추세 제거
: 시계열이 지속적으로 증가하거나 감소할 때 적용 
#### 차분이란? 
: 연이은 관측값들의 차이를 계산 
-> 시계열 데이터에서 연속된 시점의 값들을 차분하여 변화량 계산 
#### 2. 계절성 제거
: 시계열이 주기적인 패턴을 가질 때 적용 
#### 계절 차분이란?
: $t$시점의 $y$값에서 t-m시점의 y값을 빼는 것 (m: 주기의 길이)
#### 3. 분산 안정화
: 시계열의 변동 폭이 점점 커지거나 작아질 때 적용 
#### 로그 변환?
: 분산의 정상성을 확보하기 위한 방법 
#### Box-Cox (박스-콕스 변환)?
: 분산의 정상성 확보하기 위한 방법 
#### 정상/비정상 여부 판단
- 단위근 검정 사용
- 비정상 -> 다음 단계
#### 비정상 유형에 따라 정상성 확보 
- 추세 -> 차분, 추세 제거
- 계절성 -> 계절 차분 or Deseasoning
- 분산 변화 → 로그변환, Box-Cox 변환
#### 정상성 확보 후 어떤 프로세스를 따르는지에 따라 모델링!!
