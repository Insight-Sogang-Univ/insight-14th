## SESSION08_시계열2

### 1. 안정적 시계열 분석

##### <span style="background-color:#FFF5F7; color:#000000; padding:4px; border-radius:5px;"> 1-1. 자기 회귀 모델
: 과거의 자기 자신 데이터가 미래 값을 설명해줄 것이라고 가정하는 것 
- 자기 자신의 값들: 독립변수(x) -> 현재 값(y) 예측
- 시계열 데이터의 자기상관성 활용 
#### AR이란?
##### $AR(p):  y_{t}=c +\phi _1y_{t-1}+\phi_2y_{t-2}+...+\phi _py_{t-p}+\varepsilon _t$
p차 자기 회귀 모형, 과거 p개 시점의 데이터의 선형 조합을 이용해 예측하는 모델     
#### 예를 들어.. 올해 내 생일의 기온이 일년 전 생일 기온의 영향을 받을 것을 표현한 통계식이라면?
- 1년 전이므로 p는 1
-  $y_t$: 올해 채원 생일의 기온
- $y_{t-1}$: 작년 채원 생일의 기온
- $\phi_1$: 작년 기온이 올해 기온에 영향을 미치는 정도
#### 이동 평균 모댈이란?
: 저번에 배운 이동 평균(평월화0랑 다르다!, 과거 시점의 잔차를 독립변수로 삼아 현재 값을 예측하는 머델 
- 과거의 잘못 예측한 것을 현재 시점에 반영
- 즉, 잔차는 실제 값과 예측 값의 차이이다~
##### $MA(q): y_t = \mu +\varepsilon _t+ \theta _1\varepsilon_{t-1}+\theta_2\varepsilon_{t-2}+...+\theta_q\varepsilon_{t-q}$
q차 이동 평균 모형, 과거 q개의 예측 오차의 선형 결합으로 예측하는 모델
#### 예를 들어.. 오늘 공부량이 어제의 밤샘 충격에 영향을 받을 것을 표현한 통계식이라면?
- - $y_t$: 오늘 공부랼
- $\mu$: 공부량
- $\varepsilon_{t-1}$: 어제(1일 전) 밤샘이라는 충격
- $\theta _1$: 충격이 영향을 미치는 정도
#### 전제 조건은?
1. 정상성
##### AR 모형의 정상성 조건 
- - 특성 방정식: $1 - \phi_1 z - \phi_2 z^2 - \cdots - \phi_p z^p = 0$
- 모든 해의 절댓값 > 1
##### MA 모형의 정상성 조건
- 잔차들은 정상성 이미 만족 => MA 모형도 항상 정상성 만족 
2. 가역성: MA 모형을 AR(∞) 형태로 재표현할 수 있는지 여부
##### MA 모형의 가역성 조건
- 특성 방정식: $1 + \theta_1 z + \cdots + \theta_q z^q = 0$
- 모든 해의 절댓값 > 1 
#### ARMA 모형이란?
: $AR(p)$ 모형과 $MA(q)$ 모형을 결합한 모형
- 과거 p개 관측 + 과거 q개 오차 활용 
- AR(p) 모형의 정상성 + MA(q) 모형의 가역성 둘디 민족해야 함 

##### <span style="background-color:#FFF5F7; color:#000000; padding:4px; border-radius:5px;"> 1-2. 차수 결정 방법 
: 두 그래프를 통해 현재 값이 과거 값과 어떤 관계가 있는지 파악 -> AR, MA, ARMA의 적절한 차수(p, q)를 결정
#### 자기 상관 계수란?
- ARMA(p, q)모형에서 사용할 q개의 오차 결정 방법
- 두 시점 사이의 상관 관계 파악
#### 부분 자기 상관 계수란?
- ARMA(p, q) 모형에서 사용할 p개의 시점 결정 방법
- 두 시점 사이의 직접적인 영향만 파악

### 2. 불안정 시계열

##### <span style="background-color:#FFDDEB; color:#000000; padding:4px; border-radius:5px;"> 2-1. ARIMA 

##### $ARMA(p, q) + \text{차분} \rightarrow ARIMA(p,d,q)$
#### 특수한 경우
1. 백색잡음 ARIMA(0,0,0)
2. 자기회귀 ARIMA(p,0,0)
3. 이동평균 ARIMA(0,0,q)
4. 랜덤워크: 상수가 없는 ARIMA(0,1,0)
5. 표류를 포함하는 랜덤워크: 상수가 있는 ARIMA(0,1,0)

##### <span style="background-color:#FFDDEB; color:#000000; padding:4px; border-radius:5px;"> 2-2. SARIMA

##### $ARIMA(p,d,q) + \text{계절성(Seasonal) 차분} \\ \rightarrow SARIMA(p,d,q)(P, D, Q)_m$
- 시계열 데이터의 계절성을 처리할 수 있는 모델
- ARIMA 모델에 계절성 변동을 반영한 모델 

#### 불안정 시계열에서 하이퍼파라미터 결정 방법
1. d, D 차분 결정 
2. p, q 결정
3. AIC, BIC 같은 정보 기준, 그리드 서치 -> 최종 선택 

### 3. 다변량 시계열 분석

##### <span style="background-color:#FFC0CB; color:#000000; padding:4px; border-radius:5px;"> 3-1. VAR
: 자신의 과거 값과 다른 변수들의 과거 값 함께 사용 => 미래를 예측
#### 활용
1. 충격 반응 함수
: 한 시계열에 특정 시점에 충격이 발생 -> 다른 시계열에 시간에 따라 어떤 영향을 줄까?
2. 예측 오차 분산 분해
: 어떤 시계열이 상대적으로 어떤 영향을 끼치고 있는지 중요도 산출, 각 변수가 미래의 움직임을 예측하는데 어느 정도 기여하는지 평가 
### VAR 모델의 장점은?
- 변수 간 양방향 영향 분석 가능
- 도메인 지식 없어도 여러 변수 함께 예측
- 동시성 문제 X
### VAR 모델의 단점은?
- 변수가 많으면 파라미터 폭발!!
- 장기적 관계 정보 손실 가능,,
- 상관관계와 인과관계가 같지 않음 

### 4. 딥러닝을 활용한 시계열 분석

#### 왜 등장했나?
전통 모델의 한계
1. 데이터 많으면 처리 속도 느림
2. 크기가 고정된 데이터만 가능
3. 비선형 패턴 포착 어려움 


### RNN 계열!

##### <span style="background-color:#FF69B4; color:#000000; padding:4px; border-radius:5px;"> 4-1. RNN 
: 이전 시점의 정보를 현재로 전달하는 순환 구조의 기본 신경망
#### 적합한 경우는?
- 짧~중간 길이 시계열
- 순차적 의존성 강한 데이터
- 자기상관이 큰 데이터
#### 부적합?
- 매우 긴 시계열 (장기 의존성 문제가 있었디!)
- 병렬 처리가 필요한 대규모 데이터 

##### <span style="background-color:#FF69B4; color:#000000; padding:4px; border-radius:5px;"> 4-2. LSTM
: 게이트 메거니즘으로 장기 기억을 유지할 수 있도록 개선한 RNN
#### 적합한 경우?
- 장기 의존성이 중요한 경우
- 중요한 과거 이벤트가 먼 미래에 영향 주는 경우 


##### <span style="background-color:#FF69B4; color:#000000; padding:4px; border-radius:5px;"> 4-3. GRU
: LSTM을 단순화, 더 빠르게 학습하면서도 비슷한 성능을 내는 모델 
#### 적합한 경우
- 장기 의존성 중요하면서 학습 속도 중요한 경우
- 데이터 상대적으로 적을 때
- 실시간 처리 필요한 경우 

### Transformer 계열!!

##### <span style="background-color:#FF69B4; color:#000000; padding:4px; border-radius:5px;"> 4-4. Transformer
: 어텐션 메커니즘, 모든 시점을 동시에 참조해 병렬 처리하는 모델 
#### 적합한 경우
- 매우 긴 시계열
- 전역 패턴이 중요한 시계열
- 대규모 데이터
- 다변량 시계열
#### 부적합
- 짧은 시계열
- 계산 자원이 제한된 경우 


### Foundation Model!!!

##### <span style="background-color:#FF69B4; color:#000000; padding:4px; border-radius:5px;"> 4-5. Foundation Model
: 거대한 데이터로 사전 학습하여 다양한 문제를 해결할 수 있는 범용 모델
- 특정 문제 적용 X
- 범용적
- 사전 학습 후 파인튜닝으로 특화 

##### <span style="background-color:#FF69B4; color:#000000; padding:4px; border-radius:5px;"> 4-6. LLM 기반 모델
: 대규모 언어 모델의 패턴 인식 능력 => 시계열에 활용하는 잡근법

##### <span style="background-color:#FF69B4; color:#000000; padding:4px; border-radius:5px;"> 4-7. 자체 거대 모델
: 대규모 시계열로 사전 학습 => 다양한 예측 문제에 범용적 적용 가능 모델
- 시계열 전용으로 처음부터 학습
- 대규모 시계열 데이터 수집 핵심 

#### 딥러닝 모델, 어떻게 선택할까???
- **단순함 vs 성능**: RNN < LSTM/GRU < Transformer
- **해석 가능 vs 정확도**: 통계 모델 > 딥러닝 (단, TFT는 어텐션으로 해석 가능)
- **학습 시간 vs 예측 성능**: 간단한 모델은 빠르고, 복잡한 모델은 정확함
- **데이터 크기**: 적으면 통계 모델/RNN, 많으면 Transformer
