# 1. 시계열 데이터란
## 시계열 데이터
= 순차적인 시간의 흐름에 따라 수집한 데이터
= 시퀀스 데이터의 일종으로, 순서가 중요한 역할을 함!

### 종단면 & 횡단면 데이터

- 종단면(시계열) 데이터: 시간의 흐름에 따라 한 변수에 대한 자료 수집
    - 시간 의존성
    - 개체/단위 추적
    - 패턴/동향 분석
    - 효과/원인 분석

- 횡단면(비시계열) 데이터: 한 시점에 여러 변수에 대한 자료 수집
    - 시간 의존성 없음
    - 다중 변수
    - 비교 분석
    - 상관관계 분석

### 단변량 & 다변량 시계열

- 단변량: 시간에 따라 연속적으로 측정된 '단일 변수'

- 다변량: 같은 시간에 기록된 '둘 이상의 변수'로 이루어진 시계열 데이터

## 시계열 분석과 그 목적
= 미래 데이터 예측. 

## 시계열 데이터의 특징

- 시간의존성: 시간에 따라 변화하는 데이터 => 데이터의 순서를 바꾸면 그 의미가 달라짐

- 계절성과 추세: 시간에 따라 장기적으로 증가/감소하는 추세 + 일정 주기로 반복하는 패턴

- 자기상관성: 이전 시점의 값이 이후 시점의 값에 영향을 미침
    - 자기 자신끼리 상관관계를 갖는 특성
    - 자기 자신이 가진 패턴/반복이 존재함

    - **자기상관성 때문에 회귀분석이 불가능함**
        1. **통계적 추론의 제약 및 오류: ** <br>
            오차항 간에는 상관이 없어야 한다는 (독립성) 가정이 깨짐. <br> 
            오차들이 서로 의존한다는 상관관계가 존재하기 때문

        2. **예측의 어려움 & 모델 복잡성 증가: ** <br>
            잔차가 패턴을 가지므로 미래 값 예측이 왜곡 -> 오차 누적으로 신뢰도 하락 <br>
            단순 회귀 모형은 적절하지 않으므로 AR, ARIMA 등의 시계열 모델로 대체 필요

## 시계열 데이터의 분석요인

1. 추세 요인: 데이터 전반에 걸쳐 지속적으로 증가/감소하는 경향

2. 계절 요인: 일정한 주기에 따라 반복되는 패턴

3. 순환 요인: 계절성보다 길고 불규칙한 주기로 반복되는 변동

4. 불규칙 요인: 추세/계절/순환을 제외하고 남은 예측 불가능한 변동. 

## 시계열 분석의 워크플로우
= 데이터 전처리(결측치 보간, 이상치 제거, 요소분해 등 포함) -> 정상성 판단 -> 모델링 및 예측

# 2. 시계열 데이터 전처리

## 이상치/결측치 처리

- 결측치 : 특정 시점에서 값이 측정되지 않았거나 누락된 상태
    - 결측 존재 시 추세/계절성이 왜곡될 수 있으므로, 적절한 대치/보간이 필요함

- 이상치 : 값이 측정되엇으나, 측정된 값이 일반적이지 않은 경우
    - 측정 오류, 이벤트 값일 수도 있음
    - 신호/중요한 의미를 가질 수 있으므로 신중한 접근이 필요함 -> 사전 지식이 필요한 부분!

### 시계열 데이터 전처리의 필요성
-> 데이터 확인/시각화 -> 결측치 확인 및 처리 -> 이상치 확인 및 처리 -> 요소 분해

### 결측값 처리
= 삭제, 단순 대치, 보간, 머신러닝 기법 등

#### 보간법
= 주어진 데이터 포인트의 값을 사용하여 두 개 이상의 데이터 포인트 사이에서 누락된 값을 수학적으로 추정
    - 결측/이상치 구간이 짧은 경우, 보간의 정확도가 높음
    
** 보간 방법과 종류

- 선형 보간법: 두 점 사이를 연결 -> 그 사이의 값을 추정. **직선 방정식**을 이용
    - 데이터 간 관계가 선형적일 때 적합
    - 빠르고 간단, 직관적
    - 급격한 변화가 있거나 비선형인 데이터에서는 부적합할 수 있음

- 다항식 보간법: 주어진 데이터 포인트를 다항식 함수로 근사하여 값을 추정
    - 비선형 관계 모델링
    - 데이터 포인트가 많을 수록 정확한 보간을 제공함
    - 고차 다항식의 경우 과적합 가능성 있음
    - 계산이 복잡하고 비효율적일 수 있음
        
- 스플라인 보간법 : 구간별 다항식을 사용해 각 구간에서 보간을 수행
    - 다소 복잡, 연속적인 미분 가능한 곡선을 제공하므로 곡선형 데이터에 유용함
    - 과적합 적음, 고차 다항식에 비해 안정
   - 높은 복잡성, 계산량 많음. 
    - 매우 비선형적인 데이터의 경우 잘 동작하지 않을 수 있음

- 조각별 상수 보간법: 데이터를 여러 구간으로 나누고, 구간별로 동일한 상수 값으로 결측 채움
    - 이전 값 유지(직전 관측치로 채우기), 다음 값 활용(직후 관측치로 채우기)
    - 최근접 이웃 보간(앞/뒤 기준 거리상 가장 가까운 값으로 채우기)

### 이상치 탐지 및 처리

#### 이상치 탐지 방법(전처리)
    
- IQR 기반 : Q1, Q3에서 IQR*1.5 범위 바깥에 있으면 이상치로 처리
    - 주기성이 없거나 안정적인 구간의 값들에서 이상치를 탐지하는 경우 사용
    - 하루 등 구간으로 나누어 IQR 계산 시 추세의 영향을 줄일 수 있다. 
    - = 추세/계절성이 심한 데이터에는 부적합
    
- Z 스코어 : 값이 평균으로부터 몇 표준편차 떨어져 있는가 (보통 3)
    
- STL 분해 : 추세/계절성/잔차로 분해
    - 계절성/추세가 뚜렷한 시계열에 적합
    - 계산 비용이 큰 문제 있음

#### 이상치 탐지 모델

- 분류/회귀 트리(CART): 데이터를 분할해서 예측값을 산출. 예측 오차가 일정 임계값 이상인 지점을 이상치로 간주

- 클러스터링 기반 이상 탐지: K=means, DBSCAN, Isolation Forest 등을 활용해 데이터를 군집화, 정상치와 이상치가 서로 다른 군집에 모이게 됨. 다변량 데이터에 유용

- 오토인코더: 인공신경망 기반 비지도 학습 모델. 모델에 입력을 넣으면 정상 패턴은 잘 복원, 이상치는 복원오차가 커져 임계값 이상이면 이상치로 판정함

### 평활화
= 노이즈를 줄이고 전체 흐름/추세를 명확하게 파악하기 위해 데이터를 부드럽게 만드는 기술

### MA (이동평균)
-> 노이즈 제거, 추세 파악을 위해 윈도우(n) 기반으로 평균을 구하는 것

- 단순 이동평균: 기간 내 모든 값 평균

- 가중 이동평균: 최근 값에 더 큰 가중치 부여

- 지수 이동평균: 최근 데이터에 **지수적으로** 더 많은 가중치 부여

## 요소분해 (시계열 분해)
= 주어진 시계열 데이터를 개별 성분으로 분해하는 과정

= 각 요소를 분리하여 각각의 패턴을 이해하기 위해 분해함!

#### 가법모형
⭑ [시계열 = 추세 + 계절성 + 불규칙성]의 가법(+)으로 설명된다고 가정

- 시계열 전체에서 계절성의 진폭이 일정할 때 적합함

#### 승법모형
⭑ [시계열 = 추세 x 계절성 x 불규칙성]의 승법(x)으로 설명된다고 가정

- 식 전체에 로그를 씌우면 가법 형태로도 만들 수 있음
- 추세가 커질수록 계절성이나 불규칙성의 효과도 비례해서 커지는 구조
- 시간에 따라 진폭이 달라질 때 사용하기 적합함

# 3. 정상성 판단
## 정상성이란
= 시계열 데이터의 평균/분산같은 통계적 특성이 시간에 따라 일정한 수준으로 유지되는 것 
= 정상성 여부에 따라 정상/비정상 시계열로 나눌 수 있음

- 정상 시계열 : 평균, 분산이 일정함. 추세가 거의 없음. 

- 비정상 시계열: 평균이 시간에 따라 변하고 분산도 커지는 구조

## 강한 정상성과 약한 정상성

- 강정상성 = 적률(평균, 분산, 왜도, 첨도)이 시간에 관계없이 모두 일정함
    - 현실에서 찾아보기 매우 어려움: 실생황에서는 잡음/외부요인 영향으로 불안정한 데이터가 많음

- 약한 정상성
    <br>아래 조건을 만족: 
    - 평균이 모든 시점에 대해 일정함
    - 분산이 모든 시점에 대해 일정함
    - 공분산이 시점에 의존하지 않고, 단지 시차에만 의존함

## 검정을 통해 정상성 파악하기

- 단위근 검정: 충격이 영구적으로 남는, 과거 충격이 누적되어 예측이 어려운 것. 

- ADF 검정: 단위근 존재 여부를 통해 시계열의 정상성 판단

- KPSS 검정: 단위근 기반, ADF와 귀무가설/대립가설이 반대! -> H0이 정상 시계열

- KPSS vs ADF: 두 검정의 결론이 다를 수 있음!
    - 확정적 추세: 추세가 완벽하게 예측 가능함 (ADF 모델 권장)
    - 확률적 추세: 추세선이 시간에 따라 이동함(랜던 워크 성격, KPSS 모델 권장)
    - 확정적 추세가 있을 경우 결과가 다르게 나올 수 있다!

## 정상성 확보를 위한 여러가지 방법들
### 정상적으로 만드는 방법들
=> 추세 제거, 계절성 제거, 분산을 일정하게 만들어야 함

**1. 추세 제거 = 차분**
    - 대부분의 비정상적 시계열에서는 값이 누적되는 구조를 가짐
    - 누적된 값을 빼줌으로서 정상적 시계열을 얻을 수 있음

**2. 계절성 제거 = 계절 차분**
    - t시점의 값에서 t-m시점의 값을 빼는 것 (m=주기의 길이)
    - ex. 12일마다 반복 => m=12인 차분을 진행
    - 계절 차분 시 백색소음과 같은 형태의 데이터로 바뀜

**3. 분산 안정화**
    
(1) 로그변환
- 밑이 10인 로그를 적용 -> 변화폭이 줄어들어 비율적인 변화가 강조됨
- 분산이 일정해지는 효과
(2) 박스-콕스 변환
- 람다 값이 데이터에 따라 달라질 수 있음

# 시계열 정리
=> 정상성 확보 → 패턴 파악 → 적절한 모델링

- 정상성 판단: 단위근 검정(ADF, KPSS등)

- 비정상인 경우 정상성 확보: 차분, 계절 차분, 로그변환, 박스콕스 변환

- 정상성 확보 후 시계열 구조에 맞는 모델 적용: 2차시에 배울 OLS, AR, MA, ARMA, SARIMA 등등
