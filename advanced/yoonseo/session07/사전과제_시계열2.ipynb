- 단변량과 다변량: 단일 변수 or 둘 이상의 변수로 구성된 시계열적 데이터

    - 자기상관성: 이전 시점의 값이 이후 시점의 값에 영향을 미치는 것
    <br> 자기상관성으로 인해 시계열 데이터에서는 회귀를 사용하지 못함!!

- 시계열 데이터의 분석 요인: 추세 요인/ 순환 요인/ 계절 요인/ 불규칙 요인

- 시계열 분석의 기본 프로세스:

    2. 정상성 판단: 정상성을 확보한 후 다음 단계를 진행
    <br> 시간에 상관없이 일정한가? - KPSS, ADF 사용
    <br> 비정상인 경우 처리 방법 - 차분(추세 제거), 계절 차분(계절성 제거), 로그 변환 or 박스콕스 변환(분산 제거)

    3. 모델링 및 예측: 프로세스에 따라 모델링 진행

# 1. 단변량 시계열 분석

## 1.1 안정적 시계열 분석

### <span style="background-color:#E1CBCB;">자기회귀 모델 (AR)</span>
= 과거 자신의 데이터/값들(=독립변수)로 현재 값을 예측
= 시계열 데이터의 자기상관성을 활용한 모델

$AR(p):  y_{t}=c +\phi _1y_{t-1}+\phi_2y_{t-2}+...+\phi _py_{t-p}+\varepsilon _t$

- AR(p) : p차 자기 회귀 모형, 과거 p개 시점의 선형 조합을 이용해 예측을 진행
    -> $AR(1):  y_{t}=c +\phi_1y_{t-1}+\varepsilon _t$

    -> (과거의 값 * 영향을 미치는 정도) + 잔차 + 상수항


### <span style="background-color:#E1CBCB;">이동평균 모델 (MA)</span>
= 평활화의 MA 기법과는 다름!!!

= 과거 시점의 잔차(=독립변수)로 현재 값을 예측

= 어제의 충격이 오늘에 영향을 미친다 => 과거의 잘못 예측한 정도가 현재 시점에 반영되는 구조

$MA(q): y_t = \mu +\varepsilon _t+ \theta _1\varepsilon_{t-1}+\theta_2\varepsilon_{t-2}+...+\theta_q\varepsilon_{t-q}$

- MA(q) : q차 이동평균 모형, 과거 q개의 예측 오차에 대한 선형 결합으로 이루어짐
    -> $MA(1):  y_{t}=\mu +\varepsilon _t+ \theta _1\varepsilon_{t-1}$

    -> 1차 이동평균 모형 = 평균 + 잔차 + (과거 잔차 * 과거 잔차가 영향을 미치는 정도)

### AR, MA 모델의 전제 조건

#### 1. 정상성
= AR, MA 모두 안정적 시계열 모델 => 정상성 만족이 그 전제임

- AR 모형의 정상성 조건: 특성 방정식의 모든 해의 절댓값이 1 이상이어야 함!
    - ADF, KPSS 는 데이터의 정상성을 파악하는 방법 <-> **모델의 정상성**을 확보하기 위해서는 특성방정식이 중요함

- MA 모형의 정상성 조건: 잔차(백색 잡음)이 이미 정상성을 만족함 => 항상 정상성을 만족함!

#### 2. 가역성
= MA 모형을 AR(inf) 형태로 재표현할 수 있는지 여부

= 특성 방정식($1 + \theta_1 z + \cdots + \theta_q z^q = 0$) 을 이용해 확인 (모든 해의 절댓값 > 1)

### ARMA 모형
= AR(p) + MA(q) -> ARMA(p, q) 결합 형태

### 차수 결정 방법: 자기상관함수
= AR, MA, ARMA 의 차수(p, q)를 결정하기 위해 현재 값과 어떤 과거 값이 관계를 가지는지 확인한다

- ACF : $ACF(k) = \text{Corr}(y_t, y_{t-k})$
    - 자기상관계수, q값(오차의 개수)를 결정
    - 시차 k에 대해 두 관측치 간의 상관관계를 나타내는 함수

- PACF : $PACF(k) = Corr(e_t, e_{t-k})$
    - p(몇개의 시점을 활용할지)의 값을 결정하는 방법
    - 두 시점 사이의 직접적인 영향만 파악

## 1.2 불안정 시계열 분석

**불안정한 데이터의 시계열 분석 프로세스**
1. 비정상 시계열 데이터 (추세 or 계절성 존재)
2. 정상성 확보 (차분 or 계절성 차분)
3. 정상 시계열
4. ARMA를 사용해 **안정적 시계열 분석**
-> ARMA 에 정상성 확보 과정이 추가된 형태로 이루어짐!!

### <span style="background-color:#E1CBCB;">1. ARIMA</span>
= 차분, 변환을 통해 AR, MA, ARMA로 정상화하는 모델 (d=차분 횟수)

* ARIMA 모형의 특수한 경우
- 상수가 없는 ARIMA(0, 1, 0) = 랜덤워크
    - 이전 값에 백색잡음(무작위 충격)만 더해짐 = 예측 불가!!!
    - 이전의 값이 영원히 영향을 줌 + 방향성은 없음
- 상수가 있는 ARIMA(0, 1, 0) = 표류를 포함하는 랜덤워크
    - 랜덤워크 모델에 상수항이 추가된 모델의 형태
    - 상수항이 있어 방향성은 존재함

### <span style="background-color:#E1CBCB;">2. SARIMA (seasonal ARIMA)</span>
= $SARIMA(p,d,q)(P, D, Q)_m$

= 시계열 데이터의 계절성을 처리할 수 있는 모델, ARIMA 모델에 계절성 변동을 반영

- 하이퍼파라미터
    - (p, d, q) : 기존 ARIMA의 파라미터. 비계절. 
    - (P, D, Q) : 계절 ARIMA 파라미터
    - m : 계절성 주기의 길이

- 계절성 데이터에서의 ACF와 PACF
    - 계절 주기가 m일 때 ACF에서 m의 배수 lag에 급등하는 형태를 가짐
    -> 계절성 차분 먼저, 이후에 비계절성 차분

- 불안정 시계열에서의 하이퍼파라미터 결정 방법
    1. d, D(차분) 결정: ADF로 정상성 확보 여부 판단 후 결정. 대부분 0 or 1
    
    2. p, q 결정: AIC, BIC 지표 기준으로 그리드 서치를 진행한 후 결정. 보통 0~3 이내

# 2. 다변량 시계열 분석

## <span style="background-color:#E1CBCB;">VAR</span>
= 자신의 과거 값과 **다른 변수들의 과거 값**을 함께 사용하여 미래를 예측

$X_t = a_{11} X_{t-1} + a_{12} Y_{t-1} + ε_{x,t}$
$Y_t = a_{21} X_{t-1} + a_{22} Y_{t-1} + ε_{y,t}$

-> 이처럼 X, Y가 서로의 과거를 함께 사용하는 구조!

- 적합한 경우
    - 서로 영향을 주고 받는 지표의 경우 (ex. 경제 지표 - GDP, 소비, 투자 등)
    - 여러 자산의 값이 상호작용하는 경우
    - 변수 간 **상호의존성**이 중요한 경우

- 부적합한 경우
    - 변수가 너무 많은 경우 파라미터 폭발 가능성 있음
    - 변수 간 관계가 명확 X
    - 장기 예측이 필요한 경우

### VAR 모델을 활용한 분석 방법

1. 충격 반응 함수(IRF): 한 시계열에 특정 시점에서 충격이 발생할 때, 다른 시계열에 시간에 따라 어떤 영향을 주는지 분석

2. 예측오차 분산 분해(Variance Decomposition) : 어떤 시계열이 상대적으로 어떤 영향을 끼치고 있는지 중요도를 산출하기 위해서 진행

### VAR 모델의 장단점
- 양방향 영향 분석 가능
- 도메인 지식이 없어도 여러 변수를 함께 예측 가능
- 동시성 문제 없음

- 변수가 많으면 파라미터 폭발, 장기간 정보 손시 ㄹ가능
- 상관관계 != 인과관계

# 3. 딥러닝을 활용한 시계열 분석

=> 전통 모델의 한계(비선형, 크기 고정, 처리 속도 해결)를 딥러닝을 통해 해결!

## RNN 계열

### 1. RNN

- 10~100 시점 정도 짧거나 중간 길이의 시계열
- 순차적 의존성이 강한 데이터
- 자기상관이 큰 데이터

<-> 매우 긴 시계열이나 병렬 처리가 필요한 경우 X

### 2. LSTM
- 장기 의존성이 중요한 100~1000시점의 데이터
- 중요한 과거 이벤트가 먼 미래에 영향을 주는 경우

### 3. GRU
- LSTM처럼 장기의존성이 필요하지만 학습 속도가 중요한 경우
- 데이터가 상대적으로 적을 때, 실시간 처리가 필요한 경우

## Transformer 계열
- 매우 긴 시계열(1000 이상), 전역 패턴이 중요한 경우
- 대규모 데이터 (병렬 연산 활용)
- 다변량 시계열 - 변수 간 상호작용 학습

<-> 짧은 시계열은 과적합 가능, 계산 자원이 제한된 경우에도 부적합

* Vanilla Transformer, Informer, Autoformer, Temporal Fusion Transformer 등이 있음

## Foundation model
= 거대한 데이터로 사전 학습하여 다양한 문제를 해결할 수 있는 범용 모델

- 특정 문제 전용이 아닌, 범용적인 모델임!
- 사전 학습 후 파인튜닝으로 특화시키는 과정을 거침

### LLM 기반 모델

- Without Adaptation - LLM이 이미 지식을 가지고 있다고 가정. 프롬프트 엔지니어링에 집중함

- Adapt LLM: llm을 시계열 데이터로 파인튜닝한 모델. 일부 레이어만 학습하거나 전체를 파인튜닝

- Adapt to LLM: 시계열을 텍스트로 변환한 후 llm을 사용하는 방식. 텍스트화

### 자체 거대 모델
= 대규모 시계열 데이터로 사전학습, 여러 예측 문제에 범용적으로 적용

TimeGPT, TimesFM, Moirai 등

## 딥러닝 모델 선택 방법
- **트레이드오프**!!
    - 단순 vs 성능: RNN < LSTM/GRU < Transformer
    - 해석 가능 vs 정확도: 통계 모델 > 딥러닝
    - 학습 시간 vs 예측 성능: 간단하면 빠르고 복잡하면 정확해짐
    - 데이터 크기: 적으면 통계/RNN, 많으면 Transformer

-> 시계열 길이에 따라 추천 모델을 고를 수도 있음!

- 데이터가 적은 경우 딥러닝 < ARIMA 등의 통계모델 일수 있음!!
- 장기의존성 학습은 GRU < **LSTM**
- 해석이 중요한 비즈니스 문제에는 TFT 고려-> 어텐션으로 해석 가능!
