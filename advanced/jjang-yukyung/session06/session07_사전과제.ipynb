## <span style="background-color:#E0F7FA; color:#1565C0; padding:4px; border-radius:5px;"> 시계열 개념 </span>

```
종단면 데이터, 횡단면 데이터

> 데이터의 구조

- 종단면 데이터: 시간에 흐름에 따라 대상을 관측한 데이터. < 시계열 데이터가 여기 속함!
  - ex) 유경기업의 2010년부터 2025년까지의 연도별 매출액
  
- 횡단면 데이터: 특정 동일 시점에 여러 대상을 관측한 데이터
  - ex) 2023년 11월 기준, 한국 내 100개 기업의 매출액

```
단변량 시계열, 다변량 시계열

> 변수의 수에 따른 분류

- 단변량 시계열: 시간의 흐름에 따른 변수가 하나인 경우
  - 과거 자기 자신을 통해 미래를 예측 (오늘 주가를 보고 내일 주가 예측)

- 다변량 시계열: 시간의 흐름에 따른 변수가 둘 이상인 경우
  - 변수 간 상호작용과 인과관계를 고려하여 예측 (환율, 금리, 유가를 모두 고려해서 주가 예측)

```
시간의존성

현재의 데이터가 과거 시점의 데이터에 영향을 받는 성질

```
계절성

일정한 빈도(1년, 분기, 월, 주 등)로 주기적인 패턴이 반복되는 성질

```
추세

장기적으로 데이터가 상승하거나 하강하는 경향

```
자기상관성

시차(Lag)를 둔 본인 데이터와의 상관관계
- 예: 어제의 날씨와 오늘의 날씨 간의 관계)

### 시계열 분해의 4가지 구성 요소

$$Y_t = T_t + S_t + C_t + I_t$$

```
추세 요인

- 장기적인 변동 경향
- 우상향, 우하향 등 데이터의 거시적 방향성

```
계절 요인

- 고정된 주기의 반복 변동
- 12개월, 7일 등 주기가 일정함 (예: 아이스크림 판매량)

```
순환 요인

- 고정되지 않은 주기의 파동
- 경기 변동(호황/불황)처럼 주기가 길고 불규칙함

```
불규칙 요인

- 예측 불가능한 우연 변동
- 노이즈(Noise), 오차(Error)에 해당하며 분석에서 제거 대상

## <span style="background-color:#E0F7FA; color:#1565C0; padding:4px; border-radius:5px;"> 시계열 전처리 </span>

> 분석 모델링에 들어가기 전, 데이터의 품질을 높이기 위한 과정

```
결측치 처리-보간법

비어있는 값의 앞뒤 데이터를 사용하여 중간값을 추정
- 선형 보간: 앞뒤 점을 직선으로 연결
- spline 보간: 곡선으로 부드럽게 연결
- 시계열적 특성 반영: 이전 주기의 값을 참고하거나 이동평균을 사용

```
이상치 탐지

- 갑작스러운 스파이크나 데이터 오류를 찾아냄
- 이동평균이나 중위수 필터를 사용해서 벗어나는 범위를 설정 및 탐지

```
평활화(smooting)-이동평균

불규칙 요인(노이즈)을 제거하여 추세를 뚜렷하게 보기 위한 방법
- 이동평균: 일정 기간의 평균을 구하여 그래프를 부드럽게 만든다

```
요소분해

- 데이터에서 추세, 계절성, 잔차를 분리해내는 전처리 과정
  - 이를 통해 시계열이 가법 모형인지 승법 모형인지 파악하고 계절성을 제거하는 데 사용됨!

## <span style="background-color:#E0F7FA; color:#1565C0; padding:4px; border-radius:5px;"> 정상성 </span>

시계열 분석(특히 AR, MA, ARIMA 등)을 수행하기 위한 기본 전제 조건

```
정상성의 정의

시계열의 확률적 성질(통계적 특성)이 시간의 흐름에 따라 변하지 않는 상태를 의미

조건
- 평균($\mu$)이 일정하다. (모든 시점에서 동일)
- 분산($\sigma^2$)이 시점에 의존하지 않고 일정하다.
- 공분산(Covariance)은 시점($t, s$)이 아닌 시차(Lag, $k$)에만 의존한다.

```
강한 정상성과 약한 정상성

- 강한 정상성: 모든 모멘트(평균, 분산 등)를 포함한 결합 확률 분포가 시간이 지나도 불변함. (조건이 너무 까다로워 현실 데이터에는 적용이 어려움)
- 약한 정상성: 일반적으로 말하는 정상성. 평균과 분산이 일정하고 자기공분산이 시차에만 의존하는 상태.

```
단위근의 시계열적 의미

시계열적 의미: 시계열 데이터가 정상성을 띠지 않는(비정상성) 대표적인 원인

자기회귀 식 $Y_t = \alpha Y_{t-1} + \epsilon_t$ 에서 계수 $\alpha = 1$인 경우

단위근이 존재하면 외부 충격(Shock)의 효과가 사라지지 않고 영원히 지속됨! 즉, 평균과 분산이 시간 흐름에 따라 발산하는 것.

```
단위근 검정

데이터가 정상인지를 통계적으로 검정하는 방법

ADF 검정:
- 귀무가설: ($H_0$): 단위근이 존재한다 (비정상 시계열이다).
- 대립가설($H_1$): 단위근이 없다 (정상 시계열이다).
- P-value < 0.05 이면 귀무가설 기각 → 정상 시계열로 판단.

```
확정적 추세와 확률적 추세

추세가 존재하면 정상성을 만족하지 못하므로 추세 종류에 따라 처리 방법이 다르다.

확정적 추세
- 시간 $t$의 함수로 표현 가능한 추세 ($y_t = \alpha + \beta t + \epsilon_t$)
- 시간이 지나도 추세선 주위로 회귀함
- 추세 제거하면 정상성이 확보됨

확률적 추세
- 무작위적인 움직임이 누적된 추세 ($y_t = y_{t-1} + \epsilon_t$)
- 충격이 오면 영구적으로 데이터 경로가 바뀜
- 차분하면 정상성 확보됨

>> 분산이 일정하지 않은 경우에는 로그 변환을 먼저 수해앟여 변동폭을 줄인 후 차분 진행!

```
정상성 확보 방법

- 그래프 확인 (Plotting): 눈으로 보기에 변동폭이 커지는지, 추세가 있는지 확인.
- 변동폭 제어: 갈수록 위아래 진동이 커진다면 $\rightarrow$ 로그 변환.
- 계절성 제거: 주기적 패턴이 뚜렷하다면 $\rightarrow$ 계절 차분 ($D=1$).
- 추세 제거: 여전히 상승/하강 추세가 있다면 $\rightarrow$ 1차 차분 ($d=1$).
- 검정 (ADF Test): 통계적으로 정상성이 확보되었는지 확인 (P-value < 0.05).
  - 참고: ARIMA 모델에서 ARIMA(p, d, q)의 **d**가 바로 여기서 수행한 차분의 횟수를 의미
